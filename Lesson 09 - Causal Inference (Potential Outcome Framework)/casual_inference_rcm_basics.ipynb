### Причинно-следственный анализ

Мир данных претерпевает значительные изменения, и область науки о данных, некогда названная «самой сексуальной работой 21-го века» журналом Harvard Business Review, теперь сталкивается с новыми вызовами. Эйфория вокруг искусственного интеллекта (ИИ) и машинного обучения (ML) привела к появлению «пены» — завышенных ожиданий и обещаний, которые не всегда соответствуют реальности. В то время как ML-модели великолепно справляются с задачами прогнозирования, например, переводом языков или распознаванием лиц, они оказываются бессильными, когда речь заходит о вопросах «что, если» (контрфактических вопросах). Именно здесь на сцену выходит **каузальный вывод**.

Каузальный вывод — это не новая область; экономисты, биостатистики и психологи десятилетиями занимались «старомодной» наукой о данных, пытаясь понять истинное влияние образования на доходы, насыщенных жиров на сердечные приступы или слов одобрения на счастье в браке. В своей основе каузальный вывод стремится ответить на вопросы о **причинно-следственных связях**, которые пронизывают как повседневные проблемы (как увеличить продажи?), так и личные дилеммы (приводит ли иммиграция к росту безработицы?). Для решения таких вопросов недостаточно полагаться на простые корреляции.

Основное правило, которое многие слышали с детства, гласит: **[«ассоциация не является причинностью»](https://en.wikipedia.org/wiki/Correlation_does_not_imply_causation)**. Это означает, что простое наблюдение, что две переменные движутся вместе (коррелируют), не позволяет сделать вывод, что одна из них вызывает другую. Эта ошибка известна как логическая ошибка **«cum hoc ergo propter hoc»** («вместе с этим, значит из-за этого»).

Простое наблюдение корреляции недостаточно для установления причинной связи по нескольким причинам:

*   **Обратная причинность (Reverse Causation)**: Иногда кажется, что А вызывает В, но на самом деле В вызывает А. Например, школы, в которых студенты используют планшеты, показывают лучшие результаты. Наивный вывод может заключаться в том, что планшеты повышают успеваемость. Однако, возможно, более богатые школы могут позволить себе планшеты, и они же могут позволить себе лучших учителей и другие ресурсы, которые на самом деле приводят к лучшим результатам. Другой пример: быстрое вращение ветряных мельниц коррелирует с сильным ветром, но это не означает, что мельницы производят ветер; скорее, ветер заставляет мельницы вращаться. Низкий уровень холестерина может быть связан с повышенной смертностью, но это не потому, что низкий холестерин опасен; скорее, некоторые серьезные заболевания, такие как рак, вызывают как низкий холестерин, так и повышенную смертность.

*   **Общий фактор (Confounder)**: Часто корреляция между переменными А и В возникает из-за третьего, «скрытого» или **«смешивающего» фактора (confounding variable)**, который влияет как на А, так и на В.
    *   Например, люди, спящие в обуви, часто просыпаются с головной болью. Это не означает, что обувь вызывает головную боль; скорее, третий фактор — употребление алкоголя перед сном — приводит к обоим этим явлениям.
    *   Аналогично, рост продаж мороженого коррелирует с увеличением числа утоплений. Здесь общий фактор — летняя жара, которая одновременно стимулирует продажи мороженого и увеличивает число людей, купающихся в воде.
    *   Исторический пример: европейцы в Средние века считали, что вши полезны для здоровья, так как они покидали больного человека до проявления симптомов. На самом деле, небольшое повышение температуры тела (например, при лихорадке) заставляло вшей искать нового хозяина; заметные симптомы проявлялись позже, создавая иллюзию причинности.

*   **Случайное совпадение (Coincidental Relationship)**: Две переменные могут коррелировать просто по случайности, без какой-либо реальной связи. Чем больше мы исследуем данных, тем выше вероятность обнаружить случайные корреляции. Например, результат последней домашней игры футбольной команды Washington Commanders перед президентскими выборами в США коррелировал с исходом выборов с 1936 по 2000 год, хотя очевидно, что футбольные матчи не влияют на выборы.

Каузальный вывод помогает нам выйти за рамки простой ассоциации, используя строгие методологии для выявления истинных причинно-следственных связей. В основе этого подхода лежит **фреймворк потенциальных исходов (Potential Outcomes Framework)**, также известный как **модель Рубина (Rubin-Causal-Model)**.

В рамках этой парадигмы для каждой единицы (например, человека, школы) рассматриваются два **потенциальных исхода**: $Y_{1i}$ — исход, если $i$-ая единица подверглась воздействию, и $Y_{0i}$ — исход, если та же единица *не получила* воздействие. Получает, что наблюдаемый исход можно записать как $Y^{obs} = Y(1) \times T + Y(0) \times (1-T)$, где $T$ - флаг воздействия

Но существует **фундаментальная проблема каузального вывода**: мы не можем наблюдать оба потенциальных исхода для одной и той же единицы одновременно, поэтому кажется, что введение **контрфактических** исходов кажется бесполезным.

Поскольку индивидуальные причинные эффекты $\tau = Y_{1i} - Y_{0i}$ невозможно измерить, каузальный вывод фокусируется на **усредненноим эффекте лечения**, который может выражаться в разных метриках:

* **ATE (Average Treatment Effect)** - $E[Y_1 - Y_0]$.

* **ATT (Average Treatment on the Treated)** - $E[Y_1 - Y_0 \mid T = 1]$

* **ATU (Average Treatment on the Untreated)** - $E[Y_1 - Y_0 \mid T = 0]$

* **CATE (Conditional Average Treatment Effect)** - $E[Y_1 - Y_0 \mid X = x]$.

* **LATE (Local Average Treatment Effect)** - $E[Y_1 - Y_0 \mid \text{Compliers}]$

Различие между простой ассоциацией $E[Y|T=1] - E[Y|T=0]$ и истинным причинным эффектом $E[Y_1 - Y_0]$ объясняется **смещением (bias)**. Математически ассоциацию можно разложить как:

$E[Y|T=1] - E[Y|T=0] = E[Y_1|T=1] - E[Y_0|T=0] = E[Y_1|T=1] - E[Y_0|T=0] + E[Y_0|T=1] - E[Y_0|T=1] = \underbrace{E[Y_1 - Y_0|T=1]}_{ATT} + \underbrace{\{ E[Y_0|T=1] - E[Y_0|T=0] \}}_{BIAS}$

Смещение показывает, насколько группы с лечением и без лечения различались бы, если бы **никто из них не получал лечения**. Например, если школы с планшетами были бы лучше, чем школы без них, даже без планшетов (т.е., $E[Y_0|T=1]$ было бы больше, чем $E[Y_0|T=0]$), то наблюдаемая ассоциация будет смещена.


**Задача каузального вывода** — найти способы устранить это смещение, чтобы группы с лечением и без него стали сопоставимы. **Если смещение равно нулю ($E[Y_0|T=0] = E[Y_0|T=1]$), то ассоциация *становится* причинностью**. Это происходит, когда группы лечения и контроля сопоставимы *до* применения лечения.

**A/B-тестирование** является **золотым стандартом** в каузальном выводе. Его сила заключается в **рандомизации**, то есть случайном назначении влияния (тестовой группы) и контрольной группы.

При случайном назначении лечения, распределение потенциальных исходов в тестовой и контрольной группах одинаково. Это означает, что **тестовая и контрольная группы сопоставимы до начала влияния**, и **смещение исчезает**. В этом случае простая разница в средних значениях между группами (наивная оценка) становится несмещенной оценкой причинного эффекта (ATE).

Несмотря на свою мощь, A/B-тесты имеют ряд ограничений, которые могут быть решены или смягчены с помощью методов каузального вывода, особенно в **наблюдательных исследованиях**, где прямое A/B-тестирование невозможно или неэтично:

1.  **Невозможность рандомизации в реальных условиях**: Иногда провести рандомизированный эксперимент нереально или неэтично (например, нельзя случайным образом заставить людей курить, чтобы изучить влияние на рак легких, или подвергать детей насилию для изучения влияния на успеваемость). В таких случаях каузальный вывод предоставляет инструменты для работы с **наблюдательными данными**, где рандомизации не было.

2.  **Проблема скрытых смешивающих факторов (Confounding)**: В наблюдательных исследованиях, в отличие от рандомизированных экспериментов, существует высокая вероятность наличия **смешивающих факторов (confounders)** — переменных, которые влияют как на получение влияния, так и на исход. Он выходит за рамки возможностей машинного обучения, ориентированного на предсказание, позволяя отвечать на вопросы «что, если» и принимать обоснованные решения. ##### Линейная регрессия

При рассмотрении причинно-следственной связи мы увидели, что для каждого человека существует два потенциальных результата: $Y_0$ — результат, который был бы у человека, если бы он не был под воздействием, и $Y_1$ — результат, который был бы у него, если бы он был под воздействием. Установка $T$ в 0 или 1 материализует один из возможных результатов и делает невозможным для нас узнать другой. Это приводит к тому, что индивидуальный эффект лечения $\tau_i = Y_{1i} - Y_{0i}$ неизвестен.

$
Y_i = Y_{0i} + T_i(Y_{1i} - Y_{0i}) = Y_{0i}(1 - T_i) + T_i Y_{1i}
$

Итак, пока давайте сосредоточимся на более простой задаче — оценке среднего причинно-следственного эффекта. Имея это в виду, мы принимаем тот факт, что некоторые люди реагируют на воздействие лучше других, но мы также принимаем, что не можем знать, кто они. Вместо этого мы просто попытаемся посмотреть, работает ли воздействие **в среднем**.

$
ATE = E[Y_1 - Y_0]
$

Это даст нам упрощенную модель с постоянным эффектом воздействия $Y_{1i} = Y_{0i} + \kappa$. Если $\kappa$ положительно, мы скажем, что воздействие в среднем оказывает положительный эффект. Даже если некоторые люди отреагируют на него плохо, в среднем воздействие будет положительным.

Также напомним, что мы не можем просто оценить $E[Y_1 - Y_0]$ по разнице средних $E[Y|T=1] - E[Y|T=0]$ из-за смещения. Смещение часто возникает, когда результаты различаются по причинам, не связанным с самим воздействием. Один из способов увидеть это — посмотреть, как они различаются по потенциальному результату $Y_0$.

$
E[Y|T=1] - E[Y|T=0] = \underbrace{E[Y_1 - Y_0|T=1]}_{ATET} + \underbrace{\{ E[Y_0|T=1] - E[Y_0|T=0]\}}_{BIAS}
$

Ранее мы видели, как можно устранить смещение с помощью A/B-тестов, или **рандомизированных контролируемых испытаний** (РКИ), как их иногда называют. РКИ обеспечивает равенство экспериментальной и неэкспериментальной групп, и поэтому смещение исчезает. Теперь пришло время увидеть, как мы можем сделать всё это с помощью рабочей лошадки причинно-следственной связи: **линейной регрессии**.

Давайте посмотрим, как работает эта красота. В приведённом ниже коде мы хотим провести сравнительный анализ онлайн и очных занятий с помощью следующей модели:

$
exam_i = \beta_0 + \kappa \ Online_i + u_i
$

Это означает, что мы моделируем результат экзамена как базовый $\beta_0$ плюс $\kappa$, если занятие онлайн. Конечно, на результат экзамена влияют дополнительные переменные (например, настроение студента в день экзамена, количество часов обучения и так далее). Но нас не интересует понимание этих взаимосвязей. Поэтому вместо этого мы используем этот член $u_i$ для обозначения всего остального, что нас не интересует. Это называется ошибкой модели.

Обратите внимание, что $Online$ — это наш показатель лечения и, следовательно, фиктивная переменная. Она равна нулю, если лечение очное, и 1, если оно онлайн. Учитывая это, мы видим, что линейная регрессия восстановит $E[Y|T=0] = \beta_0$ и $E[Y|T=1] = \beta_0 + \kappa $. $\kappa$ будет нашим ATE.

````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import graphviz as gr
%matplotlib inline
````

````python
data = pd.read_csv("regression_df_one.csv").query("format_blended==0")

result = smf.ols('falsexam ~ format_ol', data=data).fit()
print(result.summary())
````

Это просто поразительно. Мы не только можем оценить ATE, но и бесплатно получаем доверительные интервалы и P-значения! Более того, мы видим, что регрессия делает именно то, что должна делать: сравнивает $E[Y|T=0]$ и $E[Y|T=1]$. Пересечение равно выборочному среднему при $T=0$, $E[Y|T=0]$, а коэффициент онлайн-формата — это выборочная разность средних $E[Y|T=1] - E[Y|T=0]$. Если добавить к пересечению ATE, то есть оценку параметра онлайн-формата, то получим выборочное среднее для исследуемой выборки: $78,5475 + (-4,9122) = 73,635263$.

**Теория регрессии**

Я не собираюсь слишком глубоко погружаться в принципы построения и оценки линейной регрессии. Однако немного теории поможет нам понять её эффективность в причинно-следственной связи. Прежде всего, регрессия решает теоретическую задачу наилучшего линейного прогнозирования. Пусть $\beta^*$ — вектор параметров:

$
\beta^* =\underset{\beta}{argmin} \ E[(Y_i - X_i'\beta)^2]
$

Линейная регрессия находит параметры, минимизирующие среднеквадратичную ошибку (MSE).

Если продифференцировать и приравнять к нулю, то линейное решение этой задачи будет иметь вид:

$
\beta^* = E[X_i'X_i]^{-1}E[X_i' Y_i]
$

Мы можем оценить эту бету, используя эквивалентную выборку:

$
\hat{\beta} = (X'X)^{-1}X' Y
$

````python
X = data[["format_ol"]].assign(intercep=1)
y = data["falsexam"]

def regress(y, X):
    return np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))

beta = regress(y, X)
beta
````

Приведённые выше формулы довольно общие. Однако имеет смысл рассмотреть случай, когда имеется только один регрессор. В причинно-следственном выводе мы часто хотим оценить причинно-следственное влияние переменной $T$ на результат $y$. Поэтому мы используем регрессию с этой единственной переменной для оценки этого влияния. Даже если мы включаем в модель другие переменные, они обычно играют лишь вспомогательную роль. Добавление других переменных может помочь нам оценить причинно-следственное влияние воздействия, но нас не очень интересует оценка их параметров.

При наличии единственной переменной-регрессора $T$ связанный с ней параметр будет определяться следующим образом:

$
\beta_1 = \dfrac{Cov(Y_i, T_i)}{Var(T_i)}
$

Если $T$ выбирается случайным образом, $\beta_1$ является ATE.

Если у нас несколько регрессоров, мы можем расширить следующую формулу для их учета. Предположим, что эти переменные вспомогательные, и нас интересует только оценка параметра $\kappa$, связанного с $T$.

$
y_i = \beta_0 + \kappa T_i + \beta_1 X_{1i} + ... +\beta_k X_{ki} + u_i
$

$\kappa$ можно получить с помощью следующей формулы:

$
\kappa = \dfrac{Cov(Y_i, \tilde{T_i})}{Var(\tilde{T_i})}
$

где $\tilde{T_i}$ — это остаток регрессии $T_i$ по всем остальным ковариатам $X_{1i}, ..., X_{ki}$. Теперь давайте оценим, насколько это здорово. Это означает, что коэффициент многомерной регрессии — это коэффициент того же регрессора **после учета влияния других переменных в модели**. В терминах причинно-следственной связи $\kappa$ — это коэффициент $T$ после использования всех остальных переменных для его прогнозирования.

Это интуитивно понятно. Если мы можем предсказать $T$, используя другие переменные, это означает, что оно неслучайно. Однако мы можем сделать так, чтобы $T$ было практически случайным, если мы учтем другие доступные переменные. Для этого мы используем линейную регрессию для прогнозирования $T$ на основе других переменных, а затем берем остатки этой регрессии $\tilde{T}$. По определению, $\tilde{T}$ не может быть предсказано другими переменными $X$, которые мы уже использовали для прогнозирования $T$. Довольно элегантно, $\tilde{T}$ — это вариант обработки, не связанный ни с одной другой переменной в $X$.

Кстати, это также свойство линейной регрессии. Остатки всегда ортогональны или некоррелированы ни с одной из переменных модели, которая их создала:

````python
e = y - X.dot(beta)
print("Orthogonality imply that the dot product is zero:", np.dot(e, X))
X[["format_ol"]].assign(e=e).corr()
````

**Регрессия для неслучайных данных**

До сих пор мы работали со случайными экспериментальными данными, но, как мы знаем, их трудно получить. Эксперименты очень дороги в проведении или просто нецелесообразны. Очень сложно убедить McKinsey & Co. предоставлять свои услуги бесплатно, чтобы мы могли раз и навсегда отделить ценность их консалтинговых услуг от того факта, что компании, которые могут себе позволить их оплачивать, и так очень состоятельны.

По этой причине мы теперь углубимся в неслучайные данные, или данные наблюдений. В следующем примере мы попытаемся оценить влияние дополнительного года обучения на почасовую заработную плату. Как вы, возможно, догадались, провести эксперимент с образованием крайне сложно. Нельзя просто рандомизировать людей на 4, 8 или 12 лет обучения. В данном случае у нас есть только данные наблюдений.

Сначала давайте оценим очень простую модель. Мы построим регрессию логарифмической почасовой заработной платы на количество лет обучения. Мы используем здесь логарифмы, чтобы наши оценки параметров имели процентную интерпретацию (если вы никогда не слышали об этом удивительном свойстве логарифма и хотите узнать, почему, перейдите по [этой ссылке](https://stats.stackexchange.com/questions/244199/why-is-it-that-natural-log-changes-are-percentage-changes-what-is-about-logs-th)). С её помощью мы можем сказать, что 1 дополнительный год обучения даёт повышение заработной платы на x%.

$
log(hwage)_i = \beta_0 + \beta_1 educ_i + u_i
$

````python
wage = pd.read_csv("regression_df_two.csv").dropna()
model_1 = smf.ols('np.log(hwage) ~ educ', data=wage.assign(hwage=wage["wage"]/wage["hours"])).fit()
print(model_1.summary())
````

Оценка $\beta_1$ составляет 0,0536 с 95% доверительным интервалом (0,039, 0,068). Это означает, что данная модель предсказывает рост заработной платы примерно на 5,3% за каждый дополнительный год обучения. Этот процентный рост согласуется с предположением об экспоненциальном влиянии образования на заработную плату: мы ожидаем, что переход от 11 до 12 лет обучения (в среднем до окончания средней школы) будет менее эффективным, чем переход от 14 до 16 лет (в среднем до окончания колледжа).

````python
from matplotlib import pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")

x = np.array(range(5, 20))
plt.plot(x, np.exp(model_1.params["Intercept"] + model_1.params["educ"] * x))
plt.xlabel("Years of Education")
plt.ylabel("Hourly Wage")
plt.title("Impact of Education on Hourly Wage")
plt.show()
````

Конечно, эта простая модель верна не потому, что мы можем её оценить. Обратите внимание, как я был осторожен в своих словах, говоря, что она **предсказывает** заработную плату в зависимости от образования. Я никогда не говорил, что это предсказание было причинно-следственным. На самом деле, к настоящему моменту у вас, вероятно, есть очень серьёзные основания полагать, что эта модель предвзята. Поскольку наши данные не были получены в результате случайного эксперимента, мы не знаем, сопоставимы ли те, кто получил больше образования, с теми, кто получил меньше. Идя ещё дальше, исходя из нашего понимания того, как устроен мир, мы совершенно уверены, что они несопоставимы. А именно, мы можем утверждать, что у тех, у кого больше лет образования, вероятно, более богатые родители, и что рост заработной платы по мере повышения уровня образования — это всего лишь отражение того, как благосостояние семьи связано с увеличением количества лет образования. Выражаясь математически, мы полагаем, что $E[Y_0|T=0] < E[Y_0|T=1]$, то есть те, у кого больше образования, в любом случае имели бы более высокий доход, даже без такого количества лет обучения. Если вы совсем недовольны образованием, то можете утверждать, что оно может даже *снижать* зарплаты, удерживая людей от работы и снижая их опыт.

К счастью, в наших данных у нас есть доступ ко множеству других переменных. Мы можем видеть образование родителей (meduc, feduc), показатель IQ (интеллектуального интеллекта) данного человека, количество лет опыта (exper) и стаж работы в текущей компании (tenure). У нас даже есть несколько фиктивных переменных для брака и афроамериканской этнической принадлежности.

Мы можем включить все эти дополнительные переменные в модель и оценить её:

$
log(hwage)_i = \beta_0 + \kappa \ educ_i + \pmb{\beta}X_i + u_i
$

Чтобы понять, как это помогает решить проблему смещения, давайте вспомним разбивку многомерной линейной регрессии.

$
\kappa = \dfrac{Cov(Y_i, \tilde{T_i})}{Var(\tilde{T_i})}
$

Эта формула утверждает, что мы можем предсказать `educ` на основе образования родителей, их IQ, опыта и так далее. После этого у нас останется версия `educ`, $\tilde{educ}$, которая не коррелирует со всеми ранее включёнными переменными. Это разрушит такие аргументы, как «люди с большим количеством лет обучения получают его потому, что у них выше IQ. Образование не приводит к более высокой заработной плате. Оно просто коррелирует с IQ, который и определяет заработную плату». Если мы включим IQ в нашу модель, то $\kappa$ станет доходностью дополнительного года обучения при сохранении IQ на прежнем уровне. Давайте немного подумаем, чтобы понять, что это значит. Даже если мы не можем использовать рандомизированные контролируемые испытания для поддержания равенства других факторов между исследуемыми и неизученными группами, регрессия может сделать это, включив те же факторы в модель, даже если данные не являются случайными.

````python
controls = ['IQ', 'exper', 'tenure', 'age', 'married', 'black',
            'south', 'urban', 'sibs', 'brthord', 'meduc', 'feduc']

X = wage[controls].assign(intercep=1)
t = wage["educ"]
y = wage["lhwage"]

beta_aux = regress(t, X)
t_tilde = t - X.dot(beta_aux)

kappa = t_tilde.cov(y) / t_tilde.var()
kappa
````

Этот только что рассчитанный нами коэффициент говорит нам, что для людей с одинаковым IQ, опытом, стажем, возрастом и так далее следует ожидать, что дополнительный год обучения будет связан с увеличением почасовой оплаты труда на 4,11%. Это подтверждает наше подозрение о том, что первая простая модель, содержащая только `educ`, была смещена. Это также подтверждает, что это смещение переоценивало влияние образования. После того, как мы учели другие факторы, предполагаемое влияние образования снизилось.

Если мы будем благоразумнее и будем использовать программное обеспечение, написанное другими, вместо того, чтобы писать всё самостоятельно, мы сможем даже установить доверительный интервал для этой оценки.

````python
model_2 = smf.ols('lhwage ~ educ +' + '+'.join(controls), data=wage).fit()
print(model_2.summary())
````

##### Смещение из-за пропущенных переменных (Omitted Variable Bias)

Остаётся вопрос: является ли этот параметр, который мы оценили, причинно-следственным? К сожалению, мы не можем сказать наверняка. Можно утверждать, что первая простая модель, построенная на регрессии зависимости заработной платы от образования, вероятно, таковой не является. Она не учитывает важные переменные, коррелирующие как с образованием, так и с заработной платой. Без их контроля оцениваемое влияние образования также учитывает влияние других переменных, не включенных в модель.

Чтобы лучше понять, как работает это смещение, предположим, что истинная модель влияния образования на заработную плату выглядит примерно так:

$
Wage_i = \alpha + \kappa \ Educ_i + A_i'\beta + u_i
$

На заработную плату влияет образование, которое измеряется размером $\kappa$ и дополнительными факторами способностей, обозначаемыми вектором $A$. Если мы исключим способности из нашей модели, наша оценка $\kappa$ будет выглядеть следующим образом:

$
\dfrac{Cov(Wage_i, Educ_i)}{Var(Educ_i)} = \kappa + \beta'\delta_{Ability}
$

где $\delta_{A}$ — вектор коэффициентов регрессии $Educ$ на $A$.

Ключевой момент здесь заключается в том, что это будет не совсем тот $\kappa$, который нам нужен. Вместо этого он добавляет дополнительный член $\beta'\delta_{A}$. Этот член представляет собой влияние пропущенной переменной $A$ на $Wage$, умноженное на влияние пропущенной переменной на включённую переменную $Educ$.

Это даёт нам более глубокое понимание природы смещения. Во-первых, смещение будет равно нулю, если пропущенные переменные не влияют на зависимую переменную $Y$. Это совершенно логично. Мне не нужно контролировать факторы, не имеющие отношения к заработной плате, пытаясь понять влияние образования на неё. Во-вторых, смещение также будет равно нулю, если пропущенные переменные не влияют на экспериментальную переменную. Это также интуитивно понятно. Если всё, что влияет на образование, включено в модель, то предполагаемое влияние образования никак не может смешиваться с корреляцией образования с чем-то ещё, что также влияет на заработную плату.

Короче говоря, мы говорим, что **нет OVB, если все смешивающие переменные учтены в модели**. Мы также можем использовать здесь знания о причинно-следственных графах. Смешивающая переменная (конфаундер) — это та, которая **обусловливает как воздействие, так и результат**. В примере с заработной платой IQ является смешивающим фактором. Люди с высоким IQ, как правило, получают больше лет образования, потому что им это проще, поэтому мы можем сказать, что IQ обуславливает образование. Люди с высоким IQ также, как правило, более продуктивны от природы и, следовательно, имеют более высокую заработную плату, поэтому IQ также обуславливает заработную плату. Поскольку конфаундеры — это переменные, влияющие как на воздействие, так и на результат, мы обозначим их стрелками, направленными к $T$ и $Y$. Здесь я обозначил их как $W$. Я также обозначил положительную причинно-следственную связь красным цветом, а отрицательную — синим.

````python
g = gr.Digraph()

g.edge("W", "T"), g.edge("W", "Y"), g.edge("T", "Y")

g.edge("IQ", "Educ", color="red"), g.edge("IQ", "Wage", color="red"), g.edge("Educ", "Wage", color="red")

g.edge("Crime", "Police", color="red"), g.edge("Crime", "Violence", color="red"),
g.edge("Police", "Violence", color="blue")

g
````

Причинно-следственные графы отлично подходят для описания нашего понимания мира и понимания того, как работает OVB. В нашем первом примере у нас есть граф, где образование влияет на заработную плату: более высокий уровень образования приводит к более высокой заработной плате. Однако IQ также влияет на заработную плату и образование: высокий IQ влияет как на уровень образования, так и на заработную плату. Если мы не будем учитывать IQ в нашей модели, часть его влияния на заработную плату будет проявляться через корреляцию с образованием. Это заставит влияние образования казаться выше, чем оно есть на самом деле. Это пример положительного смещения.

В качестве ещё одного примера, но с отрицательным смещением, рассмотрим причинно-следственный граф влияния полиции на уровень насилия в городах. Обычно мы видим, что в городах с более высокой численностью полиции также наблюдается более высокий уровень насилия. Означает ли это, что полиция является причиной насилия? Что ж, возможно, но я не думаю, что стоит здесь углубляться в эту дискуссию. Но также существует высокая вероятность того, что существует сбивающая с толку переменная, из-за которой мы видим смещенную версию влияния полиции на уровень насилия. Возможно, увеличение численности полиции снижает уровень насилия. Однако третья переменная – преступность – приводит как к росту насилия, так и к увеличению числа полицейских. Если мы не будем её учитывать, влияние преступности на насилие будет передаваться через полицию, создавая впечатление, что она увеличивает насилие. Это пример негативного смещения.

Причинно-следственные графи также могут показать, что как регрессионные, так и рандомизированные контролируемые исследования верны для смещений, связанных с влиянием факторов. РКИ делает это, разрывая связь смещающего фактора с экспериментальной переменной. Если сделать $T$ случайным, то, по определению, ничто не может его вызвать.

````python
g = gr.Digraph()

g.edge("W", "Y"), g.edge("T", "Y")

g.edge("IQ", "Wage", color="red"), g.edge("Educ", "Wage", color="red")

g
````

Регрессия, с другой стороны, делает это, сравнивая влияние $T$, сохраняя конфаундер $W$ на фиксированном уровне. При регрессии $W$ не перестаёт вызывать $T$ и $Y$. Он просто остаётся фиксированным, поэтому не может влиять на изменения $T$ и $Y$.

````python
g = gr.Digraph()

g.node("W=w"), g.edge("T", "Y")
g.node("IQ=x"), g.edge("Educ", "Wage", color="red")

g
````

Теперь вернемся к нашему вопросу, является ли параметр, который мы оценили для влияния `educ` на заработную плату, причинным? Мне жаль, что приходится вам это говорить, но это будет зависеть от нашей способности аргументировать за или против того факта, что все конфаундеры были включены в модель. Лично я думаю, что нет. Например, мы не включили семейное богатство. Даже если бы мы включили семейное образование, его можно было бы рассматривать только как косвенный показатель богатства. Мы также не учли такие факторы, как личные амбиции. Возможно, именно амбиции являются причиной как большего количества лет образования, так и более высокой заработной платы, поэтому они являются конфаундером. Это показывает, что **причинно-следственные выводы с неслучайными или наблюдаемыми данными всегда следует воспринимать с долей скепсиса**. Мы никогда не можем быть уверены, что все конфаундеры были учтены.

##### Включение дополнительных переменных

Мы видели, как добавление дополнительных элементов управления в нашу регрессионную модель может помочь выявить причинно-следственные связи. В современном мире больших данных это может легко превышать 1000 переменных. Как оказалось, это не только не нужно, но и может помешать выявлению причинно-следственных связей. Например, `credit_limit` представляет собой кредитный лимит клиента до того, как он или она опоздал. `risk_score` соответствует предполагаемому риску клиента до доставки электронного письма. Логично предположить, что кредитный лимит и риск, вероятно, являются очень хорошими предикторами платежей. Но чем это может быть полезно?

Во-первых, давайте разберёмся, почему мы можем не обнаружить статистическую значимость в методе лечения, даже если он есть. Возможно, как в этом случае, метод лечения оказывает очень малое влияние на результат. Если задуматься, то, что заставляет людей выплачивать свои долги, по большей части, зависит от факторов, находящихся вне контроля отдела взыскания задолженностей. Люди платят свои долги, потому что находят новую работу, управляют своими финансами, доходом и так далее. С точки зрения статистики, можно сказать, что **изменчивость платежей объясняется гораздо больше другими факторами, помимо электронной почты**.

Чтобы наглядно это понять, мы можем построить график зависимости платежей от экспериментальной переменной электронной почты. Я также нарисовал красным цветом аппроксимированную линию модели выше. Для наглядности я добавил немного шума к переменной электронной почты, чтобы она не разбивалась о ноль или единицу.

````python
sns.scatterplot(x="email", y="payments",
                alpha=0.8,
                data=data.assign(email=data["email"] + np.random.normal(0, 0.01, size=len(data["email"]))))
plt.plot(np.linspace(-0.2, 1.2), model.params[0] + np.linspace(-1, 2) * model.params[1], c="C1")
plt.xlabel("Email")
plt.ylabel("Payments");
````

Мы видим, насколько сильно варьируются платежи в группах. Визуально это выглядит так, будто в обеих группах они колеблются от чуть менее 400 до 1000. Если влияние электронного письма составляет, скажем, 5 или 10 реалов, неудивительно, что его будет трудно найти среди всей этой изменчивости.

К счастью, регрессия может помочь нам снизить эту изменчивость. Хитрость заключается в использовании дополнительных переменных. **Если переменная хорошо предсказывает результат, это объясняет большую часть её дисперсии**. Если риск и кредитный лимит хорошо предсказывают платежи, мы можем контролировать их, чтобы упростить определение влияния электронного письма на платежи. Если мы помним, как работает регрессия, это имеет интуитивное объяснение. Добавление дополнительных переменных в регрессию означает сохранение их постоянства при рассмотрении лечения. Таким образом, рассуждение заключается в том, что если мы рассматриваем схожие уровни риска и кредитного лимита, дисперсия переменной отклика «платежи» должна быть меньше. Или, другими словами, если риск и кредитная линия очень хорошо предсказывают платежи, клиенты с аналогичным риском и кредитной линией также должны иметь схожие уровни платежей, следовательно, с меньшей дисперсией.

Чтобы продемонстрировать это, давайте воспользуемся методом частичного разделения, разделив регрессию на два этапа. Сначала мы регрессируем воздействие (электронную почту) и результат (платежи) по дополнительным элементам управления: кредитному лимиту и оценке риска. Затем мы регрессируем остаток воздействия по остаткам платежей, полученным на шаге 1. (Это чисто педагогическое упражнение, на практике вам не придётся разбираться со всеми этими сложностями).

````python
model_email = smf.ols('email ~ credit_limit + risk_score', data=data).fit()
model_payments = smf.ols('payments ~ credit_limit + risk_score', data=data).fit()

residuals = pd.DataFrame(dict(res_payments=model_payments.resid, res_email=model_email.resid))

model_treatment = smf.ols('res_payments ~ res_email', data=residuals).fit()
````

Это снижает дисперсию зависимой переменной. Регрессируя платежи по кредитному лимиту и риску и получая остатки для этой модели, мы создаём новую зависимую переменную с гораздо меньшей изменчивостью, чем исходная. Последняя модель также выявляет $ATE$ с достоверной оценкой стандартной ошибки.

Из любопытства мы также можем проверить, что модель, предсказывающая воздействие, не должна снижать его дисперсию. Это связано с тем, что электронная почта по своей природе случайна, поэтому её невозможно предсказать.

````python
print("Payments Variance", np.var(data["payments"]))
print("Payments Residual Variance", np.var(residuals["res_payments"]))

print("Email Variance", np.var(data["email"]))
print("Email Residual Variance", np.var(residuals["res_email"]))

print(model_treatment.summary())
````

Обратите внимание, как дисперсия платежей изменилась с 10807 до 5652. Мы уменьшили её почти вдвое, учитывая риск и кредитные лимиты. Также обратите внимание, что нам не удалось снизить вариативность в письмах. Это логично, поскольку риск и кредитная линия не предсказывают письма (ничто не предсказывает, по определению случайности).

Теперь мы видим нечто гораздо более обоснованное. Эта новая оценка говорит нам, что следует ожидать, что клиенты, получившие письмо, заплатят в среднем на 4,4 реала больше, чем клиенты в контрольной группе. Эта оценка теперь статистически отличается от нуля. Мы также можем наглядно показать, насколько снизилась дисперсия в каждой контрольной группе.

````python
sns.scatterplot(x="res_email", y="res_payments", data=residuals)
plt.plot(np.linspace(-0.7, 1), model_treatment.params[0] + np.linspace(-1, 2) * model_treatment.params[1], c="C1")
plt.xlabel("Email Residuals")
plt.ylabel("Payments Residuals");
````

Как я уже говорил, мы сделали это из педагогических соображений. На практике можно просто добавить контрольные элементы в регрессионную модель вместе с экспериментальными, и оценки будут точно такими же.

````python
model_2 = smf.ols('payments ~ email + credit_limit + risk_score', data=data).fit()
print(model_2.summary())
````

Подводя итог, можно сказать, что всякий раз, когда у нас есть переменная, которая хорошо предсказывает исход, даже если она не является конфаундером, добавление её в нашу модель — хорошая идея. Это помогает снизить дисперсию оценок эффекта лечения. Вот как эта ситуация выглядит с помощью причинно-следственных графиков.

````python
g = gr.Digraph()
g.edge("X", "Y"), g.edge("T", "Y")
g.node("T", color="gold")

g.node("email", color="gold")
g.edge("credit_limit", "payments")
g.edge("risk_score", "payments")
g.edge("email", "payments")

g
````

**Вредные переменные**

В качестве второго мотивирующего примера рассмотрим сценарий испытания препарата в двух больницах. Обе проводят рандомизированные испытания нового препарата для лечения определённого заболевания. Интересующий нас результат — количество дней госпитализации. Если лечение эффективно, оно сократит количество дней пребывания пациента в больнице. В одной из больниц политика в отношении случайного лечения заключается в том, чтобы назначать его 90% пациентов, а 10% — плацебо. В другой больнице политика иная: препарат назначается 10% пациентов, выбранных случайным образом, а 90% — плацебо. Вам также сообщают, что больница, которая назначает 90% основного препарата и 10% плацебо, обычно получает более тяжёлые случаи заболевания.

````python
hospital = pd.read_csv("./data/confounded_df_two.csv")
hospital.head()
````

Поскольку вы имеете дело с рандомизированными данными, вашим первым побуждением будет просто провести регрессию результатов лечения.

````python
hosp_1 = smf.ols('days ~ treatment', data=hospital).fit()
print(hosp_1.summary())
````

Но вы обнаружили некоторые противоречивые результаты. Как лечение может увеличивать количество дней в больнице? Ответ заключается в том, что мы проводим два разных эксперимента. Тяжесть заболевания положительно коррелирует с увеличением количества дней в больнице, и поскольку больница с более тяжёлыми случаями также выписывает больше препарата, препарат начинает положительно коррелировать с увеличением количества дней в больнице. Если мы сравним обе больницы, то увидим, что $E[Y_0|T=0]<E[Y_0|T=1]$, то есть потенциальный исход у нелеченых пациентов в среднем ниже, чем у леченных, поскольку в больнице с менее тяжёлыми случаями находится больше нелеченых пациентов. Другими словами, тяжесть заболевания выступает в качестве конфаундера, определяя, в какую больницу обратится пациент, и, следовательно, вероятность получения препарата.

Есть два способа исправить это. Первый, который противоречит цели использования данных из обеих больниц, — это просто рассмотреть $ATE$ в каждой больнице отдельно.

````python
hosp_2 = smf.ols('days ~ treatment', data=hospital.query("hospital==0")).fit()
print(hosp_2.summary())
````

````python
hosp_3 = smf.ols('days ~ treatment', data=hospital.query("hospital==1")).fit()
print(hosp_3.summary())
````

В данном случае мы получили интуитивно понятный результат. Похоже, что препарат действительно сокращает количество дней, проведённых в больнице. Однако, поскольку мы рассматриваем каждую больницу индивидуально, данных недостаточно. В результате мы не можем получить статистически значимые результаты.

Другой подход, использующий возможности регрессии, заключается в контроле тяжести заболевания путём включения её в модель.

````python
hosp_4 = smf.ols('days ~ treatment + severity', data=hospital).fit()
print(hosp_4.summary())
````

Далее возникает вопрос: следует ли нам также включать в модель больницу? В конце концов, мы знаем, что больницы являются причиной лечения, верно? Что ж, это верно, но как только мы учитываем тяжесть заболевания, больница перестаёт коррелировать с количеством дней госпитализации, являющихся результатом. И мы знаем, что для того, чтобы быть конфаундером, переменная должна быть причиной как лечения, так и исхода. В данном случае у нас есть переменная, которая является причиной только лечения.

Но, может быть, её контроль снижает дисперсию, верно? Что ж, опять же, это неверно. Чтобы контроль снижал дисперсию, он должен быть хорошим предиктором исхода, а не лечения, как в данном случае.

Тем не менее, мы, возможно, хотим контролировать его, верно? Это не повредит... Или может?

````python
hosp_5 = smf.ols('days ~ treatment + severity + hospital', data=hospital).fit()
print(hosp_5.summary())
````

Удивительно, но это может быть болезненно!

Добавление больницы к тяжести в качестве контрольного фактора увеличило дисперсию нашей оценки $ATE$. Как это возможно? Ответ кроется в формуле для стандартной ошибки коэффициента регрессии.

$
\hat{\sigma}^2 = \dfrac{1}{n-2} \sum( y_i - \hat{y}_i )^2
$

$
\text{Var}(\hat{\beta}_2) = \dfrac{\sigma^2}{\sum(x_i - \bar{x})^2}
$

Из этой формулы видно, что стандартная ошибка обратно пропорциональна дисперсии переменной $X$. Это означает, что если $X$ не меняется существенно, будет сложно оценить её влияние на исход. Это также интуитивно понятно. Доведите это до крайности и представьте, что вы хотите оценить эффект препарата. Вы проводите тест с участием 10 000 человек, но только один из них получает лечение. Это значительно усложнит определение $ATE$, поскольку нам придётся сравнивать одного человека со всеми остальными. Другими словами, нам нужна высокая вариабельность лечения, чтобы облегчить определение его влияния.

Что касается того, почему включение больниц в модель увеличивает погрешность нашей оценки, то это потому, что они хорошо предсказывают лечение, а не исход (после учета тяжести). Таким образом, прогнозируя лечение, мы фактически снижаем его дисперсию. Чтобы увидеть это, мы снова можем прибегнуть к разделению нашей регрессии на два этапа.

````python
model_treatment = smf.ols('treatment ~ severity + hospital', data=hospital).fit()
model_days = smf.ols('days ~ severity + hospital', data=hospital).fit()

residuals = pd.DataFrame(dict(res_days=model_days.resid, res_treatment=model_treatment.resid))

model_treatment = smf.ols('res_days ~ res_treatment', data=residuals).fit()

print(model_treatment.summary())
````

````python
print("Treatment Variance", np.var(hospital["treatment"]))
print("Treatment Residual Variance", np.var(residuals["res_treatment"]))
````

Итак, суть в том, что нам следует добавить параметры, которые коррелируют как с воздействием, так и с исходом (конфаундер), например, с тяжестью в модели выше. Также следует добавить переменные, которые являются хорошими предикторами исхода, даже если они не являются конфаундерами, поскольку они снижают дисперсию наших оценок. Однако нам **НЕ** следует добавлять переменные, которые являются просто хорошими предикторами воздействия, поскольку они увеличат дисперсию наших оценок.

**Плохие переменные. Смещение выборки (Selection Bias)**

Вернемся к примеру с письмом о взыскании задолженности. Помните, что письмо было случайным образом назначено клиентам. Мы уже объяснили, что такое `credit_limit` и `risk_score`. Теперь рассмотрим оставшиеся переменные. `opened` — фиктивная переменная, указывающая, открыл ли клиент письмо или нет. `agreement` — ещё одна фиктивная переменная, которая указывает, обращался ли клиент в отдел взыскания задолженности для обсуждения условий погашения задолженности после получения письма. Какая из следующих моделей, по вашему мнению, более уместна? Первая — это модель с переменной обработки, а также `credit_limit` и `risk_score`; вторая добавляет фиктивные переменные `opened` и `agreement`.

````python
email_1 = smf.ols('payments ~ email + credit_limit + risk_score', data=data).fit()
print(email_1.summary())
````

````python
email_2 = smf.ols('payments ~ email + credit_limit + risk_score + opened + agreement', data=data).fit()
print(email_2.summary())
````

Первая модель обнаруживает статистически значимые результаты для электронного письма, а вторая — нет. Но, возможно, вторая модель верна, и для электронного письма эффект отсутствует. В конце концов, эта модель учитывает больше факторов, поэтому она должна быть более устойчивой, верно? К настоящему моменту вы, вероятно, уже знаете, что это не так. Осталось выяснить, что именно.

Мы знаем, что ДОЛЖНЫ добавить конфаундеры. Мы также знаем, что добавление переменных, которые очень хорошо предсказывают исход, — хорошая идея. Это не обязательно, но приятно. Мы также знаем, что добавлять переменные, которые предсказывают только воздействие, — плохая идея. Опять же, это не смертный грех, но его желательно избегать. Так что же это за переменные `opened` и `agreement`? Оказывается, они не относятся ни к одному из вышеперечисленных.

Если задуматься, `opened` и `agreement` определённо коррелируют с электронным письмом. В конце концов, вы не можете открыть письмо, если не получили его, и мы также говорили, что соглашение учитывает только повторные переговоры, произошедшие после отправки письма. Но **они не являются причиной письма. Наоборот, они вызваны им.**

Когда нужно понять, с какими переменными имеем дело, предпочтительно размышлять об их причинно-следственной связи. Давайте сделаем это здесь.

````python
g = gr.Digraph()

g.edge("email", "payments")
g.edge("email", "opened")
g.edge("email", "agreement")
g.edge("opened", "payments")
g.edge("opened", "agreement")
g.edge("agreement", "payments")

g.edge("credit_limit", "payments")
g.edge("credit_limit", "opened")
g.edge("credit_limit", "agreement")
g.edge("risk_score", "payments")
g.edge("risk_score", "opened")
g.edge("risk_score", "agreement")

g
````

Мы знаем, что ничто не является причиной электронных писем, поскольку они изначально случайны. И мы знаем (или, по крайней мере, у нас есть веские основания полагать), что кредитный лимит и риск приводят к платежам. Мы также считаем, что электронные письма являются причиной платежей. Что касается `opened`, мы считаем, что оно действительно приводит к платежам. Интуитивно понятно, что люди, открывшие письмо с требованием о взыскании задолженности, более охотно ведут переговоры и выплачивают свои долги. Мы также считаем, что `opened` приводит к соглашениям по тем же причинам, что и к платежам. Более того, мы знаем, что `opened` вызвано электронными письмами, и у нас есть основания полагать, что у людей с разным уровнем риска и кредитным лимитом разная частота открытия писем, поэтому кредитный лимит и риск также являются причиной `opened`. Что касается `agreement`, мы также считаем, что оно вызвано `opened`. Если рассматривать переменную отклика «платежи», мы можем представить её как результат воронки:

$
электронное письмо -> открыто -> соглашение -> оплата
$

Мы также считаем, что разные уровни риска и линии имеют разную склонность к достижению соглашения, поэтому мы отметим их как также приводящие к `agreement`. Что касается электронной почты и согласия, можно утверждать, что некоторые люди просто читают тему письма, и это повышает вероятность заключения соглашения. Дело в том, что электронная почта также может стать причиной заключения соглашения, минуя этап `opened`.

На этом графике мы видим, что `opened` и `agreement` находятся на причинно-следственной связи между электронной почтой и платежами. Таким образом, если мы будем контролировать их с помощью регрессии, то получим: «Это эффект электронной почты, при этом «открыто» и «согласие» останутся неизменными». Однако оба эти фактора являются частью причинно-следственного эффекта электронной почты, поэтому мы не хотим их фиксировать. Вместо этого мы могли бы утверждать, что электронная почта увеличивает платежи именно потому, что повышает процент согласия. Если мы зафиксируем эти переменные, мы устраним часть истинного эффекта переменной «электронная почта».

Используя обозначение потенциального результата, можно сказать, что из-за рандомизации $E[Y_0|T=0] = E[Y_0|T=1]$. Однако даже при рандомизации, если мы контролируем этап «согласие», экспериментальная и контрольная группы перестают быть сопоставимыми. На самом деле, интуитивно поразмыслив, мы можем даже догадаться, чем они отличаются:

$
E[Y_0|T=0, agreement=0] > E[Y_0|T=1, agreement=0]
$

$
E[Y_0|T=0, agreement=1] > E[Y_0|T=1, agreement=1]
$

Первое уравнение явно показывает, что мы считаем, что те, кто не получил электронное письмо и соглашение, лучше тех, кто получил электронное письмо, но не получил соглашение. Это связано с тем, что если воздействие оказывает положительный эффект, те, кто не заключил соглашение **даже после получения электронного письма**, вероятно, получают меньше платежей по сравнению с теми, кто также не заключил соглашение, но и не получил дополнительного стимула в виде электронного письма. Что касается второго уравнения, то те, кто заключил соглашение, даже не получив воздействия, вероятно, лучше тех, кто заключил соглашение, но получил дополнительное стимул в виде электронного письма.

Это может показаться очень запутанным при первом прочтении, но убедитесь, что вы его поняли. При необходимости перечитайте. Затем можно провести аналогичное рассуждение с открытой переменной. Попробуйте сделать это самостоятельно.

Этот вид смещения настолько распространён, что у него есть своё название. В то время как смещение (confounding) — это смещение, возникающее из-за отсутствия контроля общей причины, **смещение выборки (selection bias) возникает, когда мы контролируем общий эффект или переменную, находящуюся между причиной и следствием**. Как правило, всегда включайте в модель конфаундеры и переменные, которые являются хорошими предикторами $Y$. Всегда исключайте переменные, которые являются хорошими предикторами только $T$, медиаторов между лечением и результатом или общего эффекта воздействия и результата.



##### Инструментальные переменные

Один из способов контроля OVB — это добавление пропущенной переменной в нашу модель. Однако это не всегда возможно, в основном потому, что у нас просто нет данных по пропущенным переменным. Например, вернёмся к нашей модели влияния образования на заработную плату:

$
\log(\mathrm{wage})_i = \beta_0 + \kappa \ \mathrm{educ}_i + \pmb{\beta}\mathrm{Ability}_i + u_i

Один из способов избежать этого — учитывать постоянные уровни способностей при оценке влияния образования на заработную плату. Мы могли бы сделать это, включив способности в нашу модель линейной регрессии. Однако у нас нет надежных показателей способностей. Лучшее, что у нас есть, — это весьма сомнительные прокси, такие как IQ.

Но не все потеряно. Здесь в игру вступают инструментальные переменные. Идея IV заключается в том, чтобы найти другую переменную, которая обуславливает воздействие, и она коррелирует с результатом только через воздействие. Другими словами, этот инструмент $Z_i$ не коррелирует с $Y_0$, но коррелирует с $T$. Иногда это называют ограничением исключения.

````python
g = gr.Digraph()

g.edge("ability", "educ")
g.edge("ability", "wage")
g.edge("educ", "wage")
g.edge("instrument", "educ")
g
````

Если у нас есть такая переменная, мы можем восстановить причинно-следственную связь $\kappa$ с помощью формулы IV. Для этого давайте подумаем об идеальном уравнении, которое мы хотим получить. Используя более общие термины, такие как $T$ для воздействия и $W$ для конфаундеров, вот что нам нужно:

$
Y_i = \beta_0 + \kappa \ T_i + \pmb{\beta}W_i + u_i
$

Однако у нас нет данных по $W$, поэтому мы можем вычислить только

$
Y_i = \beta_0 + \kappa\ T_i + v_i, \quad v_i = \pmb{\beta}W_i + u_i
$


Поскольку $W$ является конфаундером, $\mathrm{Cov}(T, v) \neq 0$. В нашем примере это будет означать, что способности коррелируют с образованием. Если это так, то регрессионный анализ даст смещенную оценку $\kappa$ из-за пропущенных переменных.

Поскольку инструмент $Z$ коррелирует с результатом только через $T$, это означает, что $\mathrm{Cov}(Z,v) = 0$, в противном случае существовал бы второй путь из $Z$ в $Y$ через $W$. Имея это в виду, можно записать:

$
\mathrm{Cov}(Z,Y) = \mathrm{Cov}(Z,\beta_0 + \kappa\ T_i + v_i) = \kappa \mathrm{Cov}(Z,T) + \mathrm{Cov}(Z, v) = \kappa \mathrm{Cov}(Z,T)
$

Разделив каждую часть на $V(Z_i)$ и переставив члены, получаем:

$
\kappa = \dfrac{\mathrm{Cov}(Y_i, Z_i)/V(Z_i)}{\mathrm{Cov}(T_i, Z_i)/V(Z_i)} = \dfrac{\text{Приведённая форма}}{\text{1-й этап}}
$

Обратите внимание, что и числитель, и знаменатель являются коэффициентами регрессии (ковариациями, делёнными на дисперсию). Числитель — это результат регрессии $Y$ на $Z$. Другими словами, это «влияние» $Z$ на $Y$. Помните, что это не означает, что $Z$ вызывает $Y$, поскольку требуется, чтобы $Z$ влиял на $Y$ только через $T$. Скорее, это лишь отражает величину влияния $Z$ на $Y$ через $T$. У числителя есть собственное название: коэффициент приведённой формы. (reduced form stage)

Знаменатель также является коэффициентом регрессии. На этот раз это регрессия $T$ на $Z$. Эта регрессия отражает влияние $Z$ на $T$ и называется коэффициентом 1-й стадии (first form stage).

Ещё один интересный способ рассмотреть это уравнение — с точки зрения частных производных. Мы можем показать, что влияние $T$ на $Y$ равно влиянию $Z$ на $Y$, умноженному на влияние $Z$ на $T$:

$
\kappa = \dfrac{\frac{\partial y}{\partial z}}{\frac{\partial T}{\partial z}} = \dfrac{\partial y}{\partial z} * \dfrac{\partial z}{\partial T} = \dfrac{\partial y}{\partial T}
$

То, что мы видим, гораздо тоньше, чем кажется большинству людей. И это гораздо круче, чем кажется большинству людей. Записывая IV таким образом, мы говорим: «Смотрите, сложно определить влияние $T$ на $Y$ из-за конфаундеров. Но я могу легко определить влияние $Z$ на $Y$, поскольку нет ничего, что вызывало бы $Z$ и $Y$ (ограничение исключения). Однако меня интересует влияние $T$ на $Y$, а не $Z$ на $Y$. Поэтому я оценю простое влияние $Z$ на $Y$ и **масштабирую его по влиянию $Z$ на $T$**, чтобы перевести эффект в единицы $T$ вместо единиц $Z$».

Мы также можем увидеть это в упрощённом случае, когда инструментом является фиктивная переменная. В этом случае оценка IV ещё больше упрощается за счёт отношения двух разностей средних значений.

$
\kappa = \dfrac{E[Y|Z=1]-E[Y|Z=0]}{E[T|Z=1]-E[T|Z=0]}
$

Это отношение иногда называют **оценкой Вальда**.

**Квартал рождения и влияние образования на заработную плату**

До сих пор мы рассматривали эти инструменты как некую волшебную переменную $Z$, обладающую чудесным свойством влиять на результат только посредством воздействия. Честно говоря, хорошие инструменты так сложно найти, что их можно считать просто чудесами. Тем не менее, у нас есть несколько интересных примеров инструментов, которые позволяют сделать ситуацию более конкретной. Мы снова попытаемся оценить влияние образования на заработную плату. В результате люди, родившиеся позже в этом году, в среднем имеют больше лет образования, чем те, кто родился в начале года.

**Первый этап**

Прежде чем использовать квартал рождения в качестве инструмента, необходимо убедиться в его валидности. Это подразумевает аргументацию в пользу двух предположений об инструментальных переменных:

1. $\mathrm{Cov}(Z, T) \neq 0$. Это означает, что у нас должен быть сильный первый этап, или что инструмент действительно влияет на переменную воздействия.
2. $Y \perp Z | T $. Это ограничение исключения, утверждающее, что инструмент Z влияет на исход Y только через воздействие T.

К счастью, первое предположение проверяемо. Из данных видно, что $\mathrm{Cov}(Z, T)$ не равен нулю. В нашем примере, если квартал рождения действительно является инструментом, как мы уже говорили, следует ожидать, что люди, родившиеся в последнем квартале года, будут иметь немного больше времени на образование, чем те, кто родился в начале года. Прежде чем проводить какие-либо статистические тесты для проверки этого, давайте просто составим график наших данных и увидим его своими глазами.

````python
group_data = (data
              .groupby(["year_of_birth", "quarter_of_birth"])
              [["log_wage", "years_of_schooling"]]
              .mean()
              .reset_index()
              .assign(time_of_birth = lambda d: d["year_of_birth"] + (d["quarter_of_birth"])/4))
````

````python
plt.figure(figsize=(15,6))
plt.plot(group_data["time_of_birth"], group_data["years_of_schooling"], zorder=-1)
for q in range(1, 5):
    x = group_data.query(f"quarter_of_birth=={q}")["time_of_birth"]
    y = group_data.query(f"quarter_of_birth=={q}")["years_of_schooling"]
    plt.scatter(x, y, marker="s", s=200, c=f"C{q}")
    plt.scatter(x, y, marker=f"${q}$", s=100, c=f"white")

plt.title("Years of Education by Quarter of Birth (first stage)")
plt.xlabel("Year of Birth")
plt.ylabel("Years of Schooling");
````

Примечательно, что существует сезонный характер продолжительности обучения, зависящий от квартала года. Визуально мы видим, что у родившихся в первом квартале года почти всегда уровень образования ниже, чем у родившихся в последнем квартале (ведь, если учесть год рождения, у родившихся в более поздние годы, как правило, уровень образования выше).

Для большей строгости мы можем провести первый этап как линейную регрессию. Сначала преобразуем квартал рождения в фиктивные переменные:

````python
factor_data = data.assign(**{f"q{int(q)}": (data["quarter_of_birth"] == q).astype(int)
                             for q in data["quarter_of_birth"].unique()})

factor_data.head()
````

Для простоты пока будем использовать в качестве инструмента только последний квартал, q4. Мы построим регрессию продолжительности обучения (воздействие) на квартал рождения (инструмент). Это покажет, действительно ли квартал рождения положительно влияет на время обучения, как мы видели на графике выше. Здесь также необходимо учитывать год рождения, и мы добавим штат рождения в качестве дополнительного контрольного фактора.

````python
first_stage = smf.ols("years_of_schooling ~ C(year_of_birth) + C(state_of_birth) + q4", data=factor_data).fit()

print("q4 parameter estimate:, ", first_stage.params["q4"])
print("q4 p-value:, ", first_stage.pvalues["q4"])
````

Похоже, что родившиеся в последнем квартале года в среднем имеют на 0,1 года больше образования, чем родившиеся в других кварталах года. P-значение близко к нулю. Это закрывает вопрос о том, влияет ли квартал рождения на увеличение или уменьшение количества лет обучения.

**Редуцированная форма**

К сожалению, мы не можем проверить второе условие IV. Мы можем только утверждать в его пользу. Мы можем выразить уверенность в том, что квартал рождения не влияет на потенциальный доход. Другими словами, время рождения человека не является показателем его личных способностей или какого-либо другого фактора, который может повлиять на разницу в доходе, кроме влияния на образование. Хороший способ сделать это — сказать, что квартал рождения практически случайно выбран, когда мы рассматриваем его влияние на доход. (Он не случаен. Есть данные, что люди склонны зачать ребёнка в конце лета или во время каких-то праздников. Но я не могу придумать ни одной веской причины, по которой эта закономерность также как-то влияет на доход, помимо образования).

Приведя доводы в пользу ограничения исключения, мы можем перейти к расчёту сокращённой формы. Сокращённая форма направлена на выяснение того, как инструмент влияет на результат. Поскольку, согласно предположению, всё это влияние обусловлено воздействием на воздействие, это прольёт свет на то, как оно влияет на результат. Ещё раз оценим это визуально, прежде чем серьёзно заняться регрессией.

````python
plt.figure(figsize=(15,6))
plt.plot(group_data["time_of_birth"], group_data["log_wage"], zorder=-1)
for q in range(1, 5):
    x = group_data.query(f"quarter_of_birth=={q}")["time_of_birth"]
    y = group_data.query(f"quarter_of_birth=={q}")["log_wage"]
    plt.scatter(x, y, marker="s", s=200, c=f"C{q}")
    plt.scatter(x, y, marker=f"${q}$", s=100, c=f"white")

plt.title("Average Weekly Wage by Quarter of Birth (reduced form)")
plt.xlabel("Year of Birth")
plt.ylabel("Log Weekly Earnings");
````

Мы снова наблюдаем сезонный характер заработка по кварталам рождения. Родившиеся позже в этом году имеют немного более высокий доход, чем родившиеся в начале года. Для проверки этой гипотезы мы снова проведем регрессию инструментальной переменной q4 на логарифм заработной платы. Мы также добавим те же дополнительные контрольные параметры, что и на первом этапе:

````python
reduced_form = smf.ols("log_wage ~ C(year_of_birth) + C(state_of_birth) + q4", data=factor_data).fit()

print("q4 parameter estimate:, ", reduced_form.params["q4"])
print("q4 p-value:, ", reduced_form.pvalues["q4"])
````

И снова мы получили значимый результат. У родившихся в последнем квартале года заработная плата в среднем на 0,8% выше. На этот раз p-значение не так близко к нулю, как раньше, но всё ещё довольно значимо, составляя всего 0,0015.

**Инструментальные переменные вручную**

Имея как редуцированную форму, так и первую стадию, мы теперь можем масштабировать эффект первой стадии по редуцированной форме. Поскольку коэффициент первой стадии был около 0,1, это умножит эффект коэффициента редуцированной формы почти на 10. Это даст нам нашу несмещенную оценку IV среднего причинно-следственного эффекта:

$
\mathrm{ATE}_{IV} = \dfrac{\text{Редуцированная форма}}{\text{1-я стадия}}
$

````python
reduced_form.params["q4"] / first_stage.params["q4"]
````

Это означает, что каждый дополнительный год обучения в школе увеличит заработную плату на 8%.

Другой способ получить оценку IV — использовать двухэтапный метод наименьших квадратов (**2SLS**). В этой процедуре мы выполняем первый этап, как и раньше, а затем запускаем второй этап, на котором заменяем экспериментальную переменную подобранными значениями первого этапа.

$
\mathrm{educ}_i = \gamma_0 + \gamma_1 \times \mathrm{q4}_i + \gamma_2\times \mathrm{yob}_i + \gamma_3\times \mathrm{sob}_i + v_i
$

$
\log(\mathrm{wage})_i = \beta_0 + \beta_1\times \mathrm{educ}_i + \beta_2\times \mathrm{yob}_i + \beta_3\times \mathrm{sob}_i + u_i
$

$
\log(\mathrm{wage})_i = \beta_0 + \beta_1 [\gamma_0 + \gamma_1 \times \mathrm{q4}_i + \gamma_2\times \mathrm{yob}_i + \gamma_3\times \mathrm{sob}_i + v_i ] + \beta_2\times \mathrm{yob}_i + \beta_3\times \mathrm{sob}_i + u_i
$

Следует отметить, что **любой дополнительный параметр, который мы добавляем на втором этапе, также должен быть добавлен на первом этапе при выполнении IV**.

````python
iv_by_hand = smf.ols("log_wage ~ C(year_of_birth) + C(state_of_birth) + years_of_schooling_fitted",
                     data=factor_data.assign(years_of_schooling_fitted=first_stage.fittedvalues)).fit()

iv_by_hand.params["years_of_schooling_fitted"]
````

Как видите, параметры абсолютно одинаковы. Этот второй подход к IV может быть полезен благодаря интуитивному пониманию. В 2SLS на первом этапе создаётся новая версия воздействия, очищенная от смещения пропущенных переменных. Затем мы используем эту очищенную версию воздействия в линейной регрессии.

На практике, однако, мы не вычисляем IV вручную. Не потому, что это сложно, а потому, что стандартные ошибки, получаемые на этом втором этапе, немного неточны. В Python мы можем использовать библиотеку [linearmodels](https://bashtage.github.io/linearmodels/) для правильного запуска 2SLS.

Формула для 2SLS немного отличается. Мы должны добавить первую стадию между \[ \] внутри формулы. В нашем случае мы добавляем `years_of_schooling ~ q4`. Добавлять дополнительные переменные на первом этапе не нужно, поскольку это сделается автоматически, если мы включим их на втором этапе. Поэтому мы добавляем `year_of_birth` и `state_of_birth` вне формулы первого этапа.

````python
def parse(model, exog="years_of_schooling"):
    param = model.params[exog]
    se = model.std_errors[exog]
    p_val = model.pvalues[exog]
    print(f"Parameter: {param}")
    print(f"SE: {se}")
    print(f"95 CI: {(-1.96*se,1.96*se) + param}")
    print(f"P-value: {p_val}")

formula = 'log_wage ~ 1 + C(year_of_birth) + C(state_of_birth) + [years_of_schooling ~ q4]'
iv2sls = IV2SLS.from_formula(formula, factor_data).fit()
parse(iv2sls)
````

Мы снова видим, что параметр точно такой же, как и те, что были у нас раньше. Дополнительным преимуществом является то, что теперь у нас есть достоверные стандартные ошибки. Имея это в виду, мы можем сказать, что ожидаем, что один дополнительный год обучения увеличит заработную плату в среднем на 8,5%.

**Множественные инструменты**

Ещё одно преимущество использования готовых решений при запуске 2SLS-метода заключается в том, что можно легко добавить несколько инструментов. В нашем примере мы будем использовать все фиктивные переменные кварталов рождения в качестве инструментов для лет обучения.

````python
formula = 'log_wage ~ 1 + C(year_of_birth) + C(state_of_birth) + [years_of_schooling ~ q1+q2+q3]'
iv_many_zs = IV2SLS.from_formula(formula, factor_data).fit()
parse(iv_many_zs)
````

При всех трёх фиктивных переменных расчётная доходность образования теперь составляет 0.1, что означает, что мы должны ожидать среднего увеличения дохода на 10% за каждый дополнительный год обучения. Сравним это с традиционной оценкой по МНК. Для этого мы можем снова использовать метод 2SLS, но теперь без первой стадии.

````python
formula = "log_wage ~ years_of_schooling + C(state_of_birth) + C(year_of_birth) + C(quarter_of_birth)"
ols = IV2SLS.from_formula(formula, data=data).fit()
parse(ols)
````

Окупаемость образования оценивается ниже при использовании OLS, чем при использовании 2SLS. Это говорит о том, что OVB может быть не столь сильным, как мы предполагали. Также обратите внимание на доверительные интервалы. 2SLS имеет гораздо более широкий доверительный интервал (CI), чем оценка OLS. Давайте рассмотрим это подробнее.

**Слабость инструментов**

При работе с IV необходимо помнить, что мы оцениваем ATE косвенно. Наши оценки зависят как от первого, так и от второго этапа. Если влияние воздействия на исход действительно сильное, то и второй этап будет сильным. Однако неважно, насколько силен второй этап, если первый этап слабый. Слабый первый этап означает, что инструмент имеет лишь очень слабую корреляцию с воздействием. Следовательно, мы не можем много узнать о воздействии с помощью инструмента.

Формулы для вычисления стандартных ошибок IV немного сложны и не столь интуитивно понятны, поэтому мы попробуем другой подход, чтобы разобраться в этой проблеме. Мы смоделируем данные, включающие воздействие $T$ с эффектом 2.0 на результат $Y$, ненаблюдаемый конфаундер $U$ и дополнительный контрольный фактор $X$. Мы также смоделируем несколько инструментов с различной интенсивностью на первом этапе.

$$
\begin{align}
X \sim & N(0, 2^2)\\
U \sim & N(0, 2^2)\\
T \sim & N(1+0,5U, 5^2)\\
Y \sim & N(2+ X - 0,5U + 2T, 5^2)\\
Z \sim & N(T, \sigma^2) \text{ для }\sigma^2 \text{ в диапазоне от 0,1 до 100}
\end{align}
$$

````python
np.random.seed(12)
n = 10000
X = np.random.normal(0, 2, n) # observable variable
U = np.random.normal(0, 2, n) # unobservable (omitted) variable
T = np.random.normal(1 + 0.5*U, 5, n) # treatment
Y = np.random.normal(2 + X - 0.5*U + 2*T, 5, n) # outcome

stddevs = np.linspace(0.1, 100, 50)
Zs = {f"Z_{z}": np.random.normal(T, s, n) for z, s in enumerate(stddevs)} # instruments with decreasing \mathrm{Cov}(Z, T)

sim_data = pd.DataFrame(dict(U=U, T=T, Y=Y)).assign(**Zs)

sim_data.head()
````

Для повторной проверки мы видим, что корреляция между Z и T действительно уменьшается.

````python
corr = (sim_data.corr()["T"]
        [lambda d: d.index.str.startswith("Z")])

corr.head()
````

Теперь мы запустим одну модель IV для каждого имеющегося у нас прибора и соберем как оценку ATE, так и стандартную ошибку.

````python
se = []
ate = []
for z in range(len(Zs)):
    formula = f'Y ~ 1 + X + [T ~ Z_{z}]'
    iv = IV2SLS.from_formula(formula, sim_data).fit()
    se.append(iv.std_errors["T"])
    ate.append(iv.params["T"])
````

````python
plot_data = pd.DataFrame(dict(se=se, ate=ate, corr=corr)).sort_values(by="corr")

plt.scatter(plot_data["corr"], plot_data["se"])
plt.xlabel("Corr(Z, T)")
plt.ylabel("IV Standard Error");
plt.title("Variance of the IV Estimates by 1st Stage Strength");
````

````python
plt.scatter(plot_data["corr"], plot_data["ate"])
plt.fill_between(plot_data["corr"],
                 plot_data["ate"]+1.96*plot_data["se"],
                 plot_data["ate"]-1.96*plot_data["se"], alpha=.5)
plt.xlabel("Corr(Z, T)")
plt.ylabel("$\hat{ATE}$");
plt.title("IV ATE Estimates by 1st Stage Strength");
````

Как видно на графиках выше, оценки сильно различаются при слабой корреляции между $T$ и $Z$. Это связано с тем, что стандартная ошибка ($SE$) также значительно увеличивается при низкой корреляции.

Ещё один момент, на который следует обратить внимание, заключается в том, что **метод 2SLS смещен**! Даже при высокой корреляции оценка параметра всё ещё не достигает истинного значения $ATE$ 2.0 . На самом деле, 2.0 даже не попадает в 95% доверительный интервал (CI). Метод 2SLS является состоятельным, то есть приближается к истинному значению параметра только при достаточно большом размере выборки. Однако мы не можем знать, насколько большим является этот размер. Мы можем лишь придерживаться некоторых эмпирических правил, чтобы понять, как ведёт себя это смещение:

1. Метод 2SLS смещен в сторону OLS. Это означает, что если OLS имеет отрицательное/положительное смещение, то и у 2SLS оно будет. Преимущество метода 2SLS заключается в том, что он, по крайней мере, является состоятельным, в отличие от OLS, в случае OVB. В приведенном выше примере наше ненаблюдаемое значение $U$ отрицательно влияет на результат, но положительно коррелирует с воздействием, что приводит к отрицательному смещению. Именно поэтому мы видим оценку ATE ниже истинного значения (отрицательное смещение).

2. Смещение будет увеличиваться с количеством добавляемых инструментов. Если добавить слишком много инструментов, метод 2SLS всё больше становится похож на метод OLS.

Помимо понимания того, как ведёт себя это смещение, последний совет — избегать некоторых **распространённых ошибок при вычислении IV**:

1. Ручное вычисление IV. Как мы видели, ручное вычисление IV приведёт к неверным стандартным ошибкам, даже если оценки параметров верны. $SE$ не будет полностью ошибочной. Однако зачем это делать, если можно использовать программное обеспечение и получить правильную $SE$?

2. Использование чего-либо другого, кроме OLS, на первом этапе. Многие специалисты по анализу данных сталкиваются с IV и думают, что могут добиться большего. Например, они видят фиктивную переменную и думают о замене первого этапа логистической регрессией. В конце концов, они предсказывают фиктивную переменную, верно? Проблема в том, что это совершенно неверно. Согласованность IV основана на свойстве, которое может обеспечить только OLS, а именно на ортогональности остатков, поэтому любое отличие от OLS на первом этапе приведет к смещению. (Замечание: существуют некоторые современные методы, использующие машинное обучение для IV, но их результаты, в лучшем случае, сомнительны).

##### Non-compliance ситуации

**Погружение в гетерогенный мир**

Ранее мы рассматривали инструментальные переменные с более традиционной точки зрения. IV рассматривалось как своего рода естественный эксперимент, который мы можем использовать. Современная практика IV, напротив, во многом опирается на знания из медицинской науки. Она разделяет мир на четыре типа субъектов в зависимости от их реакции на инструментальную переменную.

1. Соглашающиеся (Compliers)
2. Никогда не принимающие (Never Takers)
3. Всегда принимающие (Always Takers)
4. Не поддающиеся (Defiers)

Это название пришло из фармацевтической науки. Представьте, что вы проводите эксперимент по изучению влияния нового препарата на какое-либо заболевание. Каждому испытуемому назначается определённое лечение: лекарство или плацебо. Соглашающиеся — это те, кто придерживается назначенного им лечения. Если им назначают плацебо, они его принимают; если им назначают лекарство, они также его принимают. Никогда не принимающие — это те, кто отказывается принимать лекарство. Даже если им назначают новое лекарство, они его не принимают. С другой стороны, «постоянно принимающие» – это те, кто каким-то образом может получить новый препарат, даже если им назначено плацебо. Наконец, не поддающиеся – это те, кто принимает лекарство, если им назначено контрольное исследование, и принимает контрольное исследование, если им назначено лечение. Их можно представить как надоедливого ребёнка, который делает всё наоборот. На практике они встречаются не так часто, поэтому мы часто будем их игнорировать.

Современный метод IV рассматривает инструмент как квазиэкспериментальный, где соблюдение рекомендаций не является безупречным. Тем самым он различает внутренне и внешне валидные причинно-следственные эффекты. Внутренне валидный эффект – это тот, который мы можем определить. Он валиден в конкретных условиях и с конкретными данными. В методе IV это будет эффект лечения у тех, кому инструмент меняет лечение. Внешняя валидность, с другой стороны, касается предсказательной силы этого причинно-следственного эффекта. Она ставит вопрос о том, можем ли мы обобщить эффект, обнаруженный в данной выборке, на другие популяции. Например, предположим, что вы провели рандомизированное контролируемое исследование (РКИ) в своём университете, чтобы выяснить, проявляют ли люди щедрость, если им предложить пожертвование. Эксперимент хорошо продуман, но вы приглашаете к участию только студентов-экономистов. Затем вы обнаруживаете, что все они — эгоистичные придурки. Это внутренне обоснованный вывод. Он справедлив для этих точек данных. Но можно ли на основании этого эксперимента сделать вывод, что человечество эгоистично? Вряд ли. Поэтому мы сомневаемся, обладает ли ваш эксперимент внешней валидностью, чтобы обобщить его результаты.

Для большей конкретности рассмотрим случай, когда вы хотите повысить вовлечённость пользователей, измеряемую покупками в приложении. Один из способов сделать это — попросить свой отдел маркетинга придумать push-уведомление, которое можно использовать для привлечения пользователей. Они предлагают блестящий дизайн и очень интересные пользовательские взаимодействия. С этим push-уведомлением вы переходите к разработке рандомизированного исследования. Вы выбираете 10 000 случайных клиентов и каждому из них назначаете push-уведомление с вероятностью 50%. Однако при проведении теста вы замечаете, что некоторые клиенты, которым было назначено получать push-уведомления, их не получают. Когда вы общаетесь с инженерами, они говорят, что это, вероятно, связано с тем, что у них старый телефон, который не поддерживает тип push-уведомлений, разработанный маркетинговой командой.

Сначала вы можете подумать, что в этом нет ничего особенного. Вместо того, чтобы использовать назначенное воздействие в качестве переменной воздействия, вы можете просто использовать полученное воздействие, верно? Оказывается, всё не так просто. Если построить причинно-следственный график для всей этой ситуации, он будет выглядеть примерно так:

На причинно-следственном графе вы видите назначение push-уведомлений. Оно изначально случайное, поэтому ничто не является причиной. Затем у вас есть узел, соответствующий факту доставки push-уведомлений. Не все, кому было назначено получать push-уведомления, сделали это, поэтому здесь имеет место несоблюдение. Точнее, есть некоторые «никогда не получающие»: те, кто не получает воздействия, даже если оно назначено. У вас также есть основания подозревать, что это несоблюдение не просто случайно. Поскольку пользователи со старыми телефонами не получают push-уведомления, можно утверждать, что доход также является причиной доставки push-уведомлений. Чем богаче человек, тем выше вероятность, что у него или у неё более современный телефон, что, в свою очередь, увеличивает вероятность получения им или ею push-уведомлений. Наконец, у вас есть переменная результата – покупка в приложении. Имейте в виду, что мы не знаем доход, поэтому не можем его контролировать. Имея это в виду, давайте рассмотрим, что произойдёт, если мы просто используем назначенное push-уведомление в качестве переменной воздействия, и если вместо этого мы используем в качестве переменной воздействия доставленное push-уведомление. В первом случае мы бы оценили причинно-следственный эффект по следующей разнице средних значений:

$
ATE = E[Y | pushAssigned=1] - E[Y | pushAssigned=0]
$

Как мы уже прекрасно понимаем, это несмещенная оценка для $E[Y_1] - E[Y_0]$, только если смещение $E[Y_0|pushAssigned=0] - E[Y_0|pushAssigned=1]$ равно нулю. Поскольку `pushAssigned` является случайным, мы знаем, что смещение равно нулю. Исправляет ли это вопрос? Не совсем. Видите ли, если мы так поступаем, мы фактически отвечаем на другой вопрос, а не на тот, который намеревались. Мы бы обнаружили **причинно-следственный эффект назначения воздействия**, а не само воздействие. Но различаются ли они, или мы можем экстраполировать причинно-следственный эффект назначения воздействия на ATE? Другими словами, является ли причинно-следственный эффект назначения лечения несмещенной оценкой ATE?

Как оказалось, нет. Из-за непринятия воздействия результаты участников, назначенных в тестовую группу, будут смещены в сторону результатов участников, назначенных на контрольную группу. Непринятие воздействия непреднамеренно меняет направление воздействия, делая тестовую и контрольную группы более схожими по результатам. Не путайте это со сходством переменных. Мы хотим, чтобы тестовая и контрольная группы были схожи по переменным. Это сделает их сопоставимыми. Мы не хотим, чтобы они были схожи по результатам, если эффект воздействия действительно есть.

Чтобы убедиться в этом, предположим, что сначала у нас есть все участники. Некоторые из них попадут в контрольную группу случайно. Но те, кто попали в контрольную группу, всё равно будут принимать воздействие. Это делает их, по сути, тестовой группой, перемешанной с контрольной. В результате такого смешения причинно-следственный эффект будет сложнее обнаружить при наличии непринятия воздействия.

По той же причине те, кто никогда не принимал воздействия, будут выглядеть немного похожими на тех, кто не получал его, поскольку они не принимают воздействия, даже если оно назначено. В этом смысле **причинно-следственный эффект назначения воздействия смещен в сторону нуля**, поскольку несоблюдение уменьшает обнаруживаемое воздействие. Другой способ увидеть это — представить себе крайний случай. Предположим, что несоблюдение воздействия очень высоко. Назначение воздействия ничего не говорит о полученном воздействии. Полученное воздействие в данном случае является чисто случайным. На языке IV это означало бы, что у нас очень слабая первая стадия. Используя $Z$ для обозначения назначения воздействия, мы получим следующее:

$
E[Y|Z=1] - E[Y|Z=0] = 0
$

Более не будет причинно-следственной связи между назначением воздействием и результатом - $Z$ будет просто бессмысленной случайной величиной.

Итак, мы исключили возможность использования причинно-следственной связи назначения в качестве способа оценки причинно-следственной связи воздействия. А как насчёт использования только полученного воздействия?

$
\mathrm{ATE} = E[Y | \mathrm{push}=1] - E[Y | \mathrm{push}=0]
$

И снова нам нужно подумать, является ли это смещением или нет. Просто взглянув на причинно-следственный график выше, мы понимаем, что это не так. У нас есть этот неучтённый фактор, доход, который, несомненно, всё испортит. Как мы уже говорили ранее, мы знаем, что сбои в доставке push-уведомлений в нашем случае вызваны устаревшими телефонами клиентов. Это означает, что, вероятно, $E[Y_0|\mathrm{push}=0] < E[Y_0|\mathrm{push}=1]$. Мы полагаем, что это так, поскольку у клиентов с меньшим бюджетом телефоны старые, что приводит к $\mathrm{push}=0$ и, кроме того, к снижению вероятности покупки в приложении $Y_0$.

Обидно! Мы не можем использовать ни назначенное воздействие, ни полученное воздействие для оценки нашего ATE. Но, к счастью, мы знаем, что можем использовать: инструментальные переменные. В данном случае назначенное воздействие — идеальный инструмент для оценки воздействия. Оно практически случайно и вызывает покупку в приложении только благодаря самому воздействию.

**Local Average Treatment Effect: LATE**

Локальный средний эффект лечения позволяет определить, для какой популяции мы можем оценить причинно-следственный эффект. Это также ещё один способ представления IV, который даёт нам другие интересные интуитивные выводы. В современном IV мы рассматриваем инструмент как начало причинно-следственной цепочки: $Z$ вызывает $T$, который вызывает $Y$. В этом контексте ограничение исключения означает, что $Z$ не вызывает $Y$, за исключением случаев его воздействия на $T$. Первый этап теперь рассматривается как причинно-следственное воздействие $Z$ на $T$. Мы также переписываем потенциальные результаты с помощью двойной индексации, где первый индекс обозначает контрфактуальное для инструмента, а второй — для воздействия.

$
\text{Потенциальный результат}=\begin{cases}
Y_i(1, 1) \ \text{if } T_i=1, \ Z_i=1\\
Y_i(1, 0) \ \text{if } T_i=1, \ Z_i=0\\
Y_i(0, 1) \ \text{if } T_i=0, \ Z_i=1\\
Y_i(0, 0) \ \text{if } T_i=0, \ Z_i=0\\
\end{cases}
$

Условия для инструментальной переменной теперь можно переписать так:

* **Независимость (Independence Assumption)** - $T_{0i}, T_{1i} \perp Z_i$.

Это означает, что инструмент назначен случайно. Иными словами, Z не коррелирует с потенциальными воздействиями, то есть группы с разными значениями инструмента сравнимы.

* **Ограничение исключения (Exclusion Restriction)** - $Y_i(1, 0)=Y_i(1, 1)=Y_{i1}$ и $Y_i(0, 0)=Y_i(0, 1)=Y_{i0}$.

Это означает, что потенциальный исход для воздействия одинаков в обеих группах инструмента. То есть инструмент влияет на исход только через воздействие.

* **Наличие первого этапа (Existence of 1st Stage)** - $E[T_{1i}-T_{0i}] \neq 0$.

Это означает, что инструмент действительно влияет на воздействие.

* **Монотонность (Monotonicity Assumption)** - $T_{i1} \geq T_{i0}$.

Это означает, что если у всех включён инструмент, уровень воздействия будет не ниже, чем если бы у всех он был выключен.


Теперь рассмотрим оценку Вальда, чтобы лучше понять IV:

$
ATE = \dfrac{E[Y|Z=1]-E[Y|Z=0]}{E[T|Z=1]-E[T|Z=0]}
$

Возьмём её первую часть, $E[Y|Z=1]$. Используя ограничение исключения, мы можем переписать Y в терминах потенциального результата следующим образом.

$
E[Y_i|Z_i=1]=E[Y_{i0} + T_{i1}(Y_{i1} - Y_{i0})|Z=1]
$

Используя независимость, мы можем убрать обусловленность по Z.

$
E[Y_i|Z_i=1]=E[Y_{i0} + T_{i1}(Y_{i1} - Y_{i0})]
$

Рассуждая аналогично, получаем:

$
E[Y_i|Z_i=0]=E[Y_{i0} + T_{i0}(Y_{i1} - Y_{i0})]
$

Теперь мы можем переписать числитель оценки Вальда следующим образом:

$
E[Y|Z=1]-E[Y|Z=0] = E[(Y_{i1}-Y_{i0})(T_{i1}-T_{i0})]
$

Используя монотонность, мы знаем, что $T_{i1}-T_{i0}$ равно 0 или 1, поэтому

$
E[(Y_{i1}-Y_{i0})(T_{i1}-T_{i0})] = E[(Y_{i1}-Y_{i0})|T_{i1}>T_{i0}]P(T_{i1}>T_{i0})
$

Используя аналогичные рассуждения для знаменателя, получаем, что

$
E[T|Z=1]-E[T|Z=0]=E[T_{i1}-T_{i0}]=P(T_{i1}>T_{i0})
$

Получается, что оценка Вальда выглядит следующим образом:

$
ATE = \dfrac{E[(Y_{i1}-Y_{i0})|T_{i1}>T_{i0}]P(T_{i1}>T_{i0})}{P(T_{i1}>T_{i0})}=E[(Y_{i1}-Y_{i0})|T_{i1}>T_{i0}]
$

То есть, ATE, оцененная по IV, — это ATE для субпопуляции, где $T_{i1}>T_{i0}$. А если говорить о принятии воздействия, то какая это популяция? Это популяция, где у тех, у кого инструмент включен, уровень воздействия выше, чем при выключенном инструменте. Другими словами, это популяция, соблюдающая режим воздействия.

Вывод заключается в том, что IV ничего не говорит о влиянии на «никогда не принимающих», «всегда принимающих» или «не поддающихся», поскольку для них воздействие не меняется! **IV определяет эффект лечения только для комплайеров**.

##### Regression Discontinuity Design

В нормальных условиях природа очень целостна и не подвержена резким скачкам. Это означает, что **когда мы видим скачки и всплески, они, вероятно, искусственные** и часто вызваны человеком. Эти события обычно сопровождаются контрфактуальными ситуациями по отношению к нормальному ходу вещей: если происходит что-то странное, это даёт нам некоторое представление о том, что произошло бы, если бы природа действовала иначе. Исследование этих искусственных скачков лежит в основе метода разрывной регрессии (RDD).

Базовая схема выглядит следующим образом. Представьте, что у вас есть переменная воздействия $T$ и потенциальные результаты $Y_0$ и $Y_1$. Воздействие T является разрывной функцией наблюдаемой текущей переменной $R$, такой что $D_i = \mathcal{1}\{R_i>c\}$

Другими словами, это означает, что воздействие равно нулю, когда $R$ ниже порогового значения $c$, и единице в противном случае. Это означает, что мы можем наблюдать $Y_1$ при $R>c$ и $Y_0$ при $R<c$. Чтобы понять это, представьте себе потенциальные результаты как две функции, которые мы не можем наблюдать полностью. И $Y_0(R)$, и $Y_1(R)$ присутствуют, просто мы их не видим. Порог действует как переключатель, позволяющий нам видеть только одну из этих функций, но никогда обе, как на изображении ниже:

image.png

Идея разрыва регрессии заключается в сравнении результатов чуть выше и чуть ниже порога, чтобы определить эффект на пороге. Это называется **резким RD**-дизайном, поскольку вероятность получения воздействия резко возрастает от 0 до 1 при достижении порога. Но можно также рассмотреть **нечеткий RD**-дизайн, где вероятность также возрастает, но менее резко.

**Алкоголь убивает вас?**

В большинстве стран он установлен в 18 лет, но в США (в большинстве штатов) сейчас он составляет 21 год. Так следует ли считать, что США проявляют чрезмерную осторожность и им следует снизить минимальный возраст употребления алкоголя? Или же другим странам следует повысить свой возраст, разрешенный к употреблению алкоголя?

Один из способов рассмотреть этот вопрос — с точки зрения [уровня смертности (Карпентер и Добкин, 2009)](https://www.aeaweb.org/articles?id=10.1257/app.1.1.164). С точки зрения государственной политики можно утверждать, что мы должны максимально снизить уровень смертности. Это соответствовало бы цели снижения смертности, вызванной употреблением алкоголя.

Чтобы оценить влияние алкоголя на смертность, мы могли бы использовать тот факт, что возраст, с которого разрешено употребление алкоголя, нарушает непрерывность в природе. В США люди младше 21 года не пьют (или пьют гораздо меньше), в то время как те, кому чуть больше 21 года, пьют. Это означает, что вероятность употребления алкоголя резко возрастает в 21 год, и это можно исследовать с помощью RDD.

**Оценка RDD**

Ключевое предположение, на котором основывается RDD, — это плавность потенциального результата при пороговом значении. Формально пределы потенциальных результатов при приближении текущей переменной к пороговому значению справа и слева должны быть одинаковыми.

$$
\lim_{r \to c^-} E[Y_{ti}|R_i=r] = \lim_{r \to c^+} E[Y_{ti}|R_i=r]
$$

Если это верно, мы можем найти причинно-следственную связь на пороговом уровне.

$$
\begin{align}
\lim_{r \to c^+} E[Y_{ti}|R_i=r] - \lim_{r \to c^-} E[Y_{ti}|R_i=r]=&\lim_{r \to c^+} E[Y_{1i}|R_i=r] - \lim_{r \to c^-} E[Y_{0i}|R_i=r] \\
=& E[Y_{1i}|R_i=r] - E[Y_{0i}|R_i=r] \\
=& E[Y_{1i} - Y_{0i}|R_i=r]
\end{align}
$$

Это, своего рода, локальный средний эффект лечения (LATE), поскольку мы можем знать его только на пороговом уровне. В этом контексте мы можем рассматривать RDD как локальное рандомизированное исследование. Для тех, кто достиг порогового значения, лечение могло пойти в любом направлении, и по случайности некоторые люди оказались ниже порогового значения, а некоторые – выше. В нашем примере в один и тот же момент времени некоторым людям чуть больше 21 года, а некоторым чуть меньше. Это определяется тем, родился ли человек на несколько дней позже или нет, что довольно случайно. По этой причине RDD предоставляет очень убедительную причинно-следственную связь. Это не золотой стандарт РКИ, но близко к нему.

Теперь, чтобы оценить эффект на пороговом уровне, нам нужно всего лишь оценить оба предела в приведенной выше формуле и сравнить их. Проще всего это сделать с помощью линейной регрессии.

Чтобы это сработало, мы взаимодействуем с фиктивной переменной, определяющей превышение порога, с текущей переменной.

$
y_i = \beta_0 + \beta_1 r_i + \beta_2 \mathcal{1}\{r_i>c\} + \beta_3 \mathcal{1}\{r_i>c\} r_i
$

По сути, это то же самое, что подбирать одну линейную регрессию выше порога, а другую — ниже. Параметр $\beta_0$ — это точка пересечения регрессии ниже порога, а $\beta_0+\beta_2$ — точка пересечения регрессии выше порога.

Здесь вступает в игру трюк с центрированием текущей переменной на пороге. После этого этапа предварительной обработки порог становится равным нулю. В результате точка пересечения $\beta_0$ становится прогнозируемым значением на пороге для регрессии ниже порога. Другими словами, $\beta_0=\lim_{r \to c^-} E[Y_{ti}|R_i=r]$. По тем же причинам, $\beta_0+\beta_2$ — это предел результата, полученного выше. Это означает, что

$
\lim_{r \to c^+} E[Y_{ti}|R_i=r] - \lim_{r \to c^-} E[Y_{ti}|R_i=r]=\beta_2=E[ATE|R=c]
$

Вот как это выглядит в коде для случая, когда мы хотим оценить влияние потребления алкоголя на смертность от всех причин в возрасте 21 года.

````python
rdd_df = drinking.assign(threshold=(drinking["agecell"] > 0).astype(int))

model = smf.wls("all~agecell*threshold", rdd_df).fit()

print(model.summary())
````

````python
ax = drinking.plot.scatter(x="agecell", y="all", color="C0")
drinking.assign(predictions=model.fittedvalues).plot(x="agecell", y="predictions", ax=ax, color="C1")
plt.title("Regression Discontinuity");
````

Если мы сделаем то же самое в отношении других причин, вот что мы получим.

````python
plt.figure(figsize=(8,8))

for p, cause in enumerate(["all", "mva", "suicide"], 1):
    ax = plt.subplot(3,1,p)
    drinking.plot.scatter(x="agecell", y=cause, ax=ax)
    m = smf.wls(f"{cause}~agecell*threshold", rdd_df).fit()
    ate_pct = 100*((m.params["threshold"] + m.params["Intercept"])/m.params["Intercept"] - 1)
    drinking.assign(predictions=m.fittedvalues).plot(x="agecell", y="predictions", ax=ax, color="C1")
    plt.title(f"Impact of Alcohol on Death: {np.round(ate_pct, 2)}%")

plt.tight_layout()
````

RDD показывает, что алкоголь увеличивает вероятность самоубийства и смерти в результате автомобильных аварий на 15%, что довольно существенно. Эти результаты убедительно доказывают, что не следует снижать возраст, с которого разрешено употребление алкоголя, если мы хотим минимизировать уровень смертности.

**Kernel Weighting**

Разрыв регрессии в значительной степени зависит от экстраполяционных свойств линейной регрессии. Поскольку мы рассматриваем значения в начале и конце двух линий регрессии, нам следует правильно установить эти пределы. В противном случае регрессия может быть слишком сосредоточена на подгонке других точек данных ценой плохого соответствия пороговому значению. В этом случае мы можем получить неверную оценку эффекта лечения.

Один из способов решения этой проблемы — присвоить более высокие веса точкам, расположенным ближе к пороговому значению. Существует множество способов сделать это, но наиболее популярным является перевзвешивание выборок с помощью **треугольного ядра**.

$
K(R, c, h) = \mathcal{1}\{|R-c| \leq h\} * \bigg(1-\frac{|R-c|}{h}\bigg)
$

Первая часть этого ядра — индикаторная функция, показывающая, насколько близко мы к порогу. Насколько близко? Это определяется параметром полосы пропускания $h$. Вторая часть этого ядра — весовая функция. По мере удаления от порога веса становятся всё меньше и меньше. Эти веса делятся на полосу пропускания. Если полоса пропускания большая, веса уменьшаются медленнее. Если полоса пропускания маленькая, веса быстро стремятся к нулю.

Для простоты понимания, вот как выглядят веса для этого ядра, применяемого к нашей задаче. Я установил здесь полосу пропускания, равную 1, то есть мы будем рассматривать только данные людей не старше 22 лет и не моложе 20 лет.

````python
def kernel(R, c, h):
    indicator = (np.abs(R-c) <= h).astype(float)
    return indicator * (1 - np.abs(R-c)/h)

plt.plot(drinking["agecell"], kernel(drinking["agecell"], c=0, h=1))
plt.xlabel("agecell")
plt.ylabel("Weight")
plt.title("Kernel Weight by Age");
````

Если применить эти веса к нашей исходной задаче, влияние алкоголя увеличится, по крайней мере, по всем причинам. Оно подскочит с 7,6627 до 9,7004. Результат остаётся весьма значительным.

````python
model = smf.wls("all~agecell*threshold", rdd_df,
                weights=kernel(drinking["agecell"], c=0, h=1)).fit()

print(model.summary())
````

````python
ax = drinking.plot.scatter(x="agecell", y="all", color="C0")
drinking.assign(predictions=model.fittedvalues).plot(x="agecell", y="predictions", ax=ax, color="C1")
plt.title("Regression Discontinuity (Local Regression)");
````

````python
plt.figure(figsize=(8,8))
weights = kernel(drinking["agecell"], c=0, h=1)

for p, cause in enumerate(["all", "mva", "suicide"], 1):
    ax = plt.subplot(3,1,p)
    drinking.plot.scatter(x="agecell", y=cause, ax=ax)
    m = smf.wls(f"{cause}~agecell*threshold", rdd_df, weights=weights).fit()
    ate_pct = 100*((m.params["threshold"] + m.params["Intercept"])/m.params["Intercept"] - 1)
    drinking.assign(predictions=m.fittedvalues).plot(x="agecell", y="predictions", ax=ax, color="C1")
    plt.title(f"Impact of Alcohol on Death: {np.round(ate_pct, 2)}%")

plt.tight_layout()
````

За исключением самоубийств, похоже, что добавление веса ядра усилило негативное влияние алкоголя. Опять же, если мы хотим минимизировать уровень смертности, нам НЕ следует рекомендовать снижение возраста, с которого разрешено употребление алкоголя, поскольку влияние алкоголя на уровень смертности очевидно.

Этот простой пример показывает, что происходит, когда метод разрывной регрессии работает идеально. Далее мы рассмотрим несколько диагностических тестов, которые следует провести, чтобы проверить, насколько мы можем доверять RDD, и обсудить очень близкую нам тему: влияние образования на заработок.

**Эффект овечьей шкуры и нечёткий RDD**

Что касается влияния образования на заработок, в экономике существует два основных взгляда. Первый — это широко известный аргумент о том, что образование увеличивает человеческий капитал, повышая производительность труда и, следовательно, заработок. Согласно этому взгляду, образование фактически меняет вас к лучшему. Другая точка зрения заключается в том, что образование — это всего лишь сигнальный механизм. Оно просто проводит вас через все эти сложные испытания и учебные задачи. Если вы справляетесь, это сигнализирует рынку, что вы хороший сотрудник. Таким образом, образование не делает вас более продуктивным. Оно лишь показывает рынку, насколько продуктивным вы были всегда. Здесь важен диплом. Если он у вас есть, вам будут платить больше. Мы называем это **эффектом овечьей шкуры**, поскольку в прошлом дипломы печатались на овечьей шкуре.

Чтобы проверить эту гипотезу, [Кларк и Марторелл](https://faculty.smu.edu/millimet/classes/eco7321/papers/clark%20martorell%202014.pdf) использовали разрыв регрессии для измерения влияния окончания 12-го класса на заработок. Для этого им пришлось подумать о некой переменной, по которой учащиеся, чей уровень выше, заканчивают обучение, а те, чей уровень ниже, — нет. Они нашли такие данные в системе образования Техаса.

Чтобы получить высшее образование в Техасе, нужно сдать экзамен. Тестирование начинается в 10-м классе, и ученики могут сдавать его несколько раз, но в конце 12-го класса им предстоит последний экзамен. Идея заключалась в том, чтобы собрать данные об учениках, сдававших эти экзамены, и сравнить тех, кто едва не провалил их, с теми, кто едва сдал. У этих учеников будет очень схожий человеческий капитал, но разные показатели квалификации. А именно, те, кто едва сдал экзамен, получат диплом.

````python
sheepskin = pd.read_csv("./data/rdd_df_two.csv")[["avgearnings", "minscore", "receivehsd", "n"]]
sheepskin.head()
````

Опять же, эти данные сгруппированы по текущей переменной. Она содержит не только текущую переменную (minscore, уже центрированную на нуле) и результат (avgearnings), но также вероятность получения диплома в этой ячейке с баллами и размер вызова (n). Так, например, из 12 студентов в ячейке с баллами -30 ниже порогового значения, только 5 смогли получить диплом (12 * 0,416).

Это означает, что в экспериментальном задании наблюдается некоторое отклонение. Некоторые студенты, не достигшие порогового значения, всё равно получили диплом. В данном случае разрыв регрессии **размытый**, а не резкий. Обратите внимание, что вероятность получения диплома не скачет с нуля до единицы при достижении порогового значения. Но она скачет примерно от 50% до 90%.

````python
sheepskin.plot.scatter(x="minscore", y="receivehsd", figsize=(10,5))
plt.xlabel("Test Scores Relative to Cut off")
plt.ylabel("Fraction Receiving Diplomas")
plt.title("Last-chance Exams");
````

Мы можем рассматривать нечёткое RD как своего рода несоответствие. Преодоление порога должно привести к тому, что все получат диплом, но некоторые студенты его не получают. Аналогично, если результат ниже порога, диплом не будет получен, но некоторые студенты всё равно его получают.

Точно так же, как и в случае с потенциальным результатом, в этой ситуации у нас есть потенциальный статус воздействия. $T_1$ — это воздействие, которое получили бы все, если бы результаты превысили порог. $T_0$ — это воздействие, которое получили бы все, если бы результаты оказались ниже порога. Как вы, возможно, заметили, мы можем рассматривать **порог как инструментальную переменную**. Как и в случае с IV, если мы наивно оцениваем эффект воздействия, он будет смещен в сторону нуля.

image.png

Вероятность получения воздействия меньше единицы, даже если оно превышает порог, делает наблюдаемый результат меньше истинного потенциального результата $Y_1$. Аналогично, результат, который мы наблюдаем ниже порога, выше истинного потенциального результата $Y_0$. Это создаёт впечатление, что эффект воздействия на пороге меньше, чем есть на самом деле, и нам придётся использовать методы IV для коррекции этого.

Точно так же, как мы предполагали гладкость потенциального результата, мы теперь предполагаем её и для потенциального воздействия. Кроме того, нам необходимо предположить монотонность, как в случае с IV ($T_{i1}>T_{i0} \ \forall i$). Это означает, что пересечение порога слева направо только увеличивает ваши шансы на получение диплома (или на отсутствие отклонений). С этими двумя предположениями мы получаем оценщик Вальда для LATE.

$$
\dfrac{\lim_{r \to c^+} E[Y_i|R_i=r] - \lim_{r \to c^-} E[Y_i|R_i=r]}{\lim_{r \to c^+} E[T_i|R_i=r] - \lim_{r \to c^-} E[T_i|R_i=r]} = E[Y_{1i} - Y_{0i} | T_{1i} > T_{0i}, R_i=c]
$$

Обратите внимание, что эта оценка является локальной в двух смыслах. Во-первых, она локальна, поскольку оценивает эффект воздействия только при пороговом значении $c$. Это локальность RD. Во-вторых, она локальна, поскольку оценивает эффект воздействия только для пациентов, соответствующих критериям. Это локальность IV.

Для оценки мы будем использовать 2-линейную регрессию. Числитель можно оценить так же, как мы делали ранее. Чтобы получить знаменатель, мы просто заменим результат на воздействие. Но сначала давайте поговорим о проверке корректности, которую необходимо провести, чтобы убедиться, что мы можем доверять нашим оценкам RDD.

**Тест МакКрари**

Одно из обстоятельств, которое может нарушить наш аргумент RDD, — это возможность людей манипулировать своим положением относительно порога. В нашем примере это могло произойти, если учащиеся, набравшие чуть меньше порога, нашли способ обойти систему и немного повысить свой результат теста. Другой пример — когда для получения государственного пособия нужно иметь доход ниже определённого уровня. Некоторые семьи могут намеренно занижать свой доход, просто чтобы иметь право на участие в программе.

В подобных ситуациях мы часто наблюдаем явление, называемое группировкой в плотности текущей переменной. Это означает, что у нас будет много сущностей, находящихся чуть выше или чуть ниже порога. Чтобы проверить это, мы можем построить график функции плотности текущей переменной и посмотреть, есть ли какие-либо пики вблизи порога. В нашем случае плотность определяется столбцом `n` в наших данных.

````python
plt.figure(figsize=(8,8))

ax = plt.subplot(2,1,1)
sheepskin.plot.bar(x="minscore", y="n", ax=ax)
plt.title("McCrary Test")
plt.ylabel("Smoothness at the Threshold")

ax = plt.subplot(2,1,2, sharex=ax)
sheepskin.replace({1877:1977, 1874:2277}).plot.bar(x="minscore", y="n", ax=ax)
plt.xlabel("Test Scores Relative to Cut off")
plt.ylabel("Spike at the Threshold");
````

Первый график показывает, как выглядит плотность наших данных. Как мы видим, нет пиков вблизи порога, что означает отсутствие группировки. Студенты не манипулируют своим положением на пороге. Для наглядности второй график показывает, как выглядела бы группировка, если бы студенты могли манипулировать своим положением на пороге. Мы бы увидели пик плотности для ячеек чуть выше порога, поскольку многие студенты находились бы в этой ячейке, едва сдавая экзамен.

Разобравшись с этим, мы можем вернуться к оценке эффекта овечьей шкуры. Как я уже говорил, числитель оценки Вальда можно оценить так же, как мы это делали в Sharp RD. Здесь мы будем использовать в качестве веса ядро с шириной полосы 15. Поскольку у нас также известен размер ячейки, мы умножим ядро на размер выборки, чтобы получить окончательный вес ячейки.

````python
sheepsking_rdd = sheepskin.assign(threshold=(sheepskin["minscore"]>0).astype(int))
model = smf.wls("avgearnings~minscore*threshold",
                sheepsking_rdd,
                weights=kernel(sheepsking_rdd["minscore"], c=0, h=15)*sheepsking_rdd["n"]).fit()

print(model.summary())
````

Это говорит нам о том, что эффект диплома составляет -97,7571, но статистически незначим (P-значение 0,5). Если мы отобразим эти результаты на графике, то получим очень непрерывную линию на пороговом значении. Более образованные люди действительно зарабатывают больше, но скачка в точке получения диплома об окончании 12-го класса не наблюдается. Это аргумент в пользу точки зрения, согласно которой образование увеличивает доход, повышая производительность труда, а не просто служит сигналом для рыночного индикатора. Другими словами, эффект овечьей шкуры отсутствует.

````python
ax = sheepskin.plot.scatter(x="minscore", y="avgearnings", color="C0")
sheepskin.assign(predictions=model.fittedvalues).plot(x="minscore", y="predictions", ax=ax, color="C1", figsize=(8,5))
plt.xlabel("Test Scores Relative to Cutoff")
plt.ylabel("Average Earnings")
plt.title("Last-chance Exams");
````

Однако из-за принципа работы смещения несоответствия, этот результат смещен в сторону нуля. Чтобы скорректировать это, нам нужно масштабировать его на первый этап и получить оценку Вальда. К сожалению, для этого нет хорошей реализации на Python, поэтому нам придётся сделать это вручную и использовать bootstrap для получения стандартных ошибок.

Приведённый ниже код вычисляет числитель оценки Вальда так же, как мы делали ранее, а также формирует знаменатель, заменяя целевую переменную на переменную обработки `receivehsd`. На последнем шаге числитель просто делится на знаменатель.

````python
def wald_rdd(data):
    weights=kernel(data["minscore"], c=0, h=15)*data["n"]
    denominator = smf.wls("receivehsd~minscore*threshold", data, weights=weights).fit()
    numerator = smf.wls("avgearnings~minscore*threshold", data, weights=weights).fit()
    return numerator.params["threshold"]/denominator.params["threshold"]

from joblib import Parallel, delayed

np.random.seed(45)
bootstrap_sample = 1000
ates = Parallel(n_jobs=4)(delayed(wald_rdd)(sheepsking_rdd.sample(frac=1, replace=True))
                          for _ in range(bootstrap_sample))
ates = np.array(ates)

sns.distplot(ates, kde=False)
plt.vlines(np.percentile(ates, 2.5), 0, 100, linestyles="dotted")
plt.vlines(np.percentile(ates, 97.5), 0, 100, linestyles="dotted", label="95% CI")
plt.title("ATE Bootstrap Distribution")
plt.xlim([-10000, 10000])
plt.legend();
````

Как видите, даже если масштабировать эффект по первой ступени, он всё ещё статистически не отличается от нуля. Это означает, что образование увеличивает доход не просто по принципу «эффекта овечьей шкуры», а за счёт повышения производительности труда.

#### Propensity Score

**Психология роста**

Позитивная психология изучает, как поведение человека способствует успешной жизни. Её можно рассматривать как пересечение книг по саморазвитию с академической строгостью статистики. Одним из известных открытий позитивной психологии является **мышление роста**. Идея заключается в том, что люди могут иметь как фиксированное, так и развивающееся мышление. Если у вас фиксированное мышление, вы верите, что способности даны вам при рождении или в раннем детстве. Таким образом, интеллект фиксирован и не может меняться на протяжении жизни. Если у вас его нет сейчас, вы не можете его развить. Следствием этой мысли является то, что вам не следует тратить время на области, в которых вы не преуспеваете, поскольку вы никогда не научитесь с ними справляться. С другой стороны, если у вас есть мышление роста, вы верите, что интеллект можно развить. Прямым следствием этого является то, что вы воспринимаете неудачи не как недостаток упорства, а как часть процесса обучения. Я не хочу спорить о том, какой из этих типов мышления правильный (хотя, вероятно, он где-то посередине). Для нас это не имеет большого значения. Важно то, что психологи обнаружили, что люди с установкой на рост, как правило, добиваются больших успехов в жизни. Они с большей вероятностью достигают поставленных целей.

Насколько мы разбираемся в причинно-следственных связях, мы научились относиться к этим утверждениям скептически. Действительно ли установка на рост побуждает людей достигать большего? Или же люди, которые добиваются большего, склонны развивать установку на рост в результате своего успеха? Что появилось первым, яйцо или курица? В нотации потенциального результата у нас есть основания полагать, что в этих утверждениях есть предвзятость. $Y_0|T=1$, вероятно, больше, чем $Y_0|T=0$, а это означает, что люди с установкой на рост добились бы большего, даже если бы у них было фиксированное мышление.

Чтобы прояснить ситуацию, исследователи разработали [Национальное исследование установок на обучение](https://mindsetscholarsnetwork.org/about-the-network/current-initatives/national-mindset-study/#). Это рандомизированное исследование, проводимое в государственных средних школах США, направленное на изучение влияния установки на рост. Суть его заключается в том, что учащиеся посещают семинар, призванный привить им установку на рост. Затем за ними в колледже наблюдают, чтобы оценить их успеваемость. Этот показатель был объединен в балл успеваемости и стандартизирован. Реальные данные этого исследования не публикуются в целях сохранения конфиденциальности учащихся. Однако у нас есть смоделированный набор данных с теми же статистическими свойствами, предоставленный [Athey and Wager](https://arxiv.org/pdf/1902.07409.pdf), поэтому мы будем использовать его.

````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from matplotlib import style
from matplotlib import pyplot as plt
import seaborn as sns
import graphviz as gr

%matplotlib inline

style.use("fivethirtyeight")
pd.set_option("display.max_columns", 6)

data = pd.read_csv("./data/propensity_score_df.csv")
data.sample(5, random_state=5)
````

Одна из возможных причин этого заключается в том, что экспериментальная переменная измеряется по факту посещения семинара студентом. Таким образом, хотя возможность участия была случайной, само участие – нет. В данном случае мы имеем дело со случаем несоблюдения требований. Одним из доказательств этого является корреляция между ожиданием студента успеха и участием в семинаре. Студенты с более высоким уровнем самооценки ожиданий успеха с большей вероятностью посещали семинар по мышлению роста.

Propensity score исходит из понимания того, что для достижения условной независимости $(Y_1, Y_0) \perp T | X$ не требуется прямого контроля конфаундеров $X$. Вместо этого достаточно контролировать балансирующий показатель $E[T|X]$. Этот балансирующий показатель часто представляет собой условную вероятность лечения $P(T|X)$, также называемую propensity score $e(x)$. Propensity score позволяет не накладывать условие на весь $X$ для достижения независимости потенциальных результатов лечения. Достаточно наложить условие на одну переменную — propensity score:

$
(Y_1, Y_0) \perp T | e(x)
$

Интуитивно propensity score — это условная вероятность получения лечения, верно? Таким образом, мы можем рассматривать это как своего рода функцию, которая преобразует $X$ в воздействие $T$. Показатель склонности создает промежуточное положение между переменной $X$ и воздействием $T$. Если мы покажем это в виде причинно-следственного графика, то это будет выглядеть следующим образом.

````python
g = gr.Digraph()
g.edge("T", "Y")
g.edge("X", "Y")
g.edge("X", "e(x)")
g.edge("e(x)", "T")
g
````

Если я знаю, чему равно $e(x)$, то один лишь $X$ не говорит мне ничего больше, что могло бы помочь мне узнать, чему будет равно $T$. Это означает, что контроль $e(x)$ действует так же, как прямой контроль $X$. Подумайте об этом с точки зрения нашей программы мышления. Группы изначально несопоставимы, потому что более амбициозные с большей вероятностью примут воздействией и добьются большего в жизни. Однако, если я возьму двух человек, одного из тестовой и одного из контрольной группы, но с одинаковой вероятностью получения воздействия, они будут сопоставимы. Если у них одинаковая вероятность получения воздействия, то единственная причина, по которой один из них получил его, а другой нет, — это чистая случайность.

Математически мы же хотим доказать, что $(Y_1, Y_0) \perp T | e(x)$. Это равносильно доказательству $E[T|e(x), X] = E[T|e(x)]$.

С одной стороны $E[T|e(x), X] = E[T|X] = e(x)$ по определению. С другой стороны, по [закону итеративных ожиданий](https://en.wikipedia.org/wiki/Law_of_total_expectation) $E[T|e(x)] = E[E[T|e(x),X]|e(x)] = E[e(x)|e(x)] = e(x)$

Следовательно, $E[T|e(x), X] = E[T|e(x)]$


Итак, мы получили propensity score. Что дальше? Как я уже говорил, нам нужно всего лишь задать для неё условие. Например, мы можем запустить линейную регрессию, которая будет зависеть только от propensity score, а не от всех фичей. А пока давайте рассмотрим метод, который использует только оценку склонности и ничего больше. Идея заключается в том, чтобы записать условную разность средних значений с propensity score.

$
E[Y|X,T=1]-E[Y|X,T=0] = E\bigg[\dfrac{Y}{e(x)}|X,T=1\bigg]P(T) - E\bigg[\dfrac{Y}{(1-e(x))}|X,T=0\bigg](1-P(T))
$

Мы можем ещё больше упростить это, но давайте рассмотрим это так, потому что это даёт нам хорошее представление о том, как работает propensity score. Первый член оценивает $Y_1$. Он берёт всех получивших воздействие и масштабирует их по обратной вероятности получения воздействия. Это приводит к тому, что те, у кого очень низкая вероятность получить воздействие, получают высокий вес. Это логично, не так ли? Если у кого-то низкая вероятность получить воздействие, этот человек похож на того, кто его не получил. Однако этот же человек всё-таки получил. Это должно быть интересно. По тем же причинам второй член учитывает не получивших воздействия и придаёт высокий вес тем, кто похож на получивших. Эта оценка называется взвешиванием по обратной вероятности воздействия (IPTW), поскольку она масштабирует каждую единицу по обратной вероятности полученного воздействия.

Теперь, когда у нас есть интуиция, мы можем упростить приведенное выше выражение.

$$
\begin{align}
E[Y|X,T=1]-E[Y|X,T=0] &= E\bigg[\dfrac{Y}{e(x)}|X,T=1\bigg]P(T) - E\bigg[\dfrac{Y}{(1-e(x))}|X,T=0\bigg](1-P(T)) \\
&=E\bigg[\dfrac{YT}{e(x)}\bigg|X\bigg] - E\bigg[\dfrac{Y(1-T)}{(1-e(x))}\bigg|X\bigg] \\
&=E\bigg[\dfrac{YT}{e(x)} - \dfrac{Y(1-T)}{(1-e(x))}\bigg|X\bigg] \\
&=E\bigg[Y\dfrac{T(1-e(x)) - e(x)(1-T)}{e(x)(1-e(x))}\bigg|X\bigg] \\
&=E\bigg[Y \dfrac{T-e(x)}{e(x)(1-e(x))}\bigg|X\bigg]
\end{align}
$$



В идеальном мире у нас была бы истинная оценка $e(x)$. Однако на практике механизм, назначающий воздействие, неизвестен, и нам необходимо заменить истинную оценку её оценкой $\hat{e}(x)$. Один из распространённых способов сделать это — использовать логистическую регрессию, но можно использовать и другие методы машинного обучения, например, градиентный бустинг (хотя это требует некоторых дополнительных шагов для предотвращения переобучения).

Здесь я буду придерживаться логистической регрессии. Это означает, что мне придётся преобразовать категориальные признаки в наборе данных в фиктивные.

````python
categ = ["ethnicity", "gender", "school_urbanicity"]
cont = ["school_mindset", "school_achievement", "school_ethnic_minority", "school_poverty", "school_size"]

data_with_categ = pd.concat([
    data.drop(columns=categ), # dataset without the categorical features
    pd.get_dummies(data[categ], columns=categ, drop_first=False)# categorical features converted to dummies
], axis=1)

from sklearn.linear_model import LogisticRegression

T = 'intervention'
Y = 'achievement_score'
X = data_with_categ.columns.drop(['schoolid', T, Y])

ps_model = LogisticRegression(C=1e6).fit(data_with_categ[X], data_with_categ[T])

data_ps = data.assign(propensity_score=ps_model.predict_proba(data_with_categ[X])[:, 1])

data_ps[["intervention", "achievement_score", "propensity_score"]].head()
````

````python
weight_t = 1/data_ps.query("intervention==1")["propensity_score"]
weight_nt = 1/(1-data_ps.query("intervention==0")["propensity_score"])
print("Original Sample Size", data.shape[0])
print("Treated Population Sample Size", sum(weight_t))
print("Untreated Population Sample Size", sum(weight_nt))
````

Мы также можем использовать propensity score для обнаружения искажений. Если у одной группы населения propensity score выше, чем у другой, это означает, что причиной воздействия является неслучайный фактор. Если тот же фактор также является причиной целевой переменной, то мы имеем искажения. В нашем случае мы видим, что студенты, которые сообщили о своей большей амбициозности, также с большей вероятностью посетят семинар по мышлению роста.

````python
sns.boxplot(x="success_expect", y="propensity_score", data=data_ps)
plt.title("Confounding Evidence");
````

Наконец, мы можем использовать нашу весовую оценку для оценки среднего эффекта лечения.

````python
weight = ((data_ps["intervention"]-data_ps["propensity_score"]) /
          (data_ps["propensity_score"]*(1-data_ps["propensity_score"])))

y1 = sum(data_ps.query("intervention==1")["achievement_score"]*weight_t) / len(data)
y0 = sum(data_ps.query("intervention==0")["achievement_score"]*weight_nt) / len(data)

ate = np.mean(weight * data_ps["achievement_score"])

print("Y1:", y1)
print("Y0:", y0)
print("ATE", ate)
````

Результаты означают, что мы должны ожидать, что достижения участников с назначенным воздействием будут на 0,38 стандартного отклонения выше достижений участников, которым не было назначено. Мы также видим, что если бы никто не получал воздествие, мы должны ожидать, что общий уровень достижений будет на 0,12 стандартного отклонения ниже, чем сейчас. По тем же причинам мы должны ожидать, что общий уровень достижений будет на 0,25 стандартного отклонения выше, если бы мы провели семинар для всех.

#### Matching

Как мы уже видели, регрессия отлично справляется с контролем дополнительных переменных при сравнении тестовой и контрольной групп. Если у нас есть независимость, $(Y_0, Y_1)\perp T | X$, то регрессия может определить ATE, контролируя X. Способ, которым регрессия это делает, можно назвать магическим. Чтобы получить некоторое представление об этом, давайте вспомним случай, когда все переменные X являются фиктивными переменными. В этом случае регрессия разбивает данные на фиктивные ячейки и вычисляет среднюю разность между тестовой и контрольной группами. Эта разность средних сохраняет X постоянными, поскольку мы делаем это в фиксированной ячейке фиктивной группы X. Это как если бы мы делали $E[Y|T=1] - E[Y|T=0] | X=x$, где $x$ является фиктивной ячейкой (например, все фиктивные переменные установлены в 1). Затем регрессия объединяет оценки в каждой из ячеек для получения окончательной ATE. Это делается путем применения к ячейке весов, пропорциональных дисперсии воздействия на эту группу.

Этот результат более нагляден в случае фиктивных переменных, но, как ни странно, регрессия также сохраняет непрерывные переменные постоянными при оценке эффекта. Кроме того, в случае непрерывных переменных ATE укажет направление, где ковариаты имеют большую дисперсию.

Если мы хотим оценить какой-либо причинно-следственный эффект, например, влияние профессионального обучения на заработок, и **воздействие** не назначается случайным образом, нам необходимо учитывать факторы, вмешивающиеся в процесс. Возможно, обучение проходят только более мотивированные люди, и их заработок будет выше независимо от обучения. Нам нужно оценить эффект программы обучения в небольших группах людей с примерно одинаковым уровнем мотивации и любыми другими имеющимися факторами, вмешивающимися в процесс.

В более общем случае, если мы хотим оценить какой-либо причинно-следственный эффект, но это сложно сделать из-за влияния некоторых переменных $X$, нам нужно провести сравнение воздействия и контроля в небольших группах, где $X$ одинаковы. Если у нас есть условная независимость $(Y_0, Y_1)\perp T | X$, то мы можем записать $ATE$ следующим образом.

$
ATE = \int(E[Y|X, T=1] - E[Y|X, T=0])dP(x)
$

Этот интеграл проходит по всему пространству распределения признаков $X$, вычисляет разность средних значений для всех этих небольших пространств и объединяет всё в $ATE$. Другой способ представить это — представить себе дискретный набор признаков. В этом случае можно сказать, что признаки $X$ занимают $K$ различных ячеек $\{X_1, X_2, ..., X_k\}$, и мы вычисляем эффект воздействия в каждой ячейке и объединяем их в $ATE$. В этом дискретном случае, преобразуя интеграл в сумму, мы можем получить оценку подклассификации:

$
\hat{ATE} = \sum^K_{k=1}(\bar{Y}_{k_{1}} - \bar{Y}_{k_{0}}) * \dfrac{N_k}{N}
$

где черта представляет среднее значение результата для обработанной $Y_{k1}$ и необработанной $Y_{k0}$ группы в ячейке $k$, а $N_{k}$ — количество наблюдений в этой же ячейке. Как видите, мы вычисляем локальную $ATE$ для каждой ячейки и объединяем их, используя средневзвешенное значение, где веса соответствуют размеру выборки ячейки.

Оценка подклассификации нечасто используется на практике, но она даёт нам хорошее интуитивное представление о том, что должна делать оценка причинно-следственной связи и как она должна контролировать факторы, препятствующие прохождению. Это позволяет нам исследовать другие виды оценок, такие как мэтчинг.

Идея очень похожа. Поскольку некий фактор $X$ делает так, что экспериментальные и не экспериментальные объекты изначально несопоставимы, можно сделать их таковыми, **сопоставив каждый экспериментальный объект с аналогичным не экспериментальным объектом**. Это как если бы мы находили близнеца, не экспериментального объекта, для каждого экспериментального объекта. Благодаря таким сравнениям, экспериментальные и не экспериментальные объекты снова становятся сопоставимыми.

Общий процесс выглядит так:

1. Определение меры “близости” между объектами;
2. Реализация метода мэтчинга, учитывающего установленную меру "близости";
3. Оценка качества полученного набора "близких" объектов (возможно, повторение шагов 1 и 2 до достижения "хорошего" набора);
4. Оценка разницы между полученными группами.

Существует значительное число различных методов мэтчинга, здесь мы упомянем только некоторые из них. Самым простым является **Exact Matching**, основная идея которого заключается в том, что для каждого объекта из тестовой выборки ищется идентичный объект в контрольной выборке. Если объект нашелся, то расстояние до него равно 0, иначе расстояние равно $\infty$.

Развитием данного метода является **Coarsened Exact Matching**, в нем непрерывные признаки превращаются в категориальные, и вместо поиска идентичного объекта осуществляется поиск похожих объектов в "загрубленном пространстве".

**Mahalanobis Distance Matching** уже непосредственно учитывает расстояние между объектами с помощью расстояния Махаланобиса. Если говорить упрощенно, то в нем учитывается близость к центру масс с учетом разброса:

$$
Distance(X_c, X_t) = \sqrt{(X_c - X_t)^T \times S^{-1} \times(X_c - X_t)}
$$

\\

Но у этих подходов есть фундаментальная проблема - **проклятие размерности**.
Легко найти людей, совпадающих по нескольким характеристикам, например, по полу. Но если добавить больше характеристик, таких как возраст, доход, город рождения и так далее, находить совпадения становится всё сложнее. В более общем смысле, чем больше признаков, тем больше расстояние между единицами и их совпадениями.

Это означает лишь то, что количество точек данных, необходимых для заполнения пространства признаков, растёт экспоненциально с ростом числа признаков, или измерений. Таким образом, если для заполнения пространства, скажем, трёх пространств признаков требуется $X$ точек данных, то для заполнения пространства четырёх признаков потребуется экспоненциально больше точек.

В контексте оценки подклассификации проклятие размерности означает, что она пострадает при наличии большого количества признаков. Большое количество признаков подразумевает наличие нескольких ячеек в $X$. Если ячеек несколько, в некоторых из них будет очень мало данных. Некоторые из них могут содержать только обработанные или только контрольные элементы, поэтому оценить $ATE$ для них будет невозможно, что нарушит нашу оценку. В контексте сопоставления это означает, что пространство признаков будет очень разреженным, а единицы будут находиться очень далеко друг от друга. Это увеличит расстояние между сопоставлениями и вызовет проблемы смещения.

Линейная регрессия проецирует все признаки $X$ в одно измерение $Y$. Затем она сравнивает воздействие и контроль на этой проекции. Таким образом, в некотором смысле линейная регрессия выполняет своего рода снижение размерности для оценки $ATE$. Это довольно элегантно.

Большинство причинно-следственных моделей имеют способы решения проблемы проклятия размерности. Линейная регрессия проецирует все признаки $X$ в одно измерение $Y$. Затем она сравнивает воздействие и контроль на этой проекции. Таким образом, в некотором смысле линейная регрессия выполняет своего рода снижение размерности для оценки $ATE$.

Подобную идею со снижением размерности перед сопоставлением воплощает в себе **Propensity Score Matching (PSM)**, которая сопоставляет элементы не на основе ковариат $X$, а на основе propensity score, который как раз и переводит пространство ковариат в 1-мерное пространство.



#### Difference-in-Difference

**Три рекламных щита на юге Бразилии**

Для маркетологов интернет-реклама привлекательна не только своей высокой эффективностью (хотя это так), но и тем, что её очень легко оценить. С помощью интернет-маркетинга вы можете узнать, какие клиенты видели рекламу, и отследить их с помощью файлов cookie, чтобы узнать, попали ли они на вашу целевую страницу или нажали кнопку загрузки. Вы также можете использовать машинное обучение, чтобы найти потенциальных клиентов, очень похожих на ваших клиентов, и показывать рекламу только им. В этом смысле интернет-маркетинг очень точен: вы таргетируетесь только на тех, кто вам нужен, и видите, реагируют ли они так, как вам хотелось бы.

Но не все восприимчивы к онлайн-маркетингу. Иногда приходится прибегать к менее точным методам, таким как телевизионная кампания или размещение рекламного щита на улице. Обычно отделы маркетинга ищут разнообразие маркетинговых каналов. Но если интернет-маркетинг — это профессиональная удочка для ловли тунца, то рекламный щит и телевидение — это гигантские сети, которые вы закидываете на косяки рыб в надежде поймать хотя бы несколько крупных. Другая проблема рекламы на билбордах и телевидении заключается в том, что гораздо сложнее оценить её эффективность. Конечно, можно измерить объём покупок или что-то ещё, что вы хотите стимулировать, до и после размещения билборда. Если наблюдается рост, это свидетельствует об эффективности маркетинга. Но как узнать, не является ли этот рост просто естественной тенденцией к повышению узнаваемости вашего продукта? Другими словами, как можно узнать контрфактическую разницу в $Y_0$ того, что произошло бы, если бы вы изначально не разместили билборды?

Один из методов, позволяющих ответить на подобные вопросы, — это **Difference-in-Difference**, или diff-in-diff. Diff-in-diff обычно используется для оценки эффекта макроэкономических вмешательств, таких как влияние иммиграции на уровень безработицы, влияние изменений в законах об оружии на уровень преступности или просто разница в вовлечённости пользователей в результате маркетинговой кампании. Во всех этих случаях у вас есть период до и после вмешательства, и вы хотите отделить влияние вмешательства от общей тенденции. В качестве мотивирующего примера рассмотрим следующий вопрос оценки эффективности внешней рекламы.

Чтобы выяснить, насколько эффективны рекламные щиты как маркетинговый канал, мы разместили три билборда в городе Порту-Алегри, столице штата Риу-Гранди-ду-Сул. Мы хотели проверить, увеличит ли это поступление средств на наш сберегательный счет. Для тех, кто не очень хорошо знаком с географией Бразилии, Риу-Гранди-ду-Сул находится на юге страны, в одном из самых развитых регионов.

Имея это в виду, мы решили также рассмотреть данные из другой южной столицы, Флорианополиса, столицы штата Санта-Катарина. Идея заключается в том, чтобы использовать Флорианополис в качестве контрольной выборки для оценки предполагаемого $Y_0$ по сравнению с Порту-Алегри. Мы разместили билборд в Порту-Алегри на весь июнь. Полученные нами данные выглядят следующим образом:

````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from matplotlib import style
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

%matplotlib inline

style.use("fivethirtyeight")

data = pd.read_csv("data/diff_in_diff_df.csv")
data.head()
````

Депозиты — это конечная переменная, которую мы хотим увеличить с помощью рекламных щитов. POA — фиктивный индикатор для города Порту-Алегри. Если он равен нулю, это означает, что образцы взяты из Флорианополиса. Jul — фиктивный индикатор для июля, то есть для периода после вмешательства. Если он равен нулю, это относится к образцам за май, то есть к периоду до вмешательства.


Чтобы избежать путаницы между временем и обработкой, здесь будет использоваться D для обозначения воздействия, а T — для времени. Пусть $Y_D(T)$ — потенциальный результат воздействия D в период T. В идеальном мире, где мы можем наблюдать контрфактическую ситуацию, мы бы оценили эффект следующим образом:

$
\hat{ATET} = E[Y_1(1) - Y_0(1)|D=1]
$

Проще говоря, причинно-следственный эффект — это результат в период после вмешательства в случае лечения за вычетом результата также в период после вмешательства, но в случае отсутствия лечения. Конечно, мы не можем измерить это, поскольку $Y_0(1)$ контрфактуален.

\\

**Difference-in-Difference** работает, заменяя отсутствующее контрфактическое значение следующим образом:

$
E[Y_0(1)|D=1] = E[Y_0(0)|D=1] + (E[Y_0(1)|D=0] - E[Y_0(0)|D=0])
$

При этом берётся обработанная единица **до вмешательства** и добавляется к ней трендовый компонент, который оценивается с использованием контрольной $E[Y_0(1)|D=0] - E[Y_0(0)|D=0]$. Другими словами, это означает, что обработанная единица **после вмешательства**, если бы она не подвергалась обработке, выглядела бы как **обработанная до вмешательства** плюс фактор роста, равный росту контрольной единицы.

Важно отметить, что это предполагает, что тенденции в экспериментальной и контрольной группах одинаковы:

$
E[Y_0(1) − Y_0(0)|D=1] = E[Y_0(1) − Y_0(0)|D=0]
$

Где левая часть — контрфактуальная тенденция. Теперь мы можем заменить предполагаемое контрфактическое значение в определении эффекта воздействия $E[Y_1(1)|D=1] - E[Y_0(1)|D=1]$

$
\hat{ATET} = E[Y(1)|D=1] - (E[Y(0)|D=1] + (E[Y(1)|D=0] - E[Y(0)|D=0])
$

Если переставить члены, то получим классическую оценку diff-in-diff.

$
\hat{ATET} = (E[Y(1)|D=1] - E[Y(1)|D=0]) - (E[Y(0)|D=1] - E[Y(0)|D=0])
$

Она получила такое название, потому что вычисляет разницу между разницей между воздействием и контролем до и после воздействия.

Обратите внимание, что предположение, которое делает метод «разность-в-разнице», следующее: Гораздо более правдоподобен, чем два других метода оценки. Он просто предполагает, что динамика роста в двух городах одинакова. Но он не требует одинакового базового уровня и нулевого тренда.

Чтобы наглядно представить, как работает метод diff-in-diff, мы можем спроецировать тренд роста из необработанной зоны в обрабатываемую зону, чтобы увидеть контрфактическую ситуацию, то есть количество депозитов, которое следует ожидать при отсутствии вмешательства.

image.png

Для оценки diff-diff обычно используют обычную линейную регрессию, ссылаясь на следующую теорему:

image.png

В нашем случае оценка diff-in-diff будет выглядеть как:

$
Y_i = \beta_0 + \beta_1 POA_i + \beta_2 Jul_i + \beta_3 POA_i*Jul_i + e_i
$

Обратите внимание, что $\beta_0$ — это базовый уровень контроля. В нашем случае — это уровень депозитов во Флорианополисе в мае. Если мы включим фиктивную переменную для обработанного города, мы получим $\beta_1$. Таким образом, $\beta_0 + \beta_1$ — это базовый уровень Порту-Алегри в мае, до вмешательства, а $\beta_1$ — это увеличение базового уровня Порту-Алегри по сравнению с Флорианополисом. Если мы отключим фиктивную переменную POA и включим фиктивную переменную июля, мы получим $\beta_0 + \beta_2$, что является уровнем Флорианополиса в июле после периода вмешательства. Тогда $\beta_2$ является тенденцией контроля, так как мы добавляем его к базовой линии, чтобы получить уровень контроля в период после вмешательства. В качестве резюме, $\beta_1$ является приростом, который мы получаем при переходе от контроля к обработанному, $\beta_2$ является приростом, который мы получаем при переходе от периода до периода после вмешательства. Наконец, если мы включим обе фиктивные переменные, мы получим $\beta_3$. $\beta_0 + \beta_1 + \beta_2 + \beta_3$ является уровнем в Порту-Алегри после вмешательства. Таким образом, $\beta_3$ является приростным воздействием при переходе с мая на июль и от Флорианополиса к POA. Другими словами, это оценка разницы в разнице.

````python
print(smf.ols('deposits ~ poa*jul', data=data).fit().summary())
````

Одна из очевидных проблем Diff-in-Diff заключается в невыполнении предположения о параллельности трендов. Если тренд роста в исследуемой группе отличается от тренда в контрольной группе, оценка будет смещенной. Это распространённая проблема с неслучайными данными, когда решение о воздействии на регион основывается на его потенциальной возможности положительного ответа или когда исследование нацелено на регионы с низкими показателями. Возьмём наш пример с маркетингом. Мы решили протестировать рекламные щиты в Порту-Алегри не для того, чтобы проверить их влияние в целом. Причина кроется в низкой эффективности продаж в этом регионе. Возможно, онлайн-маркетинг там неэффективен. В этом случае рост, который мы наблюдаем в Порту-Алегри без рекламного щита, может быть ниже, чем в других городах. Это приведёт к недооценке влияния рекламного щита.

#### Synthetic Control

При анализе diff-in-diff у нас были данные по клиентам из двух разных городов: Порту-Алегри и Флорианополиса. Данные охватывают два разных периода: до и после маркетинговой интервенции в Порту-Алегри, направленной на увеличение депозитов клиентов. Чтобы оценить эффект воздействия, мы провели регрессионный анализ, который дал нам оценку разницы в разнице и её стандартную ошибку.

В этом случае у нас было много выборок, поскольку данные были дезагрегированы. Но что, если у нас есть только агрегированные данные на уровне города? Например, представим, что у нас есть только средний уровень депозитов в обоих городах до и после интервенции.

|город|до|после|
|--|--|--|
|FL|171,64|206,16|
|POA|46,01|87,06|

Мы всё равно сможем вычислить оценку diff-in-diff, предполагая линейные тренды между городами. Однако Флорианополис может быть не так похож на Порту-Алегри, как хотелось бы. Например, Флорианополис известен своими прекрасными пляжами и общительным населением, в то время как Порту-Алегри больше известен своими барбекю и прериями. Проблема здесь в том, что невозможно точно знать, подходит ли контрольная группа.

Чтобы обойти это, мы воспользуемся так называемым [**«важнейшим нововведением в литературе по оценке политики за последние несколько лет»**](https://www.aeaweb.org/articles?id=10.1257/jep.31.2.3) синтетическим контролем. Он основаны на простой, но эффективной идее. Нам не нужно искать ни одного элемента в исходном наборе данных, очень похожего на воздействованного. Вместо этого мы можем создать свой собственный элемент, объединив несколько неиспользованных элементов, фактически создав синтетический контроль. Синтетический контроль настолько эффективен и в то же время настолько интуитивен, что о нем даже была опубликована статья, правда, не в научном журнале, а в [Washington Post](https://www.washingtonpost.com/news/wonk/wp/2015/10/30/how-to-measure-things-in-a-world-of-competing-claims/).

Чтобы увидеть это в действии, рассмотрим задачу оценки влияния налогообложения сигарет на их потребление. Для справки: этот вопрос долгое время обсуждался в экономике. Одна сторона утверждает, что налоги увеличат стоимость сигарет, что снизит спрос на них. Другая сторона утверждает, что, поскольку сигареты вызывают зависимость, изменение их цены не сильно повлияет на спрос. С экономической точки зрения, можно сказать, что спрос на сигареты неэластичен по цене, а повышение налогообложения — это лишь способ увеличить доходы государства за счёт курильщиков. Чтобы разобраться в этом, рассмотрим некоторые данные по США по этому вопросу.

Доходы, полученные в результате реализации закона, были направлены на различные программы в области охраны окружающей среды и здравоохранения, а также на антитабачную рекламу.

Чтобы оценить его эффективность, мы можем собрать данные о продажах сигарет в нескольких штатах за несколько лет. В нашем случае мы получили данные за период с 1970 по 2000 год из 39 штатов. Чтобы ответить на вопрос, повлияла ли Предложение 99 на потребление сигарет, мы используем период до вмешательства для построения синтетического контрольного значения. Мы объединим другие штаты, чтобы **создать фиктивный штат, очень близко соответствующий тенденции в Калифорнии**. Затем мы посмотрим, как этот синтетический контроль поведет себя после вмешательства.

Для большей формальности предположим, что у нас есть $J+1$ единиц. Без потери общности предположим, что единица 1 — это единица, на которую повлияло вмешательство. Единицы $j=2,...,J+1$ представляют собой совокупность единиц, не подвергавшихся воздействию, которую мы будем называть «донорским пулом». Также предположим, что имеющиеся данные охватывают T периодов времени, с $T_0$ периодов до воздействия. Для каждой единицы j и каждого момента времени t мы наблюдаем результат $Y_{jt}$. Для каждой единицы j и каждого периода t определим $Y^N_{jt}$ как потенциальный результат без воздействия, а $Y^I_{jt}$ — как потенциальный результат с воздействием. Тогда эффект для единицы, подвергшейся воздействию $j=1$, в момент времени t, при $t>T_0$, определяется как

$
\tau_{1t} = Y^I_{1t} - Y^N_{1t}
$

Поскольку единица $j=1$ подверглась лечению, $Y^I_{1t}$ является фактическим, а $Y^N_{1t}$ — нет. В таком случае возникает проблема: как оценить $Y^N_{1t}$? Обратите внимание, как эффект воздействия определяется для каждого периода, то есть он может меняться со временем. Он не обязательно должен быть мгновенным. Он может накапливаться или рассеиваться. Проще говоря, задача оценки эффекта воздействия сводится к **оценке того, что произошло бы с результатом воздействия единицы $j=1$, если бы она не подвергалась лечению**.

image.png

Чтобы оценить $Y^N_{1t}$, следует помнить, что комбинация единиц в донорском пуле может гораздо лучше аппроксимировать характеристики воздействованной единицы, чем любая единица в отдельности. Таким образом, синтетический контроль определяется как средневзвешенное значение единиц в контрольном пуле. Учитывая веса $\pmb{W}=(w_2, ..., w_{J+1})$, оценка синтетического контроля $Y^N_{1t}$ равна

$
\hat{Y}^N_{1t} = \sum^{J+1}_{j=2} w_j Y_{jt}
$

Итак, синтетический контроль можно рассматривать как линейную регрессию, а это также означает, что мы можем оценить его веса с помощью МНК.

Чтобы оценить эффект воздействия с помощью синтетического контроля, мы попытаемся создать «фиктивный блок», напоминающий блок, подвергавшийся воздействию, до периода вмешательства. Затем мы посмотрим, как этот «фиктивный блок» поведет себя после вмешательства. Разница между синтетическим контролем и блоком, который он имитирует, и есть эффект воздействия.

Чтобы сделать это с помощью линейной регрессии, мы найдем вес с помощью МНК. Мы минимизируем квадратное расстояние между средневзвешенным значением единиц в пуле доноров и единицей, подвергшейся воздействию, за период до вмешательства.

Для этого первым делом нам необходимо преобразовать единицы (в нашем случае штаты) в столбцы, а время — в строки. Мы создадим синтетический контроль, очень похожий на Калифорнию в период до вмешательства, и посмотрим, как он будет вести себя в период после вмешательства. По этой причине важно выбрать только период до вмешательства. Здесь признаки, по-видимому, имеют схожий масштаб, поэтому мы не будем с ними ничего делать. Если признаки имеют разные масштабы, один в тысячах, а другой в десятичных долях, больший признак будет наиболее важным при минимизации разницы. Чтобы избежать этого, важно сначала масштабировать их.

````python
features = ["cigsale", "retprice"]

inverted = (cigar.query("~after_treatment") # filter pre-intervention period
            .pivot(index='state', columns="year")[features] # make one column per year and one row per state
            .T) # flip the table to have one column per state

inverted.head()
````

Теперь мы можем определить нашу переменную $Y$ как штат Калифорния, а $X$ — как остальные штаты.

````python
y = inverted[3].values # state of california
X = inverted.drop(columns=3).values  # other states
````

Затем мы запускаем регрессию. Наличие точки пересечения эквивалентно добавлению ещё одного состояния, где каждая строка равна 1. Можно сделать и так, но, думаю, это сложнее, поэтому я просто опущу это. Регрессия вернёт набор весов, минимизирующих квадратичную разность между обработанной единицей и единицами в пуле доноров.

````python
from sklearn.linear_model import LinearRegression
weights_lr = LinearRegression(fit_intercept=False).fit(X, y).coef_
weights_lr.round(3)
````

Эти веса показывают нам, как построить синтетический контроль. Мы умножим результат состояния 1 на -0,436, состояния 2 на -1,038, состояния 4 на 0,679 и так далее. Этого можно добиться с помощью скалярного произведения матрицы состояний в пуле и весов.

````python
calif_synth_lr = (cigar.query("~california")
                  .pivot(index='year', columns="state")["cigsale"]
                  .values.dot(weights_lr))
````

Что-то не так. Что привлекает ваше внимание на этой картинке? Во-первых, после вмешательства продажи сигарет в группе синтетического контроля выше, чем в Калифорнии. Это свидетельствует о том, что вмешательство успешно снизило спрос на сигареты. Во-вторых, обратите внимание, насколько точно подобран период до вмешательства. Синтетический контроль точно соответствует штату Калифорния. Это признак того, что наша модель синтетического контроля, вероятно, переобучается. Ещё одним признаком является огромная дисперсия переменной результата синтетического контроля после вмешательства. Обратите внимание, что она не следует плавным закономерностям. Вместо этого она колеблется вверх и вниз.

Если задуматься, почему это происходит, вспомните, что в нашем пуле доноров 38 штатов. Таким образом, в нашей линейной регрессии есть 38 параметров, с которыми можно экспериментировать, чтобы максимально приблизить пул до вмешательства к пулу вмешательства. Это тот случай, когда даже при большом значении $T$ значение $N$ также велико, что даёт слишком большую гибкость нашей модели линейной регрессии. Если вы знакомы с регуляризованными моделями, знайте, что для решения этой проблемы можно использовать гребневую или лассо-регрессию. Здесь мы рассмотрим другой, более традиционный способ избежать переобучения — ограничить наш синтетический контроль только интерполяцией. Для этого мы ограничим веса положительными и суммируем их до единицы. Теперь синтетический контроль будет представлять собой выпуклую комбинацию единиц в пуле доноров. Это общая идея, давайте немного её формализуем. Синтетический контроль по-прежнему определяется как

$
\hat{Y}^N_{jt} = \sum^{J+1}_{j=2} w_j Y_{jt}
$

но теперь мы будем использовать веса $\pmb{W}=(w_2, ..., w_{J+1})$, которые минимизируют

$
||\pmb{X}_1 - \pmb{X}_0 \pmb{W}|| = \bigg(\sum^k_{h=1}v_h \bigg(X_{h1} - \sum^{J+1}_{j=2} w_j X_{hj} \bigg)^2 \bigg)^{\frac{1}{2}}
$

при условии, что $w_2, ..., w_{J+1}$ положительны и в сумме дают единицу. Обратите внимание, что $v_h$ отражает важность каждой переменной при минимизации разницы между тестовой и синтетической контрольной выборками. Различные $v$ дадут разные оптимальные веса. Один из способов выбора $V$ — сделать так, чтобы каждая переменная имела нулевое среднее и единичную дисперсию. Более сложный способ — выбрать $V$ таким образом, чтобы переменные, помогающие лучше предсказать $Y$, получили большую важность. Поскольку мы хотим упростить код, мы просто зададим одинаковое значение важности для каждой переменной.

Для реализации этого сначала определим указанную выше функцию потерь.

````python
from typing import List
from operator import add
from toolz import reduce, partial

def loss_w(W, X, y) -> float:
    return np.sqrt(np.mean((y - X.dot(W))**2))
````

Поскольку мы используем одинаковую важность для каждого признака, нам не нужно беспокоиться о $v$.

Теперь, чтобы получить оптимальные веса, мы воспользуемся оптимизацией квадратичного программирования SciPy. Мы ограничим весовые коэффициенты суммой 1 с помощью

```python
lambda x: np.sum(x) - 1
```

Кроме того, мы установим границы оптимизации в диапазоне от 0 до 1.

````python
from scipy.optimize import fmin_slsqp

def get_w(X, y):

    w_start = [1/X.shape[1]]*X.shape[1]

    weights = fmin_slsqp(partial(loss_w, X=X, y=y),
                         np.array(w_start),
                         f_eqcons=lambda x: np.sum(x) - 1,
                         bounds=[(0.0, 1.0)]*len(w_start),
                         disp=False)
    return weights

calif_weights = get_w(X, y)
print("Sum:", calif_weights.sum())
np.round(calif_weights, 4)
````

````python
calif_synth = cigar.query("~california").pivot(index='year', columns="state")["cigsale"].values.dot(calif_weights)
````

Если мы сейчас построим график результатов синтетического контроля, мы увидим гораздо более сглаженный тренд. Также обратите внимание, что в период до вмешательства синтетический контроль больше не воспроизводит в точности результаты, полученные в результате воздействия. Это хороший знак, поскольку он указывает на отсутствие переобучения.

````python
plt.figure(figsize=(10,6))
plt.plot(cigar.query("california")["year"], cigar.query("california")["cigsale"], label="California")
plt.plot(cigar.query("california")["year"], calif_synth, label="Synthetic Control")
plt.vlines(x=1988, ymin=40, ymax=140, linestyle=":", lw=2, label="Proposition 99")
plt.ylabel("Per-capita cigarette sales (in packs)")
plt.legend();
````

Имея в распоряжении синтетический контроль, мы можем оценить эффект воздействия как разницу между результатами:

$
\tau_{1t} = Y^I_{jt} - Y^N_{jt}
$

В нашем конкретном случае эффект становится всё сильнее с течением времени.

````python
plt.figure(figsize=(10,6))
plt.plot(cigar.query("california")["year"], cigar.query("california")["cigsale"] - calif_synth,
         label="California Effect")
plt.vlines(x=1988, ymin=-30, ymax=7, linestyle=":", lw=2, label="Proposition 99")
plt.hlines(y=0, xmin=1970, xmax=2000, lw=2)
plt.title("State - Synthetic Across Time")
plt.ylabel("Gap in per-capita cigarette sales (in packs)")
plt.legend();
````

К 2000 году, похоже, Предложение 99 снизило продажи сигарет на 25 пачек. Это, конечно, здорово, но вы, возможно, задаетесь вопросом: как узнать, статистически ли это?

Поскольку размер нашей выборки очень мал (39), нам придётся действовать более разумно, чтобы определить, является ли наш результат статистически значимым, а не просто случайным везением. Здесь мы воспользуемся идеей точного критерия Фишера. Его принцип очень прост. Мы переставляем местами экспериментальную и контрольную единицы. Поскольку у нас есть только одна экспериментальная единица, это означает, что для каждой единицы мы делаем вид, что она экспериментальная, а остальные — контрольные.

|итерация|1|2|...|39|
|----|-|-|-|-|
|1|treated|0|0|0|
|2|0|treated|0|0|
|...|0|0|0|0|0|0|
|39|0|0|0|treated|

В итоге у нас будет один синтетический контроль и оценки эффекта для каждого штата. В результате мы имитируем, что воздействие действительно проводилось в другом штате, не в Калифорнии, и смотрим, каким был бы предполагаемый эффект для этого лечения, которое не проводилось. Затем мы проверяем, достаточно ли выражен эффект в Калифорнии по сравнению с другим фиктивным воздействием. Идея заключается в том, что для штатов, которые фактически не подвергались воздействию, как только мы притворимся, что они были, мы не сможем обнаружить никакого значимого эффекта.

Для реализации этой идеи создана функция, которая принимает в качестве входных данных штат и оценивает синтетический контроль для этого штата. Эта функция возвращает фрейм данных с одним столбцом для штата, одним для года, одним для результата `cigsale` и синтетическим результатом для этого штата.

````python
def synthetic_control(state: int, data: pd.DataFrame) -> np.array:

    features = ["cigsale", "retprice"]

    inverted = (data.query("~after_treatment")
                .pivot(index='state', columns="year")[features]
                .T)

    y = inverted[state].values # treated
    X = inverted.drop(columns=state).values # donor pool

    weights = get_w(X, y)
    synthetic = (data.query(f"~(state=={state})")
                 .pivot(index='year', columns="state")["cigsale"]
                 .values.dot(weights))

    return (data
            .query(f"state=={state}")[["state", "year", "cigsale", "after_treatment"]]
            .assign(synthetic=synthetic))
````

````python
synthetic_control(1, cigar).head()
````

````python
# Чтобы получить результат для всех состояний, мы распараллеливаем вычисления на 8 процессов.
from joblib import Parallel, delayed

control_pool = cigar["state"].unique()

parallel_fn = delayed(partial(synthetic_control, data=cigar))

synthetic_states = Parallel(n_jobs=8)(parallel_fn(state) for state in control_pool)

synthetic_states[0].head()
````

Используя синтетический контроль для всех штатов, мы можем оценить разницу между синтетическим и истинным состоянием для всех штатов. Для Калифорнии это эффект воздействия. Для других штатов это похоже на эффект плацебо, где мы оцениваем эффект лечения синтетического контроля, хотя лечение фактически не проводилось. Если мы нанесем на график все эффекты плацебо вместе с эффектом воздействия в Калифорнии, мы получим следующий рисунок.

````python
plt.figure(figsize=(12,7))
for state in synthetic_states:
    plt.plot(state["year"], state["cigsale"] - state["synthetic"], color="C5",alpha=0.4)

plt.plot(cigar.query("california")["year"], cigar.query("california")["cigsale"] - calif_synth,
        label="California");

plt.vlines(x=1988, ymin=-50, ymax=120, linestyle=":", lw=2, label="Proposition 99")
plt.hlines(y=0, xmin=1970, xmax=2000, lw=3)
plt.ylabel("Gap in per-capita cigarette sales (in packs)")
plt.title("State - Synthetic Across Time")
plt.legend();
````

В глаза бросаются два аспекта этого рисунка. Во-первых, мы видим, что дисперсия после вмешательства выше, чем до вмешательства. Это ожидаемо, поскольку синтетический контроль призван минимизировать разницу в период до вмешательства. Другой интересный аспект заключается в том, что есть некоторые единицы, которые мы не можем хорошо подогнать даже в период до вмешательства. Это также ожидаемо. Например, если в некоторых штатах очень высокое потребление сигарет, никакая выпуклая комбинация других штатов никогда не будет им соответствовать.

Поскольку эти единицы так плохо подходят, рекомендуется исключить их из анализа. Один из способов сделать это объективно — установить пороговое значение для ошибки до вмешательства:

$
MSE = \frac{1}{N}\sum\bigg(Y_t - \hat{Y}^{Synth}_t\bigg)^2
$

и удалить единицы с высокой ошибкой. Если мы продолжим в том же духе и построим тот же график, то получим следующее.

````python
def pre_treatment_error(state):
    pre_treat_error = (state.query("~after_treatment")["cigsale"]
                       - state.query("~after_treatment")["synthetic"]) ** 2
    return pre_treat_error.mean()

plt.figure(figsize=(12,7))
for state in synthetic_states:

    # remove units with mean error above 80.
    if pre_treatment_error(state) < 80:
        plt.plot(state["year"], state["cigsale"] - state["synthetic"], color="C5",alpha=0.4)

plt.plot(cigar.query("california")["year"], cigar.query("california")["cigsale"] - calif_synth,
        label="California");

plt.vlines(x=1988, ymin=-50, ymax=120, linestyle=":", lw=2, label="Proposition 99")
plt.hlines(y=0, xmin=1970, xmax=2000, lw=3)
plt.ylabel("Gap in per-capita cigarette sales (in packs)")
plt.title("Distribution of Effects")
plt.title("State - Synthetic Across Time (Large Pre-Treatment Errors Removed)")
plt.legend();
````

Удалив шум, мы можем увидеть, насколько экстремальным является эффект в штате Калифорния. Это изображение показывает, что если бы мы предположили, что воздействие было оказано в любом другом штате, мы почти никогда не получили бы столь экстремального эффекта, как тот, который мы получили в Калифорнии.

Это изображение само по себе является формой вывода, но мы также можем вывести из этих результатов P-значение. Всё, что нам нужно сделать, — это посмотреть, во сколько раз полученный эффект ниже эффекта Калифорнии.

````python
calif_number = 3

effects = [state.query("year==2000").iloc[0]["cigsale"] - state.query("year==2000").iloc[0]["synthetic"]
           for state in synthetic_states
           if pre_treatment_error(state) < 80] # filter out noise

calif_effect = cigar.query("california & year==2000").iloc[0]["cigsale"] - calif_synth[-1]

print("California Treatment Effect for the Year 2000:", calif_effect)
np.array(effects)
````

Если мы хотим проверить одностороннюю гипотезу о том, что эффект в Калифорнии ниже нуля, мы можем оценить P-значение как долю раз, в которую эффект в Калифорнии превышает все предполагаемые эффекты.

$
PV=\frac{1}{N}\sum \mathcal{1}\{\hat{\tau}_{Calif} > \hat{\tau}_j\}
$

Как оказалось, эффект лечения в Калифорнии в 2000 году составил -24,8, что означает, что вмешательство снизило потребление сигарет почти на 25 пачек. Из всех 34 рассчитанных нами эффектов плацебо только один превышает эффект, обнаруженный в Калифорнии. Таким образом, P-значение составит 1/35.

````python
np.mean(np.array(effects) < calif_effect)
````

Такие методы, как propensity score, линейная регрессия и мэтчинг, очень хорошо подходят для контроля вмешивающихся факторов в неслучайных данных, но они основаны на ключевом предположении: условной несмешиваемости.

(Y_0, Y_1) \perp T | X
$

Они требуют, чтобы все вмешивающиеся факторы были известны и измерены, чтобы мы могли обусловить их и сделать обработку максимально приближенной к случайной. Одна из основных проблем заключается в том, что иногда мы просто не можем измерить вмешивающийся фактор. Например, возьмем классическую задачу экономики труда: определить влияние брака на заработок мужчин. В экономике хорошо известно, что женатые мужчины зарабатывают больше холостых. Однако неясно, является ли эта связь причинно-следственной или нет. Возможно, более образованные мужчины чаще вступают в брак и чаще имеют высокооплачиваемую работу, что означает, что образование является фактором, искажающим влияние брака на заработок. Для этого конфаундера мы могли бы измерить образование человека в исследовании и провести регрессионный анализ, контролирующий это. Но другим конфаундером может быть красота. Возможно, более красивые мужчины чаще вступают в брак и чаще имеют высокооплачиваемую работу. К сожалению, красота — одна из таких характеристик, как интеллект. Её мы не можем измерить достаточно точно.

Это ставит нас в сложную ситуацию, поскольку при наличии неизмеряемых конфаундеров возникает смещение. Один из способов решения этой проблемы — использование инструментальных переменных,но создание хороших инструментов — непростая задача, требующая большого творческого подхода. Вместо этого давайте воспользуемся структурой наших панельных данных.

Мы уже видели, как панельные данные позволяют нам заменить предположение о неизмеряемости конфаундеров предположением о параллельных тенденциях. Но как именно это помогает в случае неизмеряемых конфаундеров? Сначала давайте взглянем на причинно-следственный график, представляющий данную ситуацию, где у нас есть повторяющиеся наблюдения во времени. Здесь мы отслеживаем одно и то же наблюдение в течение 4 временных периодов. Брак (воздействие) и доход (результат) меняются с течением времени. В частности, брак активируется (с 0 до 1) в периоды 3 и 4, а доход увеличивается в те же периоды. Красота, неизмеряемый фактор, одинаков во всех периодах (смелое заявление, но разумное, если период составляет всего несколько лет). Итак, как мы можем знать, что причина увеличения дохода связана с браком, а не просто с увеличением фактора красоты? И, что ещё важнее, как мы можем контролировать этот невидимый фактор?

image.png

Фокус в том, чтобы увидеть, что, увеличивая единицу измерения и отслеживая её динамику с течением времени, мы уже контролируем всё, что фиксировано во времени. Это включает любые фиксированные во времени неизмеряемые факторы. Например, на графике выше мы уже знаем, что рост дохода с течением времени не может быть связан с ростом красоты, просто потому что красота остаётся неизменной (в конце концов, она фиксирована во времени). Суть в том, что, хотя мы не можем контролировать красоту, поскольку не можем её измерить, мы всё равно можем использовать панельную структуру, так что это больше не проблема.

Другой способ представить это — рассматривать эти фиксированные во времени конфаундеры как атрибуты, специфичные для каждой единицы. Это было бы эквивалентно добавлению промежуточного узла единицы в наш причинно-следственный граф. Теперь обратите внимание, как контроль самой единицы уже блокирует backdoor между результатом и любыми ненаблюдаемыми, но фиксированными во времени конфаундерами.

image.png

Подумайте об этом. Мы не можем измерить такие атрибуты, как красота и интеллект, но мы знаем, что человек, обладающий ими, — это один и тот же человек во времени. Механизм такого контроля очень прост. Всё, что нам нужно сделать, — это создать фиктивные переменные, указывающие на этого человека, и добавить их в линейную модель. Именно это мы имеем в виду, когда говорим, что можем контролировать самого человека: мы добавляем переменную (в данном случае фиктивную), которая обозначает этого конкретного человека. При оценке влияния брака на доход с использованием этой фиктивной переменной в нашей модели регрессия обнаруживает влияние брака **при сохранении переменной, соответствующей человеку, фиксированной**. Добавление этой единичной фиктивной переменной и есть то, что мы называем моделью с фиксированными эффектами.

**Фиксированные эффекты**

Чтобы сделать ситуацию более формальной, давайте сначала рассмотрим имеющиеся у нас данные. Следуя нашему примеру, попробуем оценить влияние брака на доход. Наши данные содержат две переменные, `married` и `lwage`, для нескольких человек (`nr`) за несколько лет. Обратите внимание, что заработная плата представлена в логарифмическом формате. Кроме того, у нас есть и другие контрольные параметры, такие как количество отработанных часов в этом году, количество лет обучения и так далее.

````python
data = wage_panel.load()
data.head()
````

В общем случае модель с фиксированными эффектами определяется как

$
y_{it} = \beta X_{it} + \gamma U_i + e_{it}
$

где $y_{it}$ — результат индивида $i$ в момент времени $t$, $X_{it}$ — вектор переменных для индивида $i$ в момент времени $t$. $U_i$ — набор ненаблюдаемых переменных для индивида $i$. Обратите внимание, что эти ненаблюдаемые переменные не меняются с течением времени, поэтому индекс времени отсутствует. Наконец, $e_{it}$ — ошибка. В примере с образованием $y_{it}$ — логарифм заработной платы, $X_{it}$ — наблюдаемые переменные, которые изменяются со временем, такие как брак и опыт, а $U_i$ — переменные, которые не наблюдаются, но постоянны для каждого индивида, такие как красота и интеллект.

Итак, помните, как я говорил, что использование панельных данных с моделью с фиксированными эффектами так же просто, как добавление фиктивной переменной для сущностей? Это правда, но на практике мы этого не делаем. Представьте себе набор данных с 1 миллионом клиентов. Если мы добавим по одной фиктивной переменной для каждого из них, у нас получится 1 миллион столбцов, что, вероятно, не очень хорошая идея. Вместо этого мы используем приём разбиения линейной регрессии на две отдельные модели. Предположим, у вас есть модель линейной регрессии с набором признаков $X_1$ и другим набором признаков $X_2$.

$
\hat{Y} = \hat{\beta_1} X_1 + \hat{\beta_2} X_2
$

Вы можете получить тот же параметр $\hat{\beta_1}$, выполнив следующие действия:

1. провести регрессию результата $y$ по второму набору признаков $\hat{y^*} = \hat{\gamma_1} X_2$
2. провести регрессию первого набора признаков по второму $\hat{X_1} = \hat{\gamma_2} X_2$
3. получить остатки $\tilde{X}_1 = X_1 - \hat{X_1}$ и $\tilde{y}_1 = y_1 - \hat{y^*}$
4. провести регрессию остатков результата по остаткам признаков $\hat{y} = \hat{\beta_1} \tilde{X}_1$

Параметр этой последней регрессии будет точно таким же, как и при регрессии со всеми признаками. Но как именно это нам поможет? Итак, мы можем разбить оценку модели с фиктивными переменными сущностей на два этапа. Сначала мы используем фиктивные переменные для прогнозирования результата и признака. Это шаги 1 и 2, описанные выше.

Итак, помните, что построение регрессии с фиктивной переменной так же просто, как оценка среднего значения для этой фиктивной переменной? Если вы не помните, давайте воспользуемся нашими данными, чтобы показать, что это верно. Давайте запустим модель, в которой мы прогнозируем заработную плату как функцию фиктивной переменной года.

````python
mod = smf.ols("lwage ~ C(year)", data=data).fit()
print(mod.summary())
````

Обратите внимание, как эта модель предсказывает средний доход в 1980 году в размере 1.3935, в 1981 году — 1.5129 (1.3935 + 0.1194) и так далее. Теперь, если мы вычислим средний доход по годам, мы получим точно такой же результат.

````python
data.groupby("year")["lwage"].mean()
````

Это означает, что если мы получим среднее значение для каждого человека в нашей группе, мы, по сути, регрессируем индивидуальную фиктивную переменную по другим переменным. Это обосновывает следующую процедуру оценки:

1. Создайте переменные с уменьшенным во времени средним значением, вычитая среднее значение для отдельного человека:
$\ddot{Y}_{it} = Y_{it} - \bar{Y}_i$
$\ddot{X}_{it} = X_{it} - \bar{X}_i$

2. Регрессия $\ddot{Y}_{it}$ по $\ddot{X}_{it}$

Обратите внимание, что при этом ненаблюдаемая переменная $U_i$ исчезает. Поскольку $U_i$ постоянна во времени, получаем $\bar{U_i}=U_i$. Если у нас есть следующая система из двух уравнений:

$$
\begin{align}
Y_{it} & = \beta X_{it} + \gamma U_i + e_{it} \\
\bar{Y}_{i} & = \beta \bar{X}_{it} + \gamma \bar{U}_i + \bar{e}_{it} \\
\end{align}
$$

И вычитаем одно из другого, получаем:

$$
\begin{align}
(Y_{it} - \bar{Y}_{i}) & = (\beta X_{it} - \beta \bar{X}_{it}) + (\gamma U_i - \gamma U_i) + (e_{it} - \bar{e}_{it}) \\
(Y_{it} - \bar{Y}_{i}) & = \beta(X_{it} - \bar{X}_{it}) + (e_{it}-\bar{e}_{it}) \\
\ddot{Y}_{it} & = \beta \ddot{X}_{it} + \ddot{e}_{it} \\
\end{align}
$$

что стирает все ненаблюдаемые переменные, которые постоянны во времени. Честно говоря, исчезают не только ненаблюдаемые переменные. Это происходит со всеми переменными, которые постоянны во времени. По этой причине нельзя включать переменные, которые постоянны во времени, поскольку они будут линейной комбинацией фиктивных переменных, и модель не будет работать.

Чтобы проверить, какие это переменные, мы можем сгруппировать наши данные по отдельным особям и получить сумму стандартных отклонений. Если она равна нулю, это означает, что переменная не меняется во времени ни для одной из особей.

````python
data.groupby("nr").std().sum()
````

Из наших данных нам необходимо удалить фиктивные переменные этнической принадлежности, «черный» и «хисип», поскольку они постоянны для каждого отдельного человека. Также необходимо удалить образование. Мы также не будем использовать профессию, поскольку она, вероятно, опосредует влияние брака на заработную плату (возможно, одинокие мужчины могут занимать более длительные должности). Выбрав необходимые признаки, пора оценить эту модель.

Для запуска нашей модели с фиксированными эффектами сначала получим среднее значение данных. Этого можно добиться, сгруппировав все данные по отдельным людям и вычислив среднее значение.

````python
Y = "lwage"
T = "married"
X = [T, "expersq", "union", "hours"]

mean_data = data.groupby("nr")[X+[Y]].mean()
mean_data.head()
````

Чтобы уменьшить данные, нам нужно установить индекс исходных данных равным индивидуальному идентификатору `nr`. Затем мы можем просто вычесть один фрейм данных из среднего фрейма данных.

````python
demeaned_data = (data
               .set_index("nr") # set the index as the person indicator
               [X+[Y]]
               - mean_data) # subtract the mean data

demeaned_data.head()
````

````python
mod = smf.ols(f"{Y} ~ {'+'.join(X)}", data=demeaned_data).fit()
print(mod.summary())
````

Если мы считаем, что фиксированный эффект устраняет смещение, вызванное всеми пропущенными переменными, то эта модель показывает, что брак увеличивает заработную плату мужчины на 11%. Этот результат весьма показателен. Важно отметить, что для моделей с фиксированным эффектом стандартные ошибки необходимо кластеризовать. Поэтому вместо того, чтобы выполнять все оценки вручную (что удобно только в педагогических целях), мы можем использовать библиотеку `linearmodels` и установить аргумент `cluster_entity` в значение True.

````python
from linearmodels.panel import PanelOLS
mod = PanelOLS.from_formula("lwage ~ expersq+union+married+hours+EntityEffects",
                            data=data.set_index(["nr", "year"]))

result = mod.fit(cov_type='clustered', cluster_entity=True)
print(result)
````

**Визуализация фиксированных эффектов**

Чтобы расширить наше понимание работы моделей с фиксированными эффектами, давайте немного отвлечёмся и рассмотрим другой пример. Предположим, вы работаете в крупной технологической компании и хотите оценить влияние маркетинговой кампании на билбордах на покупки в приложении. Если посмотреть на данные за прошлые годы, то можно увидеть, что отдел маркетинга, как правило, тратит больше на размещение билбордов в городах с низким уровнем покупок. Логично, правда? Им не пришлось бы тратить много рекламы, если бы продажи стремительно росли. Если запустить регрессионную модель на основе этих данных, то окажется, что более высокие затраты на маркетинг приводят к снижению количества покупок в приложении, но только потому, что маркетинговые инвестиции смещены в сторону регионов с низким уровнем расходов.

````python
toy_panel = pd.DataFrame({
    "mkt_costs":[5,4,3.5,3, 10,9.5,9,8, 4,3,2,1, 8,7,6,4],
    "purchase":[12,9,7.5,7, 9,7,6.5,5, 15,14.5,14,13, 11,9.5,8,5],
    "city":["C0","C0","C0","C0", "C2","C2","C2","C2", "C1","C1","C1","C1", "C3","C3","C3","C3"]
})

m = smf.ols("purchase ~ mkt_costs", data=toy_panel).fit()

plt.scatter(toy_panel.mkt_costs, toy_panel.purchase)
plt.plot(toy_panel.mkt_costs, m.fittedvalues, c="C5", label="Regression Line")
plt.xlabel("Marketing Costs (in 1000)")
plt.ylabel("In-app Purchase (in 1000)")
plt.title("Simple OLS Model")
plt.legend();
````

Хорошо разбираясь в причинно-следственных связях, вы решаете запустить модель с фиксированными эффектами, добавляя показатель города в качестве фиктивной переменной. Модель с фиксированными эффектами учитывает специфические для города характеристики, которые не изменяются во времени, поэтому, если город менее открыт для вашего продукта, это будет учтено. Запустив эту модель, вы наконец увидите, что увеличение затрат на маркетинг приводит к увеличению количества покупок в приложении.

````python
fe = smf.ols("purchase ~ mkt_costs + C(city)", data=toy_panel).fit()

fe_toy = toy_panel.assign(y_hat = fe.fittedvalues)

plt.scatter(toy_panel.mkt_costs, toy_panel.purchase, c=toy_panel.city)
for city in fe_toy["city"].unique():
    plot_df = fe_toy.query(f"city=='{city}'")
    plt.plot(plot_df.mkt_costs, plot_df.y_hat, c="C5")

plt.title("Fixed Effect Model")
plt.xlabel("Marketing Costs (in 1000)")
plt.ylabel("In-app Purchase (in 1000)");
````

Уделите минуту, чтобы оценить, что показывает изображение выше о том, как работает фиксированный эффект. Обратите внимание, что фиксированный эффект соответствует **одной линии регрессии на каждый город**. Также обратите внимание, что линии параллельны. Наклон линии отражает влияние маркетинговых расходов на покупки в приложении. Таким образом, **фиксированный эффект предполагает, что причинно-следственная связь постоянна для всех сущностей**, которыми в данном случае являются города. Это может быть как недостатком, так и преимуществом, в зависимости от вашей точки зрения. Это недостаток, если вам интересно найти причинно-следственную связь для каждого города. Поскольку модель конечных элементов предполагает, что эта связь постоянна для всех сущностей, вы не обнаружите никакой разницы в причинно-следственной связи. Однако, если вы хотите определить общее влияние маркетинга на покупки в приложении, панельная структура данных — очень полезный инструмент для исследования фиксированных эффектов.

**Временные эффекты**

Подобно тому, как мы разработали фиксированный эффект для индивидуального уровня, мы можем разработать фиксированный эффект для временного уровня. Если добавление фиктивной переменной для каждого индивидуума контролирует фиксированные индивидуальные характеристики, то добавление временной фиктивной переменной будет контролировать переменные, которые фиксированы для каждого периода времени, но могут меняться со временем. Одним из примеров такой переменной является инфляция. Цены и зарплата, как правило, растут со временем, но инфляция в каждый период времени одинакова для всех сущностей. Чтобы привести более конкретный пример, предположим, что количество браков увеличивается со временем. Если соотношение заработной платы и брака также меняется со временем, у нас будет время в качестве сбивающего фактора. Поскольку инфляция также влияет на рост заработной платы со временем, часть положительной связи, которую мы видим между браком и заработной платой, будет просто потому, что оба они растут со временем. Чтобы скорректировать это, мы можем добавить фиктивную переменную для каждого периода времени. В «линейных моделях» это так же просто, как добавить «TimeEffects» в нашу формулу.

````python
mod = PanelOLS.from_formula("lwage ~ expersq+union+married+hours+EntityEffects+TimeEffects",
                            data=data.set_index(["nr", "year"]))

result = mod.fit(cov_type='clustered', cluster_entity=True, cluster_time=True)
print(result)
````

В этой новой модели влияние брака на заработную плату значительно снизилось — с 0,1147 до 0,0476. Тем не менее, этот результат значим на уровне 99%, поэтому мужчины по-прежнему может ожидать увеличения заработка в результате брака.

**Когда панельные данные вам не помогут**

Использование панельных данных и моделей с фиксированными эффектами — чрезвычайно мощный инструмент для причинно-следственной связи. Когда у вас нет ни случайных данных, ни хороших инструментов, модель с фиксированными эффектами настолько же убедительна, насколько это возможно для причинно-следственной связи с неэкспериментальными данными. Тем не менее, стоит отметить, что это не панацея. Бывают ситуации, когда даже панельные данные вам не помогут.

Самый очевидный пример — наличие конфаундеров, которые меняются со временем. Фиксированные эффекты могут устранить смещение только от характеристик, которые постоянны для каждого человека. Например, предположим, что вы можете повысить свой уровень интеллекта, читая книги и употребляя в пищу много полезных жиров. Это позволит вам получить более высокооплачиваемую работу и жену. Модель с фиксированными эффектами не сможет устранить это смещение из-за неучтенного конфаундинга интеллекта, поскольку в данном примере интеллект меняется со временем.

Другой менее очевидный случай, когда модель с фиксированными эффектами не работает, — это **обратная причинно-следственная связь**. Например, предположим, что не брак влияет на ваш доход. Скорее всего, увеличение заработка увеличивает ваши шансы на брак. В этом случае корреляция положительная, но на первом месте стоит доход. Он будет меняться со временем и в одном направлении, поэтому фиксированные эффекты не смогут это контролировать.
