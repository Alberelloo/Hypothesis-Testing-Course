## Сущность мета-анализа

**Мета-анализ** представляет собой **статистический анализ, объединяющий и/или интегрирующий количественные данные из множества независимых исследований, посвященных одному и тому же исследовательскому вопросу**. Это статистический (или количественный) анализ, который отличается от традиционного нарративного обзора, где заключения делаются на основе качественной оценки исследований. По сути, мета-анализ — это «анализ анализов» (как определил Джин Гласс в 1976 году), который проводит операцию слияния данных (data fusion), давая оценку оценкам. Анализ обычно проводится на уровне исследования, то есть является «исследованием исследований».

**Фундаментальная цель мета-анализа** заключается в следующем:
*   **Вычисление объединенного размера эффекта** (combined effect size) по всем включенным исследованиям. Это позволяет получить усредненную или комбинированную оценку эффекта по всем работам.
*   **Повышение статистической мощности** (statistical power) и способности измерять слабые сигналы. Индивидуальные исследования могут обладать низкой мощностью для достижения статистической значимости при небольшом, но клинически важном эффекте. Объединение множества исследований увеличивает размер выборки, позволяя выявить такие эффекты.
*   **Разрешение неопределенностей или несоответствий** между результатами отдельных исследований. Мета-анализ позволяет выявлять факторы, объясняющие различия в результатах исследований, часто посредством стратифицированного или регрессионного анализа, что может генерировать новые гипотезы для будущих исследований.
*   **Интеграция и обобщение существующих знаний**, что играет ключевую роль в обосновании предложений по грантам, формировании рекомендаций по лечению и влиянии на политику здравоохранения. Он служит агрегатором для серии результатов тестов, повышая уровень доказательности и позволяя более уверенно внедрять статистически значимые инициативы.

**Сильные стороны и ограничения**

Мета-анализ, несмотря на свою методологическую строгость, не лишен недостатков и требует критического подхода к интерпретации результатов.

* **Сильные стороны мета-анализа**

  1.  **Повышенная объективность и воспроизводимость:**
    *   Мета-анализ привносит в процесс литературного обзора такую же **методологическую строгость**, которая требуется от экспериментальных исследований.
    *   В отличие от нарративных обзоров, выбор исследований в мета-анализе основан не на произволе автора, а на **публично разделяемых стандартах** и четко определенных критериях включения/исключения.
    *   Мета-анализ требует, чтобы все **субъективные решения**, принимаемые авторами (например, при кодировании модераторов), были **сделаны публичными**, что делает их открытыми для критики и повышает прозрачность. Это облегчает выявление некорректно проведенных мета-анализов по сравнению с нарративными обзорами, где специфическое содержание часто скрыто.

  2.  **Обобщение результатов и повышение статистической мощности:**
    *   Объединение данных из нескольких исследований **увеличивает совокупный размер выборки**, что приводит к более **высокой статистической мощности**, сужению доверительных интервалов (CI) и снижению числа ложноположительных результатов. Это позволяет выявлять эффекты, которые могли быть статистически незначимыми в отдельных небольших исследованиях.
    *   Мета-анализ эффективно объединяет все собранные данные, создавая эффект более масштабного и надежного эксперимента.
    *   Современный мета-анализ выходит за рамки простого объединения размеров эффектов, позволяя **тестировать изменчивость результатов** сверх ожидаемой из-за случайной выборки. Характеристики исследований (например, используемые инструменты измерения, популяции, дизайн) могут быть закодированы и использованы для снижения дисперсии оценки, что позволяет **статистически корректировать некоторые методологические недостатки**.
    *   Это смещает акцент с отдельных исследований на их совокупность, подчеркивая **практическую значимость размера эффекта**, а не только статистическую значимость.

  3.  **Возможность анализа подгрупп и модераторов:**
    *   Мета-анализ не игнорирует качественные различия между исследованиями, а **кодирует их как переменные-модераторы**, что позволяет эмпирически тестировать их влияние.
    *   Мета-регрессия играет **критически важную роль в учете эффектов ковариат**, особенно при наличии категориальных переменных, которые могут быть использованы для анализа подгрупп.
    *   Исследователи могут выявить факторы, объясняющие различия в результатах исследований, проводя стратифицированный анализ или мета-регрессию.

* **Опасения и ограничения мета-анализа**

  1.  **«Мусор на входе — мусор на выходе» (Garbage In, Garbage Out):**
    *   **Мета-анализ не может исправить недостатки дизайна или предвзятость (bias) в исходных исследованиях**. Если первичные исследования методологически некорректны, валидность мета-анализа будет поставлена под сомнение.
    *   Поэтому крайне важно включать только **методологически обоснованные исследования** (подход «синтез лучших доказательств»).
    *   Авторы мета-анализов должны кодировать характеристики качества исследований, чтобы оценить, влияет ли качество на размер эффекта. Если такая зависимость существует, низкокачественные исследования могут быть исключены или их влияние учтено.

  2.  **Гетерогенность (Heterogeneity):**
    *   **Гетерогенность** — это изменчивость истинных эффектов между исследованиями, обусловленная различиями в методологии, характеристиках выборки или других факторах.
    *   **Предположение однородности** (homogeneity), лежащее в основе моделей фиксированных эффектов (fixed effect models), заключается в том, что все исследования оценивают один и тот же основной параметр популяции. Однако это часто нереалистично, поскольку исследования подвержены различным источникам гетерогенности.
    *   При наличии гетерогенности **не существует единого размера эффекта для популяции** (δ), и каждое исследование оценивает свой собственный специфический эффект (δᵢ). В таких случаях комбинированная оценка может не отражать истинного значения для какой-либо конкретной популяции.
    *   **Чрезмерная межисследовательская изменчивость** может сделать комбинированную оценку бессмысленной, для чего были созданы методы выявления гетерогенности: **тест Кохрана Q**, **статистика I²** ...

  3.  **Предвзятость публикаций (Publication Bias), «Проблема запертого ящика» (File Drawer Problem):**
    *   **Предвзятость публикаций** возникает из-за тенденции публиковать исследования, демонстрирующие статистически значимые положительные результаты, в то время как исследования с отрицательными или незначимыми результатами остаются неопубликованными.
    *   Это приводит к **искаженным (преувеличенным) выводам** в мета-анализах, основанных только на опубликованных работах, создавая серьезную **ошибку базовой частоты**.
    *   **«Проблема запертого ящика»** (file drawer problem) описывает ситуацию, когда исследования с «неинтересными» результатами остаются «запертыми в ящиках» исследователей и не попадают в научную литературу.
    *   **Проблема заключается не только в журнальных практиках**, но и в **сомнительных исследовательских практиках** самих авторов, которые могут подгонять статистические модели до достижения значимости.

  4.  **«Эффект яблока и апельсина» (Combining Apples and Oranges):**
    *   Это распространенная критика, утверждающая, что мета-анализ **объединяет несравнимые исследования**. Если методологии исследований слишком сильно различаются, выводы могут быть ошибочными.
    *   **Ответ:** Авторы мета-анализов должны четко определить **критерии включения и исключения** исследований, чтобы обеспечить их сравнимость и теоретическую однородность. Мета-анализ не игнорирует качественные различия, а кодирует их как модераторы, чтобы эмпирически проверить их влияние. Однако **слабые стандарты включения** действительно могут привести к вводящим в заблуждение выводам.

\\

Таким образом, мета-анализ является мощным инструментом для синтеза научных доказательств, но его применение требует тщательного соблюдения методологии, критической оценки первичных исследований и внимательного учета потенциальных источников смещения и гетерогенности.

##### Выбор модели

При проведении мета-анализа одним из фундаментальных решений является выбор между моделями **фиксированных эффектов (Fixed-Effect)** и **случайных эффектов (Random-Effects)**.

\\

**Модель фиксированных эффектов (Fixed-Effect Model)**

Эта модель **предполагает, что все включенные исследования оценивают один и тот же основной истинный эффект ($δ$)**. Таким образом, наблюдаемые различия между оценками эффектов в разных исследованиях объясняются исключительно случайной ошибкой выборки. Это означает, что если бы каждое первичное исследование имело бесконечный размер выборки, оценка эффекта в нем точно совпадала бы с $δ$. Модель фиксированных эффектов **подходит, когда исследования считаются гомогенными**, то есть однородными по методам, популяциям и другим характеристикам.
  
В этой модели используется **Inverse-Variance Weighting**, где вес ($w_i$), присваиваемый каждому исследованию $i$, **обратно пропорционален дисперсии оценки эффекта этого исследования** ($d_i$). Это означает, что более точные исследования (с меньшей стандартной ошибкой и, следовательно, меньшей дисперсией) получают больший вес.

$$w_i = \frac{1}{\text{Var}(d_i)} = \frac{1}{\text{SE}(d_i)^2} \quad$$
  
Объединенная оценка эффекта ($\overline{T}$) рассчитывается как **взвешенное среднее оценок эффектов отдельных исследований**:

$$\overline{T} = \frac{\sum w_i T_i}{\sum w_i} \\
\text{где $T_i$ — это оценка размера эффекта для $i$-го исследования}
$$

Для проверки предположения о гомогенности используется **статистика Q-теста Кохрана**. Она измеряет общую вариабельность в наборе размеров эффектов и следует распределению хи-квадрат с $k-1$ степенями свободы, где $k$ — количество исследований. Большие значения $Q_T$ указывают на то, что наблюдаемые исследования, вероятно, происходят из нескольких популяций (т.е., присутствует гетерогенность).

$$Q_T = \sum w_i (T_i - \overline{T})^2 = \sum w_i (T_i)^2 - \frac{(\sum w_i T_i)^2}{\sum w_i} \quad$$

\\

**Модель случайных эффектов (Random-Effects Model)**

В отличие от фиксированных эффектов, эта модель **предполагает, что каждое исследование оценивает свой собственный, уникальный популяционный размер эффекта ($δ_i$)**. Эти индивидуальные истинные эффекты ($δ_i$) **считаются случайной выборкой из более широкого распределения популяционных эффектов**, которое имеет среднее значение ($δ$) и дисперсию между исследованиями ($τ²$). Таким образом, модель случайных эффектов учитывает два источника вариабельности: внутриисследовательскую вариабельность (ошибка выборки) и межисследовательскую вариабельность (гетерогенность). Модель случайных эффектов является более распространенным выбором, поскольку на практике исследования редко бывают идеально идентичными, и гетерогенность часто присутствует.

Наиболее распространенным подходом для оценки межисследовательской дисперсии (τ²) — **метод ДерСимониана-Лэйрда (DerSimonian-Laird)**, который основан на статистике $Q$-теста:

1. Сначала вычисляется **$Q$-статистика** (известная как $Q_T$) с использованием весов фиксированных эффектов $w_i = 1/\text{SE}(d_i)^2$:

$$Q = \sum_{i=1}^k w_i (d_i - \overline{d})^2$$

2.  Затем оценивается межисследовательская дисперсия $\tau^2$ с помощью следующей формулы:

$$\tau^2 = \frac{Q - (k-1)}{C} \\
\text{де $k$ — количество исследований, а $C$ — это коэффициент, вычисляемый как}
$$

$$C = \sum_{i=1}^k w_i - \frac{\sum_{i=1}^k w_i^2}{\sum_{i=1}^k w_i} \quad$$
  
  Если $Q < k-1$, то $\tau^2$ устанавливается равным 0, поскольку отрицательная дисперсия не имеет физического смысла.
  **Проблема** с этим методом состоит в том, что оценка $\tau^2$ зависит от $Q$-статистики, которая, в свою очередь, основана на предположении об отсутствии межисследовательской вариабельности (как в модели фиксированных эффектов).

В модели случайных эффектов веса для каждого исследования ($w_i^*$) корректируются с учетом как внутриисследовательской дисперсии ($s_i^2$, которая равна $\text{SE}(d_i)^2$), так и оцененной межисследовательской дисперсии ($\tau^2$).

$$w_i^* = \frac{1}{s_i^2 + \tau^2} = \frac{1}{\text{SE}(d_i)^2 + \tau^2} \quad
\text{}
$$

Эти веса $w_i^*$ для модели случайных эффектов стремятся к более равномерному распределению весов между исследованиями по сравнению с весами фиксированных эффектов. Они уменьшают вес наиболее точных исследований и увеличивают вес менее точных.

Объединенная оценка эффекта в модели случайных эффектов ($\overline{d}^*$) также является взвешенным средним:

$$\overline{d}^* = \frac{\sum w_i^* d_i}{\sum w_i^*} \quad$$

**Модель случайных эффектов является более гибкой и чаще всего рекомендуется**, поскольку она учитывает неизбежную гетерогенность между исследованиями. Включение дополнительного источника вариабельности приводит к **более широким доверительным интервалам**, что часто считается более консервативным подходом. Обычно рекомендуют **проводить оба анализа** (случайных эффектов как основной, фиксированных — как вторичный). Если результаты согласуются, это повышает уверенность в выводах. Если нет, это указывает на необходимость дальнейшего изучения причин расхождения, например, из-за "small study effects" (когда небольшие исследования имеют смещенные результаты) или систематической ошибки публикации.

После выбора модели, необходимо выбрать метод анализа наших данных, который зависит от типа данных.

###### Мета-анализ для дихотомических исходов

При работе с дихотомическими исходами (бинарными данными), такими как наличие/отсутствие события, успех/неудача, или выживание/смерть, используются специфические меры эффекта: отношение шансов (Odds Ratio, OR), относительный риск (Relative Risk, RR) и разница рисков (Risk Difference, RD).

Для мета-анализа дихотомических исходов часто используются следующие входные данные для каждого исследования:
*   **Четыре показателя в виде таблицы 2x2:**
    *   $a$: количество событий в экспериментальной группе.
    *   $b$: количество отсутствия событий в экспериментальной группе.
    *   $c$: количество событий в контрольной группе.
    *   $d$: количество отсутствия событий в контрольной группе.
*   Или, эквивалентно, общие размеры выборок ($n_e = a+b$, $n_c = c+d$) и доли событий ($p_e = a/n_e$, $p_c = c/n_c$).

\\

**1. Метод Мантеля-Хензеля**

Представьте, что у вас есть не одна, а целый набор таблиц 2x2. Каждая из них описывает взаимосвязь между вмешательством и исходом в разных условиях: например, в разных клинических центрах, среди мужчин и женщин или, что наиболее актуально для мета-анализа, в разных независимых исследованиях. Задача — получить из этого набора одну, обобщенную картину. Именно для этого и был создан метод Мантеля-Хензеля.

Метод Мантеля-Хензеля (Mantel-Haenszel, MH) — это мощный и элегантный статистический инструмент, предназначенный для **анализа стратифицированных данных** или, иными словами, для **объединения результатов из $k$ таблиц сопряженности одинаковой структуры**. В контексте мета-анализа эти "страты" — это и есть отдельные первичные исследования. Таким образом, MH позволяет синтезировать данные across studies, учитывая при этом, что каждое исследование вносит свой, уникальный вклад в общую картину.

Для работы метода необходимы "сырые" данные по каждому исследованию. А именно, для $i$-го исследования мы должны располагать классической таблицей 2x2 с четырьмя ключевыми цифрами:
*   $a_i$: количество участников в экспериментальной группе, у которых наступило интересующее событие.
*   $b_i$: количество участников в экспериментальной группе, у которых событие *не* наступило.
*   $c_i$: количество участников в контрольной группе, у которых событие наступило.
*   $d_i$: количество участников в контрольной группе, у которых событие *не* наступило.
Сумма $n_i = a_i + b_i + c_i + d_i$ дает нам общий размер выборки для этого исследования.

В отличие от других методов, которые сначала вычисляют отношение шансов (OR) для каждого исследования в отдельности, а затем усредняют эти логарифмированные значения, подход Мантеля-Хензеля более прямой. Он **усредняет сами отношения шансов (не логарифмированные)**, но делает это особым, взвешенным образом, который минимизирует влияние случайных вариаций в небольших исследованиях. Объединенная оценка отношения шансов по Мантелю-Хензелю ($OR_{MH}$) вычисляется по формуле, которая по своей сути является взвешенным средним отношений шансов во всех стратах (исследованиях).

$$
\text{OR}_{\text{MH}} = \frac{\sum_{i=1}^k w_i^{\text{(MH)}} \cdot OR_i}{\sum_{i=1}^k w_i^{\text{(MH)}}}
$$

Однако чаще ее записывают в компактном и вычислительно удобном виде:

$$
\text{OR}_{\text{MH}} = \frac{\sum_{i=1}^k \frac{a_i d_i}{n_i}}{\sum_{i=1}^k \frac{b_i c_i}{n_i}}
$$


**В числителе** мы суммируем по всем исследованиям величину $(a_i * d_i) / n_i$. Эту величину можно интерпретировать как "поддержку" идеи о том, что лечение эффективно (т.е., что события в экспериментальной группе и не-события в контрольной группе происходят чаще).
**В знаменателе** мы суммируем величину $(b_i * c_i) / n_i$, которая, наоборот, представляет собой "поддержку" нулевой гипотезы об отсутствии эффекта. Итоговое отношение этих двух сумм и дает нам объединенное отношение шансов. Если значение $OR_MH$ больше 1, это указывает на положительный эффект лечения в целом по всем исследованиям.

**Деление на $n_i$ - ключевой элемент метода. Деление на общий размер выборки исследования выполняет роль **веса**. Исследования с большей выборкой вносят больший вклад в числитель и знаменатель, то есть имеют больший "вес" в финальном результате. При этом метод естественным образом и устойчиво работает с небольшими числами и даже с нулевыми ячейками в таблицах (в большинстве случаев), что делает его чрезвычайно robust и практичным для мета-анализа реальных данных.

Таким образом, метод Мантеля-Хензеля предоставляет нам не просто среднее значение, а **взвешенную сводную меру ассоциации**, которая интегрирует информацию из всех доступных исследований в единую, статистически обоснованную оценку общего эффекта.

\\

**2. Метод Пето**

Представьте, что вы проводите мета-анализ по серьезному, но к счастью, редкому побочному эффекту лекарства. В большинстве исследований это событие не произошло ни с одним пациентом, а в тех немногих, где произошло, счет идет буквально на единицы. Стандартные методы спотыкаются о нулевые ячейки таблиц 2x2, а предположение о нормальности распределения нарушено. Что делать? Здесь на сцену выходит метод Пето — нежный гигант в мире статистики, созданный специально для работы с хрупкими, редкими событиями.

Метод Пето — это остроумный статистический маневр, позволяющий получить оценку **объединенного отношения шансов (OR)**, когда мы имеем дело с дихотомическими исходами, но стандартные методы дают сбой. Его главная суперсила — работать там, где данные скудны: при малом числе событий, пустых ячейках таблиц и несбалансированных группах. Он делает это, задаваясь простым вопросом: "Насколько наблюдаемое число событий в экспериментальной группе отличается от того, что мы ожидали бы увидеть, если бы воздействие было абсолютно бесполезным?"

Как и Мантелю-Хензелю, методу Пето нужны "сырые" данные по каждому исследованию. Для i-го исследования мы должны иметь:
*   $a_i$: число событий в экспериментальной группе.
*   $b_i$: число не-событий в экспериментальной группе.
*   $c_i$: число событий в контрольной группе.
*   $d_i$: число не-событий в контрольной группе.
*   $n_i$: общий размер выборки исследования ($a_i + b_i + c_i + d_i$).


Вместо того чтобы напрямую вычислять отношение шансов для каждого исследования, метод Пето идет обходным путем. Он работает не с самим OR, а с **количеством событий в экспериментальной группе**. В мире, где воздействие не работает (то есть истинное OR = 1), это количество должно быть примерно таким, как диктует нам теория вероятностей. Любое систематическое отклонение от этого ожидания и есть сигнал о эффекте воздействия.

Метод работает в 3 шага:

1.  **Шаг первый: Расчет отклонения для каждого исследования.**

  Для каждого исследования мы вычисляем, насколько реальность разошлась с нашими "нейтральными" ожиданиями.
  *   **Наблюдаемое ($O_i$)**: Это просто $a_i$ — сколько событий на самом деле произошло в экспериментальной группе.
  *   **Ожидаемое ($E_i$)**: Это то, сколько событий *должно было бы* произойти в экспериментальной группе, если бы воздействие не работало. Считается по формуле:
  $$
  E_i = (n_{1i} * m_{1i}) / N_i \\
  \text{где $n_{1i}$ — размер экспериментальной группы, $m_{1i}$ — общее число событий в обеих группах, $N_i$ — общая численность исследования}
  $$
        
  *   **Отклонение $(O - E)_i$**: Разница между первым и вторым. Это "сердце" метода Пето.
  Отрицательное значение говорит, что событий было меньше, чем ожидалось (возможно, воздействие работает и защищает от события). Положительное — что больше (возможно, воздействие вредит).

2.  **Шаг второй: Взвешивание и объединение.**

  Чтобы учесть разный набор выборок в исследованиях метод Пето взвешивает каждое отклонение на его **дисперсию (V_i)**.

  $$V_i = (n_{1i} * n_{2i} * m_{1i} * m_{2i}) / (N_i^2 * (N_i - 1))$$

  Теперь мы можем получить объединенное, взвешенное отклонение по всем исследованиям: просто суммируем все $(O - E)_i$ и всю дисперсию $V_i$.

3.  **Шаг третий: Превращение отклонения в отношение шансов.**
  
  Финальный шаг — преобразовать это суммарное отклонение в привычную нам метрику — **отношение шансов**. Поскольку отношение шансов обычно логарифмируют, чтобы сделать распределение симметричным, формула выглядит так:
  
  $$ln(OR_{Peto}) = \sum (O - E)_i / \sum V_i$$

  Чтобы получить итоговое $OR_{Peto}$, мы просто возводим экспоненту в эту степень: $OR_{Peto} = exp(ln(OR_{Peto}))$.

  Суть метода Пето в том, что он **обходит проблему нулевых ячеек**. Для расчета $(O - E)$ и $V$ не требуется вычислять OR для каждого отдельного исследования, где эти вычисления могли бы быть невозможны или крайне неустойчивы. Он работает напрямую с частотами, агрегирует их и только в самом конце выдает интерпретируемый результат.

###### Мета-анализ для непрерывных исходов

Для непрерывных исходов (например, измеряемые величины, баллы по шкалам) используются такие меры эффекта, как **разница средних (Mean Difference, MD)** и **стандартизированная разница средних (Standardized Mean Difference, SMD)**.

Для каждого исследования обычно требуются:
*   Размеры выборок для экспериментальной ($n_e$) и контрольной ($n_c$) групп.
*   Средние значения для экспериментальной ($\overline{Y}_e$) и контрольной ($\overline{Y}_c$) групп.
*   Стандартные отклонения для экспериментальной ($s_e$) и контрольной ($s_c$) групп.

\\

**Разность средних (Mean Difference, MD)**

MD используется, когда **первичные исследования имеют сопоставимые меры исхода** (т.е., измеряют одно и то же явление в одних и тех же единицах). По сути это просто разность средних между экспериментальной и контрольной группами:

$$\text{MD} = \overline{x}_e - \overline{x}_c \quad$$

**Стандартная ошибка MD** считается с помощью **объединенное стандартное отклонение** ($s_p$):
$$
\text{SE}_{\text{MD}} = s_p \sqrt{\frac{1}{n_e} + \frac{1}{n_c}} \\
\text{ где } s_p = \sqrt{\frac{(n_e-1)s_e^2 + (n_c-1)s_c^2}{n_e+n_c-2}}
$$

\\

**Стандартизированные разности средних (SMD: Cohen's d, Hedges' g)**

SMD используется, когда **первичные исследования применяют разные меры одного и того же конструкта** (например, разные шкалы для измерения депрессии). Результаты выражаются в единицах стандартных отклонений, что делает их безразмерными и сопоставимыми.

*   **Оценка Cohen's g (или Glass's g)** Это простейшая мера эффекта, основанная на разностях средних:

$$
\mathbf{g = \frac{\overline{Y}_e - \overline{Y}_c}{s_p}} \\
\text{где $\overline{Y}_e$ и $\overline{Y}_c$ — средние значения экспериментальной и контрольной групп соответственно}, \\
\text{$s_p$ — объединенное стандартное отклонение}
$$



*   **Коррекция на смещение для малых выборок (Hedges' d)**

Однако **$g$ является смещенной оценкой**, особенно для малых выборок, и, как правило, завышает истинный популяционный эффект. Для получения несмещенной оценки SMD, **$g$ корректируется для малых выборок, в результате чего получается Hedges' d**.

$$d = g \left(1 - \frac{3}{4m - 1}\right) \\
\text{где $m = n_e + n_c - 2$}
$$


со стандартной ошибкой

$$
\text{SE}_{\text{SMD}} = \sqrt{\frac{n_e+n_c}{n_e n_c} + \frac{\text{SMD}^2}{2(n_e+n_c)}}
$$

#### Регрессионный мета-анализ

Это метод мета-анализа, который **использует регрессионный анализ для объединения, сравнения и синтеза результатов исследований**, корректируя при этом эффекты доступных ковариат (модераторов) на зависимую переменную (размер эффекта). Она позволяет исследовать, как как категориальные, так и **непрерывные модераторы** влияют на размер эффекта.

Модель обычно принимает форму нормальной линейной регрессии:
$$
\mathbf{y_{tk} = x_{tk}'\beta + \epsilon_{tk}} \\
\text{где $y_{tk}$ — зависимая переменная (например, размер эффекта) для $k$-го исследования и $t$-го лечения},\\
\text{$x_{tk}'$ — вектор ковариат на уровне группы/исследования}, \\
\text{$\beta$ — вектор коэффициентов регрессии, и $\epsilon_{tk}$ — член ошибки}
$$

Тогда коэффициенты $\beta$ показывают, как изменение значения ковариаты (модератора) влияет на размер эффекта.

В регрессионном мета-анализе можно так же в зависимости от предпоссылок согласно собранным иследованиям выбирать между моделью фиксированных и смешанных эффектов:  

*   **Фиксированных эффектов:** $\mathbf{y_{tk} = x_{tk}'\beta + \epsilon_{tk}}$.
*   **Смешанных эффектов:** $\mathbf{y_{tk} = x_{tk}'\beta + w_{tk}'\gamma_k + \epsilon_{tk}} \text{, $w_{tk}'\gamma_k$ — эффект, отражающий вариабельность исследований}$

Обычно использую взвешенную регрессию (WLS), где в качестве веса используется обратная величина дисперсии исследования.

### Синтез значимостей

Мета-анализ становится особенно полезным в нескольких ключевых сценариях:

*   **Для подтверждения сильных эффектов:** Если несколько тестов с общей гипотезой показывают статистически значимый и положительный эффект в одном направлении, мета-анализ лишь ещё больше убеждает в наличии эффекта и позволяет увереннее внедрять инициативы.
*   **Для выявления скрытых эффектов:** Одна из наиболее интересных и ценных ситуаций, когда мета-анализ проявляет себя, это когда отдельные тесты **не достигают статистической значимости (p-value находится у порога значимости), но при этом абсолютно все указывают на лучшее состояние нового дизайна или фичи**. Например, если два независимых теста дают p-value = [0.051, 0.051], по отдельности они не позволяют отклонить нулевую гипотезу, но их комбинация кажется крайне маловероятной случайностью. Мета-анализ позволяет проверить, можно ли отклонить нулевую гипотезу, рассматривая эти p-value вместе.


Но спешить сразу мета-анализировать свои тесты не стоит. Надо иметь ввиду несколько условий, чтобы сделать всё правильно:

* Во-первых, тесты должны быть согласованы, когеренты. Иначе можно одинаковые по названию, но разные по смыслу и содержанию пилоты некорректно обобщить.

* Во-вторых, надо максимально сохранять сравнимость групп из теста в тест. Чтобы вычленить эффект от воздействия, выборки должны быть рандомизированы для соблюдения влияния одних и тех же факторов не только на обе группы, но и между тестами.

В общем случае перед применением мета-анализа надо думать, например, cравнимы ли аудитории, чтобы не получилось, что вы сравниваете в тесте мягкое и пушистое.

Во-вторых, надо ответить на вопрос, участвуют ли люди из параллельного/предыдущего теста в другом тесте с той же $H_0$. От этого зависит, с какой общей $H_0$ вы имеете дело – с независимой или зависимой.

* **Независимая общая  $H_0$**

  Это когда у нас из теста в тест фичи идёт разная конфигурация юзеров/объектов в A/B. В этом случае мы работаем с комбинацией статистик/p-value (и вероятностью таковую получить, если верна $H_0$). Для этого случая мы рассмотрим метод комбинированной вероятности Фишера (Fisher’s method, Fisher's combined probability test) и метод weighted z-score (Stouffer's, Liptak).

* **Зависимая общая $H_0$**

  Это когда из теста в тест пользователи могут повторяться. В случае зависимости – со связанными статистиками/p-value от одного и тоже набора объектов (юзеры, например) в A/B из теста в тест. Пример: последовательное (sequential) тестирование. В этом случае рассмотрим метод взвешенного гармонического среднего (weighted harmonic mean p-value).

Но не всё так просто, к сожалению. Дело в том, что таким подходом можно пользоваться лишь при незначительном объёме экспериментов. Если количество экспериментов будет расти, то:

* При расчёте методом Монте-Карло нам понадобится много ресурсов на генерацию множества вариантов.

* При аналитическом расчёте также понадобится много вычислительных ресурсов, так как интегралы будут многомерными.

Именно поэтому Фишер предложил универсальный способ, основанный на свойстве распределения Хи-квадрат:

Основная проблема в том, что произведение комбинации p-value (когда верна $H_0$) **не распределяется равномерно и не сводится к какому-либо известному распределению**. Фишер заметил, что **сумма нескольких значений $\chi^2$ сама распределяется как $\chi^2$** с соответствующим количеством степеней свободы. Он предложил преобразовать каждое p-value таким образом, чтобы оно соответствовало $\chi^2$ распределению. Он обнаружил, что если взять **натуральный логарифм вероятности, изменить его знак и удвоить**, мы получим эквивалентное значение $\chi^2$ с **2 степенями свободы (k=2)**.

Математически это выглядит так:
* Для равномерно распределенной случайной величины p, случайная величина **-2 ln(p)** будет иметь **экспоненциальное распределение**, которое является частным случаем **$\chi^2$ распределения с 2 степенями свободы**.
* Если у нас есть *k* независимых p-value ($p₁, p₂, ..., pₖ$), каждое из которых преобразовано в $\chi^2$ с 2 степенями свободы, то их сумма - $X² = -2 \sum ln(pᵢ)$.
* Эта суммарная статистика $\chi^2$ будет распределена как **$\chi^2$ с 2k степенями свободы**.
* Полученное значение $\chi^2$ сравнивается с критическими значениями $\chi^2$ распределения с 2k степенями свободы.

В нашем примере расчёт происходил бы так:
*   Для $p₁ = 0.051: -2 * ln(0.051) ≈ -2 * (-2.976) ≈ 5.952$
*   Для $p₂ = 0.051: -2 * ln(0.051) ≈ 5.952$
*   Суммарная статистика $\chi^2 = 5.952 + 5.952 = 11.904$
*   Количество степеней свободы = 2 * $k$ = 2 * 2 = 4
*   Значение $\chi^2 = 11.904$ с 4 степенями свободы даёт мета p-value **меньше 0.05**, что является статистически значимым.

\\

Несмотря на свою полезность, метод Фишера имеет несколько важных ограничений:
*   **Не учитывает направление эффекта**: Он оперирует только p-value без привязки к абсолютному значению или направлению разницы между группами. Это означает, что мета-p-value может быть статистически значимым, даже если в одном тесте B оказался лучше A, а в другом — хуже. Интерпретация такого результата становится проблематичной.
*   **Не учитывает разный объем выборок в базовой форме**: В базовой формуле метода Фишера не предусмотрено прямое введение весов, зависящих от объёмов выборок в каждом тесте. Хотя существуют предложения по введению весов, это усложняет распределение статистики, и для таких случаев рекомендуются другие, более специализированные методы.

#### Метод Weighted Z-score

Этот метод представляет собой иной подход к мета-анализу результатов тестов, предложенный социологом Шоуффером (Shouffer), а также другими исследователями, такими как Липтак (Liptak), Цзэчжун (Zezhong) и Зайкин (Zaykin).

Метод Weighted Z-score **учитывает как статистику, так и размер выборок в каждом из тестов**. Это позволяет избежать проблем, которые могут возникнуть с методом Фишера, когда стат. значимый мета-p-value получается при разнонаправленных эффектах в отдельных тестах (например, в одном случае B лучше A, в другом — хуже), что затрудняет интерпретацию. Weighted Z-score, напротив, включает знак статистики, указывающий направление эффекта.

Метод строится на основе $Z$-нормального распределения, используемого, например, в one-sample z-тесте. Если представить бесконечно большую выборку, разбитую на две части ($S_1$ и $S_2$), и использовать определение среднего, можно прийти к формуле, где веса ($w$) равны корню из объема выборки ($w = \sqrt{n}$).

В общем виде для *$k$* тестов формула для **комбинированной $Z$-статистики** выглядит следующим образом:
$$Z_{combined} = \frac{\sum_{i=1}^{k} (w_i \cdot z_i)}{\sqrt{\sum_{i=1}^{k} w_i^2}}$$
где:
*   $w_i$ — вес *i*-го теста, зависящий от размера выборки.
*   $z_i$ — z-статистика *i*-го теста.

Эта комбинированная $Z$-статистика имеет стандартное нормальное распределение, что делает ее легко интерпретируемой. Этот подход легко переносится на классический **two-sample t-test**.


Метод Weighted Z-score отлично подходит для учета различного объема выборок в независимых тестах, вводя веса. Алгоритм для расчета доверительного интервала и комбинированной абсолютной разницы, включает следующие шаги:

1.  **Нормализация весов ($w_{normalized\_i}$)**
    $$w_{normalized\_i} = \frac{w_i}{\sum w_i}$$
2.  **Расчет комбинированных средних для A и B ($\mu_{A}^{combined}$, $\mu_{B}^{combined}$)**:

    Это по сути взвешенные средние, учитывающие различия в количестве аудитории.
  $$\mu_{A}^{combined} = \sum_{i=1}^{n} (w_{normalized\_i} \cdot \mu_{A_{i}})$$
  $$\mu_{B}^{combined} = \sum_{i=1}^{n} (w_{normalized\_i} \cdot \mu_{B_{i}})$$
3.  **Вычисление комбинированной дельты (разницы)**:
    $$\Delta = \mu_{B}^{combined} - \mu_{A}^{combined}$$
4.  **Расчет дисперсии и стандартной ошибки эффекта**: Комбинированная дисперсия также является взвешенной.
    $$SE_{combined} = \sqrt{\sum_{i=1}^{n} (w_{normalized\_i}^2 \cdot SE_i^2)}$$
5.  **Построение доверительного интервала для комбинированного эффекта**:
    $$[\Delta - Z_{\alpha/2} \cdot SE_{combined}, \quad \Delta + Z_{\alpha/2} \cdot SE_{combined}]$$

\\

Метод Weighted Z-score имеет ряд преимуществ перед методом Фишера:
*   **Более привычная и интерпретируемая статистика (z-статистика)**, особенно для сравнения средних, что является частым сценарием.
*   **Наличие знака у статистики, который показывает направление эффекта**. Это критично, поскольку Фишер оперирует только p-value, не учитывая, в какую сторону направлен эффект в каждом тесте, что может привести к значимому мета-p-value при противоречивых результатах.
*   **Возможность визуализации**, которая более понятна и стабильна независимо от количества тестов.
*   **Уточнение эффектов через доверительные интервалы в абсолютных значениях**.
*   **Выше мощность критерия**: исследования, например, симуляции Уитлока (Whitlock), показали, что взвешенная версия Z-метода превосходит подход Фишера по мощности. Важно отметить, что если веса одинаковы, результаты идентичны Фишеру.


**Основной недостаток** заключается в том, что для других статистических критериев, помимо сравнения средних, возможно, придется использовать p-value, и в этих случаях метод Фишера может быть более логичным.

#### Harmonic Mean p-value

Метод взвешенного гармонического среднего p-значения (HMP) был предложен в 2019 году Дэниелом Уилсоном. Он представляет собой подход к объединению p-значений, который особенно актуален, когда **тесты зависимы друг от друга**.

Часто возникает ситуация, когда пользователи могут повторяться из теста в тест в рамках одной и той же гипотезы, или когда проверяется однородность подобранных групп через множество тестов. Например, если тест не показал статистической значимости, его могут повторить, и часть пользователей из предыдущей тестовой группы может попасть в новую тестовую группу. В таких случаях p-значения этих тестов становятся зависимыми от одних и тех же пользователей.

При таких сценариях простой усреднённый p-value может быть неинформативным или даже вводить в заблуждение, особенно если размеры аудиторий в каждом тесте сильно различаются. Например, если один тест с большим объемом аудитории не значим, а второй с малым объемом значим, их простое среднее может быть статистически значимым, но это не будет отражать реальной картины из-за повтора теста или разницы в размерах выборок.

Метод HMP помогает решить эту проблему, предоставляя способ **объединения зависимых p-значений с учётом весов**, которые могут отражать, например, размер выборок.


Вывод метода HMP основан на **правдоподобии (likelihoodist)**:

1.  **Сравнение правдоподобий (likelihood)**: Метод начинается со сравнения двух правдоподобий, что является отправной точкой в **последовательном тестировании (Sequential Testing)**. Последовательное тестирование, такое как SPRT и mSPRT, контролирует FWER в серии тестов, что актуально для множественного тестирования.
2.  **Зависимость статистик**: В последовательном тестировании статистики правдоподобий ($R$) в каждом последующем тесте **зависимы** от данных предыдущих дней, поскольку они используют всю полноту накопленной информации.
3.  **Распределение R при $H_0$**: Если верна общая нулевая гипотеза для всех тестов правдоподобий, распределение статистики $R$ будет **LogGamma**.
4.  **Аппроксимация p-value**: При определенных условиях (в частности, при параметре $v=2$, что приводит к $\alpha=1$ в распределении LogGamma), p-value может быть аппроксимировано как $1/R$.
5.  **Формирование HMP**: Заменяя R на $1/p_{value}$ в формуле усреднения R для зависимых данных $\frac{1}{L} \sum_{l=1}^{L} R_l$, получаем формулу гармонического среднего p-значений.

В общем виде формула для **гармонического среднего p-значения** с учётом весов выглядит так:
$$p_{0} = \frac{1}{\sum_{i=1}^{L} \frac{w_i}{p_i}}$$

где:
*   $p_0$ — комбинированное HMP
*   $L$ — количество тестов
*   $p_i$ — p-значение $i$-го теста
*   $w_i$ — **нормализованный вес** $i$-го теста

Вес $w_i$ может определяться через размер выборок в тестах. Например, в одном из примеров вес для теста рассчитывается как доля его аудитории от общей аудитории всех тестов. Однако, можно ввести веса и не на основе размера выборки, а исходя из **важности показателей** (например, социодемографических или метрических).

Важно отметить, что HMP, обозначаемое как $p_0$, **нельзя интерпретировать буквально как p-value**, за исключением случаев, когда оно **достаточно мало** (например, 0.05 и менее). При больших значениях $p_0$ приравнивание его к p-value может привести к **большому количеству ошибок первого рода**. Для расчета точного p-value на базе гармонического среднего для множества тестов существуют специальные формулы и таблицы.


Метод HMP особенно полезен в следующих сценариях:
*   **Проверка однородности** подобранных групп через множество зависимых тестов.
*   В качестве **дополнительной оценки** в случаях, когда несколько метрик показывают положительный результат, но требуется общая оценка значимости, например, по экономическим причинам.
*   Для борьбы с проблемой **повтора тестов** («дождались прокраса»), когда пользователи могут повторяться, а размеры аудиторий различны.

## Визуализация и интерпретация графиков

В мета-анализе используются несколько визуализаций для представления данных и обнаружения потенциальных проблем. Две основные и наиболее часто используемые визуализации — это воронкообразные графики (funnel plots) и лесные графики (forest plots). Также могут использоваться гистограммы распределения размеров эффекта и двумерные таблицы для изучения модераторов.

**1. Воронкообразный график (Funnel Plot)**

image.png


Воронкообразный график — это график, предназначенный для проверки наличия **публикационного смещения (publication bias)**. Он часто используется в систематических обзорах и мета-анализах.

**Компоненты и интерпретация:**
*   **Оси:**
    *   **Ось Y:** Представляет **точность исследования** (study precision), например, обратную стандартную ошибку или количество экспериментальных субъектов. Чем выше значение на оси Y, тем выше точность исследования. Рекомендуется использовать стандартную ошибку.
    *   **Ось X:** Показывает **результат исследования** (treatment effect measure), например, измеренный средний эффект препарата. График традиционно рисуется с мерой эффекта лечения на горизонтальной оси, что облегчает интерпретацию асимметрии.
*   **Расположение точек:** Каждая точка на графике представляет отдельное исследование.
*   **Идеальная форма (отсутствие смещения):** В отсутствие публикационного смещения предполагается, что исследования с высокой точностью будут располагаться близко к среднему значению, а исследования с низкой точностью будут равномерно распределены по обе стороны от среднего, образуя приблизительно **форму перевернутой воронки**. Симметричная перевернутая воронка указывает на «хорошо себя ведущий» набор данных, в котором публикационное смещение маловероятно.
*   **Отклонение от формы (наличие смещения):** **Отклонение от воронкообразной формы может указывать на публикационное смещение**. Асимметричная воронка предполагает взаимосвязь между оценкой эффекта лечения и точностью исследования. Это может указывать на публикационное смещение или систематическое различие между исследованиями высокой и низкой точности (так называемые «эффекты малых исследований»). Асимметрия также может возникнуть из-за использования неподходящей меры эффекта.


**Критика воронкообразных графиков:**
*   Могут создавать ложное впечатление о публикационном смещении, если высокоточные исследования отличаются от низкоточных исследований по размеру эффекта (например, из-за разных исследуемых популяций).
*   Вид графика может значительно меняться в зависимости от шкалы на оси Y (например, обратная квадратная ошибка или размер выборки).
*   Исследователи плохо справляются с визуальным распознаванием публикационного смещения по воронкообразным графикам.

**2. Лесной график (Forest Plot)**

image.png

Лесной график — это тип графика, используемый в исследованиях для отображения **результатов одной темы из нескольких исследований построчно**. Он позволяет легко увидеть вариации и тенденции в отношении вмешательства или лечения. Лесные графики обеспечивают способ отображения большинства результатов мета-анализа в одной фигуре.

**Компоненты и интерпретация:**
*   **Метки исследований (Study Labels):** Расположены в левой части графика, идентифицируют каждое соответствующее исследование и содержат имя автора исследования и год.
*   **Размер эффекта (Effect Size):** Представляет наблюдаемый эффект результата в каждом исследовании. Он отображается **точкой (или квадратом/прямоугольником/кругом)**. Положение точки на графике указывает, является ли эффект отрицательным, нейтральным или положительным.
    *   **Размер эффекта больше 1:** Исследование обнаружило положительный эффект.
    *   **Размер эффекта меньше 1:** Исследование обнаружило отрицательный или более слабый эффект.
*   **Доверительные интервалы (Confidence Intervals - CI):** Каждое исследование показывает **горизонтальную линию**, отображающую диапазон возможных результатов. Более узкая линия указывает на более высокую точность размера эффекта.
*   **Линия отсутствия эффекта (No Effect Line):** Это **вертикальная линия**, представляющая отсутствие эффекта (либо ноль, либо единица, в зависимости от используемой меры эффекта).
    *   **Если доверительный интервал исследования пересекает эту линию:** Исследование не является статистически значимым.
    *   **Если доверительный интервал исследования не пересекает эту линию:** Исследование обнаружило статистически значимый эффект.
*   **Общий эффект (Overall Effect - Diamond Shape):** Расположен в нижней части графика, представлен **алмазообразной фигурой**, суммирующей все результаты исследований (объединенный эффект). Центр алмаза показывает общий эффект, а ширина алмаза представляет его доверительный интервал.
    *   **Если алмаз не пересекает линию отсутствия эффекта:** Объединенный результат является статистически значимым.
    *   **Если алмаз пересекает линию отсутствия эффекта:** Общего эффекта нет, или он не статистически значим.
*   **Согласованность исследований:** Если все линии (доверительные интервалы) расположены близко друг к другу, это означает, что исследования согласуются по данной теме; в противном случае существуют различные выводы.

    *   `fail.h` (количество не-респондентов в группе галоперидола)
    *   `fail.p` (количество не-респондентов в группе плацебо)
    *   `drop.h` (количество выбывших из группы галоперидола)
    *   `drop.p` (количество выбывших из группы плацебо)
*   **Преобразование данных для PythonMeta**:
    *   **Почему и зачем**: Пакет **PythonMeta**, выбранный для проведения мета-анализа, требует определенный формат входных данных, а именно четыре переменные: `resp.h`, `resp.p`, `T.h` (общее количество участников в группе галоперидола) и `T.p` (общее количество участников в группе плацебо).
    *   **Как**: Соответственно, исходный набор данных был **модифицирован** для обеспечения совместимости с требованиями PythonMeta.
*   **Искомый результат (Effect Size)**:
    *   **Почему и зачем**: Авторы стремились измерить **клиническое улучшение**, которое было выбрано в качестве меры эффекта (effect size) для синтеза доказательств.
    *   **Как**: В качестве меры эффекта было выбрано **отношение рисков ($RR$)**. Значение $RR$ больше единицы указывало на эффективность галоперидола по сравнению с плацебо.

**Методы мета-анализа**

*   **Используемые модели**:
  *   **Модели с фиксированными эффектами (Fixed-effect models)**: Предполагают, что существует единый, истинный размер эффекта для всех включенных исследований.
  *   **Модели со случайными эффектами (Random-effects models)**: Позволяют размеру эффекта варьироваться от исследования к исследованию, учитывая как внутри-исследовательскую, так и меж-исследовательскую изменчивость.
  *   **Почему и зачем**: Авторы выбрали запустить обе модели **для демонстрационных целей**, хотя их априорное предположение состояло в том, что модель со случайными эффектами более подходит. Это стандартная практика в мета-анализе для всесторонней оценки.
*   **Конкретные методы оценки**:
  *   Для модели с фиксированными эффектами по умолчанию использовался метод **Мантеля-Хензеля (Mantel–Haenszel, MH)**.
  *   Для модели со случайными эффектами для оценки меж-исследовательской дисперсии ($τ^2$) использовался метод **Дерсимониана и Лэрда (DerSimonian and Laird, DL)**.

**Генерация основных результатов**
*   **Действия**:
    *   **Выбор типа исхода и размера эффекта**: Был выбран **бинарный** тип исхода ("CATE" в PythonMeta) и **отношение рисков (RR)** в качестве желаемого размера эффекта.
    *   **Запуск моделей**: Были запущены как **модель с фиксированными эффектами**, так и **модель со случайными эффектами**.
    *   **Выбор методов**: Для фиксированных эффектов был выбран **MH**, а для случайных эффектов — **DL**.
    *   **Генерация графиков**: Основными результатами этого шага были **лесные графики (forest plots)** и **воронкообразные графики (funnel plots)**. Авторы отметили, что они **модифицировали стандартные скрипты Python** для создания более "чистых" и информативных визуализаций.
*   **Почему и зачем**: Этот шаг позволяет получить базовые результаты мета-анализа, визуализировать индивидуальные и сводные размеры эффекта, а также оценить гетерогенность между исследованиями. Выбор отношения рисков ($RR$) был обусловлен тем, что это стандартный способ измерения клинического улучшения для бинарных исходов. Запуск обеих моделей (фиксированных и случайных эффектов) позволяет сравнить результаты при разных предположениях о гомогенности эффектов.

```python
def showresults(fixed_results, random_results):
    text = "{0:<20} {1:<8} {2:18} {3:10} {4}"\
           .format("Study ID", "n", "ES[95% CI]","Fixed Wt(%)", "Random Wt(%)\n\n")
    for i in range(1, len(fixed_results)):

        NewId=(fixed_results[i][0]).replace("<sub>","")

        if (fixed_results[i][0]).startswith("<sub>"):
            Id= NewId
        else:
            Id=fixed_results[i][0]
        text += "{0:<20} {1:<6} {2:5.2f}[{3:<6.2f}{4:<6.2f}]   "\
                "{5:<11.2f} {6:<15.2f} \n".format(  # for each study
                                Id,   # study ID
                                fixed_results[i][5],   # total num
                                fixed_results[i][1],   # effect size
                                fixed_results[i][3],   # lower of CI
                                fixed_results[i][4],   # higher of CI
                                100*(fixed_results[i][2]/fixed_results[0][2]),   # Fixed weight
                                100*(random_results[i][2]/random_results[0][2]))# Random weight


    text += "\n{0:<20} {1:<7} {2:5.2f}[{3:<6.2f}  {4:<6.2f}] Z={5:<.3} P-value={6:<.20}  \n".format(  # for total effect
        "Fixed Effect Model",  # total effect size model  name
        fixed_results[0][5],  # total N (all studies)
        fixed_results[0][1],  # total effect size
        fixed_results[0][3],  # total lower CI
        fixed_results[0][4],  # total higher CI
        fixed_results[0][10],
        fixed_results[0][11]
        )

    text += "{0:<20} {1:<7} {2:5.2f}[{3:<6.2f}  {4:<6.2f}] Z={5:<.3} P-value={6:<.20} \n".format(  # for total effect
        "Random Effect Model",  # total effect size model name
        random_results[0][5],  # total N (all studies)
        random_results[0][1],  # total effect size
        random_results[0][3],  # total lower CI
        random_results[0][4],  # total higher CI
        random_results[0][10],
        random_results[0][11],
        )
    text += "\n{0:<d}studies included (N={1:<d})\n".format( len(fixed_results) - 1, fixed_results[0][5])
    text += "\nQuantifying Heterogeneity: \n" "Tau\u00b2={0:.3f} I\u00b2={1:s}\n".format(
        random_results[0][12],
        str(round(random_results[0][9], 2)) + "%")

    text += "\nTest of heterogeneity:\n" "Q={0:.2f} p-value={1:s} \n".format(

        fixed_results[0][7],  # Q test value
        fixed_results[0][8],  # p value for Q test
       )
    text+= "\n{0:s}\n {1:s}\n {2:s}\n {3:s}".format("Details on meta-analysis method:",
                                                    "-Mantel-Haenszel method",
                                                    "-DerSimonian and Laird for Tau\u00b2",
                                                    "-Continuity correction of 0.5 in studies with zero cell")
    return text
def main(setting):
    d = PMA.Data()  # Load Data class
    m = PMA.Meta()  # Load Meta class
    d.datatype = setting["datatype"]  # set data type, 'CATE' for binary data or 'CONT' for continuous data
    studies = d.getdata(d.readfile("Haloperidol.txt"))  # get data from a data file
    m.subgroup = d.subgroup
    m.datatype = d.datatype  # set data type for meta-analysis calculating
    m.models = setting["models"][0]  # set effect models: 'Fixed'
    m.algorithm = setting["algorithm"]  # set algorithm, based on datatype and effect size
    m.effect = setting["effect"]  # set effect size:RR/OR/RD for binary data; SMD/MD for continuous data
    results = m.meta(studies)  # performing the analysis
    m.models = setting["models"][1]  #set effect models: Random"
    res_random = m.meta(studies)
    print(showresults(results, res_random))

if __name__ == '__main__':
    settings = {"datatype": "CATE",  # for CATEgorical/count/binary/dichotomous data
                "models": ["Fixed", "Random"],  # models: Fixed or Random
                "algorithm": "MH",  # algorithm: MH, Peto or IV
                "effect": "RR"}  # effect size: RR, OR, RD
    main(settings)
```

```python
def showstudy(studies):
    text = "{0:22}{1:20}\n".format( "Haloperidol ", " Placebo")
    text += "{0:80}{1:10}{2:10}{3:10}{4:10}\n".format("   ","  Events","Total","Events","Total")
    hal=0
    pla=0
    for i in range(len(studies)):
        text += "{0:10}{1:13}{2:10}{3:10}\n".format(
        str(studies[i][0]),   #event num of group1
        str(studies[i][1]),   #total num of group1
        str(studies[i][2]),   #event num of group2
        str(studies[i][3]),)   #total num of group2
    text += "{0:25}{1:10}".format("446","372")

    return text

def studyID(studies,random_results):
    text="{0:23}\n".format( "Study ID")
    for i in range(len(studies)):
        text += "{0:10}\n".format(studies[i][4])        #study ID
    text+="Fixed effect Model\n""Random effect Model\n""Tau\u00b2={0:.3f} I\u00b2={1:s} p-value={2:s}".format(
     random_results[0][12],
     str(round(random_results[0][9], 2)) + "%",
     random_results[0][11]
     )
    return text

#Fixed and randome effect weights

def descripweight(fixed_results,random_results):
    text = "{0:<16} {1:<}".format("  Fixed Wt(%)" ,"Random Wt(%)\n")
    for i in range(1, len(fixed_results)):
        text +="{0:15.2f}  {1:15.2f}\n".format(
                                100*(fixed_results[i][2]/fixed_results[0][2]),   # Fixed weight

                                100*(random_results[i][2]/random_results[0][2])) # Random weight
    return text

#Show effect size and 95%confidence interval
def descrip(fixed_results,random_results):
    text = "{}".format("   ES [95% CI]\n")
    for i in range(1, len(fixed_results)):
        text +="{:8.2f}  [{:.2f}   {:.2f}]\n" .format(
                                fixed_results[i][1],   # effect size
                                fixed_results[i][3],   # lower of CI
                                fixed_results[i][4],)   # higher of CI
    text += "  {0:5.2f}  [{1:.2f}  {2:6.2f}]{3:>18} {4:>20}\n  {5:5.2f}  [{6:.2f}  {7:6.2f}]{8:>18} {9:>20}".format(

        # for total effect
        fixed_results[0][1],  # total effect size
        fixed_results[0][3],  # total lower CI
        fixed_results[0][4],  # total higher CI
        "100",
        "----",
        random_results[0][1],  # total effect size
        random_results[0][3],  # total lower CI
        random_results[0][4],  # total higher CI
        "----",
        "100",
        )

    return text


#calculating prediction intervals ,it is not included in the package
def tausq(random_results):   #Tausquared
    tausq=(random_results[0][12])
    return tausq

def weight(rults):  #weight
    weight=[]
    for i in range(1,len(rults)):
        W=rults[i][2]
        weight.append(W)
    print(weight)

def totalweight(rults):  #total weight
    TW=rults[0][2]
    return TW

def Yi(rults):  #log effect size
    logES=[]
    for i in range(1,len(rults)):
        Yi=math.log(rults[i][1])
        logES.append(Yi)
    print(logES)

def VM(rults):
    VM=1/rults[0][2]
    return VM

def totalWY(rults): #total weight*log effect size
    totalWY=0
    for i in range(1,len(rults)):
        WY=(rults[i][2])*(math.log(rults[i][1]))
        totalWY+=WY
    return totalWY


def M(rults):  #summary mean
    M=totalWY(rults)/totalweight(rults)
    return M

def PI(rults):
    K=(len(rults)-1)
    df=K-2
    t=scipy.stats.t.ppf(q=1-.05/2,df=df)
    MR=M(rults)
    Tausq=tausq(rults)
    VMR=VM(rults)
    UpperPI=math.exp(MR+t*(math.sqrt(Tausq+VMR)))
    LowerPI=math.exp(MR-t*(math.sqrt(Tausq+VMR)))
    return ("Prediction Interval:"+ str([float("{:.2f}".format(LowerPI)),float("{:.2f}".format(UpperPI))]))


def main(setting):
    d = PMA.Data()  # Load Data class
    m = PMA.Meta()  # Load Meta class
    f = PMA.Fig()  # Load Fig class


    d.datatype = setting["datatype"]
    studies = d.getdata(d.readfile("Haloperidol.txt"))
    m.subgroup = d.subgroup
    m.datatype = d.datatype  # set data type for meta-analysis calculating
    m.models = setting["models"][0]  # set effect models: 'Fixed'
    m.algorithm = setting["algorithm"]  # set algorithm, based on datatype and effect size
    m.effect = setting["effect"]  # set effect size:RR/OR/RD for binary data; SMD/MD for continuous data
    results = m.meta(studies)  # performing the analysis
    m.models = setting["models"][1]  #set effect models: Random"
    res_random = m.meta(studies)
    f.forest(res_random)                    #show forest plot
    plt.text(6,1.5, descrip(results,res_random),fontsize=8, style='normal', ha="left",
         va='bottom',multialignment="left",linespacing=2.55,weight="bold")

    plt.text(10,2.5,descripweight(results,res_random),fontsize=8, style='normal', ha="left",
         va='bottom',multialignment="left",linespacing=2.55,weight="bold")

    plt.text(8.3,0.1, PI(res_random),fontsize=8, style='normal', ha="right",
         va='bottom', multialignment="left",linespacing=2.5,weight="bold")
    plt.plot((math.log(res_random[0][1]),0),((res_random[0][1]),(len(res_random)+200)),linestyle="--", lw=1)
    plt.yticks([])
    plt.text(-13,0.8,studyID(studies,res_random),multialignment="left",fontsize=8,weight="bold",linespacing=2.5)
    plt.text(-18,2.8,showstudy(studies),multialignment="right",fontsize=8,weight="bold",linespacing=2.5)
    plt.title("Risk Ratio",loc="center")
    plt.xlabel("{:<38}{:<s}".format("  Favours Placebo","Favours Haloperidol"),fontsize=10)



    #draw PI
    K=(len(res_random)-1)
    df=K-2
    t=scipy.stats.t.ppf(q=1-.05/2,df=df)
    UpperPI=(M(res_random)+t*(math.sqrt(tausq(res_random)+VM(res_random))))
    LowerPI=(M(res_random)-t*(math.sqrt(tausq(res_random)+VM(res_random))))
    plt.plot((LowerPI,UpperPI),(0.4,0.4),color="black",lw=3)

    #the total effect size diamond for fixed effect model
    x=[math.log(results[0][1]),
           math.log(results[0][3]),
           math.log(results[0][1]),
           math.log(results[0][4]),
           math.log(results[0][1])]
    y=[3.3,3,2.7,3,3.3]
    if (results[0][9]<50) :
        plt.fill(x,y,color="black", lw=1)  #filled: I2<50
    else :
        plt.plot(x,y, 'black', lw=1)       #empty: I2>50
    ax=plt.gca()
    ax.spines['top'].set_visible(False)

if __name__ == '__main__':
    settings = {"datatype": "CATE",  # for CATEgorical/count/binary/dichotomous data
                "models": ["Fixed", "Random"],  # models: Fixed or Random
                "algorithm": "MH",  # algorithm: MH, Peto or IV
                "effect": "RR"}  # effect size: RR, OR, RD
    main(settings)
```

*   **Почему и зачем**: **Эффект малых исследований** (small-study effects) возникает, когда малые исследования показывают иной, часто больший, эффект лечения по сравнению с крупными исследованиями, что может быть признаком публикационного смещения. Этот шаг позволяет оценить наличие такого эффекта и его потенциальное влияние на обобщенные результаты.

```python
def logES(rults):
    logES=[]
    for i in range(1,len(rults)):
        Yi=math.log(rults[i][1])   # effect size:rults[i][1]
        logES.append(Yi)
    return logES

#this is the standard error of the studies.In random effect models with DL method the standard error is  the square root of the the variance and the variance is the reciprocal of the weights.So in random models we should use SE(rults)as the standard error
#it can be used in IV fixed model as well
def SE(rults):
    SE=[]
    for i in range(1,len(rults)):
        VM=math.sqrt(1/rults[i][2])
        SE.append(VM)
    return SE

#to obtain the standard error in FIXED MH,IV method,we should use this function as standard error in the plot
def SEFix(rults):
    SE=[]
    for i in range(1,len(rults)):
        se=rults[i][6]
        SE.append(se)
    return SE


#create funnel plot and use log effect size on the X axis
def funnelplot(rults):
    fig=plt.figure(figsize=(9,8))
    x=logES(rults)
    y=SEFix(rults)


    plt.plot(x,y,"o",lw=1)

    plt.xlim(min(x)-4,max(x)+4)  #the xlim can change manually
    plt.ylim(max(y)+0.04,0)
    ymax=max(y)+0.1
    tes=(rults[0][1])  #total effect size
    logtes=math.log(tes)

    plt.plot ([logtes,logtes],[0,max(y)],color="blue", linestyle="--", lw=1)
    plt.plot ([logtes,logtes-(1.96*max(y))],[0,max(y)],color="blue", linestyle="--", lw=1)
    plt.plot ([logtes,logtes+(1.96*max(y))],[0,max(y)],color="blue", linestyle="--", lw=1)
    ax = gca()

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.ylabel("Standard Error",fontsize=14,rotation=90 )
    plt.xlabel("Favours Placebo logRR Favours Haloperidol",fontsize=10 )



def main(setting):
    d = PMA.Data()  # Load Data class
    m = PMA.Meta()  # Load Meta class
    f = PMA.Fig()  # Load Fig class


    d.datatype = setting["datatype"]
    studies = d.getdata(d.readfile("Haloperidol.txt"))
    m.datatype = d.datatype
    m.models = setting["models"]
    m.algorithm = setting["algorithm"]
    m.effect = setting["effect"]
    results = m.meta(studies)
    funnelplot(results)                 #This is the funnel plot we created which seems more correct
    #f.funnel(results)                  #This is the original funnel plot of the package

if __name__ == '__main__':
    settings = {"datatype": "CATE",  # for CATEgorical/count/binary/dichotomous data
                "models": "FIXED" , # models: Fixed or Random
                "algorithm": "MH",  # algorithm: MH, Peto or IV
                "effect": "RR"}  # effect size: RR, OR, RD

    main(settings)

#Contour enhanced plot
def CEfunnelplot(rults):      #to draw contour enhanced funnel plot

    fig=plt.figure(figsize=(9,8))
    ax=gca()
    x=logES(rults)
    y= SEFix(rults)

    plt.xlim(min(x)-5,max(x)+2)  #the xlim can change manually
    plt.ylim(max(y)+0.1,0)
    ax.set_facecolor(color="whitesmoke")
    tes=(rults[0][1])
    logtes=math.log(rults[0][1])

    plt.plot ([0,(-2.58*max(y))],[0,max(y)], linestyle="None", lw=1)
    plt.plot ([0,(2.58*max(y))],[0,max(y)], linestyle="None", lw=1)

    trianglex = [ -2.58*max(y), 0, 2.58*max(y)]
    triangley = [ max(y),0, max(y)]
    for i in range(3):
        plt.plot(trianglex, triangley,color="lightgray")
    plt.fill(trianglex, triangley,"lightgray")

    plt.plot ([0,(-1.96*max(y))],[0,max(y)], linestyle="None", lw=1)
    plt.plot ([0,(1.96*max(y))],[0,max(y)], linestyle="None", lw=1)

    trianglex = [ -1.96*max(y), 0,1.96*max(y)]
    triangley = [ max(y),0, max(y)]
    for i in range(3):
        plt.plot(trianglex, triangley,color="darkgrey")
    plt.fill(trianglex, triangley,"darkgrey")

    plt.plot ([0,(-1.65*max(y))],[0,max(y)], linestyle="None", lw=1)
    plt.plot ([0,(1.65*max(y))],[0,max(y)], linestyle="None", lw=1)

    trianglex = [ -1.65*max(y), 0,1.65*max(y)]
    triangley = [ max(y),0, max(y)]
    for i in range(3):
        plt.plot(trianglex, triangley,color="white")
    plt.fill(trianglex, triangley,"white")

    plt.plot ([logtes,logtes],[0,max(y)],color="blue", linestyle="--", lw=1)
    plt.plot ([logtes,logtes-1.96*max(y)],[0,max(y)],color="blue", linestyle="--", lw=1)
    plt.plot ([logtes,logtes+1.96*max(y)],[0,max(y)],color="blue", linestyle="--", lw=1)
    plt.plot(x,y,"o",lw=1,color="k")

    ax = gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    colors = ['whitesmoke', 'lightgray', 'darkgrey',"white"]
    lines = [Line2D([0], [0], color=c, linewidth=3, linestyle='None', marker="s") for c in colors]
    labels = ["p < 1%","1%< p <5%","5%< p <10%","p > 10%"]
    plt.legend(lines, labels,shadow=True)

    plt.xlabel("logRR",fontsize=14 )
    plt.ylabel("Standard Error",fontsize=14,rotation=90 )
    plt.savefig("Contour Enhanced Funnel plot.jpg")

def main(setting):
    d = PMA.Data()
    m = PMA.Meta()
    d.datatype = setting["datatype"]
    studies = d.getdata(d.readfile("Haloperidol.txt"))
    m.datatype = d.datatype
    m.models = setting["models"]
    m.algorithm = setting["algorithm"]
    m.effect = setting["effect"]
    results = m.meta(studies)
    CEfunnelplot(results)

if __name__ == '__main__':
    settings = {"datatype": "CATE",  # for CATEgorical/count/binary/dichotomous data
                "models": "FIXED" , # models: Fixed or Random
                "algorithm": "MH",  # algorithm: MH, Peto or IV
                "effect": "RR"}  # effect size: RR, OR, RD

    main(settings)
```
