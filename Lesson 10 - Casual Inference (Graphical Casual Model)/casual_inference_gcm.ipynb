### Модель генерации данных

Один из способов понимания науки в целом заключается в том, что учёные верят в существование регулярных законов, управляющих работой Вселенной.


Эти законы являются примером «процесса генерации данных» (DGP). Законы работают за кулисами, делая то, что они делают, независимо от того, знаем мы о них или нет. Мы не можем видеть их напрямую, но мы видим данные, которые они получают. Мы видим, что если отпустить мяч, он падает на землю. Это наше наблюдение, наши данные. Гравитация — часть процесса генерации данных для этого мяча. Это основополагающий закон.

Мы довольно хорошо представляем себе, как работает гравитация. По сути, если у нас есть два объекта, находящихся на расстоянии друг от друга, один с массой $m_1$ и другой с массой $m_2$, то сила $F$, притягивающая их друг к другу, равна
$$
F = G \frac{m_1 m_2}{r^2}
$$

Это уравнение — физический закон - процесс, происходящий за кулисами и определяющий движение объектов, независимо от того, знаем мы о нём или нет.

Помимо того, что существуют такие фундаментальные законы, учёные также считают, что мы можем узнать об этих законах из эмпирических наблюдений. Мы не всегда знали об уравнении гравитации — Исааку Ньютону пришлось его вывести. Но до того, как у нас появилось уравнение Ньютона, у нас были данные. Огромное количество данных о движущихся объектах, в частности, данные о движении планет по орбитам.

Наука утверждает: «Мы видим, как эти планеты движутся. Мы знаем, что должен быть какой-то закон, объясняющий, почему они движутся именно так. Держу пари, мы можем использовать то, что видим, чтобы понять этот закон». И мы могли бы! Или, по крайней мере, Ньютон мог. Учитывая наблюдение, что планеты имеют эллиптические орбиты, Ньютон показал, что уравнение выше может генерировать эти орбиты.

Мы можем сделать вывод о том, как устроен мир, основываясь на наших данных, поскольку такие вещи, как описательные уравнения, являются частью процесса генерации данных (DGP). Если в космосе движутся две планеты, их движение фактически определяется этим уравнением, и мы наблюдаем их движение как данные. Если бы мы не знали уравнения, то, имея достаточно данных для наблюдения за реальным движением и достаточного понимания других частей DGP, чтобы мы могли выделить влияние таких факторов, как импульс, и определить влияние гравитации.

DGP в социальных науках, как правило, не так хорошо себя ведут и точны, как в физических науках. Тем не менее, если мы считаем, что данные наблюдений основаны хотя бы на некоторых регулярных законах, мы говорим о наличии DGP.

DGP особенно важно, когда дело доходит до вопросов причинно-следственных связей: хорошая модель генерации данных отличается не только наличием ключевых аспектов процесса, так и верным указанием связей между ними. Более того, если были заданы только связи, то модель способна отвечать максимум на вопросы **"ассоциаций" (assosiation)**, для **причинно-следственности (causation)** дополнительно необходимы и направления связей. Причинно-следственность является более сложной задачей, так как необхоимо ещё и понимание о природе взяимосвязей параметров модели (дует ли ветер от того, что шатаются деревья или наоборот?)

Обычно DGP принято представлять в виде графов, где вершины - параметры, которые участвуют в процессе генерации данных, а ребра - взаимосвязи между ними.

Вот один из примеров такой модели для "влияния интернет образования на отчисление со школы" (переменными $U$ обычно обозначают неизвестные параметры (то есть знаем, что то-то влияет, но либо не знаем, что конкретно, либо собирем в 1 переменную сразу много-много факторов)

### Направленный Ациклический Граф

Ориентированный граф $G$ представляет собой множество вершин, соединенных стрелками. Математически это множество вершин $V$ и множество ребер $E$ упорядоченных пар вершин, где порядок соответствует направлению стрелки. Каждая вершина или узел будет соответствовать случайной величине. Если упорядоченная пара $(X,Y) \in E$, то существует стрелка, направленная из $X$ в $Y$, и мы говорим, что $X$ является родительским элементом $Y$, а $Y$является дочерним элементом $X$. Ориентированный путь между двумя вершинами — это путь, все стрелки которого указывают в одном направлении.
$Z$ называется потомком $X$, если существует ориентированный путь, ведущий из $X$ в $Z$. Ориентированный путь, начинающийся и заканчивающийся в одной и той же вершине, называется циклом. Ориентированный граф ацикличен, если в нем нет циклов. Ориентированный граф $G$ без цикла называется **ориентированным ациклическим графом (DAG)**.

Здесь мы запрещаем любые циклы  и сосредоточимся только на DAG. DAG $G$ с вершинами $V = (V_1, \cdots, V_k)$ часто ассоциируется с марковской моделью (байесовкой сетью). Если $P$ — распределение над $V$ с функцией вероятности $f$, мы говорим, что $P$ является **марковским** для $G$, или что $G$ представляет $P$, если

$$
f(v) = \prod_i f(v_i \mid \pi_i) \quad \pi_i \text{ - родительские ноды } V_i
$$

Когда распределение $P$ является марковским для DAG $G$, граф кодирует часть структур условной независимости. Это означает, что мы можем прочитать условные независимости в структуре графа, и эти условные независимости должны быть истинны для распределения.

Тип условных отношений независимости, который может быть выражен графом, наиболее важен, поскольку он требует только факторизации распределения без дополнительных предположений о конкретных факторах $f(v_i \mid \pi_i)$.

Свойство факторизации эквивалентно следующему направленному **локальному марковскому свойству (LMP)**: распределение $P$ является марковским для DAG $G$ тогда и только тогда, когда для любой переменной $V$

$$
V ⊥ nd(V) \mid \pi_V
$$

$\pi_V$ — родительские переменные $V$, а $nd(V)$ — все остальные переменные, не являющиеся потомками $V$.

Это подтверждает, что $V$ может влиять только на своих потомков, а также его зависимость от предков осуществляется только через его прямых родителей
$\pi_V$. Таким образом, если $\pi_V$, $V$ независим от своих не-потомков.

### Идентификация

Абель и Энни Васкес проснулись во вторник утром и обнаружили, что их верного пса Рекса нет у изножья кровати, как обычно. В отчаянии они забегали по дому, звали его по имени, свистели, предлагали лакомства. Наконец, один из них открыл входную дверь, и Рекс сидел на траве, виляя хвостом. Они привели его обратно.

«Я всё время говорю тебе, чтобы ты закрыла собачью дверь на защёлку», — сказал Абель.

«Он всё равно никогда не пользуется этой дверью. Держу пари, он выпрыгивает из открытого окна в подвале», — сказала Энни.

«Конечно», — сказал Абель, закатив глаза. В ту ночь он обязательно закрыл собачью дверь на защёлку.

Абель и Энни Васкес проснулись в среду утром. Солнце грело, запахи кофемашины уже доносились до их спальни, а Рекс исчез.

На этот раз им не потребовалось много времени, чтобы найти его снова на лужайке, купающегося в росе. Маленький клочок шерсти был стерт до дыр – как бы он ни выбирался, каждую ночь он явно делал это одинаково.

В ту ночь они заперли собачью дверцу на засов и заднюю дверь на два засова. Утром в четверг Рекса не стало.

В четверг вечером они заперли собачью дверцу на засов, заднюю дверь на два засова и закрыли жалюзи. В пятницу Рекса не стало.

В пятницу вечером они заперли собачью дверцу на засов, заднюю дверь на два засова, закрыли жалюзи и заблокировали вентиляционное отверстие. В субботу утром Рекс съел двух бабочек, прежде чем они нашли его на улице.

«С меня хватит», – сказал Абель. В ту ночь они заперли и закрыли все двери, все окна, заколотили камин, заделали щель между досками в гараже, заткнули слив в ванной и устроили Рексу очень строгий выговор.

Воскресным утром Энни спустилась вниз и обнаружила расстроенного Абеля, сидящего на крыльце и смотрящего на свою собаку, которая снова сбежала из дома.

«Не понимаю… Я перекрыла все возможные пути его бегства. Все щели, все щели, все дыры. Все возможные выходы были заблокированы. Но вот он!»

Энни выпила кофе. «О, я открыла окно в подвале перед сном. Я же говорила, что он выберется оттуда».

На следующую ночь они оставили всё остальное открытым, но закрыли окно в подвале. На следующее утро, наконец, наконец, наконец, они проснулись, и Рекс всю ночь просидел дома.

**Идентификация** — это процесс выяснения того, какая часть вариации в данных отвечает на ваш исследовательский вопрос. Мы называем это идентификацией, потому что мы гарантируем, что наши расчёты выявляют единственный интересующий нас теоретический механизм. Другими словами, мы преодолели все вводящие в заблуждение подсказки и определили виновника.

Исследовательский вопрос ведёт нас от теории к гипотезе, гарантируя, что проверяемая гипотеза действительно расскажет нам что-то о теории. Идентификация ведёт нас от гипотезы к данным, гарантируя, что у нас есть способ проверить эту гипотезу на основе данных, а не случайно проверить какую-то другую.

Возьмём случай Абеля и Энни. Что мы видим в реальных данных? Мы видим, что в разные ночи разные части дома опечатываются, и мы также видим, что Рекс каждую ночь выходит наружу. Исследовательский вопрос Энни заключается в том, является ли открытое окно подвала причиной выхода Рекса.

Мы — и Абель — знаем, что существует множество способов выбраться из дома. Итак, одни только эти данные не дают ответа на исследовательский вопрос Энни. По мере того, как Абель закрывает всё больше и больше вещей, он отбрасывает часть вариаций. Каждый раз, когда он это делает, количество мест, где Рекс мог бы выбраться, помимо окна, сокращается, поэтому у Энни остаётся всё меньше и меньше вариантов, чтобы ошибиться насчёт окна, но при этом остаются другие объяснения того, как собака выбралась.

Но как только мы перекрыли буквально все остальные возможные пути выхода из дома, и окно стало единственным, что осталось, остаётся только один вывод: именно открытое окно позволило собаке выскочить. Это наблюдение дало ответ на наш вопрос, просто исключив все остальные возможные объяснения.

Процесс исключения других возможных объяснений также помогает нам находить ответы на исследовательские вопросы в эмпирических исследованиях.

Идентификация требует статистических процедур для надлежащего устранения нежелательных колебаний. Но, что не менее важно, она опирается на теорию и предположения о том, как устроен мир, чтобы выяснить, каковы эти нежелательные объяснения и какие статистические процедуры необходимы. В частности, она опирается на теорию и предположения о том, как выглядит DGP. Нам необходимо сделать заявление о том, что мы уже знаем, чтобы иметь хоть какую-то надежду узнать что-то новое.

Не слишком ли важно осознавать, что нам приходится делать некоторые предположения для определения наших ответов? Но это необходимо. Представьте, что Авель и Энни попытались бы выяснить, как собака выбралась, не имея никаких предположений или знаний о процессе генерации данных. Список альтернативных объяснений, с которыми им пришлось бы иметь дело, включал бы: «собака может телепортироваться», «вселенная – это голограмма» и «в наших стенах есть невидимые отверстия, которые видят только собаки». Они так и не пришли бы к какому-либо выводу и серьёзно рисковали бы стать философами.
Итак, вот наша цель. Если мы хотим определить ту часть наших данных, которая даёт ответ на наш исследовательский вопрос, мы должны:

1. Используя теорию и то, что мы уже знаем о мире из предыдущих исследований или наблюдений, нарисовать максимально точную картину того, как выглядит DGP;
2. Использовать этот DGP, чтобы выяснить причины, по которым наши данные могут выглядеть именно так, но не отвечать на наш исследовательский вопрос;
3. Найти способы исключить эти альтернативные причины и таким образом выявить необходимые нам вариации.

Если мы хотим проследить связь выпивания вина и продолжительности жизни, то по данной модели нас будут интересовать следующие пути:

* Wine $\rightarrow$ Lifespan
* Wine $\rightarrow$  Drugs $\rightarrow$ Lifespan

И каждый из этих путей содержит свою историю — вот почему мы видим связь между вином и продолжительностью жизни в данных. Вино может влиять на употребление наркотиков, что влияет на продолжительность жизни (Wine $\rightarrow$  Drugs $\rightarrow$ Lifespan). Или люди, которые пьют вино, как правило, богаче, а более богатые люди живут дольше (Wine $\leftarrow$ Income $\rightarrow$ Lifespan). Или люди, которые пьют вино, как правило, богаче, а более богатые люди в среднем имеют лучшее здоровье, а здоровье увеличивает продолжительность жизни (`Wine $\leftarrow$ Income $\leftarrow$ U1 $\rightarrow$ Health $\rightarrow$ Lifespan). История для каждого пути. Но какие истории важны для нас?

Теперь, когда у нас есть список путей, что мы можем с ними сделать?

Я обещал вам, что мы сможем использовать эти пути, чтобы найти ответ на ваш исследовательский вопрос. И мы можем это сделать. Следующий шаг в использовании найденных нами путей — выяснить, как именно они соотносятся с вашим исследовательским вопросом.

Это подводит нас к концепции «хороших» и «плохих» путей.

\\

**Хорошие пути** — это причины, по которым переменные «влияния» и «результат» связаны и которые, по вашему мнению, должны учитываться для вашего исследовательского вопроса.

**Плохие пути** — это пути, которые не должны учитываться, другими словами, альтернативные объяснения.

Обычно это выглядит так: «каждый путь, на котором все стрелки направлены от влияния, является «хорошим путем», а все остальные — «плохим путем». Пути, на которых все стрелки направлены от влияния, также называются **front door**. Пути, на которых хотя бы одна стрелка указывает на влияние, являются **back door**. Поэтому обычно все пути front door являются «хорошими», а все back door — «плохими».

В качестве примера, на DAG'е выше есть два пути, на которых все стрелки направлены от воздействия: `Wine  →  Lifespan` и `Wine  →  Drugs  →  Lifespan`. Это все способы, которыми изменение «вина» может привести к изменению «продолжительность жизни». Итак, это наши «пути парадного входа». И если наш исследовательский вопрос звучит так: «Влияет ли вино на продолжительность жизни?», то эти front door — хорошие пути, а другие — плохие.

Всё может быть не так просто, в зависимости от вашего исследовательского вопроса. Например, если вместо вопроса «Влияет ли вино на продолжительность жизни?» вы хотите узнать: «Влияет ли вино на продолжительность жизни, помимо того, как оно может влиять на приём лекарств?»

С этим исследовательским вопросом путь `Wine  →  Drugs  →  Lifespan` внезапно становится для нас неинтересным. Если посмотреть на исходные данные, одна из причин, по которой вино и продолжительность жизни связаны, — это наркотики. Но эта часть эффекта нас больше не интересует! В этот момент мы можем считать это плохим путём.

Подобные исследовательские вопросы, где единственный хороший воздействие → результат, исследуют «прямые эффекты». Они не интересуются «косвенными эффектами», которые идут окольными путями через такие переменные, как наркотики.

Итак, у нас есть все пути, и мы выяснили, какие из них хорошие, а какие плохие. Ключ к ответу на наш исследовательский вопрос — найти способ извлечь из данных только хорошие пути, не отвлекаясь на плохие.

Наличие пути на диаграмме означает, что между переменными в начале и конце пути существует связь, объясняемая переменными, расположенными вдоль этого пути. `Wine  →  Drugs  →  Lifespan` означает, что вино и продолжительность жизни будут связаны частично из-за того, как вино влияет на наркотики. `Wine  →  Income  →  Lifespan` означает, что вино и продолжительность жизни будут связаны частично из-за того, как доход влияет как на вино, так и на продолжительность жизни.

Однако это верно только для **открытого пути**. Путь считается открытым, если все переменные на этом пути могут варьироваться. Путь считается **закрытым**, если хотя бы одна из переменных на этом пути не имеет вариаций.

Возьмем в качестве примера `Wine  →  Drugs  →  Lifespan`. Если в нашем наборе данных есть пьющие и не пьющие вино, употребляющие и не употребляющие наркотики, а также люди с более короткой и более длинной продолжительностью жизни, то все переменные на этом пути имеют вариации. Их значения различаются.

Но что, если мы выберем набор данных, в котором никто не употребляет наркотики? В этом случае вариаций по наркотикам нет, и, следовательно, никакая связь между вином и продолжительностью жизни не может быть обусловлена наркотиками. Путь закрылся.

В этом и заключается основная идея: пути становятся закрытыми и, следовательно, больше не представляют угрозы для нашей идентификации, если мы можем удалить всю вариацию, обусловленную переменной на этом пути. Выбор выборки, в которой нет вариации по этой переменной, как мы сделали с выборкой, в которой никто не употреблял наркотики, — один из способов сделать это, хотя и не самый лучший. Есть и другие способы.

Пока что предположим, что у нас есть какой-то способ контролировать переменную. Другими словами, у нас есть способ статистически скорректировать наши данные, чтобы удалить всю вариацию по переменной и, таким образом, закрыть путь с этой переменной. Сохраняя эту переменную постоянной и неизменной, если хотите. Мы также могли бы сделать это, как только что упоминалось, выбрав набор данных, в котором переменная не варьируется. Или выбрав набор необработанных наблюдений с очень близкими (или абсолютно одинаковыми) значениями этой переменной.

**Если мы можем контролировать хотя бы одну переменную на каждом из наших «плохих» путей, не контролируя ничего на одном из «хороших» путей, мы нашли ответ на наш исследовательский вопрос.**

Чтобы понять, как контролировать переменные (математически строже говорить идентифицировать условную и безусловную зависимость переменных), нам необходимо ввести понятие **d-separation**, но прежде пока поговорим о том, какие бывают минимально интересные структуры в таком графе.

### Типы структур

Без разницы каким сложным не был бы DAG, он всегда состоит из комбинации 3 базовых структур, которые нам понадобятся:

1. **Chain**

Типичным примером такой структуры является `Mediation`
image.png

В цепочке вся информация из A для прогнозирования C была включена в B. Мы имеем следующую условную независимость $A⊥C|B$.
.
2. **Fork**

image.png

В разветвлении обусловливание общего конфаундера $A$, $B$ и $C$ условно независимы. Но они не являются предельно независимыми. Другими словами, $B⊥C|A$ но $B⊥C$, как правило, неверно.

3. **Collider**

image.png

В коллайдере $A$ и $B$ независимы, а распределение $C$ зависит от значений $A$ и $B$ вместе.

Эта закономерность наиболее интересна, поскольку даёт результат, несколько противоречащий интуиции: $A$ и $B$ независимы при условии наличия $C$ хотя без наблюдения $C$ они независимы. (так называемый [парадокс Берксона](https://ru.wikipedia.org/wiki/Парадокс_Берксона)). Чтобы это осознать, можно подумать о следующем примере: мы знаем, что можно поступить в ВУЗ как по знаниям (а-ля знание предмета), так и по заслугам (напрмиер, высокие достижения в спорте). Тогда можем представить эту модель как `Знание математики` → `Поступление в ВУЗ` ← `Умение играть в баскетбол`. Сами по себе `Знание математики` и `Умение играть в баскетбол` вроде как независимы, но как только мы включаем 3-ю переменную, то они обретают условную зависимость от неё. Упрощенно, если мы знаем исход `Поступление в ВУЗ` и значение какой-то одной из переменной, то мы можем предсказать, какое значение принимает вторая переменная.


### $d$-separation

Говорят, что множество вершин $S$ блокирует путь $p$ если выполняется одно из следующих двух условий:

1. $p$ содержит хотя бы один узел, выпускающий стрелку (средний узел в chain или ), который находится в $S$.
2. $p$ содержит хотя бы один узел коллизии (средний узел в шаблоне коллайдера), который находится вне
$S$ и не имеет потомков в $S$.

Если $S$ блокирует все пути из $X$ в $Y$, говорят, что оно "**$d$-разделяет** $X$ и $Y$". Два множества $A$ и $B$ называются **$d$-separated** $S$, если любая пара $X$ и $Y$ из этих двух множеств $d$-разделена $S$.

В этом случае имеет место следующее направленное глобальное марковское свойство:
$$
A ⊥ B |S
$$

$S$ может быть пустым множеством, и независимость становится безусловной.

Первое условие $d$-разделения, по сути, утверждает, что информация может передаваться по chain или fork, и единственный способ заблокировать её — это наложить условие на узел, испускающий стрелку. Второе условие несколько противоположно: оно гласит, что для шаблона коллайдера информация естественным образом блокируется, если только не наблюдается узел коллизии или любой из его потомков, что создаёт ранее описанный эффект «объяснения». Чтобы сделать два набора переменных $X$ и $Y$
независимыми, нам нужно наложить условие на блокирующий набор $S$, который блокирует все пути от $X$ до $Y$.

Таким образом, $d$-разделение становится очень естественным и интуитивно понятным, как только мы освоим три основных шаблона и поймём частный случай шаблона коллайдера.

### $do$-оператор

Поскольку факторизация распределения $P$ кодирует локальное влияние случайных величин друг на друга, мы можем напрямую анализировать распределение вероятностей, сгенерированное вмешательством. Мы называем это сгенерированное распределение распределением после вмешательства. Операция вмешательства называется $do$-оператором.

Если мы знаем, как работать с $do$-оператором, и можем вычислить
$P(Y|do(X=x))$ для всех $x$, то мы можем ответить на любой причинно-следственный вопрос об изменении $X$ на $Y$. Например, ATE изменения $X$ с $X_0$ на $X_1$ равно

$$
P(Y|do(X=x_1)) - P(Y|do(X=x_0))
$$

Вычисления с использованием $do$-оператора называются $do$-исчислением. Цель $do$-исчисления — преобразовать выражение с $do$-оператором в выражение без $do$-оператора, включающее только наблюдаемые переменные.

Чтобы работать с $do$-оператором, нам необходимо понять распределение после вмешательства. Следующая теорема описывает распределение после вмешательства $P(V|do(X=x))$, т.е. новое распределение вероятностей управляет $V$ после того, как мы вручную зафиксируем
$X$ равным $x$.

Для любого распределения $P$ (с функцией вероятности $f$) Маркова к DAG $G$, распределение после вмешательства $P^∗$ сгенерированное вмешательством $do(X=x)$ имеет свою функцию вероятности $f^∗$
задаваемую усеченной факторизацией:

$$
f^∗ (v_1, \cdots, v_k) = \prod_{i \mid V_i \notin X} f(v_i \mid \pi_i) \mid _{X=x}
$$

где $f(v_i \mid \pi_i) \mid _{X=x}$ исходная условная вероятность $v_i$
при наличии его родителей $\pi$ с любыми родителями в $X$ фиксируется в соответствии с вмешательством $X=x$.

Графически, эффект $do(X=x)$ заключается в удалении всех стрелок от родительских узлов $X$ к $X$ в $G$, а затем в выполнении условия $X=x$. Это же условие эквивалентно следующему:

$$
P (v_1, \cdots, v_k \mid do(X=x)) = \frac{P (v_1, \cdots, v_k)}{P(X=x \mid \pi_X)}
$$

где $P(X=x|pi_X)$  - удалённое ребро.

Проблема причинно-следственной связи в DAG, по-видимому, решена, поскольку мы вывели распределение после вмешательства, используя только факторы $f(v_i \mid \pi_i)$ исходного распределения. Зная эти факторы, мы можем легко выполнить любое $do$-исчисление.

Концептуально ответ — да. Но на практике эта формула сама по себе редко может быть полезна в качестве формулы для $P(Y|do(X=x))$. Чтобы использовать усеченную факторизацию, нам необходимо наблюдать все случайные величины в DAG или, по крайней мере, подграф, содержащий всех предков и потомков $X$ и $Y$. Но в реальности мы всегда ограничены количеством измерений, которые можем собрать, и в DAG почти всегда есть узлы, которые не наблюдаются. В граф добавляется множество вершин и рёбер для учёта неизвестных зависимостей и регистрации наличия определённого неизвестного фактора-конфаундера. Часто бывает так, что при изучении причинно-следственной связи между двумя узлами $X$ и $Y$ многие детали в графе нерелевантны и могут быть упрощены или проигнорированы. По этой причине желательна простая стратегия идентификации, требующая лишь небольшого набора наблюдений.

В рамках модели потенциальных исходов мы видели, что ключевым моментом является задание набора ковариатов $X$, который удовлетворяет условию отсутствия конфаундеров (~рандомизация в A/B тестах). В этой парадигме, с использованием DAG и пониманием распределения после вмешательства, аналогичная идея заключается в поиске набора $S$ такого, что $P(Y|do(X=x),S=s)=P(Y|X=x,S=s)$. Это называется **back door критерием**. Другая стратегия, называемая **front door критерием**, определяет множество $S$ таким образом, что $P(Y|do(X=x),S=s)=P(Y|do(S=s))$, где $P(Y|do(S=s))$ можно легко определить.

Back door и front door критерии являются частными случаями или более общими методами, основанными на правилах выполнения операций. Эти два критерия имеют концептуальное значение и могут быть непосредственно применены для выявления причинно-следственных связей во многих простых графах. Они также помогают развить интуицию, облегчающую обсуждение более общих стратегий.

### Back door критерий

Единственное изменение, которое операция $do$ вносит в $P$ помимо приведения $X$ к фиксированному значению, — это удаление всех стрелок от родительских элементов $X$ $\pi_X$ к $X$. Обозначим распределение после вмешательства как
$P(⋅|do(X=x))$. Из альтернативной форулы $do$-оператора можно непосредственно вывести два простых факта:

1. $P(\pi_X=t|do(X=x))=P(\pi_X=t)$;
2. $P(⋅|do(X=x),\pi_X=t)=P(⋅|X=x,\pi_X=t)$, где $P(⋅|do(X=x),\pi_X=t)$ — условное распределение распределения после вмешательства с учетом $\pi_X=t$:
$$
P(⋅|do(X=x),\pi_X=t)= \frac{P(⋅,\pi_X=t |do(X=x))}{P(\pi_X=t|do(X=x))}
$$

Первый результат тривиально подразумевает, что операция $do$ не влияет на родительские элементы. Второй результат говорит, что если мы дополнительно условим $\pi_X$, то условное распределение
$P(⋅|do(X=x),\pi_X=t)$ будет таким же, как
$P(⋅|X=x,\pi_X=t)$, поскольку наличие стрелок между $\pi_X$ и $X$ не имеет значения, когда нам уже известны значения $\pi_X$ и $X$.

Если нас интересует
$P(Y|do(X=x)$, то

$$
P(Y|do(X=x)) = \displaystyle \sum_t P(Y|X=x,\pi_X=t) \times P(\pi_X=t \mid do(X=x))
$$

Вместе с
$P(Y|do(X=x),\pi_X=t)=P(Y|X=x,\pi_X=t)$ и $P(\pi_X=t|do(X=x))=P(\pi_X=t)$, мы доказали следующее:

$$
P(Y|do(X=x)) = \displaystyle \sum_t P(Y|X=x,\pi_X=t) \times P(\pi_X=t)
$$

Последнее выражение успешно превращает выражение $do$ в выражение без $do$ посредством обусловливания на множестве $\pi_X$. Все back door пути блокируются $\pi_X$, поскольку путь через $\pi_X$ может быть либо chain, либо fork, а не коллайдером. Оказывается, вместе с $d$-separation этот результат можно легко обобщить, используя всех предков $X$ в качестве корректирующего множества, используя любые множества, которые блокируют все пути бэкдора без участия потомков $X$. Это **back door критерий**.

**Back door критерий**

Набор $S$ в DAG $G$ допустим или достаточен для корректировки, если выполняются два условия:

1. Ни одна вершина в $S$ не является потомком $X$.

2. $S$ блокирует все обходные пути (все пути заканчиваются стрелкой, указывающей на $X$) из $X$ в $Y$.

Для любого допустимого набора $S$

$$
P(Y|do(X=x)) = \displaystyle \sum_t P(Y|X=x,S=s) \times P(S=s)
$$
обеспечивает стратегию идентификации для $P(Y|do(X=x))$.

В частности, когда $S$ — пустое множество, $P(Y|do(X=x))=P(Y|X=x)$.

Мы используем график ниже для иллюстрации использования back door критерия. Пунктирная линия с двумя стрелками означает, что мы не предполагаем никаких подробностей о пути между двумя вершинами, кроме их связности. В данном случае возможны два допустимых множества $S$.

Первый вариант — заблокировать $C$, учитывая, что все back door пути должны проходить через $C$ для достижения $Y$, и тот факт, что $C$ имеет исходящую стрелку к $Y$, поэтому путь через $C$ к $Y$ может быть только chain или fork.

Другой вариант — заблокировать $A$ и $B$. $B$ блокирует все back door пути через
$B$, поскольку он не является коллайдером. $A$ блокирует все back door пути через
$A$, поскольку $A$ не может быть коллайдером для путей типа $X$ ⟵ $A$ ← − − − → $C$ ⟶ $Y$.

Других возможных путей бэкдора нет, поэтому все они заблокированы.

При использовании критерия обхода, помимо очевидной необходимости наблюдения за $X$ и $Y$, нам нужно только дополнительно наблюдать за $C$ или $A$ и $B$, чтобы определить причинно-следственное влияние $X$ на $Y$.


### Front door критерий

Back door критерий не работает напрямую для следующей базовой модели сбивающего фактора, когда сбивающий фактор $U$ не наблюдается.

image.png

На самом деле, если это всё, что нам известно о взаимосвязи $X$ и $Y$, и мы знаем, что существует некий сбивающий фактор $U$, который не наблюдается, то мы принципиально не можем отличить истинный причинно-следственный эффект от сбивающего эффекта. Эта модель, также называемая «образцом дуги» (Pearl, 1995), лежит в основе защиты Фишером утверждения о том, что курение не обязательно вызывает рак лёгких, несмотря на то, что исследование 1950 года показало положительную связь между курением табака и раком лёгких. Он выдвинул гипотезу о сбивающем факторе (генетическом признаке), который связан с раком лёгких, но также повышает вероятность получения удовольствия от курения. Причинно-следственная связь между употреблением табака и раком лёгких была окончательно признана медицинским сообществом только после того, как было обнаружено, что остатки табака, также известные как смола, попадая в лёгкие, обволакивают реснички, вызывая их прекращение работы и, в конечном итоге, гибель, и позволяют токсичным частицам табачного дыма напрямую проникать в альвеолы. Эти токсичные частицы включают канцерогены, такие как бензол, акриламид и акрилонитрил, которые часто встречаются в смоле. Другими словами, был открыт механизм, посредством которого курение табака может вызывать рак легких, заключающийся в образовании токсичной смолы в легких. Считается даже, что 90% случаев заболевания легких можно отнести к курению.

Вышеуказанный вывод, основанный на механизме, можно представить в виде следующей диаграммы. Здесь $X$ оказывает влияние на $Y$ исключительно через $M$.

image.png

Стратегию идентификации можно описать в два этапа. Нет back door пути от $X$ к $M$ , и единственный back door от $M$ к $Y$ может быть заблокирован $X$. Другими словами, как $P(M|do(X=x))$, так и $P(Y|do(M=m))$ могут быть идентифицированы с использованием back door критерия. Второй шаг основан на аргументе, что $X$ может влиять на $Y$ только через $M$, поэтому интуитивно
$$
P(Y \mid do(X=x)) = \displaystyle \sum_m P(M=m \mid do(X=x))P(Y \mid do(M=m))
$$

**Front door критерий**

Получается, что выражение $P(Y|do(X=x))$ можно получить, заменив $P(M=m|do(X=x))$ и $P(Y|do(M=m))$ на соответствующие выражения.

Множество узлов $M$ удовлетворяет критерию «входа» для упорядоченной пары $(X,Y)$, если:

1. $M$ перекрывает все направленные пути из $X$ в $Y$;
2. все back door пути из $M$ в $Y$ блокируются $X$.
3. нет путей бэкдора из $X$ в $M$;

Если такой набор $M$ существует,
$$
$P(Y|do(X=x))= \displaystyle \sum_m P (M=m|X=x)\left(\displaystyle \sum_xP(Y|X=x′,M=m)P(X=x′)\right)
$$

является допустимой стратегией идентификации.

Условия 2 и 3 гарантируют, что два подмеханизма $M$ → $Y$ и $X$ → $M$ идентифицируются по back door критерию. Следующая лемма показывает, что условия 1 и 2 вместе обосновывают $do$ распространение.

Для DAG $G$ существует направленный путь из $X$ в $M$ и затем $Y$
. Достаточным условием для
$$
P(Y|do(X=x))=\displaystyle \sum_m P(M=m|do(X=x))P(Y|do(M=m))
$$

является то, что

1. $M$ перехватывает все направленные пути из $X$ в $Y$;
2. все back door пути из $M$ в $Y$ блокируются $X$;


Front door критерий и, в более общем смысле, идея определения причинно-следственной связи с использованием комбинаций $do$-распространения вместе с цепочкой идентифицируемых $do$-выражений весьма вдохновляют. Он позволяет нам искать стратегию идентификации в очень сложных графах, используя принцип «разделяй и властвуй». Если мы можем найти цепочку
$M_0=X,M_1,\dots,M_k=Y$ с каждым $P(M_k|do(M_{k−1}=m))$ идентифицируемым, и распространение можно рекурсивно применить для преобразования $P(Y|do(X=x))$ в механизм, использующий только подмеханизмы $P(M_k|do(M_{k−1}=m)), тогда $P(Y|do(X=x)) можно идентифицировать.

### Уверенность в модели генерации данных

В традиционном подходе к каузальному выводу, понимание DGP является краеугольным камнем: исследователи начинают с построения причинной диаграммы, которая должна отображать их представление о DGP, чтобы затем выявить все пути от воздействия к исходу и перекрыть "плохие" пути, оставляя только причинные. Этот подход является "структурным".

Однако в сложных ситуациях, особенно в социальных науках, полное знание и отображение истинного DGP часто **недостижимо**. Попытка построения полной диаграммы может привести к чрезмерным упрощениям, а также к упущению важных переменных, что может помешать распознаванию или корректному закрытию back-door путей. В таких случаях диаграмма может содержать обширные области, обозначенные как "НЕ ЗНАЮ, КАКИЕ-ТО ВЕЩИ". Возникает вопрос: как проводить каузальный вывод, если DGP остается в значительной степени неизвестным?.

Когда полное моделирование DGP оказывается нецелесообразным, исследователи могут применить подходы, направленные на **"принципиальное незнание"**, признавая наличие неучтенных факторов. Вместо того чтобы пытаться исчерпывающе перечислить и измерить все "закулисные" переменные, эти методы направлены на их непрямой контроль или полное обход их влияния:

*   **Фокусировка на front door критерии**: Один из подходов заключается в выделении нескольких front door путей от воздействия к исходу, которые не зависят от неучтенных смешивающих факторов, вместо попытки закрыть все неизвестные "закулисные" пути. Это позволяет контролировать "НЕ ЗНАЮ, КАКИЕ-ТО ВЕЩИ", не контролируя их напрямую.
*   **Использование объемных контрольных переменных**: Некоторые контрольные переменные могут одновременно устранять вариации, связанные со многими другими факторами. Например, в вопросе воспитания контроль за домом, в котором вы выросли, автоматически контролирует такие факторы, как район или город, поскольку они варьируются только между разными домами.
*   **Метод фиксированных эффектов (Fixed Effects)**: В панельных данных, когда индивиды наблюдаются многократно, контроль за индивидуальным фиксированным эффектом позволяет учесть все неменяющиеся во времени характеристики индивида (например, воспитание, личность), независимо от того, измеримы они или нет.
*   **Создание сопоставимых групп**:
    *   **Рандомизация**: В экспериментах случайное присвоение участников к группам воздействия и контроля обеспечивает их сопоставимость по всем "закулисным" переменным, избавляя от необходимости их прямого контроля.
    *   **Целенаправленные сравнения в наблюдательных исследованиях**: В отсутствие рандомизации можно искать подмножества групп воздействия и контроля, которые по предположению сопоставимы по неконтролируемым параметрам. Примером может служить сравнение сообществ, получивших финансирование, с сообществами, которые подавали заявки, но не получили его, предполагая полу-случайное распределение.
    *   **Сравнение изменений во времени**: Вместо сравнения средних уровней, можно сравнивать *изменения во времени* в группе воздействия с *изменениями во времени* в контрольной группе (например, метод Difference-in-Differences). Это делает сравнение более правдоподобным, поскольку требуется лишь, чтобы *изменения* "НЕ ЗНАЮ, КАКИЕ-ТО ВЕЩИ" были одинаковыми, а не их абсолютные уровни.

\\

Не менее важным аспектом является и валидация сделанных предположений для модели DGP. Но зачастую доказать, что какая-то предпоссылка является верной не представляется возможности, либо крайне проблематично, поэтому вместо того чтобы доказывать истинность предположений, целью является проверка их *ложности*. Для этого используются следующие подходы:

*   **Тесты на робастность (Robustness Tests)**: Это методы проверки, могут ли предположения быть опровергнуты, или как изменяются результаты при ослаблении предположений. Часто это включает проверку связей, которые *не должны существовать* согласно допущениям модели. Если такая связь обнаруживается, это является свидетельством против предположений. Например, если после контроля за переменной (например, употреблением вина) две другие переменные (наркотики и доход) все еще связаны, это указывает на ошибку в исходных допущениях о причинной диаграмме.
*   **Плацебо-тесты (Placebo Tests)**: Это специфическая форма тестов на робастность, при которой "воздействие" имитируется там, где его на самом деле не было. Если в результате такого теста обнаруживается эффект, это указывает на некорректность предположений, поскольку "эффект чего-то, что является ничем", не должен существовать. Пример: имитация политики по экономии электричества при более низком пороге потребления, чтобы проверить, различаются ли группы, которые фактически не получили никакого воздействия.
*   **Частичная идентификация (Partial Identification)**: Вместо того чтобы вынуждать себя делать все необходимые предположения для получения единственной точечной оценки эффекта, подход частичной идентификации позволяет сделать только те предположения, в которых есть уверенность. Затем, для оставшихся неизвестных аспектов, используется **диапазон возможных значений**, что приводит к **диапазону возможных оценок** для эффекта (например, "эффект не выше 5%"). Этот метод, тесно связанный с анализом чувствительности, позволяет исследователям более честно отражать неопределенность.

\\

Верификация каузальной модели, особенно в контексте неполного знания о процессе генерации данных, играет **ключевую роль в обеспечении надежности** полученных результатов. Признавая неизбежность "незнания" всех аспектов DGP, исследователи используют арсенал методов, которые позволяют:

*   **Минимизировать влияние ненаблюдаемых смешивающих факторов** (через косвенный контроль, фиксированные эффекты или умный выбор контрольных групп).
*   **Систематически проверять уязвимость своих предположений** с помощью эмпирических тестов (тесты на робастность, плацебо-тесты).
*   **Честно отражать остающуюся неопределенность** в оценках эффекта, а не принудительно давать точечные оценки на основе слишком сильных предположений.

В конечном итоге, эти методы позволяют исследователям делать максимально обоснованные каузальные выводы, даже в условиях сложной и не до конца познанной реальности, повышая **применимость моделей к реальным задачам**. "Интуиция" исследователя выступает в качестве последней проверки на **правдоподобность** результатов, обеспечивая, чтобы выводы не противоречили здравому смыслу и экспертным знаниям.
