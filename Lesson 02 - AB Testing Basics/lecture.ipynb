Сегодня мы поговорим о фундаменте любого серьезного продуктового развития и маркетинговых экспериментов — об A/B-тестировании. В мире, где каждый клик и каждая транзакция имеют измеримую ценность, способность точно установить причинно-следственную связь между изменением и результатом становится золотым ключом к успеху. Именно здесь на сцену выходит **рандомизированное контролируемое испытание (РКИ)**, или, как мы чаще называем его в нашем домене, **A/B-тест**.

### Идеальное тестирование

Когда мы проводим тестирование какой-либо гипотезы, идеальным исходом было бы иметь под рукой ещё несколько точных копий нашей вселенной, и тогда исследование эффекта выглядело бы как:
1. В одной вселенной вводим изменение и собираем результаты;
2. Сравниваем результаты из 'измененной' вселенной со вселенной, где ничего не изменялось
3. Разница между выводами и будет наиточнейней оценкой эффекта изменений.

К сожалению, точных копий у нас нет, поэтому нам приходится использовать менее точные, но всё-таки имеющие доказательную силу методы проверки гипотез.

**Концепция тестирования и уровни доказательности РКИ**

В основе любого научного исследования, особенно в медицине, лежит понятие **иерархии доказательности** (hierarchy of evidence), которая ранжирует методы получения результатов по их надежности и способности исключать систематические ошибки. Эта иерархия, или **уровни доказательности (LOEs/ELs)**, является эвристикой, помогающей оценить относительную силу выводов, полученных в ходе экспериментальных исследований. Хотя изначально эти концепции разрабатывались для медицинских исследований, их принципы универсальны и прекрасно применимы в аналитике продуктовых экспериментов, таких как A/B-тестирование.

Рандомизированное контролируемое испытание (РКИ) — это форма научного эксперимента, разработанная для контроля факторов, которые не находятся под прямым экспериментальным управлением. В контексте клинических исследований, РКИ обычно сравнивает новое лечение с существующим стандартом или плацебо. В A/B-тестировании это могут быть две разные версии одной страницы (например, "А" и "Б"), два разных алгоритма выдачи товаров, или даже различные подходы к ценообразованию.

**Ключевая идея РКИ заключается в случайном распределении участников** (в нашем случае — пользователей) по различным группам, получающим разные воздействия. Этот процесс рандомизации позволяет достичь **статистического контроля** над известными и неизвестными факторами, которые могут влиять на результаты, но не могут быть напрямую контролируемы. Например, пользователи торговой площадки вроде Ozon или Wildberries сильно различаются по покупательской способности, предпочтениям, техническим навыкам и множеству других параметров. Без рандомизации невозможно было бы отделить эффект нового дизайна кнопки от влияния того, что пользователи с определенными характеристиками чаще попадали в группу с новым дизайном.

Именно благодаря способности минимизировать **смещение выбора (selection bias)** и **смещение распределения (allocation bias)**, а также балансировать известные и неизвестные прогностические факторы, РКИ считается **золотым стандартом** испытаний. В клинических исследованиях к этому добавляется **ослепление (blinding)**, когда участники, исследователи и аналитики не знают, к какой группе отнесены те или иные испытуемые, что снижает другие формы предвзятости. В A/B-тестировании это также реализуется, когда пользователи не знают, какую версию сайта или функции они видят.

В различных иерархиях доказательности РКИ и их систематические обзоры занимают наивысшие позиции:
*   Согласно **Guyatt и Sackett (1995)**, систематические обзоры и мета-анализы РКИ с окончательными результатами, а также сами РКИ с окончательными результатами, стоят на вершине.
*   **Оксфордский центр доказательной медицины (2009, 2011)** также помещает систематические обзоры рандомизированных контролируемых исследований (с однородностью результатов) и отдельные РКИ (с узким доверительным интервалом) на верхние уровни (1a и 1b соответственно).
*   Подход **GRADE (Grading of Recommendations Assessment, Development and Evaluation)**, используемый более чем 100 организациями, включая Всемирную организацию здравоохранения, оценивает высокое качество доказательств как "высокое", если есть большая уверенность в том, что истинный эффект близок к предполагаемому. Рандомизированные испытания без существенных ограничений считаются высококачественными доказательствами.

**Почему РКИ превосходит наблюдательные исследования**

Основное преимущество РКИ перед **наблюдательными исследованиями (observational studies)** заключается в его способности устанавливать **причинно-следственные связи**, а не просто корреляции. В наблюдательных исследованиях, как следует из названия, исследователи лишь наблюдают за происходящими событиями, не вмешиваясь в них. Это создает значительные трудности в определении, что именно вызвало наблюдаемый эффект, поскольку результаты могут быть искажены **смешивающими факторами (confounding variables)**.

Представьте, что вы анализируете данные Wildberries и видите, что пользователи, которые посещают страницы с "трендовыми" товарами, совершают больше покупок. Является ли это следствием того, что "трендовые" товары сами по себе стимулируют покупки, или же дело в том, что более активные и мотивированные покупатели склонны искать именно такие товары? В наблюдательном исследовании отделить одно от другого крайне сложно.

**Пример из электронной коммерции**

Вернемся к Wildberrries. Если бы мы просто наблюдали, что пользователи, которым показывается новый, яркий баннер о скидках, покупают больше, чем те, кому он не показывается, это было бы наблюдательным исследованием. Мы не могли бы быть уверены, что именно баннер стал причиной увеличения продаж. Возможно, пользователи, которые видели баннер, были более активными и уже готовы были к покупке.
В A/B-тесте мы делаем иначе:
1.  **Случайно делим пользователей на две группы:** контрольную (А) и экспериментальную (Б).
2.  **Группе А показываем стандартный интерфейс (без нового баннера).**
3.  **Группе Б показываем интерфейс с новым баннером.**
4.  **Сравниваем ключевые метрики** (например, конверсию в покупку) между этими группами.

Благодаря рандомизации, единственное систематическое различие между группами — это наличие или отсутствие баннера. Любые другие различия (демография, история покупок, активность) будут равномерно распределены между группами случайным образом. Таким образом, если мы видим статистически значимое различие в конверсии, мы можем с высокой степенью уверенности заявить, что именно баннер **вызвал** это изменение. Это и есть **"золотой стандарт"**.

Несмотря на то, что некоторые исследования 2000-х годов предполагали, что наблюдательные исследования могут давать схожие результаты с РКИ, более поздние мета-анализы указывают на то, что **расхождения могут возникать, и различия в предполагаемой величине эффекта лечения очень распространены**. РКИ предоставляют **убедительные доказательства того, что изучаемое воздействие вызывает эффект на поведение пользователей или на бизнес-метрики**.

**Ограничения на практике**

Несмотря на все свои преимущества, РКИ, или A/B-тесты, не лишены недостатков и ограничений, особенно при их применении в сложной динамичной среде, такой как e-commerce. Критический взгляд на методологию позволяет применять РКИ более осознанно и эффективно:

1.  **Время и затраты**

Проведение масштабных РКИ может быть чрезвычайно дорогим и трудоемким процессом. Такие тесты часто занимают месяцы или даже годы, прежде чем их результаты будут опубликованы, что делает данные менее актуальными к моменту публикации. В быстро меняющемся мире e-commerce, где продукты развиваются еженедельно, длительные тесты могут означать упущенные возможности или даже устаревание тестируемой гипотезы. Для оценки редких событий (например, отток очень специфического сегмента пользователей) или незначительных побочных эффектов (негативных реакций на изменение) требуются крайне большие выборки, что делает РКИ непрактичными и вынуждает прибегать к наблюдательным исследованиям.

2.  **Конфликт интересов**

Как показывают исследования, финансирование РКИ коммерческими организациями (например, производителями лекарств в медицине или крупными компаниями в IT) может быть связано с более частыми "положительными" для спонсора результатами. Это может быть обусловлено предвзятостью публикаций (publication bias), когда положительные результаты с большей вероятностью доходят до публикации. В контексте A/B-тестов это может проявляться в "подгонке" тестов под желаемый результат или остановке тестов, которые демонстрируют негативные тенденции, что ставит под вопрос объективность выводов.

3.  **Этические дилеммы**

В A/B-тестировании этические вопросы также могут возникать, хотя и не столь драматично, как в медицинских испытаниях. Например, **концепция "терапевтического заблуждения" (therapeutic misconception)**, когда участники исследования могут считать, что они обязательно получают наилучшее для себя лечение, также применима и к A/B-тестам. Если мы тестируем новую функцию, которая потенциально может ухудшить пользовательский опыт или привести к финансовым потерям для пользователя (например, новая модель ценообразования), этично ли сознательно подвергать часть аудитории этому риску? В случае, если инновация оказывается явно вредоносной, этические соображения могут вынудить досрочно прекратить тест, что делает невозможным получение полной картины ее влияния.

4.  **Проблемы с генерализуемостью (внешней валидностью)**

РКИ, особенно те, что тщательно контролируются (т.н. "объяснительные" или "explanatory" РКИ), проверяют эффективность в идеальных лабораторных условиях с тщательно отобранными участниками. Это может ограничивать применимость их выводов к реальным, менее контролируемым условиям и более разнообразным группам пользователей. Прагматические РКИ (pRCTs) стремятся тестировать эффективность в повседневной практике, но это усложняет дизайн и анализ. Тест, проведенный на сегменте русскоязычных пользователей в Москве, может не дать тех же результатов при запуске для англоязычной аудитории в Нью-Йорке.

5.  **Методологические критические замечания**

Существуют критические замечания относительно того, что не все исследовательские вопросы могут быть адекватно решены с помощью РКИ из-за практических или этических проблем. Также, РКИ, особенно те, что фокусируются на "среднем эффекте" лечения, могут не учитывать **внутригрупповую изменчивость** — то есть, почему одни пользователи реагируют на изменение одним образом, а другие — совсем иначе.

6.  **Сетевые эффекты и "загрязнение"**

В сложных экосистемах, таких как большие платформы электронной коммерции (например, Ozon или Wildberries), могут возникать **сетевые эффекты**. Это означает, что изменение, введенное для одной группы пользователей, может косвенно повлиять на другую группу, даже если она находится в контроле. Например, если мы тестируем новый алгоритм ранжирования товаров для покупателей, который приводит к резкому росту продаж у определенных продавцов, это может изменить поведение других продавцов (например, они начнут снижать цены), что в свою очередь повлияет на покупателей в контрольной группе. В таких случаях чистое разделение на "контроль" и "эксперимент" нарушается, и интерпретация результатов РКИ становится значительно сложнее. Игнорирование сетевых эффектов может привести к некорректным выводам о причинно-следственных связях.

 \\

Несмотря на эти ограничения, **РКИ остаются самым надежным инструментом для установления причинно-следственных связей**. Критический анализ этих ограничений позволяет нам, аналитикам, более осознанно подходить к дизайну экспериментов, интерпретации их результатов и принятию решений, понимая не только силу метода, но и его границы.

Как дорожная карта, РКИ является нашим компасом в лабиринте гипотез. Он не всегда указывает на самый короткий путь, и порой его использование сопряжено с трудностями, но он неизменно ведет к самой надежной и обоснованной точке назначения — к **пониманию истинного влияния наших изменений**.

### Пайплайн тестирования гипотез

Эффективное научное исследование требует тщательно формализованного подхода. Этот процесс включает в себя ряд последовательных шагов, от первоначальной гипотезы до окончательного внедрения результатов, и предполагает сложную систему взаимодействия между различными командами.

В общем случае тестирование предстоявляет из себя следующую цепочку событий:

```
1. Формулирование гипотезы
2. Прототипирование дизайна тестирования
3. Сбор данных
4. Анализ данных
5. Формулирование итогов гипотезы
6. Подтверждение/Опровержение тестируемой гипотезы
7. Внедрение/ Отказ от внедрения
```

1.  **Формулирование гипотезы**

Начальный этап любого исследования – это формулировка проверяемой гипотезы. Гипотеза должна быть максимально точно сформулирована, чтобы не вносить недопонимае в трактовке другим коммандам, которые также участвуют в процессе тестирования. Как минимум гипотеза должна содержать:
    * Контекст, почему хотим вносить изменение
    * Вносимое изменение
    * Целевая аудитория
    * Ожидаемый эффект
  
Например,
```
* 'Контекст': На странице карточки товара Wildberries показывает динамику цены.
Сейчас там отображается средняя цена без учёта скидок кошелька, доставки и скидок по привелегии.
Тем не менее, Wildberries имеет СПП - скидка со стороны платформы, которая не учитывается в цене.
Как итог может быть ситуация, что на этой неделе скидка увеличилась -> цены будут выше (при прочих равных),
что негативно скажется на ощущениях пользователя (хотя продавец цены не поднимал фактически и не виноват в этом).
* 'Изменение': Будем в динамике указывать чистую цену, выставленную продавцом
* 'Целевая аудитория': Все пользователи площадки WB
* 'Ожидаемый эффект': Более справедливое отображение цены увеличит конверсию в заказ,
  т.к. не будет неоправданных завышений в динамике
```


2.  **Дизайн исследования**

После формулировки гипотезы разрабатывается дизайн исследования, который является критически важным для минимизации систематической ошибки и обеспечения внутренней валидности результатов. На этом этапе досконально продумывается, как будет тестироваться выдвинутая гипотеза: выбирется метод тестирования, список метрик, статистический критерий, задаются минимальные статистические параметры теста.

Например,
```
* 'Метод тестирования': A/B тест
* 'Метрики' : конверсия из просмотра в заказ при просмотре динамики цены
* 'Статистический критерий': t-test с применением дельта метода
* 'Параметры теста':
  * ошибка 1 рода: 5%
  * ошибка 2 рода: 20%
```

3.  **Сбор данных**

На этапе сбора данных реализуется разработанный дизайн. Это включает в себя набор участников, их распределение по группам согласно процедурам рандомизации и сокрытия распределения, а также проведение вмешательств и измерение исходов в соответствии с протоколом. Важно обеспечивать **верность вмешательству** (predictable use of the treatment in the same way every time), контролировать уровень пропущенных данных и отсева участников, а также учитывать потенциальные мешающие переменные.

4.  **Анализ данных**

После сбора данных следует этап их анализа для проверки гипотезы. Здесь важно предобработать данные и верно обсчитать метрики.

5.  **Формулирование итогов гипотезы **

После предобработки данных и получения итоговых метрик необходимо их правильно провалидировать с учётом возможных ошибок, которые были заложены в дизайн теста. Итогом в идеальном случае должнна быть с метриками с тестовых и контрольной группах и вывод на основе статистического критерия.

6. **Подтверждение/Опровержение тестируемой гипотезы**

На основе отчёта из предыдущего шага команда аналитики выносит вердит подтверждается ли гипотеза или нет.

7. **Внедрение/ Отказ от внедрения**

В зависимости от успеха проведенного исследования уже решается, будет ли фича 'катиться в прод' или необходимо её пересмотреть с учётом недостатков, выявленных при помощи эксперимента.

\\

Так как исследование - комплексный процесс, то в нем участвуют сразу несколько команд, отвечающие за свою составляющую. В общем виде взаимодействие может выглядеть следующим образом:

*   **Продуктовая команда:** Инициирует гипотезы, исходя из потребностей пользователей, бизнес-целей или наблюдаемых проблем. Определяет, какое изменение или новая функция может улучшить ключевые метрики.
*   **Аналитическая команда:** Формализует гипотезы в измеряемые параметры. Совместно с продуктовой командой разрабатывает дизайн эксперимента (например, A/B-тестирование, которое является формой РКИ в цифровой среде), определяет необходимые метрики, рассчитывает размер выборки и планирует методы анализа данных. Также обеспечивает прозрачность и точность результатов.
*   **Команда разработки:** Реализует техническую часть эксперимента. Это может включать разработку новых функций, механизмов рандомизации пользователей, инструментов сбора данных и интеграцию с аналитическими системами. Обеспечивает техническую возможность проведения эксперимента в соответствии с дизайном.
*   **Итерационный цикл:** После внедрения и сбора данных, аналитики проводят анализ, результаты которого возвращаются продуктовой команде для принятия решений о дальнейшем развитии продукта. Этот цикл повторяется, создавая культуру принятия решений, основанных на данных.

### Виды тестов

Тесты можно поделить по основным 3 критериям:
  1. **Цель тестирования** - что хотим проверить
  2. **Направленность альтернативной гипотезы** - эффект в какую сторону хотим смотреть
  3. **Стратегия проведения** - как рассчитывается тест

#### Цель тестирования  

1. **Superiority Tests (тест на превосходство)**

Тесты на превосходство — это односторонние проверки гипотез, в которых нулевая и альтернативные гипотезы сформулированы так, чтобы можно было проверить, превосходит ли одна группа другую более чем на заданный предел. Таким образом, нам может быть интересно показать, что новый алгоритм ранжирования рекомендаций превосходит другой более чем на незначимую для бизнеса величину.


**Цель**: доказать, что новое средство (алгоритм, препарат, дизайн) статистически лучше эталонного минимум на заданную величину.

**Формулировка:**

$$
\begin{equation}
    \begin{cases}
      H_0: \mu_{new} - \mu_{ref} \le \delta\\
      H_1: \mu_{new} - \mu_{ref} > \delta
    \end{cases}
\end{equation}
$$


**Примеры:**
* A/B-тест: новый дизайн сайта vs старый. μ = конверсия.
* Клиническое: новая вакцина дает более высокий средний уровень антител.
* Алгоритм рекомендательной системы демонстрирует большую среднюю глубину сессии.

\\

2. **Non-inferiority Tests (тест не меньшей эффективности)**

Тесты не меньшей эффективности - это односторонние проверки гипотез, в которых нулевая и альтернативные гипотезы выстраиваются так, чтобы проверить, является ли одна группа почти такой же хорошей (не намного хуже), чем другая. Например, нам может быть интересно показать, что новый, менее затратный метод работы бэкенда (для примера дольше хранить кэш) не хуже существующего.


**Цель:** доказать, что новое решение не уступает старому более, чем на заранее заданную «не-худшую» границу $\delta > 0$.

**Формулировка:**

$$
\begin{equation}
    \begin{cases}
      H_0: \mu_{new} - \mu_{ref} \le -\delta \\
      H_1: \mu_{new} - \mu_{ref} > -\delta
    \end{cases}
\end{equation}
$$

**Примеры:**  

* Новый дженерик-препарат по эффективности «не хуже» оригинала более чем на $\delta$=10%.
* Обновлённый компрессор в промышленной установке должен работать «не хуже» текущего по выходной мощности ($\delta$ = –5 кВт).
* Мобильное приложение новой версии должно иметь «не худшую» среднюю скорость загрузки страниц ($\delta$ = –200 мс).

\\

3. **Equivalence Tests (тест на эквивалентность)**  

Тесты эквивалентности используются для убедительного доказательства эквивалентности двух методов. Традиционный метод проверки гипотез эквивалентности заключается в проведении двух односторонних тестов (TOST). Нулевая гипотеза о неэквивалентности отвергается в пользу альтернативной гипотезы об эквивалентности, если отвергаются оба односторонних теста. Однако, в отличие от обычных двусторонних тестов, частота ошибок первого рода устанавливается непосредственно на номинальном уровне (обычно 0,05) — она не делится пополам. Таким образом, для проведения теста проводятся два односторонних теста на уровне значимости α. Если оба теста отвергаются, альтернативная гипотеза подтверждается на уровне значимости α. Обратите внимание, что p-значение теста — это максимальное из p-значений двух тестов.

**Цель:** показать, что разница между новинкой и референсом лежит внутри предопределённого интервала (–$\delta$, +$\delta$).


**Формулировка:**

$$
\begin{equation}
    \begin{cases}
      H_0: \left(\mu_{new} - \mu_{ref} \le -\delta \right) \cup \left(\mu_{new} - \mu_{ref} \ge +\delta \right) \\
      H_1: -\delta < \mu_{new} - \mu_{ref} < +\delta
    \end{cases}
\end{equation}
$$


**Примеры:**  

* Два способа измерения давления: нужно доказать, что их показания эквивалентны в ±3 мм рт. ст.
* Алгоритм сжатия нового формата видео даёт качество, эквивалентное старому в пределах ±2 % PSNR.
* Пилотный сайт-клон должен давать эквивалентную конверсию в пределах ±0.5 %.

\\

4. **Futility Tests (тест на бесперспективность)**  

**Цель:** на промежуточном этапе остановить «безнадёжный» эксперимент, если достижение клинически/бизнес-значимого эффекта маловероятно.


**Формулировка:**

$$
\begin{equation}
    \begin{cases}
      H_0:  \text{effect} \ge \Delta_{min}  \\
      H_1: \text{effect} < \Delta_{min}
    \end{cases}
\end{equation}
$$


**Примеры:**  

* Фарма-исследование: на 50% набора пациентов препарат не показывает обещанного улучшения — прекращаем.
* A/B-тест: на половине трафика новая функция вовсе не повышает конверсию к порогу +2 % → останавливаем запуск.
* Инженерный пилот: новый двигатель не достигает требуемой мощности $\Delta_{min}$ = 100 кВт при 75% испытаний — закрываем проект.

#### Направленность альтернативной гипотезы

1. **Односторонние тесты (One-sided)**

Это статистические тесты, в которых альтернативная гипотеза предполагает, что параметр больше или меньше определенного значения, то есть использует знак «<» или «>». Соответственно в зависимости от направленности эффекта бывает как **левосторонний (left-tailed)**, так и **правосторонний (right-tailed)**.

Их нужно применять, когда
* Важно определить, насколько один показатель превышает или уступает другой.
* Интересует только одна сторона отклонения — например, только увеличение или только уменьшение.


Примеры:

1. Проверка, превышает ли новый препарат стандартную дозу

$$
\begin{equation}
    \begin{cases}
      H_0: \mu_{new} \leq \mu_{stand} \\
      H_1: \mu_{new} > \mu_{stand}
    \end{cases}
\end{equation}
$$

2. Проверка, что новое устройство работает быстрее старого

$$
\begin{equation}
    \begin{cases}
      H_0: \text{time}_{new} \geq \text{time}_{stand} \\
      H_1: \text{time}_{new} < \text{time}_{stand}
    \end{cases}
\end{equation}
$$

3. Проверка, что уровень удовлетворенности клиентов повысился

$$
\begin{equation}
    \begin{cases}
      H_0: \text{satisfaction}_{new} \leq \text{satisfaction}_{stand} \\
      H_1: \text{satisfaction}_{new} > \text{satisfaction}_{stand}
    \end{cases}
\end{equation}
$$

\\

2. **Двухсторонние тесты (Two-sided)**

Это тесты, в которых альтернативная гипотеза утверждает, что параметр не равен определенному значению, то есть использует знак «≠».

Их нужно применять, когда
* Важно определить, есть ли различие в любую сторону — увеличение или уменьшение.
* Эффект может проявляться в обе стороны и нужно проверить наличие различия в целом.

Примеры:

1. Проверка, отличается ли среднее значение от заданного уровня:

$$
\begin{equation}
    \begin{cases}
      H_0: \mu = \mu_0 \\
      H_1: \mu \neq \mu_0
    \end{cases}
\end{equation}
$$

2. Проверка, есть ли разница между двумя методами лечения:

$$
\begin{equation}
    \begin{cases}
      H_0: \mu_1 = \mu_2 \\
      H_1: \mu_1 \neq \mu_2
    \end{cases}
\end{equation}
$$

3. Проверка, отличается ли уровень холестерина после диеты от исходного:

$$
\begin{equation}
    \begin{cases}
      H_0: \text{level}_{new} = \text{level}_{stand} \\
      H_1: \text{level}_{new} \neq \text{level}_{stand}
    \end{cases}
\end{equation}
$$

#### Стратегия проведения

1. **Фиксированный План (Fixed-Horizon Design)**

**Что это:** Классический, самый простой дизайн. Все параметры испытания (размер выборки, точки анализа, первичные конечные точки) фиксируются до начала набора выборки и не меняются в процессе исследования. Анализ данных проводится только один раз – после того, как будет достигнут запланированный обхём выборки и будут собраны все данные.

**Зачем нужен:** Простота планирования, проведения, анализа и регуляторного одобрения. Минимизация риска предвзятости из-за множественных проверок данных. Прямая интерпретация результатов.

**Основные методы**:
* $t-$test
* $z$-test
* $\chi^2$-test
* Fisher-exact test
...

**Когда лучше применять:**

* Когда эффект ожидается большим и хорошо предсказуемым.
* Когда исследование относительно короткое и затраты на длительное наблюдение высоки.
* Когда ресурсы (время, деньги) позволяют набрать сразу всю большую выборку.
* Когда риски хорошо изучены, и нет острой этической необходимости в досрочной остановке.
* Когда требуется максимальная простота и прозрачность для регуляторов.

**Примеры:**

1. Изменение цвета/текста кнопки CTA: Тест новой кнопки "Купить" (зеленая vs синяя, "Добавить в корзину" vs "Купить сейчас") на странице товара. Размер выборки рассчитан на ожидаемое изменение конверсии в корзину/покупку. Анализ – строго через 2 недели.

2. Внедрение новой формы быстрой оплаты: Сравнение конверсии на этапе оплаты между старой многошаговой формой и новой упрощенной (например, через 1 клик). Тест длится фиксированные 3 недели, анализ только в конце по метрике "успешные оплаты".

3. Изменение лимита на бесплатную доставку или минимальной суммы для cashback: Тест влияния повышения/понижения порога на средний чек или активность. Параметры фиксированы, анализ – после сбора данных за квартал.

\\

2. **Последовательные Тесты (Sequential Designs)**

**Что это:** Дизайны, в которых данные анализируются неоднократно по мере поступления (например, после набора каждых 50, 100, 150 пациентов или через определенные промежутки времени). Позволяют досрочно остановить исследование из-за явной эффективности, явной неэффективности или неприемлемого вреда до набора полной выборки.

**Зачем нужен:**
  * **Этическая выгода:** Прекращение испытания, если лечение явно эффективно (чтобы быстрее дать его всем) или явно вредно/неэффективно (чтобы не подвергать пациентов риску без пользы).
  * **Экономическая выгода:** Экономия времени и ресурсов при досрочной остановке. Особенно актуально для долгих или дорогих исследований (например, с выживаемостью как конечной точкой).

**Основные методы (ключевые концепции):**

* (m)SPRT - (mixed) Sequential Probability Ratio Test
* AVI - Always Valid Inference
* GST - Group Sequential Testing
...


**Когда лучше применять:**

* Когда конечная точка является "ценной" (выживаемость, тяжелые события) и есть этический императив минимизировать воздействие неэффективного/вредного лечения или быстро внедрить эффективное.

* Когда исследование очень длительное или дорогостоящее, и досрочная остановка дает значительную экономию.

* Когда исходный эффект лечения не вполне ясен, и есть значительная неопределенность.

* Когда размер требуемой выборки очень велик.

**Примеры:**

1. Тест новой программы лояльности: Промежуточные анализы запланированы через 1, 2 и 3 месяца. Метрики: LTV (Lifetime Value), retention rate, частота покупок/операций. Если через 1 месяц видно значимое падение LTV у новой группы, тест останавливают досрочно, чтобы минимизировать убытки. Если виден явный рост – можно ускорить rollout.

2. Тест новой скоринговой модели для кредита/овердрафта: Анализы после обработки каждых 1000 заявок. Ключевые метрики: одобрение/отказ, процент просрочек (delinquency rate). Если на промежуточном анализе просрочки в тестовой группе значительно выше контрольной, тест останавливают досрочно из-за неприемлемого риска. Если одобрение значительно выше без роста риска – можно внедрять быстрее.

3. Тест нового алгоритма детекции мошенничеств: Промежуточные анализы после выявления каждых 50 подтвержденных случаев мошенничества. Метрики: точность (precision), полнота (recall), % ложных срабатываний (false positive rate). Если на раннем анализе FPR неприемлемо высок (слишком много легитимных транзакций блокируется), тест останавливают для доработки модели. Если точность и полнота резко улучшились без роста FPR – алгоритм внедряют.

\\

3. **Адаптивные Тесты (Adaptive Designs)**

**Что это:** Наиболее гибкий тип дизайнов. Позволяют заранее запланировать модификацию параметров исследования на основе промежуточных данных (обычно "слепых" по воздействию, чтобы минимизировать предвзятость). Модификации основаны на предопределенных правилах и требуют строгого статистического контроля.

**Зачем нужен:**

  * **Повышение гибкости и эффективности:** Возможность реагировать на информацию, полученную в ходе исследования, для оптимизации дизайна.
  * **Этическая и экономическая выгода:** Более быстрая идентификация лучших доз/групп, прекращение бесперспективных ветвей, корректировка размера выборки под реальную изменчивость/эффект.

**Основные методы (ключевые концепции):**

* Адаптация размера выборки (Sample Size Re-Estimation - SSR): Увеличение или уменьшение общего числа участников на основе промежуточных оценок эффекта или вариабельности данных.

* Адаптация дизайна (Adaptive Seamless Designs): Объединение фаз (напр., IIb и III) в одно исследование. На раннем этапе выбираются перспективные дозы/группы для продолжения на позднем этапе с большей выборкой.

* Переключение первичной конечной точки: Изменение главной измеряемой переменной (с заранее оговоренных альтернатив) на основе промежуточных данных (редко, требует крайне осторожного подхода).

* Популяционная адаптация (Adaptive Enrichment): Сужение или изменение критериев включения (напр., пользователей с определенным параметром) на основе промежуточных данных для фокусировки на подгруппе, где наиболее вероятно будет эффект.

* Байесовские адаптивные дизайны: Используют байесовскую статистику для непрерывного обновления вероятности успеха по мере поступления данных и принятия адаптивных решений.

**Когда лучше применять:**

* Когда высокая неопределенность в ключевых параметрах (размер эффекта, вариабельность).

* Когда нужно оптимизировать дозу или выбрать лучшую из нескольких экспериментальных групп эффективно.

* Когда исследуется болезнь с неоднородным ответом (напр., онкология с биомаркерами), и нужно сфокусироваться на респондерах.

* Когда ресурсы ограничены, и требуется максимальная "отдача" от исследования.

* Когда разработка должна быть ускорена (напр., для заболеваний с высокой неудовлетворенной потребностью).


**Примеры:**

1. Адаптивное многорукое тестирование (Multi-Armed Bandit - MAB) для рекламных креативов или промо-предложений: Тест 5 разных баннеров на главной странице. Алгоритм динамически перераспределяет трафик в пользу вариантов, показывающих лучшую CTR (или конверсию) на ранних этапах. Менее эффективные варианты постепенно получают меньше трафика. Цель – максимизировать общую эффективность во время теста.

2. Адаптация размера выборки (SSR) для сложной метрики: Тест нового дизайна каталога товаров. Первичная метрика – конверсия в покупку с глубиной просмотра >5 страниц (редкое событие). На промежуточном анализе (через 1 неделю) оценивается реальная вариабельность и наблюдаемый эффект. Если эффект слабее ожидаемого, размер выборки автоматически увеличивается по предопределенному правилу, чтобы сохранить мощность теста к концу.

3. Адаптивное обогащение / таргетирование: Тест персонализированной скидки. Изначально тест запущен для всей аудитории. Промежуточный анализ выявляет, что предложение особенно эффективно для пользователей с высокой исторической частотой покупок. ###  Формализация A/B тестирования

Процесс формализации A/B-тестирования представляет собой систематический подход для оценки эффективности изменений, основанный на статистическом анализе. Он позволяет принимать математически обоснованные решения о внедрении новых функций или доработок.

\\

Пайплайн A/B-тестирования включает в себя несколько ключевых этапов:
1. **Постановка гипотез (Нулевая $H_0$ и Альтернативная $H_1$)**

В основе любого A/B-теста лежит проверка статистических гипотез.

* **Нулевая гипотеза $H_0$** - это утверждение об отсутствии эффекта от тестируемого изменения. Например, $H_0$ может гласить, что среднее значение метрики в тестовой группе (T) равно среднему значению в контрольной группе (C), то есть $\mu_t = \mu_c$. Или, в абсолютной постановке, разница средних равна нулю: $\mu_t - \mu_c = 0$.

  **Это гипотеза, которую мы стремимся опровергнуть.**
* **Альтернативная гипотеза $H_1$** - это утверждение о наличии эффекта или различии между тестовой и контрольной группами. Например, H₁ может утверждать, что $\mu_t \neq \mu_c$ (двусторонняя гипотеза), или $\mu_t > \mu_c$ (односторонняя гипотеза), или что отношение $(\mu_t - \mu_c) / \mu_c \neq 0$ для относительного эффекта.
  **Это то, что мы хотим подтвердить.**

Чёткая формулировка гипотез критически важна, поскольку она определяет, что именно мы измеряем и какие выводы сможем сделать. Использование как абсолютных, так и относительных метрик для постановки гипотез повышает наглядность и интерпретируемость результатов для бизнеса.

2. **Определение уровней ошибок первого ($\alpha$) и второго ($\beta$) рода**

Определение этих уровней ошибок позволяет контролировать риски и гарантировать, что принимаемые решения не являются случайными. Это баланс между риском внедрить неэффективное изменение и риском пропустить потенциально полезное. Подробнее их смотрим чуть позже.

3. **Выбор подходящего статистического критерия**

После формулировки гипотез и определения уровней ошибок необходимо выбрать статистический критерий, который будет использоваться для проверки гипотез. Подробнее их рассмотрим далее, но общий принцип - выбираем такой критерий, который:
1. подходит под распределение данных - у критериев есть свои предпоссылки, которые не должны нарушаться;
2. на данных выдаёт ошибки первого и второго рода в рамках номинальных (установленных в шаге 2);
3. по возможности является самым мощным из всех возможным альтернатив;


4. **Расчет размера выборки**
После формулировки гипотез, определения уровней ошибок и выбора статистического критерия, который будет использоваться для проверки гипотез, надо определить необходимый размер выборки.Размер напрямую зависит от того, насколько большие или маленькие ээффекты мы хотим уметь детектировать. Подробрее об этом будет рассказано далее.

5. **Расчет статистики на данных**

После сбора данных, метрики для контрольной и тестовой групп вычисляются, и на их основе рассчитывается статистическая значимость. Это включает:
1. Вычисление точечных оценок метрик (например, средних) для каждой группы.
2. Расчет разницы или отношения метрик между группами.
3. Вычисление тестовой статистики и соответствующего p-value.
4. Построение доверительных интервалов для оценки эффекта.


6. **Анализ результатов и принятие решения**

На этом этапе анализируются p-value и доверительные интервалы для принятия решения о принятии или отклонении нулевой гипотезы.

* Использование p-value:
    * Если p-value меньше или равно установленному уровню значимости ($\alpha$), то нулевая гипотеза отклоняется, и делается вывод о наличии статистически значимого эффекта.
* Использование доверительных интервалов:
    * Если ноль не попадает в доверительный интервал для разницы метрик, это эквивалентно отклонению нулевой гипотезы и означает наличие статистически значимого эффекта.
    * Доверительные интервалы предпочтительнее p-value для интерпретации результатов, так как они дают не только информацию о наличии эффекта, но и о его численной величине и точности (например, "+10±5М рублей").
* Интерпретация "серых" (не статистически значимых) результатов:
    * Даже если результат не является статистически значимым, доверительный интервал может дать ценные инсайты. Например, он позволяет понять, какой максимальный эффект мог бы быть, если бы он действительно существовал, или наоборот, исключить слишком большие эффекты.
    * Важно не просто констатировать отсутствие эффекта, а заявлять, что "если эффект и есть, то он находится в определённых границах".

#### Типы метрик

Когда мы сформулировали у себя в голове гипотезу, необходимо подумать, а какие метркии мы должны отслеживать, чтобы её подтвердить или опровергнуть? С одной стороны очевидно, что одно и тоже можно мерить с помощью разных метрик, но как же выбрать наилучшую из них ?

\\

**Продуктовая точка зрения**

С продуктовой точки зрения нам необходимо понимать две вещи:

1. **Простота метрик**

Чем **проще** метрика для объяснения - тем лучше для вас, т.к. результаты всегда легче сообщать во всем понятным терминах (особенно хорошо управленцы почему-то понимают денежные метрики). Для примера сравните две формулировки:
  * 'Данные подтверждают гипотезу, т.к. в тестовой группе средняя продолжительность визита пользователей статистически значимо сократилась.'
  * 'Данные подтверждают гипотезу, т.к. взвешенная сумма среднего времени сессии, количества просмотренных страниц и коэффициента взаимодействия с элементами сайтастатистически значимо сократилась.'

Поэтому чем метрика ближе бизнесу и легче для понимания и интерпретирования эффекта, тем более привлекательна она для выбора.

2. **Уровень прокраса**

Разные метрики **"прокрашиваются по-разному"**, то есть для разных метрик необходимо разное количество данных, чтобы получить статистически значимые отличия. Например, мы хотим померить разницу конверсий в открытии отзывов. Отзывы открывают довольно редко — скажем, базовая конверсия составляет около 0.0005 (0,05%). В таком случае, чтобы обнаружить даже небольшое увеличение этой конверсии, нам потребуется очень большой объем данных. Допустим, в тестовой группе конверсия увеличилась до 0.0007 (0,07%). При этом, из-за очень низкой начальной конверсии, чтобы добиться статистической значимости с уровнем α=0.05, потребуется собрать миллионы показов отзывов — поскольку число событий (открытий отзывов) очень мало, и большая выборка необходима для выявления даже небольшого отличия.  

В то же время, если мы измеряем изменение в более часто встречающейся метрике, например, скрол экрана до отзывов, у которых базовая конверсия составляет, скажем, 0.05 (5%), то для обнаружения такого же относительного увеличения потребуется значительно меньший объем данных.

То есть **чувствительность** метрик обратно пропорциональна выборке, необходимой для детекции эффекта.

Чтобы понять, какая метрика более или менее чувствительна, обычно прибегают к следующим вещам:

1. анализ уже прошедных A/B тестов, где можно провалидировать, был ли прокрас метрики или нет.
2. поиск прокси-метрик (proxy metric). Это метрики, которые сильно скорелированны с необходимой метрикой и сонаправлены, при этом они более чувствительны. Например, Стоимость привлечения клиента (CAC) может быть прокси-метрикой эффективности маркетинга и долгосрочной прибыльности. Тут в основном сложность, чтбоы найти такие метрики, но Microsoft ([понятный обзор статьи](https://www.youtube.com/watch?v=fSRKOr3L6AI)) в своей статье показывают, как это сделать проще и правильнее.


\\

**Техническая точка зрения**

С аналитической точки зрения мы так же должны дял себя понимать, что метрики так же делятся по 2 разрезам:


1. **Цель метрик**

Базово можно поделить метрики на 4 типа:
  * **Success** - метрики, которые хотим улучшить;
  * **Guardrail** - метрики, которые не должны упасть более чем на определённый уровень;
  * **Deterioration** - метрики, которые не должны упасть вообще;
  * **Quaility** - метрики здоровья тестирования;

Зачастую типы метрик определяются типом гипотезы, которая выдвигается в тесте и будет покрываться этими метриками. Так, success метрики зачастую проверяются superiority тестами, guardrail - non-inferiority, deterioration - inferiority тесты.

Тем не менее, зачастую можно встреить A/B тесты, которые включают сразу несколь метрик.

Рассмотрим игрушечный [пример](https://arxiv.org/pdf/2402.11609) от Spotify. Пусть мы хотим провести тест нового рекомендательного алгоритма музыки, тогда мы хотим смотреть на увеличение количества минут прослушивания музыки. При этом мы хотим убедиться, что общее потребление не увеличивается за счёт потребления подкастов -> смотрим на количество минут прослушивания подкастов. Обе метрики также включены в качестве метрик ухудшения качества, чтобы гарантировать, что они не меняются в направлении, противоположном ожидаемому и ожидаемому. Кроме того, в качестве deterioration метрики включена доля пользователей, столкнувшихся со сбоями. Единственной метрикой качества в данном случае является метрика несоответствия отношения выборки.

| Тип метрики | Наименование метрики | Тестируемое направление изменения |
| --- | --- | --- |
| **Success** | Прослушанных минут музыки | Увеличение |
| **Guardrail** | Прослушанных минут подкастов | Увеличение |
| **Deterioration** | Прослушанных минут музыки | Снижение |
| **Deterioration** | Прослушанных минут подкастов | Снижение |
| **Deterioration** | Доля юзеров с багами | Увеличение |
| **Quality**| Sample Mismatch Ratio (SRM)| N/A |

2. **Гранулярность метрик**

Это градация чисто техническая, которая необходима для правильного обсчёта статистик. Так мы будем понимать под **объектом** то, на что направлено тестируемое действие; и с одного объекта в течение эксперимента можно получить много значений. (например, покупатель сделает несколько покупок). Мы делим метрики на следующие типы п оотношени юк группировке относительно объекта:

ㅤㅤ2.1. **User-level Метрики**

Эти метрики вычисляются для каждого уникального объекта в отдельности, а затем усредняются (или берётся медиана и т.д.) по всем объектам в группе. Значение метрики привязано к уникальному идентификатору объекта.

$$
\text{Metric}_\text{user level} = \frac{1}{N} \sum_N \text{User Metric}
$$

Например, к таким метрикам относятся:

* Средняя выручка на пользователя (ARPU - Average Revenue Per User): Общая выручка группы / Количество уникальных пользователей;

* Среднее количество сессий на пользователя: Общее количество сессий / Количество уникальных пользователей;

* Среднее время, проведенное на платформе за период (на пользователя): Общее время всех пользователей / Количество уникальных пользователей;

* Процент платящих пользователей (Paying User Rate): (Количество пользователей с ≥1 покупкой) / Количество уникальных пользователей;

* Среднее количество покупок на пользователя
* Среднее время, проведенное на сайте одним пользователем

С такими метриками мы работаем в рамках обычных статистических тестов без каких-либо изменений.

ㅤㅤ2.2. **Ratio Метрики**

Эти метрики вычисляются как отношение количества событий одного типа к количеству событий другого типа (или к общему количеству событий) в рамках определенного периода или группы. Значение метрики не привязано напрямую к уникальному объекту; оно отражает плотность или интенсивность событий.

$$
\text{Metric}_\text{ratio} = \frac{\sum_N \text{Event A}}{\sum_N \text{Event B}}
$$


Например, к таким метрикам относятся:

* CTR (Click-Through Rate): Клики / Показы

* Conversion Rate (Коэффициент конверсии): Покупки / Посещения (или Покупки / Сессии, Заявки / Просмотры товара)

* Bounce Rate (Процент отказов): Сессии с одним событием / Все сессии

* Средняя глубина просмотра: Просмотренные страницы / Сессии

* Среднее время на сайте/в приложении: Общее время / Сессии

с метриками отношений дела же обстоят зачастую по-другому. Большинство стандартных статистических критериев имеют предпосылку, что данные должны быть независимы ('i.i.d'). В случае с метриками отношений мы их усредняем на общем уровне, когда числитель и знаменатель явно могут быть скорелированны, и в таком случае применение обычных тестов мы получим неверные оценки p_value и ложные выводы. В таких случаях обычно прибегают либо к тестам, где нет предпоссылки i.i.d., либо же преобразуют метрики методом линеаризации (подробнее в 10 занятии разберем), либо же считают дисперсию [дельта методом](https://habr.com/ru/companies/X5Tech/articles/740476/).

#### Матрица ошибок (Confusion matrix), ошибки 1 и 2 рода

##### Код для демонстрации

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import norm

def plot_errors_distibutions(delta=1.0, sigma=1.0, alpha=0.05):
    """
    Рисует плотности N(0, sigma^2) и N(delta, sigma^2),
    ставит порог c = q_{1-alpha} по H0, и заштриховывает:
      - ошибку первого рода (Type I): под H0, область x > c
      - ошибку второго рода (Type II): под H1, область x < c
    """
    # 1) критический порог при уровне alpha (односторонний тест)
    c = st.norm.ppf(1 - alpha, loc=0, scale=sigma)

    # 2) ось x для графика
    x_min = -4*sigma
    x_max = delta + 4*sigma
    x = np.linspace(x_min, x_max, 1000)

    # 3) плотности
    pdf0 = st.norm.pdf(x, loc=0, scale=sigma)     # H0: N(0,σ²)
    pdf1 = st.norm.pdf(x, loc=delta, scale=sigma) # H1: N(δ,σ²)

    # 4) рисуем плотности
    plt.figure(figsize=(14, 8))
    plt.plot(x, pdf0, label=r'H0: $N(0,\sigma^2)$', color='C0')
    plt.plot(x, pdf1, label=fr'H1: $N(\delta={delta},\sigma^2)$', color='C1')

    # 5) вертикальная линия в точке c
    plt.axvline(c, color='k', linestyle='--', label=fr'критерий $z_{{1-\alpha}}$')

    # 6) ошибка I рода: f0(x) заштриховать справа от c
    ix1 = x >= c
    plt.fill_between(x[ix1], pdf0[ix1], color='C0', alpha=0.3,
                     label=r'$\alpha$ (Type I): $P_{H_0}(X>c)$')

    # 7) ошибка II рода: f1(x) заштриховать слева от c
    ix2 = x <= c
    plt.fill_between(x[ix2], pdf1[ix2], color='C1', alpha=0.3,
                     label=r'$\beta$ (Type II): $P_{H_1}(X≤c)$')

    # 8) аннотации
    plt.title('Иллюстрация ошибок I и II рода (односторонний тест)')
    plt.xlabel('Значение метрики X')
    plt.ylabel('Плотность вероятности')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()


def power_one_sided(delta, n, sigma, alpha):
    """
    power для одностороннего теста H0: mu=0 vs H1: mu>0
    delta: эффект (mu1 - mu0)
    n: размер выборки
    sigma: стандартное отклонение
    alpha: уровень значимости
    """
    c = norm.ppf(1 - alpha)            # критическая точка для X̄*√n/σ
    # P( X̄*√n/σ > c | μ = delta ) = 1 - Φ( c - delta*√n/σ )
    return 1 - norm.cdf(c - delta * np.sqrt(n) / sigma)


def calculate_and_plot_mde(n:int = 100,
                           sigma:float = 1.0,
                           alpha:float = 0.05,
                           target_power:float = 0.8):

  # Сетка delta
  delta_vals = np.linspace(0, 5.0, 1001)
  power_vals = [power_one_sided(d, n, sigma, alpha) for d in delta_vals]

  # Находим MDE
  delta_vals = np.array(delta_vals)
  idx = np.where(np.array(power_vals) >= target_power)[0]
  if len(idx) > 0:
      mde = delta_vals[idx[0]]
  else:
      mde = np.nan

  # Рисуем
  plt.figure(figsize=(14,6))
  plt.plot(delta_vals, power_vals, label='power(Δ)')
  plt.axhline(target_power, color='k', linestyle='--', label=f'target power={target_power}')
  if not np.isnan(mde):
      plt.axvline(mde, color='r', linestyle='--', label=f'MDE≈{mde:.2f}')
      plt.scatter([mde], [target_power], color='r')
  plt.title('Функция мощности и MDE (односторонний тест)')
  plt.xlabel('Δ (эффект)')
  plt.ylabel('power = 1−β(Δ)')
  plt.legend()
  plt.grid(True)
  plt.show()

  print(f"При n={n}, σ={sigma}, α={alpha}, target power={target_power} → MDE ≈ {mde:.3f}")
````

##### Объяснение

Никакое тестирование не исключает ошибок, ведь полученный результат - одна из возможных реализацией случайности, произошедшей в нашей реальности. Тем не менее, статистика на то и наука, что позволяет работать в условиях случайности и неопределенности, поэтому с помощью неё мы можем точно гарантировать, что случайные величины будут находиться в рамках определенных значений.

Рассмотрим все возможные варианты, которые возможны при тестировании гипотезы:
1. Изменение не несёт эффекта на самом деле, а мы эффект нашли (Ошибка)
2. Изменение не несёт эффекта на самом деле, а мы эффект не нашли (Правильно)
3. Изменение несёт эффект на самом деле, и мы эффект нашли (Правильно)
4. Изменение несёт эффект на самом деле, а мы эффект не нашли (Ошибка)

Эти исходы будет удобно удобно изобразить в виде матрицы 2x2, где по вертикали отложим реальный факт (что на самом деле), а по горизонтали - найденный факт (что показал тест). Такая матрица называется **[матрицей ошибок](https://en.wikipedia.org/wiki/Confusion_matrix#cite_ref-22) (confusion matrix)**

| Реальный \ Найденный | Эффекта нет | Эффект есть|
| --- | --- | --- |
| Эффекта нет | Правильно | Ошибка |
| Эффект есть | Ошибка | Правильно |

Таким образом мы можем допустить всего 2 ошибки (отсюда ошибки 1 и 2 рода).

В более строгом статистическом виде матрица ошибок записывается как

| Верная \ Подтвержденная | $H_0$ | $H_1$|
| --- | --- | --- |
| $H_0$ | Правильно | Ошибка |
| $H_1$ | Ошибка | Правильно |

* **Ошибку первого рода**, которая обозначается $\alpha$, зачастую называют False Positive в терминах ML и ложноположительным исходом в биологии и медицине. По своему смыслу она описывает ситуацию, когда при реальном отсутствии эффекта мы говорим, что эффект есть. Тогда формально её можно записать как $\alpha = P(H_1|H_0)$.

  Стандартным значением для онлайн A/B тестов считается 5% ($alpha$ = 0.05), но так же зачастую фигурируют значения 0.01, 0.005, 0.001.

* **Ошибку второго рода**, обозначаемую $\beta$, так же называют называют False Negative в ML и ложноотрицательным исходом в биологии и медицине. По своему смыслу она описывает ситуацию, когда при реальном наличии эффекта мы говорим, что эффекта нет. Тогда формально её можно записать как $\beta = P(H_0|H_1)$.

В терминах A/B тестирования ошибку 1 рода так же называют уровнем значимости (и говорят, что проверяется гипотеза на уровне значимости $\alpha$). Ошибка второго рода не имеет явного названия, но она тесна связана с **мощностью** теста - способностью теста находить эффект там, где он есть (1 - $\beta$).

Чтобы легче понять эти определения, рассмотрим парочку разных примеров:

1. Компьютерная безопасность

Наличие уязвимостей в вычислительных системах приводит к тому, что приходится, с одной стороны, решать задачу сохранения целостности компьютерных данных, а с другой стороны — обеспечивать нормальный доступ легальных пользователей к этим данным. В данном контексте возможны следующие нежелательные ситуации:

ㅤ* когда авторизованные пользователи классифицируются как нарушители (ошибки первого рода);

ㅤ* когда нарушители классифицируются как авторизованные пользователи (ошибки второго рода);

2. Фильтрация спама

Ошибка первого рода происходит, когда механизм блокировки/фильтрации спама ошибочно классифицирует легитимное email-сообщение как спам и препятствует его нормальной доставке. В то время как большинство «антиспам»-алгоритмов способны блокировать/фильтровать большой процент нежелательных email-сообщений, гораздо более важной задачей является минимизировать число «ложных тревог» (ошибочных блокировок нужных сообщений).

Ошибка второго рода происходит, когда антиспам-система ошибочно пропускает нежелательное сообщение, классифицируя его как «не спам». Низкий уровень таких ошибок является индикатором эффективности антиспам-алгоритма.

3. Тесты на беременность

Так у нас есть две гипотезы: нулевая $H_0$ и альтернативная $H_1$. Пусть без потери общности альтернативная гипотеза подразумевает, что значение метрики в контрольной группе выросло на $+\delta$. Тогда мы пожем построить два распределения метрики: по одному на каждую из групп. Предварительны мы их стандартизируем, чтобы ошибки имели легко интерпретируемые значения.

**Стандартизация** - трансформация метрики, при которой среднее распределения сдвигается в ноль, а стандартное отклонение становится единицей:

$$
X_{stand} = \frac{X_i - \bar{X}}{\sigma}
$$

* **Ошибка 2 рода ($\beta$)** - площадь под кривой стандартизированного распределения метрики при верной $H_1$, которая левее (1-$\alpha$) квантиля (критической области) распределения $H_0$;

#### MDE & sample size calculation

На визуализации ошибки 2 рода в предыдущем блоке видно, как онf меняется при изменении $\delta$ - чем больше разница, тем меньше ошибка 2 рода. Интуитивно это кажется довольно понятным: чем более явной будет разница, тем легче её обнаружить, а значит выще мощность теста и меньше ошибка второго рода. Получается, что если мы заранее задаём ошибку 2 рода, то есть какая-то теоретическая $\delta_{max}$, больше модуля которой тест не сможет статистически обнаружить.

Вот эта $\delta_{max}$ так же называется **MDE (Minimum Detectable Effect)**. О нём проще говорить в терминах мощности, что MDE - это наименьший «различимый» эффект $\delta$, при котором при заданном уровне значимости $\alpha$ и требуемой мощности ($1-\beta$) мы всё ещё будем с вероятностью большей  $1-\beta$ отвергать $H_0$. Проще говоря, это такой эффект, при котором мощность теста достигает нужного уровня.

Заметим, что если мы зафиксируем $\alpha$, $\beta$ и $\delta$, то можно влиять на mde "ужимая" сами распределения метрик - ведь при фиксированном эффекте у нас пересечение распределений будет меньше, отчего и мощность увеличится. Ужимание распределения есть не что иное, как уменьшение дисперсии распределения, а как можно уменьшить дисперсию статистики? Верно, нужно набрать как можно больше наблюдений, чтобы выборка стала как можно репрезентативнее относительно генеральной совокупности. Таким образом появляется важная зависимость: чем больше размер выборки (sample size) в группах, тем больше MDE (то есть они прямо пропорциональны).

1. Определяем **доступный размер выборки**, а потом считаем по формулам **MDE**;

  Бывает ситуация, когда решение должно происходить от размера доступной выборки.

  1. Предположим, что бизнес-заказчику важна репутация своего продукта, поэтому он готов эксперементировать не более чем на X% выборки. Тогда логичным решением будет посчитать MDE и констатировать ему, что на выделенной группе пользователей не сможем увидеть эффекте менее, чем **MDE**. Далее по-хорошему соотнести расчитанный **MDE** с ожиданиями от внедрения изменения, и если ожидаемый эффект меньше, то необходимо либо настаивать на увеличении выборки до тех пор, пока не получим достаточно малый **MDE** (по сути 2 поход), либо вообще не проводить тест из-за его статистической незначимости.

  2. Продукт довольно молодой и готов экспериментировать сразу на всех своих пользователях, в таком случае (раз не можем увелить размер выборки) нужно просто посчитать MDE и держать в голове, что если нет изменения, то либо оно при текущем наборе пользователей крайне мало,, либо же эффекта нет.


2. Определяем **MDE**, а потом считаем по формулям **размер выборок**

  Этот путь исходит от того, что мы понимаем либо приблизительный эффект, которое изменение может за собой нести (например из профессионального опыта), либо же минимально экономически выгодный эффект, меньше которого мы не готовы принимать. Отсюда мы можем принять **MDE** за эти значения и уже выститывать достаточные размер выборки.

#### Статистические критерии

Теперь переходим непосредственно к статистическим критериям - лакмусовым бумажкам, которые говорят, а является ли **статистически значимой** разница в метриках между тестовой и контрольной группой.

[Статистически значимая разница](https://en.wikipedia.org/wiki/Statistical_significance) - разница, которая не является результатом случайности. "Статистичность" воплащается в p_value - вероятности получить такую и более экстримальную разницу при верной $H_0$.

Есть несколько категорий критериев:
  * Параметрические - критерии, которые применяются к данным с определёнными распределениями
  * Непараметрические - критерии, которые не делают предположений о форме распределения данных

Вообще критериев существует огромное множество ([тык](https://www.statskingdom.com/index.html) и [тык](https://www.biostathandbook.com/testchoice.html)), но в онлайн A/B зачастую применяются следующие:

##### Параметрические тесты

###### z-тест

Z-тест — это параметрический статистический тест, который используется для проверки гипотез о средних значениях одной или двух генеральных совокупностей. Z-тест традиционно применяется, когда известна дисперсия генеральной совокупности или когда размер выборки достаточно велик (как правило, более 30 наблюдений), чтобы выборочное стандартное отклонение могло служить надежной оценкой генерального. В последнем случае, благодаря Центральной Предельной Теореме (ЦПТ), распределение выборочных средних аппроксимируется нормальным распределением, что позволяет использовать Z-статистику.

**Z-тест используется для проверки следующих типов гипотез:**
*   Сравнение среднего одной выборки с известным значением генеральной совокупности.
*   Сравнение средних двух независимых выборок.
*   Сравнение доли в одной выборке с известной долей генеральной совокупности.
*   Сравнение долей двух независимых выборок.

\\

**Предпосылки Z-теста**

Для корректного применения Z-теста необходимо выполнение нескольких предпосылок:
1.  **Нормальность распределения выборочных средних**:
    *   Если исходные данные генеральной совокупности распределены нормально, то выборочные средние также будут распределены нормально.
    *   Если исходные данные не распределены нормально, но размер выборки достаточно велик, то согласно **[Центральной Предельной Теореме (ЦПТ)](https://ru.wikipedia.org/wiki/Центральная_предельная_теорема)**, распределение выборочных средних будет аппроксимироваться нормальным распределением. Это означает, что z-тест может быть использован даже для данных из ненормального распределения, например, экспоненциального.
2.  **Независимость наблюдений**: Все наблюдения в выборках должны быть независимыми друг от друга. Если данные зависимы (например, повторные измерения на одних и тех же объектах), то z-тест могут работать некорректно, приводя к завышенной ошибке первого рода.
3.  **Известная дисперсия генеральной совокупности или большой размер выборки**: Традиционно Z-тест требует знания стандартного отклонения генеральной совокупности $\sigma$. Однако на практике $\sigma$ часто неизвестна. В таких случаях, если размер выборки достаточно велик, выборочное стандартное отклонение $s$ может использоваться в качестве хорошей оценки $\sigma$.

\\

**Формулировка Z-критерия**

Рассмотрим наиболее распространенный случай Z-теста — **сравнение средних двух независимых выборок**.

**Нулевая гипотеза ($H_0$):** Средние значения двух генеральных совокупностей равны, или их разность равна некоторому гипотетическому значению ($\mu_1 - \mu_2 = D_0$). Чаще всего $D_0=0$, что означает отсутствие эффекта.
$H_0: \mu_1 = \mu_2$ (или $\mu_1 - \mu_2 = 0$)

**Альтернативная гипотеза ($H_1$):** Средние значения отличаются.
$H_1: \mu_1 \neq \mu_2$ (двусторонний тест)
$H_1: \mu_1 > \mu_2$ (односторонний тест, правый хвост)
$H_1: \mu_1 < \mu_2$ (односторонний тест, левый хвост)

**Z-статистика для сравнения средних двух независимых выборок:**

$$
Z = \frac{(\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}
$$

Где:
*   $\bar{X}_1$ и $\bar{X}_2$ — выборочные средние первой и второй групп соответственно.
*   $n_1$ и $n_2$ — размеры выборок первой и второй групп соответственно.
*   $\mu_1 - \mu_2$ — гипотетическая разность средних (часто $0$ под $H_0$).
*   $\sigma_1^2$ и $\sigma_2^2$ — дисперсии генеральных совокупностей первой и второй групп. Если они неизвестны, для больших выборок используются выборочные дисперсии $s_1^2$ и $s_2^2$ в качестве оценок.

\\

**Критерий принятия решения**

Решение о принятии или отклонении нулевой гипотезы основывается на сравнении Z-статистики с критическими значениями из стандартного нормального распределения или на сравнении **p-value** с уровнем значимости $\alpha$.

1.  **Метод p-value**:
    *   **p-value** — это вероятность получить наблюдаемые или более экстремальные результаты, если нулевая гипотеза верна.
    *   Если **p-value < $\alpha$**, нулевая гипотеза отвергается. Это означает, что наблюдаемые различия статистически значимы, и мы заключаем, что эффект существует.
    *   Если **p-value $\ge \alpha$**, нулевая гипотеза не отвергается. Это означает, что наблюдаемые различия не являются статистически значимыми, и мы не можем утверждать наличие эффекта.

2.  **Метод доверительного интервала (ДИ)**:
    * Доверительный интервал предоставляет диапазон значений, в котором с заданной вероятностью ( 1 - $\alpha$%) лежит истинное значение параметра генеральной совокупности.
    * Для сравнения двух средних, ДИ для разности средних строится следующим образом
    
$$\left(\bar{X}_1 - \bar{X}_2 \right) \pm Z_{1-\alpha/2} \cdot \sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}
$$
ㅤㅤㅤㅤ* Если значение 0 (или гипотетическая разность) **не** попадает в доверительный интервал, нулевая гипотеза отвергается.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71']

def two_sample_z_test(control: np.ndarray, test: np.ndarray, pop_std: float = None, alpha: float = 0.05) -> Dict[str, float]:
    n1, n2 = len(control), len(test)
    m1, m2 = np.mean(control), np.mean(test)
    diff = m2 - m1

    if pop_std is not None:
        s1 = s2 = pop_std
    else:
        s1 = np.std(control, ddof=1)
        s2 = np.std(test, ddof=1)

    se = np.sqrt(s1**2/n1 + s2**2/n2)
    z_score = diff / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return {'difference': diff, 'z_score': z_score, 'p_value': p_value, 'std_error': se}

def generate_data(mu_ctrl: float, mu_test: float, sigma: float, sample_size: int, seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    if seed is not None:
        np.random.seed(seed)
    control = np.random.normal(loc=mu_ctrl, scale=sigma, size=sample_size)
    test = np.random.normal(loc=mu_test, scale=sigma, size=sample_size)
    return control, test

def run_simulation(n_sim: int, sample_size: int, mu_ctrl: float, mu_test: float, sigma: float, pop_std: float = None, seed: int = None) -> np.ndarray:
    rng = np.random.default_rng(seed)
    p_values = np.empty(n_sim)
    for i in range(n_sim):
        ctrl = rng.normal(mu_ctrl, sigma, sample_size)
        test = rng.normal(mu_test, sigma, sample_size)
        res = two_sample_z_test(ctrl, test, pop_std)
        p_values[i] = res['p_value']
    return p_values

def plot_combined_distributions(
    control: np.ndarray,
    test: np.ndarray,
    pvalues_aa: np.ndarray,
    pvalues_ab: np.ndarray,
    alpha: float = 0.05,
    subtitle: str = ''
) -> None:
    """Объединённая визуализация распределений метрик и p-значений"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # 1. График распределения метрик
    sns.histplot(control, color=COLORS[0], alpha=0.6, kde=True, label='Контроль', ax=ax1)
    sns.histplot(test, color=COLORS[1], alpha=0.6, kde=True, label='Тест', ax=ax1)
    ax1.set_title('Распределение метрик в группах', fontsize=14)
    ax1.set_xlabel('Значение метрики', fontsize=12)
    ax1.set_ylabel('Плотность', fontsize=12)
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # 2. График распределения p-значений
    estimated_first_type_error = np.mean(pvalues_aa < alpha)
    estimated_power = np.mean(pvalues_ab < alpha)
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax2.plot(X, Y_aa, label='A/A')
    ax2.plot(X, Y_ab, label='A/B')
    ax2.plot([alpha, alpha], [0, 1], '--k', alpha=0.8)
    ax2.plot([0, alpha], [estimated_first_type_error]*2, '--b', alpha=0.8,
             label=f'Ошибка I рода = {estimated_first_type_error:.3f}')
    ax2.plot([0, alpha], [estimated_power]*2, '--r', alpha=0.8,
             label=f'Мощность = {estimated_power:.3f}')
    ax2.plot([0, 1], [0, 1], '--k', alpha=0.8)

    ax2.set_title('Распределение p-значений', fontsize=14)
    ax2.set_xlabel('p-value', fontsize=12)
    ax2.set_ylabel('Доля p-value < x', fontsize=12)
    ax2.legend(fontsize=12)
    ax2.grid()

    plt.suptitle(subtitle, y=1.02, fontsize=14)
    plt.tight_layout()
    plt.show()


params = {
    'mu_ctrl': 1000,
    'mu_test': 1005,
    'sigma': 100,
    'pop_std': 100,
    'sample_size': 10_000,
    'n_sim': 1_000,
    'alpha': 0.05,
    'seed': 66
}

# 1. Генерация данных
control, test = generate_data(
    mu_ctrl=params['mu_ctrl'],
    mu_test=params['mu_test'],
    sigma=params['sigma'],
    sample_size=params['sample_size'],
    seed=params['seed']
)

# 3. Симуляции
pvals_aa = run_simulation(
    n_sim=params['n_sim'],
    sample_size=params['sample_size'],
    mu_ctrl=params['mu_ctrl'],
    mu_test=params['mu_ctrl'],
    sigma=params['sigma'],
    pop_std=params['pop_std'],
    seed=params['seed']
)

pvals_ab = run_simulation(
    n_sim=params['n_sim'],
    sample_size=params['sample_size'],
    mu_ctrl=params['mu_ctrl'],
    mu_test=params['mu_test'],
    sigma=params['sigma'],
    pop_std=params['pop_std'],
    seed=params['seed'] + 1
)

# 4. Объединённая визуализация
subtitle = (f"Размер выборки: {params['sample_size']} | "
            f"Эффект: {params['mu_test']-params['mu_ctrl']} | "
            f"σ: {params['sigma']} | "
            f"α: {params['alpha']}")

plot_combined_distributions(
    control=control,
    test=test,
    pvalues_aa=pvals_aa,
    pvalues_ab=pvals_ab,
    alpha=params['alpha'],
    subtitle=subtitle
)
````

###### t-тест

t-тест — параметрический статистический тест, используемый для проверки гипотез о средних значениях генеральных совокупностей, когда дисперсия генеральной совокупности неизвестна. Он основан на t-распределении Стьюдента, которое учитывает неопределённость, вызванную оценкой дисперсии по выборке. t-тест особенно важен для малых выборок (n<30), но применим и для больших выборок.


**t-тест используется для проверки следующих гипотез**

* Сравнение выборочного среднего с гипотетическим значением генеральной совокупности $\mu_0$.
* Сравнение средних двух несвязанных групп.
* Сравнение средних измерений, полученных от одних и тех же объектов (зависимые выборки).

\\

**Предпосылки t-теста**

Для корректного применения t-теста необходимо выполнение нескольких предпосылок:
1.  **Нормальность распределения выборочных средних**:
    *   Если исходные данные генеральной совокупности распределены нормально, то выборочные средние также будут распределены нормально.
    *   Если исходные данные не распределены нормально, но размер выборки достаточно велик (n>30), то согласно **[Центральной Предельной Теореме (ЦПТ)](https://ru.wikipedia.org/wiki/Центральная_предельная_теорема)**, распределение выборочных средних будет аппроксимироваться нормальным распределением. Ежели выборка менее 30 человек, то t-статистика так сконструирована, чтобы делать поправки на большую диспперсию, из-за чего он более устойчив к нарушению этой предпосылки.
2.  **Независимость наблюдений**:
    *   **Классический двухвыборочный t-тест (Student's t-test)** предполагает, что сравниваемые выборки независимы. Это означает, что наблюдения в одной группе не должны влиять на наблюдения в другой группе. В A/B-тестировании это обычно достигается случайным разделением пользователей из генеральной совокупности на контрольную и тестовую группы.
    *   Однако на практике встречаются **зависимые выборки**. Например, когда одна и та же группа людей измеряется до и после воздействия. В таких случаях применение стандартного двухвыборочного t-теста для независимых выборок может привести к некорректным результатам, так как предположение о независимости нарушается. Для зависимых выборок используется **одновыборочный t-тест для попарных разностей** (Paired t-test), который рассматривает разницу между измерениями одной и той же единицы наблюдения.
3.  **Гомогенность дисперсий**     

    *   Классический t-тест предполагает, что дисперсии генеральных совокупностей, из которых взяты выборки, равны.
    *   Если дисперсии генеральных совокупностей различаются, особенно при несбалансированных размерах выборок (**сплит не 50 на 50**) или **малых выборках**, применение классического t-теста может привести к увеличению ошибки первого рода (ложноположительных срабатываний). В таких ситуациях рекомендуется использовать **t-тест Уэлча (Welch's t-test)**, который не предполагает равенства дисперсий. Важно отметить, что для **больших выборок** и сбалансированного сплита (50 на 50) классический t-тест часто ведет себя корректно даже при разных дисперсиях.

\\

**Формулировка t-критерия**

*   **Одновыборочный T-тест (сравнение среднего выборки с заданной константой)**

    $$t = \frac{\bar{X} - \mu_0}{s / \sqrt{n}}$$
    *   $\bar{X}$ — выборочное среднее.
    *   $\mu_0$ — заданное значение для сравнения.
    *   $s$ — выборочное стандартное отклонение.
    *   $n$ — размер выборки.

    Для этой статистики степени свободы $ddf = n-1$

*   **Двухвыборочный T-тест для независимых выборок (классический критерий Стьюдента при равных дисперсиях)**

    $$t = \frac{\bar{X}_1 - \bar{X}_2}{s_p \sqrt{1/n_1 + 1/n_2}}$$
    *   $\bar{X}_1$, $\bar{X}_2$ — выборочные средние двух групп.
    *   $s_p$ — объединенное стандартное отклонение.
    *   $n_1$, $n_2$ — размеры выборок.
    *   $s_p = \sqrt\frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}$
    *   Для большого размера выборки (когда $n_1=n_2=n$) и если обе выборки из одной генеральной совокупности с $\sigma$, то знаменатель упрощается до $\sqrt{2 \sigma^2 / n}$.

    Для этой статистики степени свободы $ddf = n_1 + n_2 - 2$

*   **Одновыборочный T-тест для попарных разностей (Paired T-test) для зависимых выборок**

    Эта задача переформулируется. Вместо сравнения средних двух выборок, создается новая выборка попарных разностей $D_i = X_{1i} - X_{2i}$. Затем применяется одновыборочный T-тест к этой выборке разностей, проверяя гипотезу о том, что среднее разностей равно нулю:
    $$t = \frac{\bar{D}}{s_D / \sqrt{n}}$$
    *   $\bar{D}$ — среднее попарных разностей.
    *   $s_D$ — стандартное отклонение попарных разностей.
    *   $n$ — количество пар наблюдений.

    Для этой статистики степени свободы $ddf = n-1$
    
    При случайном подборе отклонение разностей будет $\sigma_D = \sqrt{2}\sigma_{\text{gen}}$. Однако при неслучайном, например, парном подборе похожих объектов, $\sigma_D$ может быть существенно меньше $\sqrt{2}\sigma_{\text{gen}}$.

\\

**Критерии принятия решений:**

Решение о принятии или отклонении нулевой гипотезы основывается на сравнении t-статистики с критическими значениями из стандартного нормального распределения или на сравнении **p-value** с уровнем значимости $\alpha$.

1.  **Метод p-value**:
    *   **p-value** — это вероятность получить наблюдаемые или более экстремальные результаты, если нулевая гипотеза верна.
    *   Если **p-value < $\alpha$**, нулевая гипотеза отвергается. Это означает, что наблюдаемые различия статистически значимы, и мы заключаем, что эффект существует.
    *   Если **p-value $\ge \alpha$**, нулевая гипотеза не отвергается. Это означает, что наблюдаемые различия не являются статистически значимыми, и мы не можем утверждать наличие эффекта.

2.  **Метод доверительного интервала (ДИ)**:
    * Доверительный интервал предоставляет диапазон значений, в котором с заданной вероятностью ( 1 - $\alpha$%) лежит истинное значение параметра генеральной совокупности.
    * Для сравнения двух средних ДИ для разности средних строится следующим образом
    
$$\left(\bar{X}_1 - \bar{X}_2 \right) \pm t_{(1-\alpha/2,ddf)} \cdot \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}
$$
ㅤㅤㅤㅤ* Если значение 0 (или гипотетическая разность) **не** попадает в доверительный интервал, нулевая гипотеза отвергается. Если 0 попадает в интервал, нулевая гипотеза не отвергается.


````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, Optional
from tqdm import tqdm

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71']


def generate_data(mu_ctrl: float, mu_test: float, sigma: float, sample_size: int,
                 paired: bool = False, correlation: float = 0.5, seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    """Генерация данных для независимого или парного теста"""
    if seed is not None:
        np.random.seed(seed)

    if not paired:
        control = np.random.normal(loc=mu_ctrl, scale=sigma, size=sample_size)
        test = np.random.normal(loc=mu_test, scale=sigma, size=sample_size)
    else:
        # Генерация коррелированных данных для парного теста
        base = np.random.normal(loc=0, scale=sigma, size=sample_size)
        control = mu_ctrl + base + np.random.normal(loc=0, scale=sigma*np.sqrt(1-correlation**2), size=sample_size)
        test = mu_test + base + np.random.normal(loc=0, scale=sigma*np.sqrt(1-correlation**2), size=sample_size)

    return control, test


def two_sample_t_test(control: np.ndarray, test: np.ndarray, paired: bool = False) -> Dict[str, float]:
    """Двухвыборочный t-тест (независимый или парный)"""
    if paired:
        result = stats.ttest_rel(control, test)
    else:
        result = stats.ttest_ind(control, test, equal_var=True)

    diff = np.mean(test) - np.mean(control)
    return {
        'difference': diff,
        't_statistic': result.statistic,
        'p_value': result.pvalue,
        'mean_control': np.mean(control),
        'mean_test': np.mean(test)
    }

def run_simulation(n_sim: int, sample_size: int, mu_ctrl: float, mu_test: float,
                  sigma: float, paired: bool = False, seed: int = None) -> np.ndarray:
    """Запуск симуляций для оценки ошибок I рода и мощности"""
    rng = np.random.default_rng(seed)
    p_values = np.empty(n_sim)

    for i in tqdm(range(n_sim)):
        if not paired:
            ctrl = rng.normal(mu_ctrl, sigma, sample_size)
            test = rng.normal(mu_test, sigma, sample_size)
        else:
            # Для парного теста генерируем коррелированные данные
            base = rng.normal(0, sigma, sample_size)
            ctrl = mu_ctrl + base + rng.normal(0, sigma*0.5, sample_size)
            test = mu_test + base + rng.normal(0, sigma*0.5, sample_size)

        res = two_sample_t_test(ctrl, test, paired=paired)
        p_values[i] = res['p_value']

    return p_values

def plot_combined_distributions(
    control: np.ndarray,
    test: np.ndarray,
    pvalues_aa: np.ndarray,
    pvalues_ab: np.ndarray,
    alpha: float = 0.05,
    subtitle: str = '',
    paired: bool = False
) -> None:
    """Объединённая визуализация распределений метрик и p-значений"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # 1. График распределения метрик
    if not paired:
        sns.histplot(control, color=COLORS[0], alpha=0.6, kde=True, label='Контроль', ax=ax1)
        sns.histplot(test, color=COLORS[1], alpha=0.6, kde=True, label='Тест', ax=ax1)
    else:
        # Для парного теста показываем разницу
        diff = test - control
        sns.histplot(diff, color=COLORS[2], alpha=0.6, kde=True, label='Разница (Тест - Контроль)', ax=ax1)

    ax1.set_title('Распределение метрик в группах', fontsize=14)
    ax1.set_xlabel('Значение метрики', fontsize=12)
    ax1.set_ylabel('Плотность', fontsize=12)
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # 2. График распределения p-значений
    estimated_first_type_error = np.mean(pvalues_aa < alpha)
    estimated_power = np.mean(pvalues_ab < alpha)
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax2.plot(X, Y_aa, label='A/A')
    ax2.plot(X, Y_ab, label='A/B')
    ax2.plot([alpha, alpha], [0, 1], '--k', alpha=0.8)
    ax2.plot([0, alpha], [estimated_first_type_error]*2, '--b', alpha=0.8,
             label=f'Ошибка I рода = {estimated_first_type_error:.3f}')
    ax2.plot([0, alpha], [estimated_power]*2, '--r', alpha=0.8,
             label=f'Мощность = {estimated_power:.3f}')
    ax2.plot([0, 1], [0, 1], '--k', alpha=0.8)

    test_type = "Парный t-тест" if paired else "Двухвыборочный t-тест"
    ax2.set_title(f'Распределение p-значений ({test_type})', fontsize=14)
    ax2.set_xlabel('p-value', fontsize=12)
    ax2.set_ylabel('Доля p-value < x', fontsize=12)
    ax2.legend(fontsize=12)
    ax2.grid()

    plt.suptitle(subtitle, y=1.02, fontsize=14)
    plt.tight_layout()
    plt.show()

def run_full_analysis(params: Dict, paired: bool = False) -> None:
    """Полный анализ для t-теста (независимого или парного)"""
    # 1. Генерация данных
    control, test = generate_data(
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        sample_size=params['sample_size'],
        paired=paired,
        seed=params['seed']
    )

    # 2. Симуляции
    pvals_aa = run_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_ctrl'],
        sigma=params['sigma'],
        paired=paired,
        seed=params['seed']
    )

    pvals_ab = run_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        paired=paired,
        seed=params['seed'] + 1
    )

    # 3. Объединённая визуализация
    test_type = "Парный t-тест" if paired else "Двухвыборочный t-тест"
    subtitle = (f"{test_type} | Размер выборки: {params['sample_size']} | "
               f"Эффект: {params['mu_test']-params['mu_ctrl']} | "
               f"σ: {params['sigma']} | "
               f"α: {params['alpha']}")

    plot_combined_distributions(
        control=control,
        test=test,
        pvalues_aa=pvals_aa,
        pvalues_ab=pvals_ab,
        alpha=params['alpha'],
        subtitle=subtitle,
        paired=paired
    )

# Параметры для анализа
indep_params = {
    'mu_ctrl': 1000,
    'mu_test': 1005,
    'sigma': 100,
    'sample_size': 10_000,  # Уменьшил размер выборки для t-теста
    'n_sim': 10_000,
    'alpha': 0.05,
    'seed': 666
}

paired_params = {
    'mu_ctrl': 1000,
    'mu_test': 1005,
    'sigma': 100,
    'sample_size': 3_000,  # Уменьшил размер выборки для t-теста
    'n_sim': 10_000,
    'alpha': 0.05,
    'seed': 666
}


# Запуск анализа для двухвыборочного t-теста
print("Анализ для двухвыборочного t-теста (независимые выборки)")
run_full_analysis(indep_params, paired=False)

# Запуск анализа для парного t-теста
print("\nАнализ для парного t-теста")
run_full_analysis(paired_params, paired=True)
````

##### Непараметрические тесты

###### $\chi^2$-тест

χ²-тест — это непараметрический статистический тест, который используется для анализа категориальных данных.

В зависимости от задачи различают два основных вида χ²-теста:
* **«Goodness-of-Fit» test (тест согласия)** — проверка, насколько распределение наблюдаемых частот соответствует теоретическому распределению.
* **Independence test (тест независимости)** — проверка взаимосвязи (независимости) двух категориальных признаков в таблице копряженности (contingency table)

\\

**Предпосылки χ²-теста**  

* Независимость наблюдений: каждая единица выборки входит в таблицу только один раз.
* Адекватный объём выборки и ожидаемые частоты: в ячейках таблицы не должно быть слишком малых ожиданий (обычно $E_i \ge 5$).
* Категории взаимоисключающие и исчерпывающие: каждая наблюдаемая единица попадает ровно в одну категорию.

\\

**Формулировки χ² критериев**

1. **Тест согласия (Goodness-of-Fit)**

**Нулевая гипотеза ($H_0$):**  наблюдаемое распределение совпадает с теоретическим (или с заданным распределением)

**Альтернативная гипотеза ($H_1$):** наблюдаемое распределение отличается от теоретического  

Статистика:


$χ2 = \sum_{i=1}^k \frac{(O_i - E_i)^2}{E_i}$

Где
* $O_i$— наблюдаемая частота в $i$-й категории,
* $E_i$ — ожидаемая теоретическая частота в $i$-й категории,
• $k$ — число категорий.  

Степени свободы: $df=k−1−m$, где $m$ — число оцененных на основе данных параметров (если параметры распределения не фиксированы заранее).

\\

2) **Тест независимости**

Рассматриваем таблицу размера $r \times c $ (строки — категории одного фактора, столбцы — другого).

**Нулевая гипотеза ($H_0$):**  два признака независимы: $P(\text{row}_i, \text{col}_j) = P(\text{row}_i) \times P(\text{col}_j)$  

**Альтернативная гипотеза ($H_1$):** признаки зависимы: $P(\text{row}_i, \text{col}_j) \neq P(\text{row}_i) \times P(\text{col}_j)$  


Статистика:
$χ2 =
\displaystyle \sum_{i=1}^r \sum_{j=1}^c \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$

Степени свободы: $df=(r−1) (c−1)$

\\

**Критерий принятия решения**

Решение о принятии или отклонении нулевой гипотезы основывается на сравнении $\chi^2$-статистики с критическими значениями из стандартного нормального распределения или на сравнении p-value с уровнем значимости $\alpha$.

1. Метод p-value:

* p-value — это вероятность получить наблюдаемые или более экстремальные результаты, если нулевая гипотеза верна.
* Если p-value <  α , нулевая гипотеза отвергается. Это означает, что наблюдаемые различия статистически значимы, и мы заключаем, что эффект существует.
* Если p-value  ≥α , нулевая гипотеза не отвергается. Это означает, что наблюдаемые различия не являются статистически значимыми, и мы не можем утверждать наличие эффекта.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, List

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

def chi2_test(observed: np.ndarray, expected: np.ndarray = None) -> Dict[str, float]:
    """
    Проведение хи-квадрат теста для таблицы сопряженности
    observed - наблюдаемые частоты (2D массив)
    expected - ожидаемые частоты (если None, рассчитываются автоматически)
    """
    chi2, p_value, dof, expected = stats.chi2_contingency(observed)

    return {
        'chi2_statistic': chi2,
        'p_value': p_value,
        'degrees_of_freedom': dof,
        'expected_frequencies': expected
    }

def generate_contingency_table(n_samples: int, p_ctrl: List[float], p_test: List[float],
                              seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    """
    Генерация таблицы сопряженности для двух групп (контроль и тест)
    p_ctrl - вероятности категорий в контрольной группе
    p_test - вероятности категорий в тестовой группе
    """
    if seed is not None:
        np.random.seed(seed)

    # Нормализуем вероятности
    p_ctrl = np.array(p_ctrl) / np.sum(p_ctrl)
    p_test = np.array(p_test) / np.sum(p_test)

    # Генерируем данные
    control = np.random.multinomial(n_samples, p_ctrl)
    test = np.random.multinomial(n_samples, p_test)

    # Создаем таблицу сопряженности (категории по строкам, группы по столбцам)
    observed = np.vstack([control, test]).T

    return observed

def run_chi2_simulation(n_sim: int, n_samples: int, p_ctrl: List[float],
                        p_test: List[float], seed: int = None) -> np.ndarray:
    """
    Запуск симуляций хи-квадрат теста
    """
    rng = np.random.default_rng(seed)
    p_values = np.empty(n_sim)
    n_categories = len(p_ctrl)

    for i in range(n_sim):
        # Генерируем данные
        p_ctrl_norm = np.array(p_ctrl) / np.sum(p_ctrl)
        p_test_norm = np.array(p_test) / np.sum(p_test)

        control = rng.multinomial(n_samples, p_ctrl_norm)
        test = rng.multinomial(n_samples, p_test_norm)
        observed = np.vstack([control, test]).T

        # Проводим тест
        res = chi2_test(observed)
        p_values[i] = res['p_value']

    return p_values

def plot_chi2_results(
    observed: np.ndarray,
    expected: np.ndarray,
    pvalues_aa: np.ndarray,
    pvalues_ab: np.ndarray,
    alpha: float = 0.05,
    subtitle: str = '',
    categories: List[str] = None
) -> None:
    """
    Визуализация результатов хи-квадрат теста
    """
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

    # 1. График наблюдаемых и ожидаемых частот
    n_categories = observed.shape[0]
    if categories is None:
        categories = [f'Категория {i+1}' for i in range(n_categories)]

    x = np.arange(n_categories)
    width = 0.35

    # Наблюдаемые частоты
    ax1.bar(x - width/2, observed[:, 0], width, label='Контроль (набл.)', color=COLORS[0], alpha=0.7)
    ax1.bar(x + width/2, observed[:, 1], width, label='Тест (набл.)', color=COLORS[1], alpha=0.7)

    # Ожидаемые частоты (для контрольной группы)
    ax1.plot(x - width/2, expected[:, 0], 'o--', color=COLORS[2], label='Ожидаемые (контроль)')

    # Ожидаемые частоты (для тестовой группы)
    ax1.plot(x + width/2, expected[:, 1], 'o--', color=COLORS[3], label='Ожидаемые (тест)')

    ax1.set_title('Наблюдаемые и ожидаемые частоты', fontsize=14)
    ax1.set_xlabel('Категории', fontsize=12)
    ax1.set_ylabel('Частота', fontsize=12)
    ax1.set_xticks(x)
    ax1.set_xticklabels(categories)
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # 2. График относительных частот
    rel_control = observed[:, 0] / np.sum(observed[:, 0])
    rel_test = observed[:, 1] / np.sum(observed[:, 1])

    ax2.bar(x - width/2, rel_control, width, label='Контроль', color=COLORS[0], alpha=0.7)
    ax2.bar(x + width/2, rel_test, width, label='Тест', color=COLORS[1], alpha=0.7)

    ax2.set_title('Относительные частоты по группам', fontsize=14)
    ax2.set_xlabel('Категории', fontsize=12)
    ax2.set_ylabel('Доля', fontsize=12)
    ax2.set_xticks(x)
    ax2.set_xticklabels(categories)
    ax2.legend()
    ax2.grid(True, linestyle='--', alpha=0.7)

    # 3. График распределения p-значений
    estimated_first_type_error = np.mean(pvalues_aa < alpha)
    estimated_power = np.mean(pvalues_ab < alpha)
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax3.plot(X, Y_aa, label='A/A (одинаковые распределения)')
    ax3.plot(X, Y_ab, label='A/B (разные распределения)')
    ax3.plot([alpha, alpha], [0, 1], '--k', alpha=0.8)
    ax3.plot([0, alpha], [estimated_first_type_error]*2, '--b', alpha=0.8,
             label=f'Ошибка I рода = {estimated_first_type_error:.3f}')
    ax3.plot([0, alpha], [estimated_power]*2, '--r', alpha=0.8,
             label=f'Мощность = {estimated_power:.3f}')
    ax3.plot([0, 1], [0, 1], '--k', alpha=0.8)

    ax3.set_title('Распределение p-значений (хи-квадрат тест)', fontsize=14)
    ax3.set_xlabel('p-value', fontsize=12)
    ax3.set_ylabel('Доля p-value < x', fontsize=12)
    ax3.legend(fontsize=10)
    ax3.grid()

    plt.suptitle(subtitle, y=1.02, fontsize=14)
    plt.tight_layout()
    plt.show()

def run_chi2_analysis(params: Dict, categories: List[str] = None) -> None:
    """
    Полный анализ для хи-квадрат теста
    """
    # 1. Генерация данных
    observed = generate_contingency_table(
        n_samples=params['n_samples'],
        p_ctrl=params['p_ctrl'],
        p_test=params['p_test'],
        seed=params['seed']
    )

    # Получаем ожидаемые частоты
    _, _, _, expected = stats.chi2_contingency(observed)

    # 2. Симуляции
    pvals_aa = run_chi2_simulation(
        n_sim=params['n_sim'],
        n_samples=params['n_samples'],
        p_ctrl=params['p_ctrl'],
        p_test=params['p_ctrl'],  # Для A/A теста используем одинаковые распределения
        seed=params['seed']
    )

    pvals_ab = run_chi2_simulation(
        n_sim=params['n_sim'],
        n_samples=params['n_samples'],
        p_ctrl=params['p_ctrl'],
        p_test=params['p_test'],
        seed=params['seed'] + 1
    )

    # 3. Визуализация
    effect_size = np.sum(np.abs(np.array(params['p_test']) - np.array(params['p_ctrl']))) / 2
    subtitle = (f"Хи-квадрат тест | Размер выборки: {params['n_samples']} | "
               f"Категории: {len(params['p_ctrl'])} | "
               f"Размер эффекта: {effect_size:.3f} | "
               f"α: {params['alpha']}")

    plot_chi2_results(
        observed=observed,
        expected=expected,
        pvalues_aa=pvals_aa,
        pvalues_ab=pvals_ab,
        alpha=params['alpha'],
        subtitle=subtitle,
        categories=categories
    )

# Параметры для анализа (пример для 4 категорий)
params = {
    'p_ctrl': [30, 40, 20, 10],  # Распределение по категориям в контрольной группе
    'p_test': [25, 35, 30, 10],   # Распределение по категориям в тестовой группе
    'n_samples': 1000,            # Количество наблюдений в каждой группе
    'n_sim': 1000,                # Количество симуляций
    'alpha': 0.05,                # Уровень значимости
    'seed': 42                    # Seed для воспроизводимости
}

categories = ['Кат. A', 'Кат. B', 'Кат. C', 'Кат. D']  # Названия категорий

# Запуск анализа
run_chi2_analysis(params, categories)
````

###### Mann–Whitney U test

U-тест (тест Манна–Уитни) — это непараметрический критерий для сравнения двух независимых выборок по уровню распределений. Он не требует нормальности данных и основан на ранжировании объединённой выборки.

U-тест Манна–Уитни применяют, когда необходимо проверить, что одна выборка сдвинута по метрике относительно другой. **Его нельзя использовать для проверки равенства средних или медиан выборок** (подробно рассматрвиается [тут](https://habr.com/ru/companies/avito/articles/709596/))

**Предпосылки U-теста**  

* Независимость наблюдений внутри и между группами.
* Однородность формы распределений (однотипные «спреды») — в отсутствии сдвига медиан распределения должны быть одинаковыми по форме.
* Минимум порядкового уровня: достаточно уметь упорядочить (ранжировать) все наблюдения.

\\

**Формулировки U-теста**  

**Нулевая гипотеза ($H_0$):** распределения в двух группах совпадают $F_T = F_C$

**Альтернативная гипотеза ($H_1$):**  распределения в двух группах различаются $F_T \neq F_C$

\\

**U-статистика**

$$
U = \displaystyle \sum_{i}^{N_T} \sum_{j}^{N_C} \left[ T_i < C_j \right]
$$

\\

**Критерий принятия решения**

Решение о принятии или отклонении нулевой гипотезы основывается на сравнении U-статистики с критическими значениями из стандартного нормального распределения или на сравнении **p-value** с уровнем значимости $\alpha$.

1.  **Метод p-value**:
    *   **p-value** — это вероятность получить наблюдаемые или более экстремальные результаты, если нулевая гипотеза верна.
    *   Если **p-value < $\alpha$**, нулевая гипотеза отвергается. Это означает, что наблюдаемые различия статистически значимы, и мы заключаем, что эффект существует.
    *   Если **p-value $\ge \alpha$**, нулевая гипотеза не отвергается. Это означает, что наблюдаемые различия не являются статистически значимыми, и мы не можем утверждать наличие эффекта.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from typing import Tuple, Dict, List

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

def mann_whitney_u_test(control: np.ndarray, test: np.ndarray) -> Dict[str, float]:
    """
    Проведение U-теста Манна-Уитни для двух независимых выборок
    Возвращает статистику теста и p-value
    """
    u_stat, p_value = stats.mannwhitneyu(control, test, alternative='two-sided')

    # Расчет эффекта (ранговая корреляция)
    n1, n2 = len(control), len(test)
    r_effect = 1 - (2 * u_stat) / (n1 * n2)

    return {
        'u_statistic': u_stat,
        'p_value': p_value,
        'effect_size': r_effect,
        'median_control': np.median(control),
        'median_test': np.median(test)
    }

def generate_data(mu_ctrl: float, mu_test: float, sigma: float,
                 sample_size: int, dist: str = 'normal', seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    """
    Генерация данных для теста (нормальное или экспоненциальное распределение)
    """
    if seed is not None:
        np.random.seed(seed)

    if dist == 'normal':
        control = np.random.normal(loc=mu_ctrl, scale=sigma, size=sample_size)
        test = np.random.normal(loc=mu_test, scale=sigma, size=sample_size)
    elif dist == 'exponential':
        control = np.random.exponential(scale=mu_ctrl, size=sample_size)
        test = np.random.exponential(scale=mu_test, size=sample_size)
    else:
        raise ValueError("Unsupported distribution type")

    return control, test

def run_u_test_simulation(n_sim: int, sample_size: int, mu_ctrl: float,
                         mu_test: float, sigma: float, dist: str = 'normal',
                         seed: int = None) -> np.ndarray:
    """
    Запуск симуляций U-теста
    """
    rng = np.random.default_rng(seed)
    p_values = np.empty(n_sim)

    for i in tqdm(range(n_sim)):
        if dist == 'normal':
            ctrl = rng.normal(mu_ctrl, sigma, sample_size)
            test = rng.normal(mu_test, sigma, sample_size)
        elif dist == 'exponential':
            ctrl = rng.exponential(mu_ctrl, sample_size)
            test = rng.exponential(mu_test, sample_size)

        res = mann_whitney_u_test(ctrl, test)
        p_values[i] = res['p_value']

    return p_values

def plot_u_test_results(
    control: np.ndarray,
    test: np.ndarray,
    pvalues_aa: np.ndarray,
    pvalues_ab: np.ndarray,
    alpha: float = 0.05,
    subtitle: str = '',
    dist: str = 'normal'
) -> None:
    """
    Визуализация результатов U-теста Манна-Уитни
    """
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

    # 1. График распределений
    sns.kdeplot(control, color=COLORS[0], label='Контроль', ax=ax1, fill=True)
    sns.kdeplot(test, color=COLORS[1], label='Тест', ax=ax1, fill=True)
    ax1.axvline(np.median(control), color=COLORS[0], linestyle='--', label='Медиана контроля')
    ax1.axvline(np.median(test), color=COLORS[1], linestyle='--', label='Медиана теста')

    ax1.set_title('Распределение значений в группах', fontsize=14)
    ax1.set_xlabel('Значение', fontsize=12)
    ax1.set_ylabel('Плотность', fontsize=12)
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # 2. График рангов (позволяет понять природу U-статистики)
    combined = np.concatenate([control, test])
    ranks = stats.rankdata(combined)
    rank_control = ranks[:len(control)]
    rank_test = ranks[len(control):]

    ax2.scatter(control, rank_control, color=COLORS[0], alpha=0.5, label='Контроль')
    ax2.scatter(test, rank_test, color=COLORS[1], alpha=0.5, label='Тест')
    ax2.set_title('Ранги значений', fontsize=14)
    ax2.set_xlabel('Исходное значение', fontsize=12)
    ax2.set_ylabel('Ранг', fontsize=12)
    ax2.legend()
    ax2.grid(True, linestyle='--', alpha=0.7)

    # 3. График распределения p-значений
    estimated_first_type_error = np.mean(pvalues_aa < alpha)
    estimated_power = np.mean(pvalues_ab < alpha)
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax3.plot(X, Y_aa, label='A/A (одинаковые распределения)')
    ax3.plot(X, Y_ab, label='A/B (разные распределения)')
    ax3.plot([alpha, alpha], [0, 1], '--k', alpha=0.8)
    ax3.plot([0, alpha], [estimated_first_type_error]*2, '--b', alpha=0.8,
             label=f'Ошибка I рода = {estimated_first_type_error:.3f}')
    ax3.plot([0, alpha], [estimated_power]*2, '--r', alpha=0.8,
             label=f'Мощность = {estimated_power:.3f}')
    ax3.plot([0, 1], [0, 1], '--k', alpha=0.8)

    ax3.set_title('Распределение p-значений (U-тест)', fontsize=14)
    ax3.set_xlabel('p-value', fontsize=12)
    ax3.set_ylabel('Доля p-value < x', fontsize=12)
    ax3.legend(fontsize=10)
    ax3.grid()

    plt.suptitle(subtitle, y=1.02, fontsize=14)
    plt.tight_layout()
    plt.show()

def run_full_u_test_analysis(params: Dict) -> None:
    """
    Полный анализ U-теста Манна-Уитни
    """
    # 1. Генерация данных
    control, test = generate_data(
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        sample_size=params['sample_size'],
        dist=params['dist'],
        seed=params['seed']
    )

    # 2. Проведение теста на реальных данных
    test_result = mann_whitney_u_test(control, test)
    # print(f"Результат U-теста:\n"
    #       f"U-статистика = {test_result['u_statistic']:.2f}\n"
    #       f"p-value = {test_result['p_value']:.4f}\n"
    #       f"Размер эффекта (r) = {test_result['effect_size']:.3f}\n"
    #       f"Медиана контроля = {test_result['median_control']:.2f}\n"
    #       f"Медиана теста = {test_result['median_test']:.2f}")

    # 3. Симуляции
    pvals_aa = run_u_test_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_ctrl'],  # Для A/A теста
        sigma=params['sigma'],
        dist=params['dist'],
        seed=params['seed']
    )

    pvals_ab = run_u_test_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        dist=params['dist'],
        seed=params['seed'] + 1
    )

    # 4. Визуализация
    effect_size = test_result['effect_size']
    subtitle = (f"U-тест Манна-Уитни | Размер выборки: {params['sample_size']} | "
               f"Распределение: {params['dist']} | "
               f"Эффект: Δмедиан = {test_result['median_test']-test_result['median_control']:.2f} | "
               f"Размер эффекта (r): {effect_size:.3f} | "
               f"α: {params['alpha']}")

    plot_u_test_results(
        control=control,
        test=test,
        pvalues_aa=pvals_aa,
        pvalues_ab=pvals_ab,
        alpha=params['alpha'],
        subtitle=subtitle,
        dist=params['dist']
    )

# Параметры для анализа (нормальное распределение)
params_normal = {
    'mu_ctrl': 100,
    'mu_test': 102,
    'sigma': 15,
    'sample_size': 1_000,
    'n_sim': 5_000,
    'alpha': 0.05,
    'dist': 'normal',
    'seed': 666
}

# Параметры для анализа (экспоненциальное распределение)
params_exp = {
    'mu_ctrl': 10,
    'mu_test': 10.5,
    'sigma': 0,
    'sample_size': 9_000,
    'n_sim': 5_000,
    'alpha': 0.05,
    'dist': 'exponential',
    'seed': 666
}

print("Анализ для нормального распределения:")
run_full_u_test_analysis(params_normal)

print("\nАнализ для экспоненциального распределения:")
run_full_u_test_analysis(params_exp)
````

###### Bootstrap

Унивесальным критерием для любой метрики является метод бутстрапирования.

Общий папйплан будет:

* Много-много раз
  1. Делаем подвыборки с возвращением
  2. Считаем метрики из полученных выборок

* На основе полученных статистик делаем вывод о принятии/отклонении гипотезы

Подробнее методологию и разные виды бутстрапа рассматривали в [первой лекции](https://github.com/Alberelloo/Hypothesis-Testing-Course/tree/main/Lesson%2001%20-%20Statistics%20in%20Python%20%26%20Bootstrap)

````python
import numpy as np
from scipy.stats import bootstrap
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, List

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

def bootstrap_compare_means(control: np.ndarray,
                          test: np.ndarray,
                          n_resamples: int = 9999,
                          confidence_level: float = 0.95) -> Dict[str, float]:
    """
    Сравнение средних двух групп с помощью бутстрепа (современный метод)
    """
    # Функция для расчета разницы средних
    def diff_means(x, y):
        return np.mean(x) - np.mean(y)

    # Подготовка данных для scipy bootstrap
    data = (test, control)

    # Расчет доверительного интервала для разницы средних
    res = bootstrap(data, diff_means, n_resamples=n_resamples,
                   confidence_level=confidence_level,
                   method='BCa',  # Bias-corrected and accelerated
                   paired=False)

    # Расчет p-value через перцентильный метод
    observed_diff = diff_means(test, control)
    combined = np.concatenate([control, test])
    bs_diffs = np.empty(n_resamples)

    for i in range(n_resamples):
        bs_control = np.random.choice(combined, size=len(control), replace=True)
        bs_test = np.random.choice(combined, size=len(test), replace=True)
        bs_diffs[i] = diff_means(bs_test, bs_control)

    p_value = np.mean(np.abs(bs_diffs) >= np.abs(observed_diff))

    return {
        'difference': observed_diff,
        'ci_lower': res.confidence_interval.low,
        'ci_upper': res.confidence_interval.high,
        'p_value': p_value,
        'mean_control': np.mean(control),
        'mean_test': np.mean(test),
        'n_resamples': n_resamples
    }

def generate_data(mu_ctrl: float, mu_test: float, sigma: float,
                 sample_size: int, dist: str = 'normal', seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    """
    Генерация данных для теста
    """
    if seed is not None:
        np.random.seed(seed)

    if dist == 'normal':
        control = np.random.normal(loc=mu_ctrl, scale=sigma, size=sample_size)
        test = np.random.normal(loc=mu_test, scale=sigma, size=sample_size)
    elif dist == 'exponential':
        control = np.random.exponential(scale=mu_ctrl, size=sample_size)
        test = np.random.exponential(scale=mu_test, size=sample_size)
    elif dist == 'uniform':
        control = np.random.uniform(low=mu_ctrl, high=mu_ctrl+sigma, size=sample_size)
        test = np.random.uniform(low=mu_test, high=mu_test+sigma, size=sample_size)
    else:
        raise ValueError("Unsupported distribution type")

    return control, test

def run_bootstrap_simulation(n_sim: int, sample_size: int, mu_ctrl: float,
                           mu_test: float, sigma: float, n_resamples: int = 999,
                           dist: str = 'normal', seed: int = None) -> np.ndarray:
    """
    Запуск симуляций бутстреп-теста
    """
    rng = np.random.default_rng(seed)
    p_values = np.empty(n_sim)

    for i in tqdm(range(n_sim)):
        if dist == 'normal':
            ctrl = rng.normal(mu_ctrl, sigma, sample_size)
            test = rng.normal(mu_test, sigma, sample_size)
        elif dist == 'exponential':
            ctrl = rng.exponential(mu_ctrl, sample_size)
            test = rng.exponential(mu_test, sample_size)
        elif dist == 'uniform':
            ctrl = rng.uniform(mu_ctrl, mu_ctrl+sigma, sample_size)
            test = rng.uniform(mu_test, mu_test+sigma, sample_size)

        res = bootstrap_compare_means(ctrl, test, n_resamples=n_resamples)
        p_values[i] = res['p_value']

    return p_values

def plot_bootstrap_results(
    control: np.ndarray,
    test: np.ndarray,
    pvalues_aa: np.ndarray,
    pvalues_ab: np.ndarray,
    alpha: float = 0.05,
    subtitle: str = '',
    dist: str = 'normal'
) -> None:
    """
    Визуализация результатов бутстреп-теста
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))

    # 1. График распределений с доверительными интервалами
    sns.kdeplot(control, color=COLORS[0], label='Контроль', ax=ax1, fill=True)
    sns.kdeplot(test, color=COLORS[1], label='Тест', ax=ax1, fill=True)

    # Добавляем доверительные интервалы для средних
    def mean_stat(x):
        return np.mean(x)

    res_ctrl = bootstrap((control,), mean_stat, n_resamples=9999)
    res_test = bootstrap((test,), mean_stat, n_resamples=9999)

    ax1.axvspan(res_ctrl.confidence_interval.low, res_ctrl.confidence_interval.high,
                color=COLORS[0], alpha=0.2, label='CI контроля')
    ax1.axvspan(res_test.confidence_interval.low, res_test.confidence_interval.high,
                color=COLORS[1], alpha=0.2, label='CI теста')
    ax1.axvline(np.mean(control), color=COLORS[0], linestyle='--', label='Среднее контроля')
    ax1.axvline(np.mean(test), color=COLORS[1], linestyle='--', label='Среднее теста')

    ax1.set_title('Распределение значений с CI средних', fontsize=14)
    ax1.set_xlabel('Значение', fontsize=12)
    ax1.set_ylabel('Плотность', fontsize=12)
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # 3. График распределения p-значений
    estimated_first_type_error = np.mean(pvalues_aa < alpha)
    estimated_power = np.mean(pvalues_ab < alpha)
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax2.plot(X, Y_aa, label='A/A (одинаковые распределения)')
    ax2.plot(X, Y_ab, label='A/B (разные распределения)')
    ax2.plot([alpha, alpha], [0, 1], '--k', alpha=0.8)
    ax2.plot([0, alpha], [estimated_first_type_error]*2, '--b', alpha=0.8,
             label=f'Ошибка I рода = {estimated_first_type_error:.3f}')
    ax2.plot([0, alpha], [estimated_power]*2, '--r', alpha=0.8,
             label=f'Мощность = {estimated_power:.3f}')
    ax2.plot([0, 1], [0, 1], '--k', alpha=0.8)

    ax2.set_title('Распределение p-значений (бутстреп)', fontsize=14)
    ax2.set_xlabel('p-value', fontsize=12)
    ax2.set_ylabel('Доля p-value < x', fontsize=12)
    ax2.legend(fontsize=10)
    ax2.grid()

    plt.suptitle(subtitle, y=1.02, fontsize=14)
    plt.tight_layout()
    plt.show()

def run_full_bootstrap_analysis(params: Dict) -> None:
    """
    Полный анализ с помощью бутстрепа
    """
    # 1. Генерация данных
    control, test = generate_data(
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        sample_size=params['sample_size'],
        dist=params['dist'],
        seed=params['seed']
    )

    # 2. Проведение теста на реальных данных
    test_result = bootstrap_compare_means(
        control,
        test,
        n_resamples=params['n_resamples']
    )

    # print(f"Результат бутстреп-теста ({params['dist']} распределение):")
    # print(f"Среднее контроль: {test_result['mean_control']:.2f}")
    # print(f"Среднее тест: {test_result['mean_test']:.2f}")
    # print(f"Разница: {test_result['difference']:.2f}")
    # print(f"95% ДИ разницы: [{test_result['ci_lower']:.2f}, {test_result['ci_upper']:.2f}]")
    # print(f"p-value: {test_result['p_value']:.4f}")
    # print(f"Количество ресэмплов: {test_result['n_resamples']}")

    # 3. Симуляции
    pvals_aa = run_bootstrap_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_ctrl'],  # Для A/A теста
        sigma=params['sigma'],
        n_resamples=params['n_resamples'],
        dist=params['dist'],
        seed=params['seed']
    )

    pvals_ab = run_bootstrap_simulation(
        n_sim=params['n_sim'],
        sample_size=params['sample_size'],
        mu_ctrl=params['mu_ctrl'],
        mu_test=params['mu_test'],
        sigma=params['sigma'],
        n_resamples=params['n_resamples'],
        dist=params['dist'],
        seed=params['seed'] + 1
    )

    # 4. Визуализация
    pooled_std = np.sqrt((np.var(control) + np.var(test)) / 2)
    effect_size = test_result['difference'] / pooled_std if pooled_std != 0 else 0

    subtitle = (f"Бутстреп сравнение средних | Размер выборки: {params['sample_size']} | "
               f"Распределение: {params['dist']} | "
               f"Эффект: Δсредних = {test_result['difference']:.2f} | "
               f"Размер эффекта (Cohen's d): {effect_size:.3f} | "
               f"α: {params['alpha']}")

    plot_bootstrap_results(
        control=control,
        test=test,
        pvalues_aa=pvals_aa,
        pvalues_ab=pvals_ab,
        alpha=params['alpha'],
        subtitle=subtitle,
        dist=params['dist']
    )

# Параметры для анализа (нормальное распределение)
params_normal = {
    'mu_ctrl': 100,
    'mu_test': 105,
    'sigma': 15,
    'sample_size': 1_000,
    'n_sim': 200,
    'n_resamples': 5_000,
    'alpha': 0.05,
    'dist': 'normal',
    'seed': 666
}

# Параметры для анализа (экспоненциальное распределение)
params_exp = {
    'mu_ctrl': 10,
    'mu_test': 10.5,  # Разница в масштабе
    'sigma': 0,     # Не используется для экспоненциального
    'sample_size': 10_000,
    'n_sim': 100,
    'n_resamples': 5_000,
    'alpha': 0.05,
    'dist': 'exponential',
    'seed': 666
}

print("=== Анализ для нормального распределения ===")
run_full_bootstrap_analysis(params_normal)

print("\n=== Анализ для экспоненциального распределения ===")
run_full_bootstrap_analysis(params_exp)
````

### Проверка корректности критерия

Как видим, каждый критерий имеет свои предпосылки для своего принения. Если они будут нарушены, то критерией не будет гарантированно держать ошибки 1 и 2 рода в рамках номинально заданных. Таким образом, перед началом A/B теста важно не только выбрать критерий, но и проверить корректность его применения.

К сожалению, не все подпосылки можно как-либо проверить, из-за чего могут появляться сомнения о применимости статистического критерия на данных. В таком случае на практике проверяют не предпосылки, а сам критерий на исторических данных. Для этого проводят серию A/A и A/B тестов на части исторических данных.

* **A/A тест**

  **A/A тест** - это тест, в котором никакой группе не даётся эффекта (поэтому A/A), из-за чего фактически будет верна нулевая гипотеза $H_0$. Проверяя критерий на этих группах мы можем оценить ошибку 1 рода.

  *  Много-много раз:
    1. Берем исторические данные на тот же период, который мы хотим держать тест;
    2. Случайным образом делим данные на тестовую и контрольную группу;
    3. Применяем выбранный статистический критерий;
  * Считаем, в скольки случаях статистический критерий отверг $H_0$ - это число и будет оценкой ошибки 1 рода этого критерия на ваших данных.

* **A/B тест**

  **A/B тест** - это тест, в котором какой-то группе даётся номинальный эффект, из-за чего фактически будет верна альтернативная гипотеза $H_1$. Проверяя критерий на этих группах мы можем оценить ошибку 2 рода и мощность.

  *  Много-много раз:
    1. Берем исторические данные на тот же период, который мы хотим держать тест;
    2. Случайным образом делим данные на тестовую и контрольную группу;
    3. Назначаем тестовой группе воздейние, равное вашему MDE;
    3. Применяем выбранный статистический критерий;
  * Считаем, в скольки случаях статистический критерий отверг $H_0$ - это число и будет оценкой мощности этого критерия на ваших данных.


Так как цель проведения A/A и A/B тестов - оценить, как будет вести себя выбранный критерий на реальных данных, то мы делаем предположение, что те условия, в которых мы проводим эти тесты, совпадают с тем, что мы только планируем запускать. Это довольно тонкий момент, т.к.:
1. Не всегда будут исторические данные (например супер молодой продукт)
2. Исторические данные отличаются (data-drift) от тех, которые мы получим (например исторические данные собраны в период гендерных праздников или была волна хайпа, что сместила фокус покупок на товар X)

В первом случае у нас нет ничего лучше, кроме как делать предположения на основе эксперных предположений и проверять их выполнение пост-фактум.
Во втроом случае мы можем попытаться выбрать наиболее похожий исторический период к нашему и проводить тесты на нём.



````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, List
from tqdm import tqdm

# Настройки визуализации
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['font.size'] = 12
COLORS = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

def generate_historical_data(mu: float = 2500, std: float = 800,
                           n_samples: int = 10000, seed: int = None) -> np.ndarray:
    """
    Генерация исторических данных

    Parameters:
    -----------
    mu : float
        Среднее значение
    std : float
        Стандартное отклонение
    n_samples : int
        Количество наблюдений
    seed : int, optional
        Seed для воспроизводимости

    Returns:
    --------
    np.ndarray
        Массив с историческими данными
    """
    if seed is not None:
        np.random.seed(seed)
    return np.random.normal(mu, std, n_samples)

def run_simulations(data: np.ndarray, sample_size: int, effect: float = 0,
                  n_iter: int = 10000, seed: int = None) -> Tuple[np.ndarray, np.ndarray]:
    """
    Проведение симуляций A/A и A/B тестов

    Parameters:
    -----------
    data : np.ndarray
        Исторические данные
    sample_size : int
        Размер выборки для каждой группы
    effect : float
        Размер эффекта для группы B
    n_iter : int
        Количество итераций симуляции
    seed : int, optional
        Seed для воспроизводимости

    Returns:
    --------
    Tuple[np.ndarray, np.ndarray]
        Массивы p-значений для A/A и A/B тестов
    """
    rng = np.random.default_rng(seed)
    pvalues_aa = np.empty(n_iter)
    pvalues_ab = np.empty(n_iter)

    for i in tqdm(range(n_iter)):
        # A/A тест
        a = rng.choice(data, size=sample_size, replace=False)
        b = rng.choice(data, size=sample_size, replace=False)
        pvalues_aa[i] = stats.ttest_ind(a, b).pvalue

        # A/B тест
        a = rng.choice(data, size=sample_size, replace=False)
        b = rng.choice(data, size=sample_size, replace=False) + effect
        pvalues_ab[i] = stats.ttest_ind(a, b).pvalue

    return pvalues_aa, pvalues_ab

def calculate_error_rates(pvalues_aa: np.ndarray, pvalues_ab: np.ndarray,
                        alpha: float = 0.05) -> Dict[str, float]:
    """
    Расчет ошибок I и II рода

    Parameters:
    -----------
    pvalues_aa : np.ndarray
        P-значения для A/A тестов
    pvalues_ab : np.ndarray
        P-значения для A/B тестов
    alpha : float
        Уровень значимости

    Returns:
    --------
    Dict[str, float]
        Словарь с оценками ошибок и их доверительными интервалами
    """
    def _ci_bernoulli(p: float, n: int) -> Tuple[float, float]:
        """Расчет доверительного интервала для пропорции"""
        t = stats.norm.ppf(1 - alpha / 2)
        std_n = np.sqrt(p * (1 - p) / n)
        return p - t * std_n, p + t * std_n

    first_type_error = np.mean(pvalues_aa < alpha)
    second_type_error = np.mean(pvalues_ab >= alpha)
    power = 1 - second_type_error

    return {
        'first_type_error': first_type_error,
        'first_type_error_ci': _ci_bernoulli(first_type_error, len(pvalues_aa)),
        'second_type_error': second_type_error,
        'second_type_error_ci': _ci_bernoulli(second_type_error, len(pvalues_ab)),
        'power': power,
        'power_ci': _ci_bernoulli(power, len(pvalues_ab))
    }

def plot_results(pvalues_aa: np.ndarray, pvalues_ab: np.ndarray,
                alpha: float = 0.05, beta: float = 0.2) -> None:
    """
    Визуализация результатов симуляций

    Parameters:
    -----------
    pvalues_aa : np.ndarray
        P-значения для A/A тестов
    pvalues_ab : np.ndarray
        P-значения для A/B тестов
    alpha : float
        Уровень значимости
    beta : float
        Целевая вероятность ошибки II рода
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # График распределения p-значений
    X = np.linspace(0, 1, 1000)
    Y_aa = [np.mean(pvalues_aa < x) for x in X]
    Y_ab = [np.mean(pvalues_ab < x) for x in X]

    ax1.plot(X, Y_aa, label='A/A тесты', color=COLORS[0])
    ax1.plot(X, Y_ab, label='A/B тесты', color=COLORS[1])
    ax1.plot([alpha, alpha], [0, 1], '--k', alpha=0.5)
    ax1.plot([0, 1], [0, 1], '--k', alpha=0.5)
    ax1.set_title('Кумулятивное распределение p-значений', fontsize=14)
    ax1.set_xlabel('p-value', fontsize=12)
    ax1.set_ylabel('Доля p-value < x', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # График распределения p-значений (гистограммы)
    bins = np.linspace(0, 1, 21)
    ax2.hist(pvalues_aa, bins=bins, color=COLORS[0], alpha=0.5, density=True, label='A/A тесты')
    ax2.hist(pvalues_ab, bins=bins, color=COLORS[1], alpha=0.5, density=True, label='A/B тесты')
    ax2.axvline(alpha, color='k', linestyle='--', alpha=0.5)
    ax2.set_title('Распределение p-значений', fontsize=14)
    ax2.set_xlabel('p-value', fontsize=12)
    ax2.set_ylabel('Плотность', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def print_error_report(error_rates: Dict[str, float]) -> None:
    """
    Вывод отчета об ошибках

    Parameters:
    -----------
    error_rates : Dict[str, float]
        Словарь с оценками ошибок
    """
    print("\nОценка ошибок теста:")
    print(f"Ошибка I рода: {error_rates['first_type_error']:.4f}")
    print(f"  Доверительный интервал: [{error_rates['first_type_error_ci'][0]:.4f}, {error_rates['first_type_error_ci'][1]:.4f}]")
    print(f"Ошибка II рода: {error_rates['second_type_error']:.4f}")
    print(f"  Доверительный интервал: [{error_rates['second_type_error_ci'][0]:.4f}, {error_rates['second_type_error_ci'][1]:.4f}]")
    print(f"Мощность теста: {error_rates['power']:.4f}")
    print(f"  Доверительный интервал: [{error_rates['power_ci'][0]:.4f}, {error_rates['power_ci'][1]:.4f}]")

# Параметры эксперимента
SAMPLE_SIZE = 1500 # Необходимый размер выборки
ALPHA = 0.05       # Допустимая вероятность ошибки I рода
BETA = 0.2         # Допустимая вероятность ошибки II рода
MU_CONTROL = 2500  # Среднее в контрольной группе
EFFECT = 100       # Ожидаемый эффект
STD = 800          # Стандартное отклонение
N_SIM = 5000       # Количество симуляций
SEED = 42          # Seed для воспроизводимости


# 1. Генерация исторических данных
historical_data = generate_historical_data(mu=MU_CONTROL, std=STD, seed=SEED)

# 2. Проведение симуляций
pvalues_aa, pvalues_ab = run_simulations(
    historical_data, SAMPLE_SIZE, EFFECT, N_SIM, SEED
)

# 3. Расчет ошибок
error_rates = calculate_error_rates(pvalues_aa, pvalues_ab, ALPHA)
print_error_report(error_rates)

# 4. Визуализация
plot_results(pvalues_aa, pvalues_ab, ALPHA, BETA)
````

**Распределение p_value**

Если тест контролирует вероятность ошибки первого рода при любом заданном уровне значимости $\alpha$, то формально это можно записать так:

$$
\mathbb{P}(\text{p_value} < \alpha | H_0) = \alpha \quad  \forall \alpha \in [0, 1]
$$

Заметим, что в левой части равенства записано выражение для функции распределения p-value. Из равенства следует, что функция распределения p-value в точке X равна X для любого X от 0 до 1. Эта функция распределения является функцией распределения равномерного распределения от 0 до 1. Мы только что показали, что статистический критерий контролирует вероятность ошибки первого рода на заданном уровне для любого уровня значимости тогда и только тогда, когда при верности нулевой гипотезы p-value распределено равномерно от 0 до 1.

При верности нулевой гипотезы p-value должно быть распределено равномерно. А как должно быть распределено p-value при верности альтернативной гипотезы? Из условия для вероятности ошибки второго рода $\mathbb{P}(\text{p_value} \geq \alpha | H_1) = \beta$ следует, что $\mathbb{P}(\text{p_value} < \alpha | H_1) = 1 - \beta$.

Получается, график функции распределения p-value при верности альтернативной гипотезы должен проходить через точку [$\alpha$, $1 - \beta$], где $\alpha$ и $\beta$– допустимые вероятности ошибок конкретного эксперимента.


### Основные проблемы тестов

A/B-тестирование – это мощный инструмент для принятия решений, но его корректность сильно зависит от правильного применения статистических методов и учета потенциальных проблем. Рассмотрим типичные проблемы и возможные способы их решения.

#### Проблема поглядывания (peeking problem)

Предположим, что мы захотели где-то в середине эксперимента посмотреть на результаты p_value и мы видим такую картину, что его значение уже несколько дней находится ниже запланированной $\alpha$. Так может есть смысл заранее остановить эксперимент  заверить, что альтернативная гипотеза верна ?

\\

image.png

Интуитивно можно привести 2 аргемента **против** решения лосрочной остановки теста:

1. Посчитанное p_value есть величина от случайно набранной выборки. Таким образом её распределение тоже есть случайная величина, а, следовательно, может в конце начать двигаться в обратном направление и закончить за её пределами

\\

image.png

2. Мы же считали корректность нашего статистического критерия при фиксированной выборке. Если мы будет считать критерий на меньшей выборке, то не будет гарантий, что ошибки 1 и 2 рода будет контролироваться в рамках номинальных.

Многократная проверка результатов A/B-теста до достижения заранее определенного размера выборки или окончания фиксированного горизонта эксперимента называется "peeking". Основная проблема с peeking заключается в том, что оно искажает результаты теста, значительно увеличивая вероятность ошибки первого рода (ложнопоположительного результата).  Если мы, например, установили допустимый уровень ошибки первого рода в 5%, то при каждом взгляде на данные мы даем себе новую возможность увидеть случайное отклонение, которое превысит пороговое значение значимости.

Каждый статистический критерий имеет заданный уровень значимости, который является вероятностью ошибки первого рода. Когда мы многократно проверяем результаты в течение эксперимента, мы по сути выполняем несколько статистических тестов. Вероятность совершить хотя бы одну ошибку первого рода во всей серии тестов (семейная ошибка первого рода, FWER) становится значительно выше, чем заданный α для одного теста.

#### Несоблюдение требований RCT

Вспомним её раз, что A/B  тест - одна из реализации более глобальной методологии **Рандомизированного контролируемого испытания (RCT)**, основной смысл которой - создание **независиымх** **гомогенных** групп для сравнения путём **полной рандомизации**. Соответственно, даже при правильно подобранном и точно спланированном тесте мы не сможем с статистической точки зрения гарантировать верность результатов, если у нас будут проблемы либо с рандомизацией, независимостью , либо с гомогенностью групп.

\\

С **гомогенностью** групп проблему можно попытаться решить 2 способами:
  1. Если у нас достаточно большой набор групп, то попытаться выровнять группы по ключевым для неё показателям либо путём выброса лишних пользователе либо же методами [мэтчинга](https://arxiv.org/pdf/1010.5586)
  2. Пррибегнуть к методологиям из причинно-следственного анализа (Casual Inference), методы которого специально разработаны для таких случаев; (будут расмотрены в лекциях 9 и 10)

Если ни один из методов не может быть применен, то самым правильным решением будет признание теста нерепрезентативным и перезапуск с продуманной гомогенностью.

\\

С независимостью групп вопрос обстоит сложнее: в теории в каждом тесте полученные группы могут быть независимы в виду индивидуальных отличий пользователей. Тем не менее, есть тесты, где эта зависимость настолько незначительна, что ей просто принебрегают, а есть и те, где она неоспоримо вносит свой вклад (такими сферами обычно являются тесты в агрегаторах такси или социальных сетях.). В случаях с наличием зависимости обычно применяют другие подходы: switchback тесты (будем рассматривать на 8 лекции) или [кластерные A/B тесты](https://medium.com/@AnalyticsAtMeta/how-meta-tests-products-with-strong-network-effects-96003a056c2c), рассмотрение которых выходит за грани курса.

\\

С **рандомизацией** проблемы в основном возникают с технической стороны из-за двух причин:
  1. Не правильно работает сплиттер (не случайно разбиваются пользователи по группам)
  2. Есть проблемы интеграции сплиттера с продуктом (сам продукт почему-то не посылает пользователей в группу)

Первая причина может возникнуть крайне редко (если аналитики и разработчики не правильно реализовали процесс назначения группы пользователю) и обычно чинится же после замечания раз и на всегда.

Вторая же проблема может после починок появляться вновь, т.к. уже зависит и от продукта, который регулярно переписывается или интегрируется с новым функционалом.

Ввиду нерегулярности второго случая необходимо при проведении теста регулярно проверять, а правильно ли назначаются группы пользователям. Для этого есть 2 подхода:

**1. Sample Mismatch Ratio (SRM)**



Это критерий, которые проверяет правильность распределения пользователей по групам. По сути на каждый день мы коммулятивно считаем количество пользователй в каждой группе и сравниваем это распределение с теоретическим за счёт какого-либо подходящего статистического критерия. Если критерий показывает прокрас (статистически значимый результат), то что-то не так происходит с распределение пользователей, и нужно оперативно смотреть, что пошло не так.

Обычно поиски начинаются с инспекции сплиттера, чтобы убедиться, что сервис правильно назначает группы пользователей, и что проблема заложена где-то на стороне продукта.

Если проблема не со сплиттером, то начинают смотреть распределения групп в основных разрезах продукта (платформа, страна, регион, версия продукта...), чтобы локализовать проблему и найти причину неправильного разбиения.

**2. A/A/B тест**

Если же мы не хотим отдельно считать SRM, то можно пробовать проводить A/A/B тест - то есть одновременно проводить обычно A/B тест, но ещё рядышком так же держать одну группу без воздействия, которую необхдоимо сравнивать с контрольной (отсюда и название A/A/B). То есть мы в тесте сравниваем две A группы между собой на отсутствие эффекта (которого нет), и только потом уже сравниваем с тестовой группой, чтбоы понять наличие и размер эффекта.

В таком подходе есть 2 момента, которые необходимо понимать:

1. Из-за необходимости ввести ещё одну контрольную группу мы будем вынуждены повысить размер выборки для теста.

Если мы можем позволить увеличить размер выборки ещё половину, то никаких проблем не возникает; если же не можем, то можно попробовать фиктивно разбить контрольную группу на 2 подгруппы, которые будем сравнивать между собой, а сравнивать тестовую группу будем уже с изначально запланированной. Такой подход может быть рабочим, если полученные подгруппы будет адекватных размеров (-> требует дополнительной проверки), а так же если в механизме рандомизации не наклдаывается никаких дополнительных параметров на участников теста (Например все пользователи этого теста получают какой-то уникальный аттрибут, из-за которого могут возникать проблемы на стороне продукта. В таком случае мы не сможем увидеть проблемы в разбиении на A/A тесте, так как обе группы имеют этот аттрибут.)

2. Раз у нас есть 2 контрольных группы, то с какой из них надо же сравнивать?


Самым естественным и правильным исходом является объединение двух контрольных групп в одну со следующим сравнением с группой B. Всякие альтернативные методы (например сравнение с обоими группами или сравнение с "лучшей" контрольной группой) являются значительно хуже (если вообще правильными) ествественного. Подробнее об этом написато [здесь](https://koch-kir.medium.com/не-стоит-проводить-а-а-в-тест-936e9e7a3b96).
