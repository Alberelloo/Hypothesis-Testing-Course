### Байесовский vs Частотный подход

**Исторический контекст и философские различия: Частотный и байесовский подходы в статистике**

Возникновение современных статистических парадигм уходит корнями в фундаментальный вопрос эпистемологии: **какова природа вероятности?** Два принципиально разных ответа на этот вопрос сформировали дихотомию, определяющую развитие статистической науки на протяжении трех столетий. **Частотный (Classical/Frequentist)** подход, ассоциируемый с именами Рональда Фишера, Ежи Неймана и Карла Пирсона, и **байесовский (Bayesian)** подход, восходящий к работам Томаса Байеса и Пьера-Симона Лапласа, представляют не просто альтернативные методы расчета, но принципиально различные философские системы интерпретации неопределенности.

**Гносеологические истоки: две концепции вероятности**

В основе расхождения лежит противоположное понимание самой категории вероятности.

* **Частотный подход**

Частотный подход, достигший расцвета в первой половине XX века, трактует вероятность как **объективную меру долгосрочной частоты события**. Эта концепция предполагает мысленный эксперимент с бесконечным числом идентичных повторений: вероятность события $P(A)$ есть предел его относительной частоты при неограниченном росте испытаний. Например, вероятность выпадения орла у "честной" монеты строго равна 0.5, поскольку лишь это значение обеспечивает сходимость относительной частоты к константе при бесконечных подбрасываниях. Такой взгляд естественным образом порождает ключевые принципы частотной статистики: параметры модели (скажем, истинная конверсия $p%) считаются фиксированными, хотя и неизвестными константами, лишенными вероятностных распределений, тогда как данные интерпретируются как случайные реализации, порожденные этими параметрами. Центральным становится понятие повторяемости эксперимента — валидность статистических выводов (оценок, p-value, доверительных интервалов) проверяется через их поведение в гипотетической бесконечной серии идентичных исследований. Хотя декларируемая объективность частично нивелируется субъективностью выбора модели и уровня значимости, частотная парадигма стремится минимизировать влияние исследователя на результат.

* **Байесовский подход**

Байесовский подход, получивший формальное основание в работе Байеса (1763) и развитый Лапласом, предлагает радикально иную эпистемологическую позицию. Здесь вероятность есть **субъективная мера степени уверенности в истинности утверждения**, основанная на доступной информации. Такой взгляд не требует мысленных повторений эксперимента — вероятность становится инструментом рационального мышления в условиях неполного знания. Например, утверждение "вероятность дождя завтра равна 70%" отражает не частоту метеорологических событий, а степень обоснованной убежденности говорящего с учетом данных метеоррадаров, синоптических карт и прошлого опыта. Это влечет кардинальное отличие в аксиоматике: параметры модели (такие как конверсия $p$) суть случайные величины, наделенные распределениями вероятностей, количественно выражающими нашу неопределенность относительно их значений. Наблюдаемые данные при этом фиксированы — они не "генерируются", а являются данностью. Статистический вывод превращается в последовательное обновление убеждений: априорное распределение (prior), кодирующее знания до эксперимента, преобразуется в апостериорное распределение (posterior) через механизм условной вероятности при поступлении новых данных. Явное включение предварительных знаний делает байесовский подход неизбежно субъективным, но одновременно предоставляет мощный аппарат для инкорпорирования экспертной информации и результатов предыдущих исследований.

**Философское противостояние и методологические следствия**

Философские основания подходов порождают контрастные принципы принятия решений. Частотная парадигма, укорененная в позитивизме, рассматривает мир как объективную реальность с фиксированными параметрами. Её методология нацелена на создание процедур с гарантированными долгосрочными свойствами: 95% доверительный интервал должен покрывать истинный параметр ровно в 95% гипотетических повторений, а критерий значимости — контролировать частоту ложных срабатываний. Ключевой инструмент — **p-value** — интерпретируется как вероятность наблюдать данные столь же или более экстремальные, чем полученные, при условии истинности нулевой гипотезы. Важно подчеркнуть: это не вероятность того, что гипотеза верна или неверна, а мера согласия данных с моделью в рамках частотной метафоры повторяемости.

Байесовская методология, близкая к философскому прагматизму и субъективному байесианству де Финетти, переносит фокус на обновление знаний. Апостериорное распределение даёт прямые вероятностные утверждения: "вероятность того, что вариант A превосходит B, составляет 92%" или "с вероятностью 90% прирост конверсии лежит в интервале [0.3%, 2.7%]". Такие формулировки естественным образом соответствуют интуитивному запросу лиц, принимающих решения. Более того, байесовский вывод когерентен в техническом смысле — соблюдение аксиом вероятности гарантирует непротиворечивость выводов, чего нельзя строго утверждать о частотных методах.

Парадокс Линдли (1957) ярко демонстрирует методологический разрыв. Рассмотрим ситуацию: частотный критерий при большом объёме данных даёт очень малое p-value (<0.01), формально требуя отвергнуть H0. Однако байесовский анализ, учитывающий априорную малую вероятность альтернативы, может дать высокую апостериорную вероятность H0 (>95%). Противоречие возникает из-за разного учёта "редкости" событий: частотник вычисляет вероятность данных при фиксированной H0, байесианец взвешивает правдоподобие обеих гипотез с учётом их априорной вероятности. Этот пример показывает, что статистическая значимость не тождественна практической значимости, а игнорирование априорных знаний может приводить к контр-интуитивным выводам.

**Практическая конвергенция и области применения**

Несмотря на философские расхождения, при определённых условиях методы демонстрируют конвергенцию. Теорема Бернштейна — фон Мизеса (1940) устанавливает, что при увеличении объёма данных апостериорное распределение сходится к нормальному с центром в частотной оценке и дисперсией, обратной информации Фишера.

$$
|| P(θ|x_1, \cdots x_n) - N(\hat{\theta}_n, n^{-1}\mathcal{I}(\theta_0)^{-1})||_{TV} \rightarrow 0
$$

Более того, многие частотные процедуры допускают байесовскую интерпретацию: например, МНК-оценка совпадает с апостериорным средним при гауссовских ошибках и равномерном априорном распределении. Однако на малых выборках или при наличии содержательных априорных знаний различия становятся критически важными.

Исторически частотные методы доминировали в прикладных науках XX века благодаря развитой теории проверки гипотез и доступности таблиц распределений. Ситуация начала меняться с компьютерной революцией: методы Монте-Карло для байесовских вычислений (Metropolis-Hastings, Gibbs sampling) преодолели аналитические ограничения, а рост мощности процессоров сделал их доступными. Сегодня выбор парадигмы определяется конкретной задачей. Частотный подход сохраняет преимущества в задачах, требующих стандартизированных протоколов: клинические испытания (регулируемые ICH E9), индустриальный контроль качества (карты Шухарта), ситуации с жёсткими требованиями к контролю ошибок I/II рода. Его сила — в воспроизводимости процедур и независимости от субъективных предпосылок, что критично в регулируемых отраслях.

Байесовский подход демонстрирует превосходство в динамичных средах, где решения принимаются последовательно. В онлайн-экспериментах (A/B/n-тесты) он позволяет непрерывно мониторить вероятность превосходства вариантов без поправок на множественные сравнения. При ограниченных данных (пилотные исследования, редкие события) включение априорной информации через соответствующие распределения (Beta для пропорций, Gamma для частот) повышает эффективность. Интерпретируемость вывода ("вероятность улучшения 95%") ускоряет коммуникацию с бизнес-аудиторией.

**Эволюционный синтез и современные тенденции**

Историческое противостояние постепенно сменяется методологическим синтезом. С одной стороны, частотные заимствуют байесовские идеи: байесовские p-value, эмпирические байесовские методы (James-Stein оценка), false discovery rate (FDR) контролируют ожидаемую долю ошибок, что ближе к байесовскому духу. С другой — байесовцы разрабатывают "объективные" априорные распределения, минимизирующие влияние исследователя. Возникающие гибридные методы (frequentist-Bayes) используют байесовские модели для генерации гипотез и частотные методы для их валидации.

В контексте A/B тестирования этот синтез проявляется в разработке адаптивных дизайнов, где байесовские методы применяются для промежуточных решений (остановка/продолжение набора), а финальный вывод верифицируется частотными критериями. Подобная конвергенция отражает общую тенденцию: современная прикладная статистика всё реже выступает как "чистая" парадигма, а скорее как инструментарий, где выбор метода определяется не доктриной, а природой данных, объёмом выборки, наличием априорной информации и требуемой формой ответа. Понимание же исторических корней и философских оснований остаётся необходимым условием для осознанного выбора в этом расширяющемся методологическом пространстве.

### Байесовская статистика

#### Теоретическая справка

В основе всей байесовской статистики лежит формула, выведенная Томасом Байесом в XVIII веке. Пусть $A$ и $B$ — два события в вероятностном пространстве, причём $P(B)>0$. Теорема Байеса утверждает, что условная вероятность события $A$, при условии что произошло $B$, выражается через  
$$
P(A\mid B) = \frac{P(B\mid A)\,P(A)}{P(B)} = \frac{P(B\mid A)\,P(A)}{\displaystyle \sum_{A' \in A}P(B\mid A')\,P(A')}
$$

* $P(A)$ называется априорной вероятностью
* $P(B\mid A)$ — условной вероятностью наблюдения $B$ при гипотезе $A$.
* $P(B)= \displaystyle\sum_{A' \in A}P(B\mid A')\,P(A')$ приводит все вероятности к единице (нормировка).

\\

В большинстве статистических задач объектом интереса является не дискретный «список гипотез», а непрерывный вектор параметров $\theta \in \Theta$ и наблюдения (данные) $D$. В этом случае мы оперируем плотностями $p(\theta)$ и $p(D\mid\theta)$, а теорема Байеса переписывается в виде

$$
p(\theta\mid D) = \frac{p(D\mid \theta)\,p(\theta)}{p(D)} = \frac{p(D\mid \theta)\,p(\theta)}{\displaystyle\int_{\Theta}p(D\mid \theta)\,p(\theta)\,d\theta}
$$


В этой формуле
* $p(\theta)$ есть априорная плотность - отражает наше начальное представление о параметре до получения данных;  
* $p(D\mid\theta)$ есть likelihood, «правдоподобие» данных - показывает, с какой «силой» каждая гипотеза $\theta$ объясняет наблюдения;  
* $p(D)=\int_{\Theta}p(D\mid \theta)\,p(\theta)\,d\theta$ - маргинальное правдоподобие (evidence) и играет роль нормировочной константы, гарантируя, что $p(\theta\mid D)$ интегрируется в единицу.

* $p(\theta\mid D)$ - апостериорная вероятность - это обновлённое распределение параметра после учёта данных.

\\

Таким образом, базовый путь байесовского вывода состоит из следующих шагов:

1. **Выбор приора**

  Приор формализует знания о параметрах до наблюдения данных. Выбор приора — ключевой аспект байесовского моделирования, т.к. неправильный его выбор может напрямую изменить результат, посчитанный по формуле Байеса. Приоры бывают 2 видов:

  * **Неинформативные** -  Минимизируют влияние на постериорное распределение тем, не дают никакой особо информации (аналогично частотному подходу). Обычно задаются распределениями, которые равномерны на необходимом премежутке данных. ($\text{Beta}(0.5, 0.5)$ или $\text{Beta}(1, 1)$)

  * **Информативные** - Кодируют экспертные знания о распределении параметра. ($\text{Beta}(50, 50)$ для 50% конверсии или $\text{Gamma}(16, 125)$ для ARPU со средним значением в 2k)

  Зачастую выбор приора является балансом между гибкостью и точность распределения в параметризации и **сопряжённости**. Так же выбранное распределение будет влиять на то, каким образом будет дальше идти подсчёт всего уровнения.

2. **Построение правдоподобия**

  На этом этапе задаётся вероятностная модель данных при данном гипотетическом значении параметра $\theta$. То есть, определяется, как данные генерируются в теории, если известно значение $\theta$. Обычно это выглядит как функция вероятности (или плотности) данных p(D ∣ θ)p(D\,|\,\theta)p(D∣θ).

Например, для задачи с бинарными исходами (успех/неуспех), это может быть распределение Бернулли:
$$
p(D\mid θ) = \theta^{x} (1 - \theta)^{n - x}
$$

где $x$ — количество успехов, $n$ — общее число испытаний.

3. **Обновление априорных знаний**

  Этот шаг — основной в байесовском выводе, он позволяет объединить наши начальные предположения и новые данные, чтобы получить полное обновлённое понимание вероятностей параметров. Используя формулу Байеса, мы вычисляем апостериорное распределение — обновлённую вероятность гипотезы $\theta$ после получения данных.  

  Этот момент самый тяжёлый,так как как-то подсчитать маргинальное правдоподобие не представляется вычислительно возможным. В этом случае есть 2 развилки:

  * Удобный выбори приора: если мы выбираем приор таким образом, чтобы он был **[сопряжён](https://ru.wikipedia.org/wiki/Сопряжённое_априорное_распределение)** с правдоподобием, то мы можем аналитически подсчитать сразу апостериорное распределение, минуя подсчёты маргинального правдоподобия. Например, для конверсий (бернуллиевской величины) часто приором берут $\text{Beta}(\alpha, \beta)$, что гарантирует, что постериор будет тогда распределён как $\text{Beta}(\alpha + x, \beta + n - x)$.

  * Численные методы: если не получается выбрать сопряжённый приор, то остаётся только применять численные методы для подсчёта маргинального правдоподобия:
    * [MCMC (Monte-Carlo Markov Chain)](https://ru.wikipedia.org/wiki/Марковская_цепь_Монте-Карло) - строит цепи Маркова, у которых стационарное распределение — искомый апостериор. Это обеспечивает возможность генерировать выборки из апостериорного распределения, по которым далее можно делать оценки всей выборки. Основными [алгоритмами](https://m-clark.github.io/docs/ld_mcmc/) являются алгоритм Гиббса, Метрополиса–Гастингса. алгоритм Ланжевена ...
    * [Вариационный вывод (Variational Inference)](https://ermongroup.github.io/cs228-notes/inference/variational/) - аппроксимирует апостериорное распределением из параметрического семейства $q_\phi(\theta)$, минимизируя расстояние Кульбака-Лейблера (KL-дивергенцию).

    * [Метод Лапласа](https://en.wikipedia.org/wiki/Laplace%27s_method) - аппроксимирует лог-апостериор квадратичной функцией около его моды, что позволяет получить приближенную нормальную аппроксимацию.

4. **Интерпретаия постериорных значений**

  Как только мы обновили приорные знания с помощью правлободобия и данных ,мы получаем апостериорное распределения параметра $\theta$. Вспоминаем, что в байесовской парадигме оцениваемые параметры не скаляры, а целые вероятностные распределения. Это, в свою очередь, имеет очень приятное последствие - более простая интерпретация. Аналогом **доверительных интервалов (confidence intervals)** в частотной подходе здесь будут **достоверные интервалы (credibility intervals)**. Давайте сравним их интепретацию:

  * **95% Доверительный интервал** - при многократном сэмплировании 95% доверительных интервалов будут покрывать действительное значение параметра $\theta$;
  * **95% Достоверный интервал** - 95% вероятность, что $\theta$ лежит в данном интервале;

  Получается, что достоверные интервалы как раз и дают ответ на вопрос, который подсознательно мы себе задаём при тестировании гипотез (какой шанс, что группа B лучше). Простота интерпретации - один из больших плюсов этого подхода, т.к. здесь он аинтуитивно понятна всем.

##### Кейс для понимания

Давай на кейсе попробуем посчитать постериор с сопряжённым приором:

Компания MM Manufacturing славится производством высокоточных весов. Их система первоклассная, обеспечивающая практически идеальную калибровку большинства изделий. Однако в последнее время возникла необычная проблема – всплеск жалоб клиентов на неидеальные веса. В ответ на это MM Manufacturing ввела дополнительный уровень человеческой проверки, чтобы гарантировать безупречность всех отправляемых клиентам весов.

Чтобы проанализировать динамику фактических производственных весов, они поручили своей команде Data Science проанализировать вероятность появления таких нестандартных весов и, что ещё важнее, отслеживать распределение таких случаев, чтобы понять, как повысить производительность. К счастью, вся продукция досконально документируется сходя с конвейера.

Подход команды основан на байесовской калибровке. Каждый месяц они обновляют функцию плотности вероятности бета-распределения, чтобы оценить аномалию в весах и её динамику с течением времени. Для этого они используют данные с конвейера, которые служат наблюдениями для окончательного определения апостериорной вероятности. Им также необходимо установить приорное распределение, которое может основываться на исторических данных, знаниях предметной области или неспецифическом, неинформативном распределении.

Для значений $\alpha$ и $\beta$ в бета-распределении наблюдаемых данных или оценки правдоподобия они выбирают следующую стратегию:

* $\alpha$ = Количество правильно откалиброванных весов + 1 (что обеспечивает $\alpha$ > 0)
* $\beta$ = Количество неправильно откалиброванных весов + 1 (что обеспечивает $\beta$ > 0)


Что касается выбора априорной вероятности, они изначально выбирают неинформативную, представленную $\text{Beta}(1, 1)$ (равномерное распределение, как показано ниже), что минимизирует влияние априорной вероятности на апостериорную и делает основную ставку на данные наблюдений.

```python
import random
import matplotlib.pyplot as plt
from scipy.stats import beta
np.random.seed(42)

# Симулируем данные: 500 наблюдений с 90% нормальных and 10% аномальных весов
normal_instances =  [random.uniform(4.85, 5.15) for i in range(450)]
anomalous_instances_1 =  [random.uniform(3, 4.85) for i in range(25)]
anomalous_instances_2 =  [random.uniform(5.15, 6) for i in range(25)]

data = np.concatenate((normal_instances, anomalous_instances_1, anomalous_instances_2))

# Наш приор Beta распределения (неинформативный приор)
prior_alpha = 1
prior_beta = 1

# Beta распределение на данных
likelihood_alpha = len(data[(data >= 4.85) & (data <= 5.15)]) + 1
likelihood_beta = len(data) - likelihood_alpha + 1

# Сччитаем параметры постериора на основе правдоподобия и приора
posterior_alpha = prior_alpha + likelihood_alpha
posterior_beta = prior_beta + likelihood_beta

# Изобразим приор, правдоподобие и постериор на бета распределении
x = np.linspace(0, 1, 1000)
prior_distribution = beta.pdf(x, prior_alpha, prior_beta)
likelihood_distribution = beta.pdf(x, likelihood_alpha, likelihood_beta)
posterior_distribution = beta.pdf(x, posterior_alpha, posterior_beta)


# Создание фигуры с двумя субплoтами
plt.figure(figsize=(18, 7))

# 2. Байесовское обновление (главный график)
x = np.linspace(0, 1, 1000)
prior = beta.pdf(x, prior_alpha, prior_beta)
likelihood = beta.pdf(x, likelihood_alpha, likelihood_beta)
posterior = beta.pdf(x, posterior_alpha, posterior_beta)

# Основные кривые
plt.plot(x, prior, color='#e74c3c', linewidth=3, label=f'Prior ~ Beta({prior_alpha},{prior_beta})')
plt.plot(x, likelihood, color='#3498db', linewidth=3, label=f'Likelihood (данные)')
plt.plot(x, posterior, color='#2ecc71', linewidth=4, label=f'Posterior ~ Beta({posterior_alpha},{posterior_beta})')

# Аннотации и оформление
plt.title('Байесовская калибровка вероятности аномалий', fontsize=14, pad=15)
plt.xlabel('Вероятность нормального измерения (θ)', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Вертикальные линии мод
for dist, color, name in [(prior, '#e74c3c', 'Prior'),
                         (likelihood, '#3498db', 'Likelihood'),
                         (posterior, '#2ecc71', 'Posterior')]:
    mode = x[np.argmax(dist)]
    plt.axvline(mode, color=color, linestyle=':', alpha=0.7)
    plt.text(mode, np.max(dist)*1.05, f'{mode:.3f}',
             ha='center', color=color, fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

# Заполнение и легенда
plt.fill_between(x, prior, color='#e74c3c', alpha=0.1)
plt.fill_between(x, likelihood, color='#3498db', alpha=0.1)
plt.fill_between(x, posterior, color='#2ecc71', alpha=0.1)
plt.legend(framealpha=1, shadow=True)

plt.tight_layout()
plt.show()
```

Как и предполагалось, наша апостериорная вероятность практически идентична правдоподобию, поэтому калибровка не сильно повлияла на неё. Это также показывает влияние равномерной приорной вероятности на апостериорную вероятность.

В следующем месяце у нас больше данных, и теперь наша приорная вероятность представляет собой апостериорную вероятность предыдущего месяца. Аналогично, мы могли бы получить некоторую информацию о внутренней системе и соответствующим образом скорректировать априорную вероятность.

Предположив, что MM Manufacturing обратила внимание и внесла некоторые изменения в систему, теперь только 6% весов являются аномальными. Теперь у нас более информативная приорная вероятность, учитывая апостериорную вероятность из наших предыдущих данных.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import beta

np.random.seed(42)
normal_instances = [random.uniform(4.85, 5.15) for i in range(470)]
anomalous_instances_1 = [random.uniform(3, 4.85) for i in range(15)]
anomalous_instances_2 = [random.uniform(5.15, 6) for i in range(15)]

data = np.concatenate((normal_instances, anomalous_instances_1, anomalous_instances_2))

# Initial prior belief about normal behavior using a Beta distribution
prior_alpha = posterior_alpha
prior_beta = posterior_beta

# Beta Distribution as inferred by Observed data
likelihood_alpha = len(data[(data >= 4.85) & (data <= 5.15)]) + 1
likelihood_beta = len(data) - likelihood_alpha + 1

# Calculate posterior parameters based on observed data and prior
posterior_alpha = prior_alpha + likelihood_alpha
posterior_beta = prior_beta + likelihood_beta

# Plot the prior, likelihood and posterior Beta distributions
plt.figure(figsize=(18, 7))
x = np.linspace(0, 1, 1000)
prior_distribution = beta.pdf(x, prior_alpha, prior_beta)
likelihood_distribution = beta.pdf(x, likelihood_alpha, likelihood_beta)
posterior_distribution = beta.pdf(x, posterior_alpha, posterior_beta)

# Улучшенная визуализация
plt.plot(x, prior_distribution, 'r-', linewidth=3, alpha=0.7,
         label=f'Prior ~ Beta({prior_alpha},{prior_beta})')
plt.plot(x, likelihood_distribution, 'b-', linewidth=3, alpha=0.7,
         label=f'Likelihood (данные)')
plt.plot(x, posterior_distribution, 'g-', linewidth=4,
         label=f'Posterior ~ Beta({posterior_alpha},{posterior_beta})')

# Добавляем моды распределений
for dist, color in [(prior_distribution, 'red'),
                   (likelihood_distribution, 'blue'),
                   (posterior_distribution, 'green')]:
    mode = x[np.argmax(dist)]
    plt.axvline(mode, color=color, linestyle='--', alpha=0.5)
    plt.text(mode, np.max(dist)*1.05, f'{mode:.3f}',
             ha='center', color=color, fontsize=10)

# Улучшаем оформление
plt.title('Байесовская калибровка вероятности аномалий', fontsize=14, pad=15)
plt.xlabel('Вероятность нормального измерения (θ)', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)

# Улучшенная легенда
legend = plt.legend(frameon=True, framealpha=1,
                   facecolor='white', edgecolor='gray',
                   fontsize=11, borderpad=1)
legend.get_frame().set_linewidth(1.0)

plt.tight_layout()
plt.show()
```

На этот раз мы видим влияние априорной вероятности на апостериорную вероятность и насколько более определённым стало распределение. Связь между априорной вероятностью, апостериорной вероятностью и вероятностью здесь гораздо более чётко видна.

Рассмотрев два описанных выше сценария, становится очевидным, что различные результаты можно использовать для получения информации о производительности системы, проведения сравнений для наблюдения улучшений и улучшения калибровки данных в широком временном спектре.

Привлекательность бета-распределения заключается в его адаптивности и универсальности при определении различных распределений, в то время как сила байесовской калибровки заключается в ее способности гибко охватывать и интегрировать сложные параметры модели.

##### Шаг 2. Построение модели-бейзлайна

Траектория кумулятивного числа случаев выглядит экспоненциальной. Это соответствует данным эпидемиологии, поскольку на ранних стадиях эпидемии оно растёт экспоненциально.

```python
t = df_country["days_since_100"].to_numpy()
confirmed = df_country["confirmed"].to_numpy()

with pm.Model() as model_exp1:
    # Intercept
    a = pm.Normal("a", mu=0, sigma=100)

    # Slope
    b = pm.Normal("b", mu=0.3, sigma=0.3)

    # Exponential regression
    growth = a * (1 + b) ** t

    # Error term
    eps = pm.HalfNormal("eps", 100)

    # Likelihood
    pm.Normal("obs", mu=growth, sigma=eps, observed=confirmed)
```

Давайте улучшим нашу модель. Наличие отрицательных случаев обусловлено использованием нормального распределения правдоподобия. Вместо этого давайте используем отрицательное биномиальное распределение (`NegativeBinomial`), которое похоже на пуассоновское, но имеет дополнительный параметр дисперсии, обеспечивающий большую гибкость в моделировании дисперсии данных.

Мы также изменим априорное распределение свободного члена (intercept), центрировав его на 100, и сузим априорное распределение наклона.

Отрицательное биномиальное распределение использует параметр избыточной дисперсии, который мы опишем с помощью гамма-распределения. В сопутствующем пакете `preliz`, библиотеке для получения априорного распределения, есть удобная утилита `maxent`, которая поможет нам параметризовать это априорное распределение, поскольку гамма-распределение не так интуитивно понятно, как нормальное.

```python
dist = pz.Gamma()
pz.maxent(dist, lower=0.1, upper=20, mass=0.95);
```

```python
t = df_country["days_since_100"].to_numpy()
confirmed = df_country["confirmed"].to_numpy()


with pm.Model() as model_exp2:
    # Intercept
    a = pm.Normal("a", mu=100, sigma=25)

    # Slope
    b = pm.Normal("b", mu=0.3, sigma=0.1)

    # Exponential regression
    growth = a * (1 + b) ** t

    alpha_param = pz.maxent(pz.Gamma(), lower=0.1, upper=20, mass=0.95, plot=False).to_pymc("alpha")

    # Likelihood
    pm.NegativeBinomial("obs", growth, alpha=alpha_param, observed=confirmed)
```

```python
with model_exp2:
    prior_pred = pm.sample_prior_predictive(random_seed=RANDOM_SEED)

prior_obs = prior_pred.prior_predictive["obs"].values.squeeze().T

fig = go.Figure()
for i in range(min(100, prior_obs.shape[1])):  # Show max 100 traces
    fig.add_trace(
        go.Scatter(
            x=list(range(len(prior_obs[:, i]))),
            y=prior_obs[:, i],
            mode="lines",
            line=dict(color="steelblue", width=0.5),
            opacity=0.4,
            showlegend=False,
        )
    )

fig.update_layout(
    title="Prior predictive",
    xaxis_title="Days since 100 cases",
    yaxis_title="Positive cases",
    yaxis=dict(range=[-100, 1000]),
    xaxis=dict(range=[0, 10]),
    template="plotly_white",
)
```

Намного лучше! Нет отрицательных случаев, и шкала выглядит более разумной. Но мы всё равно можем включить больше знаний из предметной области.

Мы знаем, что:

* Свободный член не может быть меньше 100 (по построению наших данных)
* Темп роста должен быть положительным

Давайте закодируем это знание: мы используем полунормальное распределение (`HalfNormal`) для свободногочлена; то же самое можно применить и для наклона, который, как мы знаем, не будет отрицательным.

```python
with pm.Model() as model_exp3:
    # Intercept
    a0 = pm.HalfNormal("a0", sigma=25)
    a = pm.Deterministic("a", a0 + 100)

    # Slope
    b = pm.HalfNormal("b", sigma=0.2)

    # Exponential regression
    growth = a * (1 + b) ** t

    # Overdispersion parameter
    alpha = pz.maxent(pz.Gamma(), lower=0.1, upper=20, mass=0.95, plot=False).to_pymc("alpha")

    # Likelihood
    pm.NegativeBinomial("obs", growth, alpha=alpha, observed=confirmed)

    prior_pred3 = pm.sample_prior_predictive(random_seed=RANDOM_SEED)
```

##### Шаг 6. Проверка соответствия модели

Подобно априорному прогнозированию, мы также можем генерировать новые данные, многократно извлекая выборки из апостериорного распределения и генерируя данные с использованием этих параметров. Этот процесс называется апостериорной прогностической проверкой и является ключевым этапом валидации байесовской модели.

Апостериорная прогностическая проверка работает следующим образом:

1. Извлекаются выборки параметров из апостериорного распределения (которые у нас уже есть из выборки MCMC).
2. Для каждого набора значений параметров генерируются новые синтетические наборы данных с использованием той же функции правдоподобия, что и в нашей модели.
3. Сравниваются эти синтетические наборы данных с нашими наблюдаемыми данными.

Это позволяет нам оценить, может ли наша модель воспроизводить ключевые особенности наблюдаемых данных. Если апостериорные прогностические выборки сильно отличаются от наших фактических данных, это говорит о том, что наша модель может упускать важные аспекты процесса генерации данных. И наоборот, если апостериорные прогностические выборки хорошо охватывают наши наблюдаемые данные, это свидетельствует о том, что наша модель улавливает основные закономерности в данных.

* Обучение на наблюдаемых данных: подгонка модели с использованием доступных обучающих данных

Давайте продемонстрируем этот подход, модифицировав нашу модель экспоненциального роста для использования узлов pm.Data.

### Байесовский A/B Тест

Во многом байесовский сеттинг A/B тестирования полностью повторяет обычный частотный, единственные отличия - подходы к определению параметра теста и непосредственно его интерпретация.


**Определение Параметров**

Для байесовского A/B нет какой-то формулы, которая связывает MDE, ошибки 1 и 2 рода и минимальный размер выборки, поэтому в данном случае необходимую длительность необходимо выявлять для каждой отдельной модели на основе исторических или синтетических данных.

**Интерпретация**

Так как в байесовском фреймворке мы получаем распределение параметра, то оценка через обычные статистические критерии не сильно принима, зато мы можем напрямую сравнивать полученные распределения метрик для тестовых и контрольной групп. Таким образом самый базовый критерий принятия решения выглядит как
$$
P(Posterior_B > Posterior_A) >= 1- \alpha
$$

То есть с вероятность не менее $1 - alpha$ эффект в тестовой группе выше.

Так же выбывает полезно привязать выбранную метрику к чему-нибудь денежному. Например, если мы оценивали конверсии, то можем их подвязать к монетизации:
$$
\text{Monetization} = (\text{Per User Value}) \times (\text{Num Users}) \times \delta- (\text{Implementation Cost})
$$
Таким образом, мы получаем из $p(\delta|X)$ распределение с оценёнными шансами на монетизацию фичи $p(\text{Monetization}|X)$.


Есть более интересные метрики, например Ожидаемые потери {метрика} с одного человека при выборе конкретной группы, но они вне рассмотрения занятия (можно посмотреть [тут](https://towardsdatascience.com/bayesian-ab-testing-part-i-conversions-ac2635f878ec/) и [тут](https://towardsdatascience.com/bayesian-ab-testing-part-ii-revenue-1fbcf04f96cd/)).

#### Байесовский A/B на конверсиях

Давайте сначала рассмотрим простой двухвариантный A/B-тест, где интересующей нас метрикой является доля пользователей, совершивших действие (например, покупку хотя бы одного товара), то есть конверсия. Наши варианты обозначены как A и B, где A относится к существующей целевой странице, а B — к новой странице, которую мы хотим протестировать. Результат, который мы хотим получить, — это определить, «лучше» ли B, чем A, что зависит от базовых «истинных» коэффициентов конверсии для каждого варианта. Мы можем сформулировать это следующим образом:

Пусть $\theta_A, \theta_B$ — истинные коэффициенты конверсии для вариантов A и B соответственно. Тогда результат конверсии посетителя в варианте A — случайная величина
Bernoulli(\theta_A), а Bernoulli(\theta_B) — для варианта B. Если предположить, что поведение посетителей на целевой странице не зависит от других посетителей (справедливое предположение), то общее количество конверсий
для варианта имеет биномиальное распределение:
$$
y \sim \sum^N \text{Bernoulli}(\theta) = \text{Binomial}(N, \theta)
$$
В байесовской модели мира мы предполагаем, что истинные коэффициенты конверсии $\theta_A, \theta_B$ неизвестны, и вместо этого каждый из них подчиняется бета-распределению. Предполагается, что коэффициенты независимы (мы бы распределили трафик между каждым вариантом случайным образом, чтобы один вариант не влиял на другой):

$$
\theta_A \sim \theta_B \sim \text{Beta}(\alpha, \beta)
$$

Наблюдаемые данные за время A/B-теста (распределение правдоподобия) таковы: количество посетителей, попавших на страницу N, и количество посетителей, купивших хотя бы один товар y:
$$
y_A \sim \text{Binomial}(n=N_A, p=\theta_A), y_B \sim \text{Binomial}(n=N_B, p=\theta_B)
$$
Таким образом, мы можем сделать выборку из совместного апостериорного распределения $g(\theta_A,\theta_B)$.

бета-распределение является сопряженным априорным распределением для биномиального распределения, поэтому нам не нужна выборка MCMC для оценки апостериорного распределения. Тем не менее, мы всё же продемонстрируем, как можно реализовать выборку в PyMC3, и это упростит расширение модели с использованием различных априорных данных, предположений о зависимостях и т. д.

Наконец, помните, что нас интересует результат, лучше ли B, чем A. На практике общепринятой мерой превосходства B является относительное повышение коэффициентов конверсии, т. е. процентная разница $\theta_B$
по сравнению с $\theta_A$:
$$
\text{reluplift}_B = \theta_B/ \theta_A -1
$$

```python
@dataclass
class BetaPrior:
    alpha: float
    beta: float

@dataclass
class BinomialData:
    trials: int
    successes: int

class ConversionModelTwoVariant:
    def __init__(self, priors: BetaPrior):
        self.priors = priors

    def create_model(self, data: List[BinomialData]) -> pm.Model:
        trials = [d.trials for d in data]
        successes = [d.successes for d in data]
        with pm.Model() as model:
            p = pm.Beta("p", alpha=self.priors.alpha, beta=self.priors.beta, shape=2)
            obs = pm.Binomial("y", n=trials, p=p, shape=2, observed=successes)
            reluplift = pm.Deterministic("reluplift_b", p[1] / p[0] - 1)
        return model
```

Теперь, когда мы определили класс, принимающий в качестве входных данных априорное распределение и наши синтетические данные, первым шагом будет выбор подходящего априорного распределения. При практическом применении необходимо учитывать несколько моментов, но в рамках данной работы мы сосредоточимся на следующем:

* Мы предполагаем, что для каждого варианта задано одинаковое априорное бета-распределение.

* Неинформативное или слабоинформативное априорное распределение возникает, когда мы задаем низкие значения альфа и бета. Например, $\alpha$ = 1, $\beta$ = 1 приводит к равномерному распределению в качестве априорного распределения. Если бы мы рассматривали одно распределение изолированно, установка этого априорного распределения означала бы, что мы ничего не знаем ни о значении параметра, ни о нашей уверенности в нём. Однако в контексте A/B-тестирования нас интересует сравнение относительного прироста одного варианта по сравнению с другим. При слабоинформативном априорном бета-распределении это относительное прирост распределения очень широкое, поэтому мы неявно говорим, что варианты могут сильно различаться.

* Сильное априорное распределение возникает, когда мы задаем высокие значения $\alpha$, $\beta$. В отличие от вышесказанного, сильное априорное распределение подразумевает, что относительное распределение эффекта невелико, т.е. наше априорное предположение заключается в том, что варианты не сильно отличаются друг от друга.

Мы иллюстрируем эти положения с помощью априорных прогностических проверок.

##### Смотрим на разные приоры

```python
weak_prior = ConversionModelTwoVariant(BetaPrior(alpha=100, beta=100))
strong_prior = ConversionModelTwoVariant(BetaPrior(alpha=10000, beta=10000))

with weak_prior.create_model(data=[BinomialData(1, 1), BinomialData(1, 1)]):
    weak_prior_predictive = pm.sample_prior_predictive(samples=10000)

with strong_prior.create_model(data=[BinomialData(1, 1), BinomialData(1, 1)]):
    strong_prior_predictive = pm.sample_prior_predictive(samples=10000)

fig, axs = plt.subplots(2, 1, figsize=(14, 7), sharex=True)
az.plot_posterior(weak_prior_predictive.prior["reluplift_b"], textsize=10, ax=axs[0], kind="hist")
axs[0].set_title(f"B vs. A аплифт", fontsize=10)
axs[0].axvline(x=0, color="red")
az.plot_posterior(strong_prior_predictive.prior["reluplift_b"], textsize=10, ax=axs[1], kind="hist")
axs[1].set_title(f"B vs. A аплифт", fontsize=10)
axs[1].axvline(x=0, color="red")
```

При слабом априорном распределении наш 94% HDI для относительного прироста вероятности для B по сравнению с A составляет примерно [-20%, +20%], тогда как при сильном априорном распределении он составляет примерно [-2%, +2%]. Это фактически «отправная точка» для распределения относительного прироста вероятности и повлияет на то, как наблюдаемые конверсии преобразуются в апостериорное распределение.

На практике выбор этих априорных распределений зависит от более широкого контекста компании, проводящей A/B-тестирование. Сильное априорное распределение может помочь предотвратить ложные прокрасы, но может потребовать больше данных для обнаружения выигрышных вариантов, если они существуют (и чем больше данных, тем больше времени потребуется для проведения теста). Слабое априорное распределение придает больший вес наблюдаемым данным, но также может привести к большему количеству ложных прокрасов из-за проблем с остановкой на ранней стадии.

Ниже мы рассмотрим результаты вывода, полученные при двух различных вариантах априорного распределения.

```python
def generate_binomial_data(
    variants: List[str], true_rates: List[str], samples_per_variant: int = 100000
) -> pd.DataFrame:
    data = {}
    for variant, p in zip(variants, true_rates):
        data[variant] = bernoulli.rvs(p, size=samples_per_variant)
    agg = (
        pd.DataFrame(data)
        .aggregate(["count", "sum"])
        .rename(index={"count": "trials", "sum": "successes"})
    )
    return agg
```

```python
true_rates = [0.23, 0.23]
weak_prior = BetaPrior(alpha=10, beta=10)
strong_prior = BetaPrior(alpha=1000, beta=1000)

generated = generate_binomial_data(variants = ['A', 'B'],
                                 true_rates = true_rates,
                                 samples_per_variant= 100_000)
data = [BinomialData(**generated[v].to_dict()) for v in ['A', 'B']]
with ConversionModelTwoVariant(priors=weak_prior).create_model(data):
    trace_weak = pm.sample(draws=5000, cores=4, chains=4)
with ConversionModelTwoVariant(priors=strong_prior).create_model(data):
    trace_strong = pm.sample(draws=5000, cores=4, chains=4)


true_rel_uplift = true_rates[1] / true_rates[0] - 1

fig, axs = plt.subplots(2, 1, figsize=(14, 7), sharex=True)
az.plot_posterior(trace_weak.posterior["reluplift_b"], textsize=10, ax=axs[0], kind="hist")
axs[0].set_title(f"True Rel Uplift = {true_rel_uplift:.1%}, {weak_prior}", fontsize=10)
axs[0].axvline(x=0, color="red")
az.plot_posterior(trace_strong.posterior["reluplift_b"], textsize=10, ax=axs[1], kind="hist")
axs[1].set_title(f"True Rel Uplift = {true_rel_uplift:.1%}, {strong_prior}", fontsize=10)
axs[1].axvline(x=0, color="red")
fig.suptitle("B vs. A Rel Uplift")
```

* В обоих случаях истинный прирост в 0% лежит в пределах 94% HDI.

* Затем мы можем использовать это относительное распределение прироста, чтобы принять решение о применении новой целевой страницы/функций в варианте B по умолчанию. Например, мы можем решить, что если 94% HDI выше 0, мы катим вариант B. В этом случае HDI включает 0, поэтому решение будет не ввыкатывать вариант B.

#### Байесовский A/B на выручке

А что, если мы хотим сравнить варианты A/B-тестирования с точки зрения генерируемого ими дохода и/или оценить, какой дополнительный доход приносит выигрышный вариант? Мы не можем использовать для этого бета-биномиальную модель, поскольку возможные значения для каждого посетителя теперь находятся в диапазоне $[0 \text{ ;} \infty]$. Пусть модель выглядит следующим образом:

Доход, генерируемый отдельным посетителем, равен доходу = вероятности покупки * средняя сумма покупки:
$$
\text{Revenue}_{i} = \text{Bernoulli}(\theta)_i \times \text{Exponential}(\lambda)_{i} \times \mathbb{I}(\text{Bernoulli}(\theta)_i = 1)
$$

Мы предполагаем, что вероятность оплаты не зависит от средней суммы, потраченной при оплате. Это типичное предположение на практике, если только у нас нет оснований полагать, что эти два параметра имеют взаимозависимость. Таким образом, мы можем создать отдельные модели для общего числа посетителей, совершающих покупки, и общей суммы, потраченной посетителями, совершающими покупки (предполагая независимость поведения каждого посетителя):
$$
c \sim \sum^N\text{Bernoulli}(\theta) = \text{Binomial}(N, \theta)
$$
$$
r \sim \sum^K\text{Exponential}(\lambda) = \text{Gamma}(K, \lambda)
$$

где
$N$ — общее число посетителей,
$K$ — общее число посетителей, совершивших хотя бы одну покупку.

Мы можем повторно использовать нашу бета-биномиальную модель, разработанную ранее, для моделирования конверсий. Для средней суммы покупки мы используем гамма-распределение (которое также является сопряженным априорным распределением гамма-вероятности). Таким образом, сетап выглядит следующим образом:
$$
\theta_A \sim \theta_B \sim \text{Beta}(\alpha_1, \beta_1) \\
\lambda_A \sim \lambda_B \sim \text{Gamma}(\alpha_2, \beta_2) \\
c_A \sim \text{Binomial}(N_A, \theta_A), c_B \sim \text{Binomial}(N_B, \theta_B) \\
r_A \sim \text{Gamma}(c_A, \lambda_A), r_B \sim \text{Gamma}(c_B, \lambda_B) \\
\mu_A = \theta_A \times \frac{1}{\lambda_A}, \mu_B = \theta_B \times \frac{1}{\lambda_B} \\
\text{reluplift}_B =\mu_B/\mu_A -1
$$

Здесь  $\mu$ представляет средний доход на пользователя, включая тех, кто не совершил покупку. Это лучший способ оценить общий эффект от выручки — некоторые варианты могут увеличить средний объём продаж, но уменьшить долю посетителей, которые вообще платят (например, если мы продвигали более дорогие товары на целевой странице).

```python
@dataclass
class GammaPrior:
    alpha: float
    beta: float

@dataclass
class RevenueData:
    visitors: int
    purchased: int
    total_revenue: float

class RevenueModel:
    def __init__(self, conversion_rate_prior: BetaPrior, mean_purchase_prior: GammaPrior):
        self.conversion_rate_prior = conversion_rate_prior
        self.mean_purchase_prior = mean_purchase_prior

    def create_model(self, data: List[RevenueData], comparison_method: str) -> pm.Model:
        num_variants = len(data)
        visitors = [d.visitors for d in data]
        purchased = [d.purchased for d in data]
        total_revenue = [d.total_revenue for d in data]

        with pm.Model() as model:
            theta = pm.Beta(
                "theta",
                alpha=self.conversion_rate_prior.alpha,
                beta=self.conversion_rate_prior.beta,
                shape=num_variants,
            )
            lam = pm.Gamma(
                "lam",
                alpha=self.mean_purchase_prior.alpha,
                beta=self.mean_purchase_prior.beta,
                shape=num_variants,
            )
            converted = pm.Binomial(
                "converted", n=visitors, p=theta, observed=purchased, shape=num_variants
            )
            revenue = pm.Gamma(
                "revenue", alpha=purchased, beta=lam, observed=total_revenue, shape=num_variants
            )
            revenue_per_visitor = pm.Deterministic("revenue_per_visitor", theta * (1 / lam))
            theta_reluplift = []
            reciprocal_lam_reluplift = []
            reluplift = []
            for i in range(num_variants):
                if comparison_method == "compare_to_control":
                    comparison_theta = theta[0]
                    comparison_lam = 1 / lam[0]
                    comparison_rpv = revenue_per_visitor[0]
                elif comparison_method == "best_of_rest":
                    others_theta = [theta[j] for j in range(num_variants) if j != i]
                    others_lam = [1 / lam[j] for j in range(num_variants) if j != i]
                    others_rpv = [revenue_per_visitor[j] for j in range(num_variants) if j != i]
                    if len(others_rpv) > 1:
                        comparison_theta = pmm.maximum(*others_theta)
                        comparison_lam = pmm.maximum(*others_lam)
                        comparison_rpv = pmm.maximum(*others_rpv)
                    else:
                        comparison_theta = others_theta[0]
                        comparison_lam = others_lam[0]
                        comparison_rpv = others_rpv[0]
                else:
                    raise ValueError(f"comparison method {comparison_method} not recognised.")
                theta_reluplift.append(
                    pm.Deterministic(f"theta_reluplift_{i}", theta[i] / comparison_theta - 1)
                )
                reciprocal_lam_reluplift.append(
                    pm.Deterministic(
                        f"reciprocal_lam_reluplift_{i}", (1 / lam[i]) / comparison_lam - 1
                    )
                )
                reluplift.append(
                    pm.Deterministic(f"reluplift_{i}", revenue_per_visitor[i] / comparison_rpv - 1)
                )
        return model
```

```python
c_prior = BetaPrior(alpha=5000, beta=5000)
mp_prior = GammaPrior(alpha=9000, beta=900)

data = [
    RevenueData(visitors=10000, purchased=1000, total_revenue=70000),
    RevenueData(visitors=10000, purchased=1000, total_revenue=70000),
]

with RevenueModel(c_prior, mp_prior).create_model(data, "best_of_rest"):
    revenue_prior_predictive = pm.sample_prior_predictive(samples=10000)
```

Подобно модели для конверсий, ширина априорного прогнозного распределения будет зависеть от силы наших априорных данных.

Далее мы сгенерируем синтетические данные для модели. Мы сгенерируем следующие сценарии:

* Одинаковая склонность к покупке и одинаковая средняя стоимость покупки.

* Низкая склонность к покупке и более высокая средняя стоимость покупки, но общий доход на посетителя тот же.

* Более высокая склонность к покупке и более высокая средняя стоимость покупки, а также общий более высокий доход на посетителя.

```python
def generate_revenue_data(
    variants: List[str],
    true_conversion_rates: List[float],
    true_mean_purchase: List[float],
    samples_per_variant: int,
) -> pd.DataFrame:
    converted = {}
    mean_purchase = {}
    for variant, p, mp in zip(variants, true_conversion_rates, true_mean_purchase):
        converted[variant] = bernoulli.rvs(p, size=samples_per_variant)
        mean_purchase[variant] = expon.rvs(scale=mp, size=samples_per_variant)
    converted = pd.DataFrame(converted)
    mean_purchase = pd.DataFrame(mean_purchase)
    revenue = converted * mean_purchase
    agg = pd.concat(
        [
            converted.aggregate(["count", "sum"]).rename(
                index={"count": "visitors", "sum": "purchased"}
            ),
            revenue.aggregate(["sum"]).rename(index={"sum": "total_revenue"}),
        ]
    )
    return agg

def run_scenario_value(
    variants: List[str],
    true_conversion_rates: List[float],
    true_mean_purchase: List[float],
    samples_per_variant: int,
    conversion_rate_prior: BetaPrior,
    mean_purchase_prior: GammaPrior,
    comparison_method: str,
) -> az.InferenceData:
    generated = generate_revenue_data(
        variants, true_conversion_rates, true_mean_purchase, samples_per_variant
    )
    data = [RevenueData(**generated[v].to_dict()) for v in variants]
    with RevenueModel(conversion_rate_prior, mean_purchase_prior).create_model(
        data, comparison_method
    ):
        trace = pm.sample(draws=5000, return_inferencedata=True, chains=2, cores=4)

    n_plots = len(variants)
    fig, axs = plt.subplots(nrows=n_plots, ncols=1, figsize=(16, 7), sharex=True)
    for i, variant in enumerate(variants):
        if i == 0 and comparison_method == "compare_to_control":
            axs[i].set_yticks([])
        else:
            az.plot_posterior(
                trace.posterior[f"reluplift_{i}"], textsize=10, ax=axs[i], kind="hist"
            )
        true_rpv = true_conversion_rates[i] * true_mean_purchase[i]
        axs[i].set_title(f"Rel Uplift {variant}, True RPV = {true_rpv:.2f}", fontsize=10)
        axs[i].axvline(x=0, color="red")
    fig.suptitle(f"Method {comparison_method}, {conversion_rate_prior}, {mean_purchase_prior}")

    return trace
```

#### Байесовский A/B/C на конверсиях

Здесь обратим внимание анализу тестов с тремя и более вариантами — например, вместо того, чтобы тестировать одну целевую страницу, мы хотим протестировать несколько идей одновременно. Как определить, есть ли среди них победитель?

Здесь можно использовать два основных подхода:

* Возьмём A в качестве «контрольного варианта». Сравниваем другие варианты (B, C и т. д.) с A по одному.

* Для каждого варианта сравниваем с max() других вариантов.

Подход 1 интуитивно понятен большинству и легко объясним. Но что, если есть два варианта, которые оба превосходят контрольный, и мы хотим узнать, какой из них лучше? Мы не можем сделать такой вывод, используя отдельные распределения прироста. Подход 2 справляется с этим случаем — он фактически пытается определить, есть ли явный победитель или явный проигравший среди всех вариантов.

Ниже мы реализуем настройку модели для обоих подходов, очистив наш предыдущий код, чтобы он был обобщён на случай с n вариантами. Обратите внимание, что мы также можем использовать эту модель для случая с 2 вариантами.

```python
class ConversionModel:
    def __init__(self, priors: BetaPrior):
        self.priors = priors

    def create_model(self, data: List[BinomialData], comparison_method) -> pm.Model:
        num_variants = len(data)
        trials = [d.trials for d in data]
        successes = [d.successes for d in data]
        with pm.Model() as model:
            p = pm.Beta("p", alpha=self.priors.alpha, beta=self.priors.beta, shape=num_variants)
            y = pm.Binomial("y", n=trials, p=p, observed=successes, shape=num_variants)
            reluplift = []
            for i in range(num_variants):
                if comparison_method == "compare_to_control":
                    comparison = p[0]
                elif comparison_method == "best_of_rest":
                    others = [p[j] for j in range(num_variants) if j != i]
                    if len(others) > 1:
                        comparison = pmm.maximum(*others)
                    else:
                        comparison = others[0]
                else:
                    raise ValueError(f"comparison method {comparison_method} not recognised.")
                reluplift.append(pm.Deterministic(f"reluplift_{i}", p[i] / comparison - 1))
        return model

def run_scenario_bernoulli(
    variants: List[str],
    true_rates: List[float],
    samples_per_variant: int,
    priors: BetaPrior,
    comparison_method: str,
) -> az.InferenceData:
    generated = generate_binomial_data(variants, true_rates, samples_per_variant)
    data = [BinomialData(**generated[v].to_dict()) for v in variants]
    with ConversionModel(priors).create_model(data=data, comparison_method=comparison_method):
        trace = pm.sample(draws=5000, return_inferencedata=True, chains=2, cores=4)

    n_plots = len(variants)
    fig, axs = plt.subplots(nrows=n_plots, ncols=1, figsize=(16, 7), sharex=True)
    for i, variant in enumerate(variants):
        if i == 0 and comparison_method == "compare_to_control":
            axs[i].set_yticks([])
        else:
            az.plot_posterior(
                trace.posterior[f"reluplift_{i}"], textsize=10, ax=axs[i], kind="hist"
            )
        axs[i].set_title(f"Rel Uplift {variant}, True Rate = {true_rates[i]:.2%}", fontsize=10)
        axs[i].axvline(x=0, color="red")
    fig.suptitle(f"Method {comparison_method}, {priors}")

    return trace
```

```python
# Подход 1
_ = run_scenario_bernoulli(
    variants=["A", "B", "C"],
    true_rates=[0.21, 0.23, 0.228],
    samples_per_variant=100000,
    priors=BetaPrior(alpha=5000, beta=5000),
    comparison_method="compare_to_control",
)
```

* Относительный аплифт для вариантов B и C показывает, что они явно лучше вариантов A (94% HDI значительно выше 0), примерно на 7–8% относительно.

* Однако мы не можем сделать вывод о наличии победителя между вариантами B и C, т.к. их HDI крайне сильно пересекаются.


```python
# Подход 2
_ = run_scenario_bernoulli(
    variants=["A", "B", "C"],
    true_rates=[0.21, 0.23, 0.228],
    samples_per_variant=100000,
    priors=BetaPrior(alpha=5000, beta=5000),
    comparison_method="best_of_rest",
)
```

* График для варианта A показывает, что он явно проигрывает вариантам B и C (94% HDI для относительного подъёма для варианта A значительно ниже 0).

* Обратите внимание, что расчёт относительного аплифта для вариантов B и C фактически игнорирует вариант A. Это связано с тем, что, например, при расчёте относительного подъёма для варианта B максимальным из других вариантов, вероятно, будет вариант C. Аналогично, при расчёте относительного подъёма для варианта C он, вероятно, сравнивается с вариантом B.

* Графики для вариантов B и C показывают, что мы пока не можем назвать явного победителя среди двух вариантов, поскольку 94% HDI всё ещё пересекается с 0. Для обнаружения разницы в коэффициентах конверсии 23% и 22,8% нам потребуется выборка большего размера.

* Одним из недостатков этого подхода является то, что мы не можем напрямую сказать, каков аплифт этих вариантов по сравнению с вариантом A (контрольным). Это число часто важно на практике, поскольку позволяет оценить общее влияние изменений, полученных в ходе A/B-тестирования, на всех пользователей. Однако мы можем получить это число приблизительно, переформулировав вопрос следующим образом: «Насколько вариант A хуже двух других?» (что показано в распределении относительного прироста для варианта A).
